## <span style="color:#2E86C1;">Introduction</span>

La **grande distribution** est un secteur hautement concurrentiel où la capacité à **anticiper la demande** constitue un avantage stratégique majeur. Walmart, en tant que leader mondial du commerce de détail, génère chaque semaine des volumes considérables de données transactionnelles à travers ses magasins répartis sur différentes régions.

Ce projet, intitulé **« Système Intelligent de Prévision des Ventes en Grande Distribution »**, vise à concevoir et déployer un pipeline complet de science des données pour **prédire les ventes hebdomadaires** de Walmart. En exploitant des données historiques enrichies de variables contextuelles — telles que la température, le prix du carburant, le taux de chômage, l'indice des prix à la consommation (CPI) et les périodes promotionnelles (*markdown*) — nous chercherons à construire des modèles prédictifs robustes et interprétables.

### Objectifs principaux

- **Comprendre** le contexte métier de la grande distribution et les facteurs qui influencent les ventes.
- **Explorer et préparer** les données en identifiant les anomalies, les valeurs manquantes et les relations entre variables.
- **Modéliser** les ventes à l'aide d'algorithmes de Machine Learning (régression, arbres de décision, forêts aléatoires, gradient boosting, etc.).
- **Évaluer** la performance des modèles à travers des métriques adaptées (MAE, RMSE, WMAE).
- **Exploiter** les résultats pour formuler des recommandations stratégiques à destination des décideurs.

### Jeux de données utilisés

| Fichier | Description |
|---|---|
| `train.csv` | Données d'entraînement : ventes hebdomadaires par magasin et département (2010-02 à 2012-10) |
| `test.csv` | Données de test pour la prédiction (2012-11 à 2013-07) |
| `stores.csv` | Informations sur les magasins (type et superficie) |
| `features.csv` | Variables contextuelles (température, carburant, CPI, chômage, promotions) |

### Démarche adoptée

Le projet est structuré en **trois grandes parties** :
1. **Compréhension et Préparation des Données** — Exploration, nettoyage et ingénierie des variables.
2. **Modélisation et Évaluation Prédictive** — Entraînement, optimisation et comparaison des modèles.
3. **Intelligence Décisionnelle et Exploitation** — Interprétation des résultats et recommandations stratégiques.

## <span id="11" style="color:#2E86C1;">1.1 Comprendre le contexte retail Walmart</span>

Le secteur du **retail (commerce de détail)** repose sur la vente directe de produits aux consommateurs à travers des magasins physiques ou des plateformes en ligne. Dans ce contexte, les entreprises génèrent de grandes quantités de données liées aux ventes, aux clients et aux opérations. L’analyse de ces données permet d’améliorer la prise de décision, d’optimiser les stocks et d’anticiper la demande.

**Walmart**, l’un des plus grands détaillants au monde, produit chaque jour un volume massif de données grâce à ses milliers de magasins et ses millions de transactions. Ces données constituent un contexte pertinent pour appliquer des méthodes de **science des données et de prévision des ventes**.

Dans ce projet, l’objectif est d’analyser les **données historiques de ventes** afin de comprendre les facteurs influençant la demande et de développer des **modèles capables de prédire les ventes futures**, permettant ainsi d’améliorer la gestion des stocks et la planification commerciale.

### <span id="111" style="color:#2E86C1;">1.1.1 Analyse du fonctionnement de la grande distribution</span>

La grande distribution constitue l'un des secteurs les plus complexes à piloter sur le plan analytique, en raison de la multiplicité des variables qui influencent la performance commerciale. Walmart, en tant que leader mondial de la distribution avec plus de 10 000 points de vente dans le monde, illustre parfaitement les défis inhérents à ce modèle économique.
### Modèle économique et logique de volume
Le modèle économique de la grande distribution repose sur un principe fondamental : générer de la rentabilité par la rotation rapide des stocks plutôt que par des marges unitaires élevées. Walmart applique cette logique à travers sa stratégie historique "Everyday Low Prices" (EDLP), qui privilégie des prix bas permanents à des promotions ponctuelles. Dans ce contexte, la capacité à anticiper précisément la demande devient un avantage concurrentiel majeur : une rupture de stock signifie une vente perdue et une dégradation de l'expérience client, tandis qu'un surstock génère des coûts de stockage, des démarques et potentiellement des destructions de marchandise.
### Organisation par départements et hétérogénéité des ventes
Au sein de chaque magasin, les ventes sont enregistrées à la granularité Store × Département × Semaine, soit une combinaison de 45 magasins et 81 départements couvrant l'ensemble des rayons — épicerie, électronique, habillement, jardinage, automobile, pharmacie, entre autres. Cette granularité fine révèle une forte hétérogénéité interne : certains départements, tels que le département 92 (épicerie générale) ou le département 95 (jardinage), concentrent une part disproportionnée du chiffre d'affaires total, tandis que d'autres affichent des volumes marginaux mais des profils saisonniers très marqués. Cette disparité impose de traiter chaque combinaison Store × Département comme une entité à part entière dans la modélisation.
### Les leviers opérationnels de la grande distribution
La performance d'un distributeur comme Walmart repose sur la maîtrise de trois leviers opérationnels interdépendants, qui constituent précisément les axes de décision que ce projet vise à soutenir.
### 1. Gestion des stocks et réapprovisionnement
La planification des stocks s'appuie sur l'anticipation de la demande future. Une prévision inexacte se traduit immédiatement par un impact financier mesurable : en cas de sous-estimation, la rupture de stock prive le magasin de ventes et dégrade la satisfaction client ; en cas de surestimation, le surstock mobilise des ressources logistiques, génère des coûts de stockage et conduit à des démarques souvent subies. Dans ce projet, l'objectif est de réduire les ruptures de stock de 20% grâce à une meilleure anticipation des volumes hebdomadaires.
### 2. Politique promotionnelle et gestion des MarkDowns
Walmart recourt à des remises temporaires, appelées MarkDowns, pour stimuler les ventes sur certains rayons, écouler des stocks excédentaires ou répondre à des dynamiques concurrentielles. Ces promotions génèrent des pics de demande dont l'intensité varie selon le type de magasin, le département concerné et la période de l'année. Anticiper l'effet d'un MarkDown sur les ventes hebdomadaires est donc un enjeu majeur pour dimensionner correctement les stocks en amont et maximiser le retour sur investissement promotionnel.
### 3. Planification des achats et gestion des délais fournisseurs
Les décisions d'achat sont structurellement prises plusieurs semaines avant la mise en rayon, en raison des délais logistiques et des contraintes fournisseurs. Elles doivent intégrer à la fois les événements prévisibles — fêtes nationales, variations saisonnières — et le contexte économique local propre à chaque zone de chalandise. Une planification des achats fondée sur des prévisions fiables permet de sécuriser les approvisionnements aux moments critiques tout en limitant les immobilisations de trésorerie.
### 4. Les facteurs externes structurant la demande
Au-delà des décisions internes, les ventes en grande distribution sont étroitement corrélées à des variables contextuelles que le dataset Walmart capture de manière explicite. Les événements de fêtes nationales, représentés par la variable IsHoliday, provoquent des pics de ventes très marqués pouvant doubler la moyenne hebdomadaire, comme c'est le cas lors de Thanksgiving en semaine 47. L'indice des prix à la consommation (CPI) influence les arbitrages d'achat des ménages, notamment sur les catégories non-essentielles. Le taux de chômage local (Unemployment) accroît la sensibilité au prix dans les zones économiquement fragiles. Le prix du carburant (Fuel_Price) affecte quant à lui la fréquence et le périmètre des déplacements en magasin. Enfin, la température (Temperature) conditionne la saisonnalité de certains rayons comme le jardinage, les articles de saison ou l'habillement.
La prise en compte de l'ensemble de ces facteurs dans le modèle prédictif est indispensable pour capturer la variabilité réelle des ventes et produire des prévisions robustes.

### <span id="112" style="color:#2E86C1;">1.1.2 Identification des décisions à soutenir (stocks, promotions...)</span>

### 1) Décisions “Stocks” à soutenir (réassort & allocation)

### A. Réassort (quand et combien commander)

- Question : “Combien commander pour Store X / Dept Y la semaine prochaine (ou 2–4 semaines) ?”

- Sortie attendue : prévision des ventes + quantité recommandée (réassort).

- KPI : taux de rupture (stockout), taux de service, surstock, coût de stockage.

>Les décisions de réassort consistent à déterminer quand et combien de produits commander pour chaque magasin et département. Grâce aux prévisions de ventes, il est possible d’estimer les quantités nécessaires pour les semaines à venir. L’objectif est d’éviter les ruptures de stock tout en limitant le surstock et les coûts de stockage.

### B. Stock de sécurité (safety stock)

- Question : “Quel stock minimum garder pour éviter les ruptures pendant pics saisonniers / holidays ?”

- Sortie : niveau de stock de sécurité recommandé par dept/magasin (selon volatilité).

- KPI : ruptures évitées, stabilité de disponibilité.

>Le stock de sécurité correspond à la quantité minimale de produits à garder pour éviter les ruptures de stock, surtout pendant les périodes de forte demande comme les fêtes ou les événements saisonniers. En analysant la variabilité des ventes, on peut recommander un niveau de stock de sécurité adapté pour chaque magasin et département. L’objectif est de maintenir une bonne disponibilité des produits pour les clients.

### C. Allocation inter-magasins (prioriser où envoyer le stock)

- Question : “Si stock limité, quels magasins doivent être servis en priorité ?”

- Sortie : classement des magasins/depts selon demande prévue + risque de rupture.

- KPI : pertes de ventes évitées, disponibilité globale.

>Lorsque le stock est limité, il est important de décider quels magasins doivent être approvisionnés en priorité. En utilisant les prévisions de ventes, on peut identifier les magasins et départements où la demande sera la plus forte et où le risque de rupture est élevé. Cela permet d’envoyer les produits là où ils seront le plus nécessaires et de réduire les pertes de ventes.

### D. Préparation des pics (événements & périodes de fêtes)

- Question : “Quelles semaines vont exploser (Noël, Thanksgiving, etc.) et sur quels depts ?”

- Sortie : calendrier de pics + “top depts à risque”.

- KPI : précision des pics, réduction des ruptures en périodes critiques.

>Les prévisions permettent aussi d’identifier les périodes où les ventes vont fortement augmenter, comme pendant les fêtes ou certains événements saisonniers. En repérant ces semaines à l’avance et les départements concernés, Walmart peut mieux préparer ses stocks. Cela aide à éviter les ruptures de produits pendant les périodes de forte demande.

### 2) Décisions “Promotions” à soutenir (planification & évaluation)

### E. Planifier les promos (quand lancer / sur quels produits)

- Question : “Quelle promo faire, à quelle semaine, dans quel magasin/département ?”

- Sortie : recommandation de périodes “propices” (là où la demande est sensible / saisonnière).

- KPI : uplift ventes (l’augmentation des ventes grâce à la promotion), marge, ROI promo(le retour sur investissement de la promotion).

>Les prévisions de ventes peuvent aussi aider à planifier les promotions. En analysant la saisonnalité et la sensibilité de la demande, il devient possible d’identifier les semaines et les départements où une promotion serait la plus efficace. Cela permet d’augmenter les ventes tout en optimisant la marge et le retour sur investissement des campagnes promotionnelles.

### F. Mesurer l’impact des promotions (uplift)

- Question : “La promo a-t-elle réellement augmenté les ventes ou juste déplacé la demande ?”

- Sortie : comparaison ventes attendues (baseline) vs ventes observées pendant promo.

- KPI : uplift net(l’augmentation réelle des ventes grâce à la promotion), cannibalisation (quand une promotion fait baisser les ventes d’un autre produit ou département), efficacité par dept/magasin.

>Cette analyse permet de mesurer si une promotion a réellement augmenté les ventes ou si elle a seulement déplacé la demande dans le temps. Pour cela, on compare les ventes prévues sans promotion avec les ventes réellement observées pendant la période promotionnelle. Cela permet d’évaluer l’efficacité de la promotion pour chaque magasin et département.

### G. Éviter les promos “inutiles” (mauvais timing)

- Question : “Est-ce qu’on a besoin d’une promo si la demande est déjà naturellement haute ?”

```python
# ── Importation des bibliothèques ─────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# ── Lecture des fichiers CSV ──────────────────────────────────────────────────
train    = pd.read_csv('../Data/train.csv')
features = pd.read_csv('../Data/features.csv')
stores   = pd.read_csv('../Data/stores.csv')

# ── Vérifier que les fichiers sont bien chargés ───────────────
print("train :", train.shape)
print("features :", features.shape)
print("stores :", stores.shape)

# ── Aperçu des datasets ───────────────────────────────────────

print("► Aperçu du dataset : TRAIN")
display(train.head())

print("► Aperçu du dataset : FEATURES")
display(features.head())

print("► Aperçu du dataset : STORES")
display(stores.head())
```

Avant de fusionner les fichiers, il est important de comprendre **comment la jointure fonctionne** et **sur quelles colonnes elle doit être effectuée**. Dans le projet Walmart, les données sont réparties dans plusieurs fichiers qui contiennent des informations différentes mais **liées entre elles par des clés communes**.

### 1️⃣ Les datasets utilisés

- **train.csv** → contient les ventes hebdomadaires (`Weekly_Sales`) par **magasin (`Store`)**, **département (`Dept`)** et **date (`Date`)**.

- **test.csv** → structure similaire à **train**, mais **sans les ventes** (utilisé pour la prédiction).

- **features.csv** → contient des **variables externes** qui influencent les ventes (température, prix du carburant, promotions, CPI, chômage, etc.).

- **stores.csv** → contient les **caractéristiques des magasins** (type de magasin et taille).

---

### 2️⃣ Les clés de jointure

Pour combiner les données, on utilise des **colonnes communes  (clés)**.

| Dataset | Clés communes |
|--------|---------------|
| train ↔ features | Store, Date, IsHoliday |
| train ↔ stores | Store |

Ces clés permettent de **relier les informations provenant de différentes tables**.

```python
# ── Jointure des données en un seul DataFrame ─────────────────────────────────

# ── Jointure des données TRAIN ─────────────────────────────────
# 1) Fusionner train avec features sur (Store, Date, IsHoliday)
df_train = train.merge(features, on=['Store', 'Date', 'IsHoliday'], how='left')
# 2) Fusionner le résultat avec stores sur (Store)
df_train = df_train.merge(stores, on='Store', how='left')

print("Dimensions du dataset train combiné :", df_train.shape)

# ── Aperçu des datasets ────────────────────────────────────────

print("► Aperçu du dataset TRAIN")
display(df_train.head(100))
```

### <span id="135" style="color:#2E86C1;">1.3.5 Traiter les valeurs manquantes</span>

Les valeurs manquantes représentent l'un des problèmes les plus fréquents en science des données. Leur présence peut **biaiser les résultats** des analyses statistiques et **dégrader les performances** des modèles prédictifs si elles ne sont pas traitées correctement.

Dans le dataset Walmart, les valeurs manquantes proviennent principalement des **variables promotionnelles** (`MarkDown1` à `MarkDown5`), qui ne sont renseignées que lorsqu'une promotion est active. D'autres variables économiques comme `CPI` et `Unemployment` peuvent également présenter des lacunes ponctuelles.

Cette section a pour objectif de :
1. **Identifier** et quantifier les valeurs manquantes par colonne 
2. **Analyser** la nature des données manquantes (aléatoires ou structurelles)
3. **Appliquer** une stratégie de traitement adaptée à chaque variable (imputation, remplacement par 0, suppression, etc.)

Le choix de la méthode de traitement dépendra de la **nature de la variable**, du **taux de valeurs manquantes** et de son **importance** pour la modélisation.

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ── 1) Identification et quantification des valeurs manquantes ───────────
print("=" * 70)
print("  1) IDENTIFICATION DES VALEURS MANQUANTES")
print("=" * 70)

for nom, df in [("TRAIN", df_train)]:
    missing = df.isnull().sum()
    missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
    resume = pd.DataFrame({
        'Manquantes': missing, 
        '% Manquant': missing_pct,
        'Type': df.dtypes
    })
    resume = resume.sort_values('% Manquant', ascending=False)
    
    print(f"\n► Dataset {nom} ({len(df)} lignes, {df.shape[1]} colonnes) :")
    cols_avec_na = resume[resume['Manquantes'] > 0]
    cols_sans_na = resume[resume['Manquantes'] == 0]
    print(f"  • {len(cols_avec_na)} colonne(s) avec des valeurs manquantes")
    print(f"  • {len(cols_sans_na)} colonne(s) complètes")
    if not cols_avec_na.empty:
        display(cols_avec_na)


# ── 2) Analyse de la nature des valeurs manquantes ───────────────────────
print("\n" + "=" * 70)
print("  2) ANALYSE DE LA NATURE DES VALEURS MANQUANTES")
print("=" * 70)

# MarkDowns : vérifier si les NaN correspondent aux périodes sans promotions
print("\n► MarkDown1 à MarkDown5 :")
md_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
md_all_null = df_train[md_cols].isnull().all(axis=1).sum()
md_any_null = df_train[md_cols].isnull().any(axis=1).sum()
md_no_null  = df_train[md_cols].notna().all(axis=1).sum()
print(f"  • Lignes avec TOUS les MarkDowns manquants : {md_all_null:,} ({md_all_null/len(df_train)*100:.1f}%)")
print(f"  • Lignes avec AU MOINS un MarkDown manquant : {md_any_null:,} ({md_any_null/len(df_train)*100:.1f}%)")
print(f"  • Lignes avec AUCUN MarkDown manquant       : {md_no_null:,} ({md_no_null/len(df_train)*100:.1f}%)")

# Vérifier si les MarkDowns manquants dépendent de la date (avant/après une certaine date)
print("\n► Période des MarkDowns disponibles :")
for md in md_cols:
    dates_dispo = df_train[df_train[md].notna()]['Date']
    if len(dates_dispo) > 0:
        print(f"  {md} : disponible du {dates_dispo.min().date()} au {dates_dispo.max().date()}")
    else:
        print(f"  {md} : aucune valeur disponible")


# ── 3) Visualisation des valeurs manquantes ──────────────────────────────
print("\n" + "=" * 70)
print("  3) VISUALISATION DES VALEURS MANQUANTES")
print("=" * 70)

fig = make_subplots(rows=1, subplot_titles=["Valeurs manquantes – TRAIN", "Valeurs manquantes – TEST"],
                    horizontal_spacing=0.15)

for idx, (nom, df) in enumerate([("TRAIN", df_train)]):
    missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
    missing_pct = missing_pct[missing_pct > 0].sort_values(ascending=True)
    
    if not missing_pct.empty:
        colors = ['#e74c3c' if v > 50 else '#f39c12' if v > 20 else '#3498db' for v in missing_pct.values]
        fig.add_trace(
            go.Bar(y=missing_pct.index, x=missing_pct.values, orientation='h',
                   marker_color=colors, text=[f'{v:.1f}%' for v in missing_pct.values],
                   textposition='outside', name=nom, showlegend=False),
            row=1, col=idx+1
        )

fig.update_layout(height=450, width=700, title_text="Pourcentage de valeurs manquantes par colonne",
                  title_font_size=15)
fig.update_xaxes(title_text="% de valeurs manquantes")
fig.show()
```

**1) Identification :**

**2) Nature des valeurs manquantes (MarkDowns) :**

- Les MarkDowns ne sont disponibles qu'à partir du **2011-11-11**, soit environ 1 an et 9 mois après le début du dataset TRAIN (2010-02-05). Les NaN avant cette date correspondent à une **absence réelle de données promotionnelles** (Walmart ne les enregistrait pas encore).
- Seulement **~23 %** des lignes disposent de l'ensemble des 5 variables MarkDown sans aucune valeur manquante.

### <span id="136" style="color:#2E86C1;">1.3.6 Détecter et gérer les valeurs aberrantes</span>

Les valeurs aberrantes (*outliers*) sont des observations qui s'écartent significativement de la distribution normale des données. Leur présence peut **fausser les statistiques descriptives** (moyenne, écart-type), **biaiser l'entraînement** des modèles prédictifs et conduire à des **prévisions erronées**.

Dans le contexte de la grande distribution, les outliers peuvent avoir des origines variées :
- **Pics de ventes exceptionnels** lors d'événements promotionnels majeurs (Black Friday, Thanksgiving)
- **Erreurs de saisie** ou anomalies dans les systèmes d'enregistrement
- **Variations économiques locales** inhabituelles (CPI, taux de chômage)
- **Comportements atypiques** de certains magasins ou départements

Cette section a pour objectif de :
1. **Détecter** les valeurs aberrantes à l'aide de méthodes statistiques (IQR, Z-score) et de visualisations (boxplots)
2. **Analyser** leur origine pour distinguer les anomalies réelles des valeurs extrêmes légitimes
3. **Décider** de la stratégie de traitement adaptée : conservation, transformation ou suppression

> ⚠️ **Note importante** : Dans le domaine de la vente au détail, certaines valeurs extrêmes sont **légitimes** (ex. ventes record pendant les fêtes). Il est donc essentiel de ne pas supprimer automatiquement tous les outliers, mais de les analyser au cas par cas.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Détection des outliers par la méthode IQR</strong>
</div>


```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.graph_objects as go

import numpy as np



print("=" * 70)
print("  1) DÉTECTION DES OUTLIERS – MÉTHODE IQR (Q1 - 1.5×IQR / Q3 + 1.5×IQR)")
print("=" * 70)

num_cols_train = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 
                  'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5',
                  'CPI', 'Unemployment', 'Size']

resultats_iqr = []

for col in num_cols_train:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    borne_inf = Q1 - 1.5 * IQR
    borne_sup = Q3 + 1.5 * IQR
    
    outliers = df_train[(df_train[col] < borne_inf) | (df_train[col] > borne_sup)]
    n_out = len(outliers)
    pct_out = (n_out / len(df_train) * 100)
    
    resultats_iqr.append({
        'Variable': col,
        'Q1': round(Q1, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Borne inf': round(borne_inf, 2),
        'Borne sup': round(borne_sup, 2),
        'Nb outliers': n_out,
        '% outliers': round(pct_out, 2)
    })

df_iqr = pd.DataFrame(resultats_iqr)
print("\n► Tableau récapitulatif des outliers (méthode IQR) – Dataset TRAIN :\n")
display(df_iqr.style.background_gradient(subset=['% outliers'], cmap='YlOrRd')
        .format({'Q1': '{:.2f}', 'Q3': '{:.2f}', 'IQR': '{:.2f}',
                 'Borne inf': '{:.2f}', 'Borne sup': '{:.2f}',
                 'Nb outliers': '{:,}', '% outliers': '{:.2f}%'}))
```

```python

# Grouper les variables par nature pour des subplots lisibles
groupes = {
    'Ventes': ['Weekly_Sales'],
    'Promotions (MarkDowns)': ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'],
    'Contexte économique': ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment'],
}

for titre_groupe, cols in groupes.items():

    n = len(cols)

    fig = make_subplots(
        rows=1,
        cols=n,
        #subplot_titles=cols,
        horizontal_spacing=0.05
    )

    for i, col in enumerate(cols):

        fig.add_trace(
            go.Violin(
                y=df_train[col],
                name=col,
                box_visible=True,
                points='outliers',
                line_color='#F39C12'
            ),
            row=1,
            col=i+1
        )

    fig.update_layout(
        title=f"Boxplots – {titre_groupe}",
        template="plotly_white",
        height=450,
        width=max(350*n, 500),
        showlegend=False
    )

    fig.show()
```

### <span id="137" style="color:#2E86C1;">1.3.7 Validation de la cohérence des données avant analyse</span>

  Avant de passer à l'analyse exploratoire, il est indispensable de réaliser une **validation finale de la cohérence des données**. Cette étape constitue un bilan de toutes les vérifications et transformations effectuées lors du nettoyage (sections 1.3.1 à 1.3.6).

  L'objectif est de confirmer que le dataset est **complet, cohérent et correctement typé** en vérifiant :

  - **Les types de données** : chaque variable est-elle dans le format attendu (`datetime64`, `category`, `float64`, `bool`, `int64`) ?
  - **L'absence de valeurs manquantes critiques** : les colonnes essentielles (Store, Dept, Date, Weekly_Sales) ne contiennent-elles aucun NaN ?
  - **Les plages de valeurs** : les variables numériques sont-elles dans des intervalles réalistes ?
  - **La cohérence croisée** : les magasins et départements du dataset TEST sont-ils bien présents dans le dataset TRAIN ?
  - **La continuité temporelle** : les dates couvrent-elles bien la période attendue sans interruption majeure ?

  Ce récapitulatif permet de valider que le dataset est **prêt pour la modélisation** et qu'aucune anomalie n'a été introduite lors des transformations précédentes.

```python
# ── Validation de la cohérence des données avant analyse ──────────────────────

resultats = []

# ══════════════════════════════════════════════════════════════════════════════
# 1) VÉRIFICATION DES TYPES DE DONNÉES
# ══════════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) VÉRIFICATION DES TYPES DE DONNÉES")
print("=" * 70)

types_attendus = {
    'Store': 'category', 'Dept': 'category', 'Date': 'datetime64[ns]',
    'IsHoliday': 'bool', 'Type': 'category', 'Size': 'int64',
    'Temperature': 'float64', 'Fuel_Price': 'float64',
    'CPI': 'float64', 'Unemployment': 'float64',
    'MarkDown1': 'float64', 'MarkDown2': 'float64',
    'MarkDown3': 'float64', 'MarkDown4': 'float64', 'MarkDown5': 'float64',
    'Weekly_Sales': 'float64'
}

for df_name, df in [("TRAIN", df_train)]:
    erreurs_type = []
    for col, dtype_attendu in types_attendus.items():
        if col in df.columns:
            if str(df[col].dtype) != dtype_attendu:
                erreurs_type.append(f"{col}: {df[col].dtype} (attendu: {dtype_attendu})")
    if erreurs_type:
        print(f"\n  ❌ {df_name} – Types incorrects :")
        for e in erreurs_type:
            print(f"     • {e}")
        resultats.append(("Types de données", df_name, "❌ ÉCHEC", f"{len(erreurs_type)} colonne(s) incorrecte(s)"))
    else:
        print(f"\n  ✅ {df_name} – Tous les types sont conformes")
        resultats.append(("Types de données", df_name, "✅ OK", "Tous les types conformes"))

# ══════════════════════════════════════════════════════════════════════════════
# 2) VALEURS MANQUANTES SUR LES COLONNES ESSENTIELLES
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  2) VALEURS MANQUANTES – COLONNES ESSENTIELLES")
print("=" * 70)

cols_essentielles_train = ['Store', 'Dept', 'Date', 'Weekly_Sales', 'IsHoliday', 'Type', 'Size']
cols_essentielles_test  = ['Store', 'Dept', 'Date', 'IsHoliday', 'Type', 'Size']

for df_name, df, cols_ess in [("TRAIN", df_train, cols_essentielles_train)]:
    na_ess = df[cols_ess].isnull().sum()
    na_prob = na_ess[na_ess > 0]
    if len(na_prob) > 0:
        print(f"\n  ❌ {df_name} – Valeurs manquantes détectées :")
        for col, n in na_prob.items():
            print(f"     • {col}: {n} NaN")
        resultats.append(("NaN essentiels", df_name, "❌ ÉCHEC", f"{len(na_prob)} colonne(s) avec NaN"))
    else:
        print(f"\n  ✅ {df_name} – Aucun NaN dans les colonnes essentielles ({len(cols_ess)} vérifiées)")
        resultats.append(("NaN essentiels", df_name, "✅ OK", "Aucun NaN"))

# ══════════════════════════════════════════════════════════════════════════════
# 3) PLAGES DE VALEURS DES VARIABLES NUMÉRIQUES
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  3) PLAGES DE VALEURS – VARIABLES NUMÉRIQUES")
print("=" * 70)

plages_attendues = {
    'Temperature':  (-30, 130),      # °F
    'Fuel_Price':   (0.5, 7.0),      # $/gallon
    'CPI':          (80, 300),       # Indice
    'Unemployment': (0.0, 25.0),      # %
    'Size':         (10000, 300000),   # pieds carrés
}

for df_name, df in [("TRAIN", df_train)]:
    print(f"\n  ► {df_name} :")
    anomalies = []
    for col, (v_min, v_max) in plages_attendues.items():
        if col in df.columns:
            val_min = df[col].min()
            val_max = df[col].max()
            hors_plage = ((df[col] < v_min) | (df[col] > v_max)).sum()
            statut = "✅" if hors_plage == 0 else f"⚠️ {hors_plage} valeurs hors plage"
            print(f"    {col:<15} [{val_min:>10.2f} – {val_max:>10.2f}]  attendu [{v_min} – {v_max}]  {statut}")
            if hors_plage > 0:
                anomalies.append(col)
    if anomalies:
        resultats.append(("Plages numériques", df_name, "⚠️ ATTENTION", f"{len(anomalies)} variable(s) hors plage"))
    else:
        resultats.append(("Plages numériques", df_name, "✅ OK", "Toutes les plages conformes"))

# ══════════════════════════════════════════════════════════════════════════════
# 4) DIMENSIONS FINALES
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  4) DIMENSIONS FINALES")
print("=" * 70)
print(f"\n  TRAIN : {df_train.shape[0]:>10,} lignes × {df_train.shape[1]} colonnes")
resultats.append(("Dimensions", "TRAIN", "✅ INFO", f"{df_train.shape[0]:,} × {df_train.shape[1]}"))

# ══════════════════════════════════════════════════════════════════════════════
# TABLEAU RÉCAPITULATIF
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  TABLEAU RÉCAPITULATIF DE VALIDATION")
print("=" * 70)

df_recap = pd.DataFrame(resultats, columns=['Vérification', 'Dataset', 'Statut', 'Détails'])
display(df_recap)
```
