## <span style="color:#2E86C1;">Introduction</span>

La **grande distribution** est un secteur hautement concurrentiel où la capacité à **anticiper la demande** constitue un avantage stratégique majeur. Walmart, en tant que leader mondial du commerce de détail, génère chaque semaine des volumes considérables de données transactionnelles à travers ses magasins répartis sur différentes régions.

Ce projet, intitulé **« Système Intelligent de Prévision des Ventes en Grande Distribution »**, vise à concevoir et déployer un pipeline complet de science des données pour **prédire les ventes hebdomadaires** de Walmart. En exploitant des données historiques enrichies de variables contextuelles — telles que la température, le prix du carburant, le taux de chômage, l'indice des prix à la consommation (CPI) et les périodes promotionnelles (*markdown*) — nous chercherons à construire des modèles prédictifs robustes et interprétables.

### Objectifs principaux

- **Comprendre** le contexte métier de la grande distribution et les facteurs qui influencent les ventes.
- **Explorer et préparer** les données en identifiant les anomalies, les valeurs manquantes et les relations entre variables.
- **Modéliser** les ventes à l'aide d'algorithmes de Machine Learning (régression, arbres de décision, forêts aléatoires, gradient boosting, etc.).
- **Évaluer** la performance des modèles à travers des métriques adaptées (MAE, RMSE, WMAE).
- **Exploiter** les résultats pour formuler des recommandations stratégiques à destination des décideurs.

### Jeux de données utilisés

| Fichier | Description |
|---|---|
| `train.csv` | Données d'entraînement : ventes hebdomadaires par magasin et département (2010-02 à 2012-10) |
| `test.csv` | Données de test pour la prédiction (2012-11 à 2013-07) |
| `stores.csv` | Informations sur les magasins (type et superficie) |
| `features.csv` | Variables contextuelles (température, carburant, CPI, chômage, promotions) |

### Démarche adoptée

Le projet est structuré en **trois grandes parties** :
1. **Compréhension et Préparation des Données** — Exploration, nettoyage et ingénierie des variables.
2. **Modélisation et Évaluation Prédictive** — Entraînement, optimisation et comparaison des modèles.
3. **Intelligence Décisionnelle et Exploitation** — Interprétation des résultats et recommandations stratégiques.

## <span id="11" style="color:#2E86C1;">1.1 Comprendre le contexte retail Walmart</span>

Le secteur du **retail (commerce de détail)** repose sur la vente directe de produits aux consommateurs à travers des magasins physiques ou des plateformes en ligne. Dans ce contexte, les entreprises génèrent de grandes quantités de données liées aux ventes, aux clients et aux opérations. L’analyse de ces données permet d’améliorer la prise de décision, d’optimiser les stocks et d’anticiper la demande.

**Walmart**, l’un des plus grands détaillants au monde, produit chaque jour un volume massif de données grâce à ses milliers de magasins et ses millions de transactions. Ces données constituent un contexte pertinent pour appliquer des méthodes de **science des données et de prévision des ventes**.

Dans ce projet, l’objectif est d’analyser les **données historiques de ventes** afin de comprendre les facteurs influençant la demande et de développer des **modèles capables de prédire les ventes futures**, permettant ainsi d’améliorer la gestion des stocks et la planification commerciale.

### <span id="111" style="color:#2E86C1;">1.1.1 Analyse du fonctionnement de la grande distribution</span>

La grande distribution constitue l'un des secteurs les plus complexes à piloter sur le plan analytique, en raison de la multiplicité des variables qui influencent la performance commerciale. Walmart, en tant que leader mondial de la distribution avec plus de 10 000 points de vente dans le monde, illustre parfaitement les défis inhérents à ce modèle économique.
### Modèle économique et logique de volume
Le modèle économique de la grande distribution repose sur un principe fondamental : générer de la rentabilité par la rotation rapide des stocks plutôt que par des marges unitaires élevées. Walmart applique cette logique à travers sa stratégie historique "Everyday Low Prices" (EDLP), qui privilégie des prix bas permanents à des promotions ponctuelles. Dans ce contexte, la capacité à anticiper précisément la demande devient un avantage concurrentiel majeur : une rupture de stock signifie une vente perdue et une dégradation de l'expérience client, tandis qu'un surstock génère des coûts de stockage, des démarques et potentiellement des destructions de marchandise.
### Organisation par départements et hétérogénéité des ventes
Au sein de chaque magasin, les ventes sont enregistrées à la granularité Store × Département × Semaine, soit une combinaison de 45 magasins et 81 départements couvrant l'ensemble des rayons — épicerie, électronique, habillement, jardinage, automobile, pharmacie, entre autres. Cette granularité fine révèle une forte hétérogénéité interne : certains départements, tels que le département 92 (épicerie générale) ou le département 95 (jardinage), concentrent une part disproportionnée du chiffre d'affaires total, tandis que d'autres affichent des volumes marginaux mais des profils saisonniers très marqués. Cette disparité impose de traiter chaque combinaison Store × Département comme une entité à part entière dans la modélisation.
### Les leviers opérationnels de la grande distribution
La performance d'un distributeur comme Walmart repose sur la maîtrise de trois leviers opérationnels interdépendants, qui constituent précisément les axes de décision que ce projet vise à soutenir.
### 1. Gestion des stocks et réapprovisionnement
La planification des stocks s'appuie sur l'anticipation de la demande future. Une prévision inexacte se traduit immédiatement par un impact financier mesurable : en cas de sous-estimation, la rupture de stock prive le magasin de ventes et dégrade la satisfaction client ; en cas de surestimation, le surstock mobilise des ressources logistiques, génère des coûts de stockage et conduit à des démarques souvent subies. Dans ce projet, l'objectif est de réduire les ruptures de stock de 20% grâce à une meilleure anticipation des volumes hebdomadaires.
### 2. Politique promotionnelle et gestion des MarkDowns
Walmart recourt à des remises temporaires, appelées MarkDowns, pour stimuler les ventes sur certains rayons, écouler des stocks excédentaires ou répondre à des dynamiques concurrentielles. Ces promotions génèrent des pics de demande dont l'intensité varie selon le type de magasin, le département concerné et la période de l'année. Anticiper l'effet d'un MarkDown sur les ventes hebdomadaires est donc un enjeu majeur pour dimensionner correctement les stocks en amont et maximiser le retour sur investissement promotionnel.
### 3. Planification des achats et gestion des délais fournisseurs
Les décisions d'achat sont structurellement prises plusieurs semaines avant la mise en rayon, en raison des délais logistiques et des contraintes fournisseurs. Elles doivent intégrer à la fois les événements prévisibles — fêtes nationales, variations saisonnières — et le contexte économique local propre à chaque zone de chalandise. Une planification des achats fondée sur des prévisions fiables permet de sécuriser les approvisionnements aux moments critiques tout en limitant les immobilisations de trésorerie.
### 4. Les facteurs externes structurant la demande
Au-delà des décisions internes, les ventes en grande distribution sont étroitement corrélées à des variables contextuelles que le dataset Walmart capture de manière explicite. Les événements de fêtes nationales, représentés par la variable IsHoliday, provoquent des pics de ventes très marqués pouvant doubler la moyenne hebdomadaire, comme c'est le cas lors de Thanksgiving en semaine 47. L'indice des prix à la consommation (CPI) influence les arbitrages d'achat des ménages, notamment sur les catégories non-essentielles. Le taux de chômage local (Unemployment) accroît la sensibilité au prix dans les zones économiquement fragiles. Le prix du carburant (Fuel_Price) affecte quant à lui la fréquence et le périmètre des déplacements en magasin. Enfin, la température (Temperature) conditionne la saisonnalité de certains rayons comme le jardinage, les articles de saison ou l'habillement.
La prise en compte de l'ensemble de ces facteurs dans le modèle prédictif est indispensable pour capturer la variabilité réelle des ventes et produire des prévisions robustes.

### <span id="112" style="color:#2E86C1;">1.1.2 Identification des décisions à soutenir (stocks, promotions...)</span>

### 1) Décisions “Stocks” à soutenir (réassort & allocation)

### A. Réassort (quand et combien commander)

- Question : “Combien commander pour Store X / Dept Y la semaine prochaine (ou 2–4 semaines) ?”

- Sortie attendue : prévision des ventes + quantité recommandée (réassort).

- KPI : taux de rupture (stockout), taux de service, surstock, coût de stockage.

>Les décisions de réassort consistent à déterminer quand et combien de produits commander pour chaque magasin et département. Grâce aux prévisions de ventes, il est possible d’estimer les quantités nécessaires pour les semaines à venir. L’objectif est d’éviter les ruptures de stock tout en limitant le surstock et les coûts de stockage.

### B. Stock de sécurité (safety stock)

- Question : “Quel stock minimum garder pour éviter les ruptures pendant pics saisonniers / holidays ?”

- Sortie : niveau de stock de sécurité recommandé par dept/magasin (selon volatilité).

- KPI : ruptures évitées, stabilité de disponibilité.

>Le stock de sécurité correspond à la quantité minimale de produits à garder pour éviter les ruptures de stock, surtout pendant les périodes de forte demande comme les fêtes ou les événements saisonniers. En analysant la variabilité des ventes, on peut recommander un niveau de stock de sécurité adapté pour chaque magasin et département. L’objectif est de maintenir une bonne disponibilité des produits pour les clients.

### C. Allocation inter-magasins (prioriser où envoyer le stock)

- Question : “Si stock limité, quels magasins doivent être servis en priorité ?”

- Sortie : classement des magasins/depts selon demande prévue + risque de rupture.

- KPI : pertes de ventes évitées, disponibilité globale.

>Lorsque le stock est limité, il est important de décider quels magasins doivent être approvisionnés en priorité. En utilisant les prévisions de ventes, on peut identifier les magasins et départements où la demande sera la plus forte et où le risque de rupture est élevé. Cela permet d’envoyer les produits là où ils seront le plus nécessaires et de réduire les pertes de ventes.

### D. Préparation des pics (événements & périodes de fêtes)

- Question : “Quelles semaines vont exploser (Noël, Thanksgiving, etc.) et sur quels depts ?”

- Sortie : calendrier de pics + “top depts à risque”.

- KPI : précision des pics, réduction des ruptures en périodes critiques.

>Les prévisions permettent aussi d’identifier les périodes où les ventes vont fortement augmenter, comme pendant les fêtes ou certains événements saisonniers. En repérant ces semaines à l’avance et les départements concernés, Walmart peut mieux préparer ses stocks. Cela aide à éviter les ruptures de produits pendant les périodes de forte demande.

### 2) Décisions “Promotions” à soutenir (planification & évaluation)

### E. Planifier les promos (quand lancer / sur quels produits)

- Question : “Quelle promo faire, à quelle semaine, dans quel magasin/département ?”

- Sortie : recommandation de périodes “propices” (là où la demande est sensible / saisonnière).

- KPI : uplift ventes (l’augmentation des ventes grâce à la promotion), marge, ROI promo(le retour sur investissement de la promotion).

>Les prévisions de ventes peuvent aussi aider à planifier les promotions. En analysant la saisonnalité et la sensibilité de la demande, il devient possible d’identifier les semaines et les départements où une promotion serait la plus efficace. Cela permet d’augmenter les ventes tout en optimisant la marge et le retour sur investissement des campagnes promotionnelles.

### F. Mesurer l’impact des promotions (uplift)

- Question : “La promo a-t-elle réellement augmenté les ventes ou juste déplacé la demande ?”

- Sortie : comparaison ventes attendues (baseline) vs ventes observées pendant promo.

- KPI : uplift net(l’augmentation réelle des ventes grâce à la promotion), cannibalisation (quand une promotion fait baisser les ventes d’un autre produit ou département), efficacité par dept/magasin.

>Cette analyse permet de mesurer si une promotion a réellement augmenté les ventes ou si elle a seulement déplacé la demande dans le temps. Pour cela, on compare les ventes prévues sans promotion avec les ventes réellement observées pendant la période promotionnelle. Cela permet d’évaluer l’efficacité de la promotion pour chaque magasin et département.

### G. Éviter les promos “inutiles” (mauvais timing)

- Question : “Est-ce qu’on a besoin d’une promo si la demande est déjà naturellement haute ?”

### <span id="131" style="color:#2E86C1;">1.3.1 Importation et jointure des données</span>

```python
# ── Importation des bibliothèques ─────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# ── Lecture des fichiers CSV ──────────────────────────────────────────────────
train    = pd.read_csv('../Data/train.csv')
features = pd.read_csv('../Data/features.csv')
stores   = pd.read_csv('../Data/stores.csv')

# ── Vérifier que les fichiers sont bien chargés ───────────────
print("train :", train.shape)
print("features :", features.shape)
print("stores :", stores.shape)

# ── Aperçu des datasets ───────────────────────────────────────

print("► Aperçu du dataset : TRAIN")
display(train.head())

print("► Aperçu du dataset : FEATURES")
display(features.head())

print("► Aperçu du dataset : STORES")
display(stores.head())
```

Avant de fusionner les fichiers, il est important de comprendre **comment la jointure fonctionne** et **sur quelles colonnes elle doit être effectuée**. Dans le projet Walmart, les données sont réparties dans plusieurs fichiers qui contiennent des informations différentes mais **liées entre elles par des clés communes**.

### 1️⃣ Les datasets utilisés

- **train.csv** → contient les ventes hebdomadaires (`Weekly_Sales`) par **magasin (`Store`)**, **département (`Dept`)** et **date (`Date`)**.

- **test.csv** → structure similaire à **train**, mais **sans les ventes** (utilisé pour la prédiction).

- **features.csv** → contient des **variables externes** qui influencent les ventes (température, prix du carburant, promotions, CPI, chômage, etc.).

- **stores.csv** → contient les **caractéristiques des magasins** (type de magasin et taille).

---

### 2️⃣ Les clés de jointure

Pour combiner les données, on utilise des **colonnes communes  (clés)**.

| Dataset | Clés communes |
|--------|---------------|
| train ↔ features | Store, Date, IsHoliday |
| train ↔ stores | Store |

Ces clés permettent de **relier les informations provenant de différentes tables**.

```python
# ── Jointure des données en un seul DataFrame ─────────────────────────────────

# ── Jointure des données TRAIN ─────────────────────────────────
# 1) Fusionner train avec features sur (Store, Date, IsHoliday)
df_train = train.merge(features, on=['Store', 'Date', 'IsHoliday'], how='left')
# 2) Fusionner le résultat avec stores sur (Store)
df_train = df_train.merge(stores, on='Store', how='left')

print("Dimensions du dataset train combiné :", df_train.shape)

# ── Aperçu des datasets ────────────────────────────────────────

print("► Aperçu du dataset TRAIN")
display(df_train.head(100))
```

  - **`Store` et `Dept`** : conversion en `category`. Bien que leurs valeurs soient numériques, ce sont des **identifiants** — calculer une moyenne de numéros de magasins n'a aucun sens. Le type `category` explicite cette nature et évite les opérations arithmétiques non pertinentes.
  - **`Type`** : conversion en `category` (type de magasin : A, B, C) pour optimiser la mémoire et refléter sa nature catégorielle.
  - **Variables numériques** (`Temperature`, `Fuel_Price`, `CPI`, `Unemployment`, `MarkDown1-5`, `Weekly_Sales`) : vérification qu'elles sont bien en `float64`.
  - **`IsHoliday`** : vérification qu'elle reste en `bool`.

  À l'issue de cette étape, le dataset sera correctement typé et prêt pour les étapes suivantes de nettoyage et d'analyse.

```python
# ── 1) Classification des variables par nature sémantique ─────────────────────

# Classification sémantique de chaque variable
def classifier_variable(col):
    if col in ['Store', 'Dept']:
        return 'Identifiant (catégoriel)'
    elif col == 'Date':
        return 'Temporelle'
    elif col == 'IsHoliday':
        return 'Booléenne'
    elif col == 'Type':
        return 'Catégorielle nominale'
    elif col in ['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI',
                 'Unemployment', 'Size']:
        return 'Numérique continue'
    elif col.startswith('MarkDown'):
        return 'Numérique continue (promotions)'
    else:
        return 'Autre'

# ── 2) Tableau récapitulatif détaillé ─────────────────────────────────────────
print("\n" + "=" * 70)
print("  TABLEAU RÉCAPITULATIF DES VARIABLES (TRAIN)")
print("=" * 70)

recap = pd.DataFrame({
    'Type Pandas'       : df_train.dtypes,
    'Nature sémantique' : [classifier_variable(c) for c in df_train.columns],
    'Nb valeurs uniques': df_train.nunique(),
    'Valeurs manquantes': df_train.isnull().sum(),
    '% Manquant'        : (df_train.isnull().sum() / len(df_train) * 100).round(2),
    'Exemple valeur'    : df_train.iloc[100]
})

display(recap)

# ── 3) Résumé par nature sémantique ──────────────────────────────────────────
print("\n" + "=" * 70)
print("  DISTRIBUTION DES TYPES DE VARIABLES")
print("=" * 70)

type_counts = recap['Nature sémantique'].value_counts()
for nature, count in type_counts.items():
    cols = recap[recap['Nature sémantique'] == nature].index.tolist()
    print(f"\n  🔹 {nature} ({count}) : {cols}")

# ── 4) Diagnostic : types à corriger ─────────────────────────────────────────
print("\n" + "=" * 70)
print("  ⚠️  DIAGNOSTIC DES TYPES À CORRIGER")
print("=" * 70)

types_attendus = {
    'Store': 'category', 'Dept': 'category', 'Date': 'datetime64[ns]',
    'Type': 'category'
}
erreurs_type = []
for col, dtype_attendu in types_attendus.items():
    dtype_actuel = str(df_train[col].dtype)
    if dtype_actuel != dtype_attendu:
        print(f"   ⚠️  {col:8s} → actuellement {dtype_actuel:15s} → devrait être {dtype_attendu}")
        erreurs_type.append(col)
    else:
        print(f"   ✅  {col:8s} → déjà {dtype_actuel}")

if not erreurs_type:
    print("\n   ✅ Tous les types sont déjà corrects.")
else:
    print(f"\n   → {len(erreurs_type)} correction(s) nécessaire(s) : {erreurs_type}")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
#  Actions correctives – Conversion des types de variables
# ══════════════════════════════════════════════════════════════════════════════

# ── Harmonisation des formats ──────────────────────────────────────────────────

for df_name, df in [("TRAIN", df_train)]:
    print("=" * 70)
    print(f"  HARMONISATION DES FORMATS – {df_name}")
    print("=" * 70)

    # ── 1) Date → datetime64 ──────────────────────────────────────────────
    if not pd.api.types.is_datetime64_any_dtype(df['Date']):
        df['Date'] = pd.to_datetime(df['Date'])
        print(f"  Date   → converti en {df['Date'].dtype}  ✅")
    else:
        print(f"  Date   → déjà {df['Date'].dtype}  ✅")

    # ── 2) Store, Dept → category (identifiants, pas des valeurs numériques)
    for col in ['Store', 'Dept']:
        if df[col].dtype != 'category':
            df[col] = df[col].astype('category')
            print(f"  {col:<6} → converti en {df[col].dtype}  ✅")
        else:
            print(f"  {col:<6} → déjà {df[col].dtype}  ✅")

    # ── 3) Type → category ────────────────────────────────────────────────
    if df['Type'].dtype != 'category':
        df['Type'] = df['Type'].astype('category')
        print(f"  Type   → converti en {df['Type'].dtype}  ✅")
    else:
        print(f"  Type   → déjà {df['Type'].dtype}  ✅")

    # ── 4) Vérification des colonnes numériques (float64) ─────────────────
    num_cols_check = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment',
                      'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
    if df_name == "TRAIN":
        num_cols_check.append('Weekly_Sales')

    for col in num_cols_check:
        if col in df.columns:
            if df[col].dtype == 'float64':
                print(f"  {col:<15} → {df[col].dtype}  ✅")
            else:
                df[col] = df[col].astype('float64')
                print(f"  {col:<15} → converti en {df[col].dtype}  ✅")

    print(f"\n► Résumé des types après harmonisation ({df_name}) :")
    display(df.dtypes.to_frame('Type').T)
    print()
```

### <span id="135" style="color:#2E86C1;">1.3.5 Traiter les valeurs manquantes</span>

Les valeurs manquantes représentent l'un des problèmes les plus fréquents en science des données. Leur présence peut **biaiser les résultats** des analyses statistiques et **dégrader les performances** des modèles prédictifs si elles ne sont pas traitées correctement.

Dans le dataset Walmart, les valeurs manquantes proviennent principalement des **variables promotionnelles** (`MarkDown1` à `MarkDown5`), qui ne sont renseignées que lorsqu'une promotion est active. D'autres variables économiques comme `CPI` et `Unemployment` peuvent également présenter des lacunes ponctuelles.

Cette section a pour objectif de :
1. **Identifier** et quantifier les valeurs manquantes par colonne 
2. **Analyser** la nature des données manquantes (aléatoires ou structurelles)
3. **Appliquer** une stratégie de traitement adaptée à chaque variable (imputation, remplacement par 0, suppression, etc.)

Le choix de la méthode de traitement dépendra de la **nature de la variable**, du **taux de valeurs manquantes** et de son **importance** pour la modélisation.

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ── 1) Identification et quantification des valeurs manquantes ───────────
print("=" * 70)
print("  1) IDENTIFICATION DES VALEURS MANQUANTES")
print("=" * 70)

for nom, df in [("TRAIN", df_train)]:
    missing = df.isnull().sum()
    missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
    resume = pd.DataFrame({
        'Manquantes': missing, 
        '% Manquant': missing_pct,
        'Type': df.dtypes
    })
    resume = resume.sort_values('% Manquant', ascending=False)
    
    print(f"\n► Dataset {nom} ({len(df)} lignes, {df.shape[1]} colonnes) :")
    cols_avec_na = resume[resume['Manquantes'] > 0]
    cols_sans_na = resume[resume['Manquantes'] == 0]
    print(f"  • {len(cols_avec_na)} colonne(s) avec des valeurs manquantes")
    print(f"  • {len(cols_sans_na)} colonne(s) complètes")
    if not cols_avec_na.empty:
        display(cols_avec_na)


# ── 2) Analyse de la nature des valeurs manquantes ───────────────────────
print("\n" + "=" * 70)
print("  2) ANALYSE DE LA NATURE DES VALEURS MANQUANTES")
print("=" * 70)

# MarkDowns : vérifier si les NaN correspondent aux périodes sans promotions
print("\n► MarkDown1 à MarkDown5 :")
md_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
md_all_null = df_train[md_cols].isnull().all(axis=1).sum()
md_any_null = df_train[md_cols].isnull().any(axis=1).sum()
md_no_null  = df_train[md_cols].notna().all(axis=1).sum()
print(f"  • Lignes avec TOUS les MarkDowns manquants : {md_all_null:,} ({md_all_null/len(df_train)*100:.1f}%)")
print(f"  • Lignes avec AU MOINS un MarkDown manquant : {md_any_null:,} ({md_any_null/len(df_train)*100:.1f}%)")
print(f"  • Lignes avec AUCUN MarkDown manquant       : {md_no_null:,} ({md_no_null/len(df_train)*100:.1f}%)")

# Vérifier si les MarkDowns manquants dépendent de la date (avant/après une certaine date)
print("\n► Période des MarkDowns disponibles :")
for md in md_cols:
    dates_dispo = df_train[df_train[md].notna()]['Date']
    if len(dates_dispo) > 0:
        print(f"  {md} : disponible du {dates_dispo.min().date()} au {dates_dispo.max().date()}")
    else:
        print(f"  {md} : aucune valeur disponible")


# ── 3) Visualisation des valeurs manquantes ──────────────────────────────
print("\n" + "=" * 70)
print("  3) VISUALISATION DES VALEURS MANQUANTES")
print("=" * 70)

fig = make_subplots(rows=1, subplot_titles=["Valeurs manquantes – TRAIN", "Valeurs manquantes – TEST"],
                    horizontal_spacing=0.15)

for idx, (nom, df) in enumerate([("TRAIN", df_train)]):
    missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
    missing_pct = missing_pct[missing_pct > 0].sort_values(ascending=True)
    
    if not missing_pct.empty:
        colors = ['#e74c3c' if v > 50 else '#f39c12' if v > 20 else '#3498db' for v in missing_pct.values]
        fig.add_trace(
            go.Bar(y=missing_pct.index, x=missing_pct.values, orientation='h',
                   marker_color=colors, text=[f'{v:.1f}%' for v in missing_pct.values],
                   textposition='outside', name=nom, showlegend=False),
            row=1, col=idx+1
        )

fig.update_layout(height=450, width=700, title_text="Pourcentage de valeurs manquantes par colonne",
                  title_font_size=15)
fig.update_xaxes(title_text="% de valeurs manquantes")
fig.show()
```

- Les MarkDowns ne sont disponibles qu'à partir du **2011-11-11**, soit environ 1 an et 9 mois après le début du dataset TRAIN (2010-02-05). Les NaN avant cette date correspondent à une **absence réelle de données promotionnelles** (Walmart ne les enregistrait pas encore).

### <span id="136" style="color:#2E86C1;">1.3.6 Détecter et gérer les valeurs aberrantes</span>

Les valeurs aberrantes (*outliers*) sont des observations qui s'écartent significativement de la distribution normale des données. Leur présence peut **fausser les statistiques descriptives** (moyenne, écart-type), **biaiser l'entraînement** des modèles prédictifs et conduire à des **prévisions erronées**.

Dans le contexte de la grande distribution, les outliers peuvent avoir des origines variées :
- **Pics de ventes exceptionnels** lors d'événements promotionnels majeurs (Black Friday, Thanksgiving)
- **Erreurs de saisie** ou anomalies dans les systèmes d'enregistrement
- **Variations économiques locales** inhabituelles (CPI, taux de chômage)
- **Comportements atypiques** de certains magasins ou départements

Cette section a pour objectif de :
1. **Détecter** les valeurs aberrantes à l'aide de méthodes statistiques (IQR, Z-score) et de visualisations (boxplots)
2. **Analyser** leur origine pour distinguer les anomalies réelles des valeurs extrêmes légitimes
3. **Décider** de la stratégie de traitement adaptée : conservation, transformation ou suppression

> ⚠️ **Note importante** : Dans le domaine de la vente au détail, certaines valeurs extrêmes sont **légitimes** (ex. ventes record pendant les fêtes). Il est donc essentiel de ne pas supprimer automatiquement tous les outliers, mais de les analyser au cas par cas.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Détection des outliers par la méthode IQR</strong>
</div>


```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.graph_objects as go

import numpy as np



print("=" * 70)
print("  1) DÉTECTION DES OUTLIERS – MÉTHODE IQR (Q1 - 1.5×IQR / Q3 + 1.5×IQR)")
print("=" * 70)

num_cols_train = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 
                  'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5',
                  'CPI', 'Unemployment', 'Size']

resultats_iqr = []

for col in num_cols_train:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    borne_inf = Q1 - 1.5 * IQR
    borne_sup = Q3 + 1.5 * IQR
    
    outliers = df_train[(df_train[col] < borne_inf) | (df_train[col] > borne_sup)]
    n_out = len(outliers)
    pct_out = (n_out / len(df_train) * 100)
    
    resultats_iqr.append({
        'Variable': col,
        'Q1': round(Q1, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Borne inf': round(borne_inf, 2),
        'Borne sup': round(borne_sup, 2),
        'Nb outliers': n_out,
        '% outliers': round(pct_out, 2)
    })

df_iqr = pd.DataFrame(resultats_iqr)
print("\n► Tableau récapitulatif des outliers (méthode IQR) – Dataset TRAIN :\n")
display(df_iqr.style.background_gradient(subset=['% outliers'], cmap='YlOrRd')
        .format({'Q1': '{:.2f}', 'Q3': '{:.2f}', 'IQR': '{:.2f}',
                 'Borne inf': '{:.2f}', 'Borne sup': '{:.2f}',
                 'Nb outliers': '{:,}', '% outliers': '{:.2f}%'}))
```

```python
# ── Validation de la cohérence des données avant analyse ──────────────────────

resultats = []

# ══════════════════════════════════════════════════════════════════════════════
# 1) VÉRIFICATION DES TYPES DE DONNÉES
# ══════════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) VÉRIFICATION DES TYPES DE DONNÉES")
print("=" * 70)

types_attendus = {
    'Store': 'category', 'Dept': 'category', 'Date': 'datetime64[ns]',
    'IsHoliday': 'bool', 'Type': 'category', 'Size': 'int64',
    'Temperature': 'float64', 'Fuel_Price': 'float64',
    'CPI': 'float64', 'Unemployment': 'float64',
    'MarkDown1': 'float64', 'MarkDown2': 'float64',
    'MarkDown3': 'float64', 'MarkDown4': 'float64', 'MarkDown5': 'float64',
    'Weekly_Sales': 'float64'
}

for df_name, df in [("TRAIN", df_train)]:
    erreurs_type = []
    for col, dtype_attendu in types_attendus.items():
        if col in df.columns:
            if str(df[col].dtype) != dtype_attendu:
                erreurs_type.append(f"{col}: {df[col].dtype} (attendu: {dtype_attendu})")
    if erreurs_type:
        print(f"\n  ❌ {df_name} – Types incorrects :")
        for e in erreurs_type:
            print(f"     • {e}")
        resultats.append(("Types de données", df_name, "❌ ÉCHEC", f"{len(erreurs_type)} colonne(s) incorrecte(s)"))
    else:
        print(f"\n  ✅ {df_name} – Tous les types sont conformes")
        resultats.append(("Types de données", df_name, "✅ OK", "Tous les types conformes"))

# ══════════════════════════════════════════════════════════════════════════════
# 2) VALEURS MANQUANTES SUR LES COLONNES ESSENTIELLES
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  2) VALEURS MANQUANTES – COLONNES ESSENTIELLES")
print("=" * 70)

cols_essentielles_train = ['Store', 'Dept', 'Date', 'Weekly_Sales', 'IsHoliday', 'Type', 'Size']
cols_essentielles_test  = ['Store', 'Dept', 'Date', 'IsHoliday', 'Type', 'Size']

for df_name, df, cols_ess in [("TRAIN", df_train, cols_essentielles_train)]:
    na_ess = df[cols_ess].isnull().sum()
    na_prob = na_ess[na_ess > 0]
    if len(na_prob) > 0:
        print(f"\n  ❌ {df_name} – Valeurs manquantes détectées :")
        for col, n in na_prob.items():
            print(f"     • {col}: {n} NaN")
        resultats.append(("NaN essentiels", df_name, "❌ ÉCHEC", f"{len(na_prob)} colonne(s) avec NaN"))
    else:
        print(f"\n  ✅ {df_name} – Aucun NaN dans les colonnes essentielles ({len(cols_ess)} vérifiées)")
        resultats.append(("NaN essentiels", df_name, "✅ OK", "Aucun NaN"))

# ══════════════════════════════════════════════════════════════════════════════
# 3) PLAGES DE VALEURS DES VARIABLES NUMÉRIQUES
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  3) PLAGES DE VALEURS – VARIABLES NUMÉRIQUES")
print("=" * 70)

plages_attendues = {
    'Temperature':  (-30, 130),      # °F
    'Fuel_Price':   (0.5, 7.0),      # $/gallon
    'CPI':          (80, 300),       # Indice
    'Unemployment': (0.0, 25.0),      # %
    'Size':         (10000, 300000),   # pieds carrés
}

for df_name, df in [("TRAIN", df_train)]:
    print(f"\n  ► {df_name} :")
    anomalies = []
    for col, (v_min, v_max) in plages_attendues.items():
        if col in df.columns:
            val_min = df[col].min()
            val_max = df[col].max()
            hors_plage = ((df[col] < v_min) | (df[col] > v_max)).sum()
            statut = "✅" if hors_plage == 0 else f"⚠️ {hors_plage} valeurs hors plage"
            print(f"    {col:<15} [{val_min:>10.2f} – {val_max:>10.2f}]  attendu [{v_min} – {v_max}]  {statut}")
            if hors_plage > 0:
                anomalies.append(col)
    if anomalies:
        resultats.append(("Plages numériques", df_name, "⚠️ ATTENTION", f"{len(anomalies)} variable(s) hors plage"))
    else:
        resultats.append(("Plages numériques", df_name, "✅ OK", "Toutes les plages conformes"))

# ══════════════════════════════════════════════════════════════════════════════
# 4) DIMENSIONS FINALES
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  4) DIMENSIONS FINALES")
print("=" * 70)
print(f"\n  TRAIN : {df_train.shape[0]:>10,} lignes × {df_train.shape[1]} colonnes")
resultats.append(("Dimensions", "TRAIN", "✅ INFO", f"{df_train.shape[0]:,} × {df_train.shape[1]}"))

# ══════════════════════════════════════════════════════════════════════════════
# TABLEAU RÉCAPITULATIF
# ══════════════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("  TABLEAU RÉCAPITULATIF DE VALIDATION")
print("=" * 70)

df_recap = pd.DataFrame(resultats, columns=['Vérification', 'Dataset', 'Statut', 'Détails'])
display(df_recap)
```

### <span id="142" style="color:#2E86C1;">1.4.2 Analyse descriptive des ventes par magasin et département</span>

  Après avoir analysé la structure globale du dataset et les distributions des variables, cette section se concentre sur l'**analyse des ventes** à deux niveaux de granularité : **par magasin** et **par département**. L'objectif est de comprendre comment le chiffre d'affaires se répartit entre les différentes entités et d'identifier les segments les plus contributifs ou les plus disparates.

  L'analyse se déroule en **cinq étapes** :

  1. **Ventes totales et moyennes par magasin** — classement des 45 magasins selon leur chiffre d'affaires cumulé et leurs ventes hebdomadaires moyennes, avec mise en évidence des écarts entre types (A, B, C).
  2. **Dispersion des ventes par magasin** — analyse de la variabilité intra-magasin (écart-type, coefficient de variation) pour identifier les magasins à comportement stable vs volatil.
  3. **Ventes par département** — identification des départements les plus contributifs au chiffre d'affaires global et des départements à faible activité.
  4. **Croisement magasin × département** — heatmap des ventes moyennes pour visualiser les combinaisons les plus performantes.
  5. **Synthèse par type de magasin** — comparaison agrégée des performances entre magasins de type A, B et C.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Ventes totales et moyennes par magasin</strong>
</div>

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ══════════════════════════════════════════════════════════════════════════
#  1) VENTES TOTALES ET MOYENNES PAR MAGASIN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) VENTES TOTALES ET MOYENNES PAR MAGASIN")
print("=" * 70)

# Agrégation par magasin
store_stats = (
    df_train
    .groupby(['Store', 'Type'], observed=True)['Weekly_Sales']
    .agg(Total='sum', Moyenne='mean', Mediane='median', Ecart_type='std', Nb_lignes='count')
    .reset_index()
    .sort_values('Total', ascending=False)
)

store_stats['CV (%)'] = (store_stats['Ecart_type'] / store_stats['Moyenne'] * 100).round(1)

print(f"\n► Tableau récapitulatif des ventes par magasin (classé par CA total décroissant) :\n")
display(
    store_stats.style
    .set_caption("Ventes hebdomadaires par magasin – Dataset TRAIN")
    .format({
        'Total': '{:,.0f} $',
        'Moyenne': '{:,.0f} $',
        'Mediane': '{:,.0f} $',
        'Ecart_type': '{:,.0f}',
        'Nb_lignes': '{:,}',
        'CV (%)': '{:.1f}%'
    })
    .background_gradient(subset=['Total'], cmap='Blues')
    .background_gradient(subset=['CV (%)'], cmap='YlOrRd')
)

# Ratio top 10 / bottom 10
top10_total = store_stats.head(10)['Total'].sum()
bot10_total = store_stats.tail(10)['Total'].sum()

print(f"\n► Top 10 magasins : {top10_total:,.0f} $ ({top10_total / store_stats['Total'].sum() * 100:.1f}% du CA total)")
print(f"► Bottom 10 magasins : {bot10_total:,.0f} $ ({bot10_total / store_stats['Total'].sum() * 100:.1f}% du CA total)")
print(f"► Ratio Top 10 / Bottom 10 : {top10_total / bot10_total:.1f}x")
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Ventes totales et moyennes par magasin**

L'analyse des ventes par magasin révèle des **disparités considérables** entre les 45 points de vente :

- **Concentration du CA** : les 10 premiers magasins concentrent une part disproportionnée du chiffre d'affaires total, tandis que les 10 derniers ne représentent qu'une fraction modeste. Le ratio Top 10 / Bottom 10 met en évidence un écart de performance significatif.
- **Lien clair avec le type de magasin** : les magasins de **Type A** (bleu) occupent systématiquement le haut du classement, tant en ventes totales qu'en ventes moyennes. Les **Type B** (orange) se situent en position intermédiaire, et les **Type C** (vert) ferment le classement.
- **Ventes moyennes vs totales** : le classement reste globalement stable entre les deux métriques, confirmant que la taille du magasin (Type/Size) est un déterminant majeur du volume de ventes.
- **Quelques inversions notables** : certains magasins de type B se rapprochent des performances moyennes de type A, suggérant des facteurs locaux (emplacement, concurrence, zone de chalandise) qui modulent l'effet de la taille.

> 💡 **Point clé** : Le type de magasin (A, B, C) est un excellent prédicteur du niveau de ventes. Toutefois, la variabilité intra-type justifie une analyse plus fine de la dispersion (étape suivante).

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Dispersion des ventes par magasin (boxplots)</strong>
</div>

```python
# ══════════════════════════════════════════════════════════════════════════
#  2) DISPERSION DES VENTES PAR MAGASIN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  2) DISPERSION DES VENTES PAR MAGASIN")
print("=" * 70)

# Boxplot des ventes par magasin, ordonné par médiane, coloré par type
store_order = (
    df_train.groupby('Store')['Weekly_Sales']
    .median()
    .sort_values(ascending=False)
    .index.tolist()
)

# Mapping Store → Type pour la couleur
store_type_map = df_train.drop_duplicates('Store').set_index('Store')['Type'].to_dict()

color_map = {'A': '#2E86C1', 'B': '#E67E22', 'C': '#27AE60'}

fig3 = go.Figure()
for store_id in store_order:
    data_s = df_train[df_train['Store'] == store_id]['Weekly_Sales']
    t = store_type_map[store_id]
    fig3.add_trace(go.Box(
        y=data_s, name=str(store_id),
        marker_color=color_map[t],
        showlegend=False,
        boxmean=True
    ))

fig3.update_layout(
    title='Distribution des ventes hebdomadaires par magasin (ordonné par médiane décroissante)',
    xaxis_title='Magasin (Store)',
    yaxis_title='Weekly Sales ($)',
    height=550, width=1200,
    template='plotly_white'
)
fig3.show()

# Coefficient de variation par magasin
print('\n► Magasins avec le plus fort coefficient de variation (CV) :')
top_cv = store_stats.nlargest(10, 'CV (%)')[['Store', 'Type', 'Moyenne', 'Ecart_type', 'CV (%)']]
display(top_cv.style.format({'Moyenne': '{:,.0f} $', 'Ecart_type': '{:,.0f}', 'CV (%)': '{:.1f}%'}))

print('\n► Magasins avec le plus faible coefficient de variation (CV) :')
bot_cv = store_stats.nsmallest(10, 'CV (%)')[['Store', 'Type', 'Moyenne', 'Ecart_type', 'CV (%)']]
display(bot_cv.style.format({'Moyenne': '{:,.0f} $', 'Ecart_type': '{:,.0f}', 'CV (%)': '{:.1f}%'}))
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Dispersion des ventes par magasin**

Les boxplots révèlent des profils de dispersion très différents selon les magasins :

- **Magasins de type A** : outre des médianes élevées, ils présentent des **boîtes larges** et de **nombreux outliers** vers le haut, reflétant la coexistence de départements à faibles et fortes ventes au sein d'un même magasin.
- **Magasins de type C** : distributions plus **resserrées** avec des médianes basses — des magasins à activité plus homogène et plus modeste.
- **Coefficient de variation (CV)** : les magasins avec un CV élevé sont ceux qui abritent à la fois des départements à très faible activité et des départements phares. Un CV faible indique un comportement plus uniforme entre départements.

> 💡 **Implication pour la modélisation** : la forte hétérogénéité intra-magasin suggère qu'un modèle global unique sera insuffisant — une approche par segment (type de magasin ou département) pourrait améliorer les prévisions.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Ventes par département</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  3) VENTES PAR DÉPARTEMENT
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  3) VENTES PAR DÉPARTEMENT")
print("=" * 70)

# Agrégation par département
dept_stats = (
    df_train
    .groupby('Dept')['Weekly_Sales']
    .agg(Total='sum', Moyenne='mean', Mediane='median', Ecart_type='std')
    .reset_index()
)
# Nombre de magasins par département
dept_store_count = df_train.groupby('Dept')['Store'].nunique().reset_index(name='Nb_magasins')
dept_stats = dept_stats.merge(dept_store_count, on='Dept')
dept_stats['Part_CA (%)'] = (dept_stats['Total'] / dept_stats['Total'].sum() * 100).round(2)
dept_stats = dept_stats.sort_values('Total', ascending=False)
dept_stats['Cumul_CA (%)'] = dept_stats['Part_CA (%)'].cumsum().round(2)

print(f'\n► {df_train["Dept"].nunique()} départements identifiés\n')
print('► Top 15 départements par chiffre d\'affaires total :\n')
display(
    dept_stats.head(15).style
    .set_caption('Top 15 départements – Dataset TRAIN')
    .format({
        'Total': '{:,.0f} $', 'Moyenne': '{:,.0f} $', 'Mediane': '{:,.0f} $',
        'Ecart_type': '{:,.0f}', 'Part_CA (%)': '{:.2f}%', 'Cumul_CA (%)': '{:.2f}%'
    })
    .background_gradient(subset=['Part_CA (%)'], cmap='Blues')
)

# ── Graphique : Top 20 départements par CA ───────────────────────────
top20_dept = dept_stats.head(20).sort_values('Total', ascending=True)

fig4 = px.bar(
    top20_dept,
    y='Dept', x='Total',
    orientation='h',
    title='Top 20 départements par ventes totales – Dataset TRAIN',
    labels={'Total': 'Ventes totales ($)', 'Dept': 'Département'},
    text=top20_dept['Total'].apply(lambda x: f'{x/1e6:.1f}M$'),
    color='Part_CA (%)',
    color_continuous_scale='Blues'
)
fig4.update_traces(textposition='outside', textfont_size=9)
fig4.update_layout(
    height=600, width=900,
    template='plotly_white',
    yaxis=dict(type='category', dtick=1),
    margin=dict(l=60, r=100)
)
fig4.show()

# ── Courbe de Pareto (concentration du CA) ──────────────────────────
fig5 = go.Figure()
fig5.add_trace(go.Bar(
    x=dept_stats['Dept'].astype(str),
    y=dept_stats['Part_CA (%)'],
    name='Part du CA (%)',
    marker_color='#3498db'
))
fig5.add_trace(go.Scatter(
    x=dept_stats['Dept'].astype(str),
    y=dept_stats['Cumul_CA (%)'],
    name='Cumul CA (%)',
    mode='lines+markers',
    marker_color='#e74c3c',
    yaxis='y2'
))
fig5.add_hline(y=80, line_dash='dash', line_color='gray',
               annotation_text='80% du CA', yref='y2')
fig5.update_layout(
    title='Diagramme de Pareto – Concentration du CA par département',
    xaxis_title='Département (classé par CA décroissant)',
    yaxis_title='Part du CA (%)',
    yaxis2=dict(title='Cumul CA (%)', overlaying='y', side='right', range=[0, 105]),
    height=500, width=1100,
    template='plotly_white',
    legend=dict(x=0.7, y=0.3)
)
fig5.show()
```

**📝 Commentaire – Ventes par département**

L'analyse par département met en lumière une **forte concentration du chiffre d'affaires** :

- **Départements dominants** : les départements en tête du classement (comme les départements alimentaires ou les rayons à forte rotation) représentent des parts de CA significatives individuellement.
- **Longue traîne** : de nombreux départements contribuent de manière marginale, avec des parts de CA inférieures à 1 %. Ces départements pourraient être regroupés pour simplifier la modélisation.
- **Couverture variable** : tous les départements ne sont pas présents dans les 45 magasins — certains ne sont actifs que dans un sous-ensemble de magasins.

> 💡 **Point clé** : La concentration du CA sur quelques départements clés suggère de porter une attention particulière à leur prévision. Les départements mineurs pourraient être modélisés ensemble ou avec des approches simplifiées.

<strong>Croisement magasin × département (heatmap)</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  4) CROISEMENT MAGASIN × DÉPARTEMENT
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  4) HEATMAP – VENTES MOYENNES PAR MAGASIN × DÉPARTEMENT")
print("=" * 70)

# Sélectionner les top 20 départements pour lisibilité
top20_depts = dept_stats.head(20)['Dept'].tolist()

pivot = (
    df_train[df_train['Dept'].isin(top20_depts)]
    .groupby(['Store', 'Dept'])['Weekly_Sales']
    .mean()
    .reset_index()
    .pivot(index='Store', columns='Dept', values='Weekly_Sales')
)

# Ordonner les magasins par ventes moyennes totales
store_order_hm = pivot.mean(axis=1).sort_values(ascending=False).index
pivot = pivot.loc[store_order_hm]

fig6 = px.imshow(
    pivot.values,
    x=[str(d) for d in pivot.columns],
    y=[str(s) for s in pivot.index],
    color_continuous_scale='YlOrRd',
    labels=dict(x='Département', y='Magasin', color='Ventes moy. ($)'),
    title='Heatmap des ventes hebdomadaires moyennes – Top 20 départements × 45 magasins',
    aspect='auto'
)
fig6.update_layout(
    height=800, width=1000,
    template='plotly_white',
    xaxis=dict(dtick=1),
    yaxis=dict(dtick=1)
)
fig6.show()

# Nombre de combinaisons Store×Dept actives
n_combos = df_train.groupby(['Store', 'Dept']).ngroups
n_max = df_train['Store'].nunique() * df_train['Dept'].nunique()
print(f'\n► Combinaisons Store×Dept actives : {n_combos:,} / {n_max:,} possibles ({n_combos/n_max*100:.1f}%)')
```

**📝 Commentaire – Heatmap magasin × département**

La heatmap croisée permet de visualiser les **interactions** entre magasins et départements :

- **Gradient net** : les magasins en haut (type A, gros CA) affichent des intensités de couleur plus fortes sur la plupart des départements, confirmant un effet de taille global.
- **Départements phares** : certaines colonnes sont uniformément chaudes (rouges), indiquant des départements à forte activité dans **tous** les magasins. Ce sont les rayons incontournables de l'enseigne.
- **Zones blanches** : les cellules vides ou très pâles correspondent à des combinaisons Store×Dept inexistantes ou à très faible activité. Toutes les combinaisons ne sont pas actives.
- **Disparités intra-département** : pour un même département, les ventes peuvent varier considérablement d'un magasin à l'autre, reflétant des différences de taille, de zone géographique ou de clientèle.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Synthèse par type de magasin</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  5) SYNTHÈSE PAR TYPE DE MAGASIN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  5) SYNTHÈSE PAR TYPE DE MAGASIN")
print("=" * 70)

# Statistiques agrégées par type
type_stats = (
    df_train
    .groupby('Type')
    .agg(
        Nb_magasins=('Store', 'nunique'),
        Nb_departements=('Dept', 'nunique'),
        Nb_observations=('Weekly_Sales', 'count'),
        CA_total=('Weekly_Sales', 'sum'),
        Moyenne=('Weekly_Sales', 'mean'),
        Mediane=('Weekly_Sales', 'median'),
        Ecart_type=('Weekly_Sales', 'std'),
        Taille_moy=('Size', 'mean')
    )
    .reset_index()
)
type_stats['Part_CA (%)'] = (type_stats['CA_total'] / type_stats['CA_total'].sum() * 100).round(1)
type_stats['CA_moy_par_magasin'] = type_stats['CA_total'] / type_stats['Nb_magasins']

print('\n► Tableau de synthèse par type de magasin :\n')
display(
    type_stats.style
    .set_caption('Synthèse des ventes par type de magasin – Dataset TRAIN')
    .format({
        'CA_total': '{:,.0f} $', 'Moyenne': '{:,.0f} $', 'Mediane': '{:,.0f} $',
        'Ecart_type': '{:,.0f}', 'Taille_moy': '{:,.0f} sq ft',
        'Part_CA (%)': '{:.1f}%', 'CA_moy_par_magasin': '{:,.0f} $',
        'Nb_observations': '{:,}'
    })
    .background_gradient(subset=['Part_CA (%)'], cmap='Blues')
)

# ── Graphiques comparatifs ───────────────────────────────────────────
fig = make_subplots(specs=[[{"secondary_y": True}]])

# ── CA total ─────────────────────────────────
fig.add_trace(
    go.Bar(
        x=type_stats['Type'],
        y=type_stats['CA_total'],
        name="CA total",
        showlegend=False,
        marker_color=[color_map[t] for t in type_stats['Type']],
        text=type_stats['CA_total'].apply(lambda x: f'{x/1e6:.1f} M$'),
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(color='white', size=12)
    ),
    secondary_y=False
)

# ── Taille moyenne ───────────────────────────
fig.add_trace(
    go.Scatter(
        x=type_stats['Type'],
        y=type_stats['Taille_moy'],
        name="Taille moyenne",
        mode="lines+markers+text",
        text=type_stats['Taille_moy'].apply(lambda x: f'{x:,.0f} sq ft'),
        textposition="top right",
        line=dict(color="#111112", width=2),
        marker=dict(size=9)
    ),
    secondary_y=True
)

fig.update_layout(
    title="Comparaison des types de magasins (A, B, C)",
    template="plotly_white",
    height=450,
    width=900
)

fig.update_yaxes(title_text="CA total ($)", secondary_y=False)
fig.update_yaxes(title_text="Taille moyenne (sq ft)", secondary_y=True)

fig.show()
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Synthèse par type de magasin**

La synthèse par type confirme la **hiérarchie nette** entre les trois catégories de magasins :

- **Type A** : les magasins les plus grands (superficie moyenne la plus élevée) génèrent la **plus grande part du CA** avec les ventes moyennes les plus élevées. Leur distribution est également la plus dispersée, avec une longue queue vers les valeurs élevées.
- **Type B** : catégorie intermédiaire en termes de taille et de performance. Les ventes sont plus modérées mais restent significatives.
- **Type C** : les plus petits magasins, avec les ventes les plus faibles et les distributions les plus resserrées.

**Observations clés :**
- La variable `Type` (et par extension `Size`) est un **prédicteur majeur** du niveau de ventes — elle devra figurer parmi les features prioritaires dans les modèles.
- Malgré la hiérarchie globale, la **variabilité intra-type** reste importante (surtout pour le Type A), ce qui justifie l'utilisation de features supplémentaires pour affiner les prévisions.
- Les distributions en violon montrent que les **médianes** diffèrent davantage que les **moyennes**, confirmant l'asymétrie des ventes dans tous les types.

> ✅ **Conclusion de la section 1.4.2** : L'analyse descriptive des ventes par magasin et département révèle des disparités marquées, tant entre magasins (effet taille/type) qu'entre départements (effet Pareto). Ces patterns structurels seront exploités dans la phase de feature engineering et de modélisation.

### <span id="143" style="color:#2E86C1;">1.4.3 Analyse des tendances temporelles et saisonnières</span>

  Cette section explore la **dimension temporelle** des ventes hebdomadaires Walmart. L'objectif est d'identifier les **tendances de fond**, les **cycles saisonniers** et l'**impact des événements commerciaux** (jours fériés) sur les performances de vente.

  L'analyse se déroule en **six étapes** :

  1. **Évolution globale des ventes hebdomadaires** — série temporelle agrégée sur l'ensemble des magasins avec moyenne mobile.
  2. **Saisonnalité mensuelle et comparaison annuelle** — profil saisonnier moyen et comparaison Year-over-Year.
  3. **Tendances par type de magasin** — décomposition temporelle selon les types A, B et C.
  4. **Impact des jours fériés** — quantification de l'effet des semaines fériées sur les ventes.
  5. **Profil saisonnier des top départements** — patterns mensuels des départements les plus contributifs.
  6. **Synthèse temporelle** — tableau récapitulatif annuel et trimestriel.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Évolution globale des ventes hebdomadaires</strong>

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

# ══════════════════════════════════════════════════════════════════════════
#  1) ÉVOLUTION GLOBALE DES VENTES HEBDOMADAIRES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) ÉVOLUTION GLOBALE DES VENTES HEBDOMADAIRES")
print("=" * 70)

# Agrégation hebdomadaire totale
weekly_total = (
    df_train.groupby('Date', observed=True)
    .agg(Total_Sales=('Weekly_Sales', 'sum'),
         Moyenne=('Weekly_Sales', 'mean'),
         Nb_obs=('Weekly_Sales', 'count'),
         Holiday=('IsHoliday', 'first'))
    .reset_index()
    .sort_values('Date')
)

# Moyenne mobile 4 semaines
weekly_total['MA_4w'] = weekly_total['Total_Sales'].rolling(window=4, center=True).mean()

print(f"\n► Période couverte : {weekly_total['Date'].min().strftime('%d/%m/%Y')} → {weekly_total['Date'].max().strftime('%d/%m/%Y')}")
print(f"► Nombre de semaines : {len(weekly_total)}")
print(f"  - Semaines fériées : {weekly_total['Holiday'].sum()}")
print(f"  - Semaines normales : {(~weekly_total['Holiday']).sum()}")
print(f"\n► Ventes hebdomadaires totales (tous magasins) :")
print(f"  - Minimum : {weekly_total['Total_Sales'].min():,.0f} $  ({weekly_total.loc[weekly_total['Total_Sales'].idxmin(), 'Date'].strftime('%d/%m/%Y')})")
print(f"  - Maximum : {weekly_total['Total_Sales'].max():,.0f} $  ({weekly_total.loc[weekly_total['Total_Sales'].idxmax(), 'Date'].strftime('%d/%m/%Y')})")
print(f"  - Moyenne : {weekly_total['Total_Sales'].mean():,.0f} $")

# ── Graphique : évolution temporelle ──────────────────────────────────
fig_ts = go.Figure()

# Bandes verticales pour les fêtes
for _, r in weekly_total[weekly_total['Holiday']].iterrows():
    fig_ts.add_vrect(
        x0=r['Date'] - pd.Timedelta(days=3),
        x1=r['Date'] + pd.Timedelta(days=3),
        fillcolor='rgba(255, 200, 50, 0.15)',
        line_width=0, layer='below'
    )

# Série principale
fig_ts.add_trace(go.Scatter(
    x=weekly_total['Date'], y=weekly_total['Total_Sales'],
    mode='lines', name='Ventes hebdo totales',
    line=dict(color='#2E86C1', width=1.5),
    hovertemplate='%{x|%d/%m/%Y}<br>Ventes: %{y:,.0f} $<extra></extra>'
))

# Moyenne mobile
fig_ts.add_trace(go.Scatter(
    x=weekly_total['Date'], y=weekly_total['MA_4w'],
    mode='lines', name='Moyenne mobile 4 sem.',
    line=dict(color='#E74C3C', width=2.5, dash='dot')
))

# Marqueurs fêtes
holidays_ts = weekly_total[weekly_total['Holiday']]
fig_ts.add_trace(go.Scatter(
    x=holidays_ts['Date'], y=holidays_ts['Total_Sales'],
    mode='markers', name='Semaines fériées',
    marker=dict(color='#F39C12', size=9, symbol='star', line=dict(width=1, color='#333'))
))

fig_ts.update_layout(
    title='Évolution des ventes hebdomadaires totales – Tous magasins confondus',
    xaxis_title='Date', yaxis_title='Ventes totales ($)',
    height=500, width=1000,
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    hovermode='x unified'
)
fig_ts.show()
```

```python
# ══════════════════════════════════════════════════════════════════════════
#  2) SAISONNALITÉ MENSUELLE ET COMPARAISON ANNUELLE
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  2) SAISONNALITÉ MENSUELLE ET COMPARAISON ANNUELLE")
print("=" * 70)

# Extraire année et mois
df_train['Year'] = df_train['Date'].dt.year
df_train['Month'] = df_train['Date'].dt.month

month_labels = ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Jun',
                'Jul', 'Aoû', 'Sep', 'Oct', 'Nov', 'Déc']

# Ventes mensuelles par année
monthly_sales = (
    df_train.groupby(['Year', 'Month'], observed=True)['Weekly_Sales']
    .agg(['mean', 'sum', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'sum': 'Total', 'count': 'Nb_obs'})
    .sort_values(['Year', 'Month'])
)

# Profil mensuel moyen (toutes années)
monthly_avg = (
    df_train.groupby('Month', observed=True)['Weekly_Sales']
    .agg(['mean', 'std'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'std': 'Ecart_type'})
)
monthly_avg['Mois'] = monthly_avg['Month'].map(dict(zip(range(1, 13), month_labels)))

mois_max = monthly_avg.loc[monthly_avg['Moyenne'].idxmax()]
mois_min = monthly_avg.loc[monthly_avg['Moyenne'].idxmin()]
print(f"\n► Profil saisonnier mensuel (moyenne sur toutes les années) :")
print(f"  - Mois le plus fort : {mois_max['Mois']} ({mois_max['Moyenne']:,.0f} $ en moyenne)")
print(f"  - Mois le plus faible : {mois_min['Mois']} ({mois_min['Moyenne']:,.0f} $ en moyenne)")
print(f"  - Amplitude saisonnière : {(mois_max['Moyenne'] / mois_min['Moyenne'] - 1) * 100:.1f}%")

# ── Graphique 1 : Comparaison Year-over-Year ─────────────────────────
fig_yoy = go.Figure()
colors_year = {2010: '#2E86C1', 2011: '#E67E22', 2012: '#27AE60'}

for year in sorted(monthly_sales['Year'].unique()):
    data_y = monthly_sales[monthly_sales['Year'] == year]
    fig_yoy.add_trace(go.Scatter(
        x=data_y['Month'], y=data_y['Moyenne'],
        mode='lines+markers', name=str(year),
        line=dict(color=colors_year.get(year, '#999'), width=2.5),
        marker=dict(size=7),
        hovertemplate=f'{year}<br>Mois %{{x}}<br>Vente moy: %{{y:,.0f}} $<extra></extra>'
    ))

fig_yoy.update_layout(
    title='Comparaison Year-over-Year des ventes moyennes mensuelles',
    xaxis=dict(title='Mois', tickvals=list(range(1, 13)), ticktext=month_labels),
    yaxis_title='Ventes moyennes ($)',
    height=450, width=900,
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_yoy.show()

# ── Graphique 2 : Profil saisonnier moyen ± écart-type ───────────────
fig_season = go.Figure()

fig_season.add_trace(go.Scatter(
    x=monthly_avg['Month'],
    y=monthly_avg['Moyenne'] + monthly_avg['Ecart_type'],
    mode='lines', line=dict(width=0), showlegend=False, hoverinfo='skip'
))
fig_season.add_trace(go.Scatter(
    x=monthly_avg['Month'],
    y=monthly_avg['Moyenne'] - monthly_avg['Ecart_type'],
    mode='lines', line=dict(width=0), showlegend=False,
    fill='tonexty', fillcolor='rgba(46, 134, 193, 0.2)', hoverinfo='skip'
))
fig_season.add_trace(go.Scatter(
    x=monthly_avg['Month'], y=monthly_avg['Moyenne'],
    mode='lines+markers', name='Moyenne mensuelle',
    line=dict(color='#2E86C1', width=3),
    marker=dict(size=9, color='#2E86C1'),
    hovertemplate='%{text}<br>Moyenne: %{y:,.0f} $<extra></extra>',
    text=monthly_avg['Mois']
))

fig_season.update_layout(
    title='Profil saisonnier moyen des ventes (± 1 écart-type)',
    xaxis=dict(title='Mois', tickvals=list(range(1, 13)), ticktext=month_labels),
    yaxis_title='Ventes moyennes ($)',
    height=450, width=900,
    template='plotly_white', showlegend=False
)
fig_season.show()
```

```python
# ══════════════════════════════════════════════════════════════════════════
#  4) IMPACT DES JOURS FÉRIÉS SUR LES VENTES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  4) IMPACT DES JOURS FÉRIÉS SUR LES VENTES")
print("=" * 70)

# Comparaison Holiday vs Non-Holiday
holiday_comp = (
    df_train.groupby('IsHoliday', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'std', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Mediane', 'std': 'Ecart_type', 'count': 'Nb_obs'})
)
holiday_comp['Période'] = holiday_comp['IsHoliday'].map({True: 'Férié', False: 'Normal'})

print(f"\n► Comparaison Holiday vs Non-Holiday :")
display(
    holiday_comp[['Période', 'Moyenne', 'Mediane', 'Ecart_type', 'Nb_obs']].style
    .set_caption('Ventes hebdomadaires – Semaines fériées vs normales')
    .format({'Moyenne': '{:,.0f} $', 'Mediane': '{:,.0f} $', 'Ecart_type': '{:,.0f}', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='Blues')
)

moy_hol = holiday_comp.loc[holiday_comp['IsHoliday'] == True, 'Moyenne'].values[0]
moy_norm = holiday_comp.loc[holiday_comp['IsHoliday'] == False, 'Moyenne'].values[0]
lift = (moy_hol / moy_norm - 1) * 100
print(f"\n► Effet férié : +{lift:.1f}% de ventes moyennes en semaine fériée")

# Détail par fête (classées par mois)
holiday_rows = df_train[df_train['IsHoliday']].copy()

def classify_holiday(date):
    m = date.month
    if m == 2: return 'Super Bowl'
    elif m == 9: return 'Labour Day'
    elif m == 11: return 'Thanksgiving'
    elif m == 12: return 'Christmas'
    else: return 'Autre'

holiday_rows['Fete'] = holiday_rows['Date'].apply(classify_holiday)

fete_stats = (
    holiday_rows.groupby('Fete', observed=True)['Weekly_Sales']
    .agg(['mean', 'sum', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'sum': 'Total', 'count': 'Nb_obs'})
    .sort_values('Moyenne', ascending=False)
)

print(f"\n► Ventes moyennes par fête :")
display(
    fete_stats.style
    .set_caption('Ventes moyennes par événement commercial majeur')
    .format({'Moyenne': '{:,.0f} $', 'Total': '{:,.0f} $', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='YlOrRd')
)

# ── Graphique : Holiday vs Normal par type de magasin ─────────────────
holiday_type = (
    df_train.groupby(['Type', 'IsHoliday'], observed=True)['Weekly_Sales']
    .mean()
    .reset_index()
)
holiday_type['Période'] = holiday_type['IsHoliday'].map({True: 'Férié', False: 'Normal'})

fig_hol_bar = px.bar(
    holiday_type,
    x='Type', y='Weekly_Sales',
    color='Période', barmode='group',
    color_discrete_map={'Férié': '#F39C12', 'Normal': '#2E86C1'},
    title='Impact des semaines fériées par type de magasin',
    labels={'Weekly_Sales': 'Ventes moyennes ($)', 'Type': 'Type de magasin'},
    text_auto='.0f'
)
fig_hol_bar.update_layout(
    height=450, width=700,
    template='plotly_white'
)
fig_hol_bar.show()
```

- **Thanksgiving en tête** : c'est de loin l'événement le plus impactant, avec des ventes moyennes nettement supérieures à tous les autres jours fériés. Cela s'explique par le Black Friday qui suit immédiatement.

```python
# ══════════════════════════════════════════════════════════════════════════
#  6) SYNTHÈSE TEMPORELLE
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  6) SYNTHÈSE TEMPORELLE")
print("=" * 70)

# Tableau récapitulatif par année
yearly_stats = (
    df_train.groupby('Year', observed=True)['Weekly_Sales']
    .agg(['sum', 'mean', 'median', 'std', 'count'])
    .reset_index()
    .rename(columns={'sum': 'CA_total', 'mean': 'Moyenne', 'median': 'Mediane',
                     'std': 'Ecart_type', 'count': 'Nb_obs'})
)
yearly_stats['CV (%)'] = (yearly_stats['Ecart_type'] / yearly_stats['Moyenne'] * 100).round(1)

print(f"\n► Tableau récapitulatif par année :")
display(
    yearly_stats.style
    .set_caption('Statistiques annuelles des ventes – Dataset TRAIN')
    .format({
        'CA_total': '{:,.0f} $', 'Moyenne': '{:,.0f} $', 'Mediane': '{:,.0f} $',
        'Ecart_type': '{:,.0f}', 'Nb_obs': '{:,}', 'CV (%)': '{:.1f}%'
    })
    .background_gradient(subset=['CA_total'], cmap='Blues')
)

# Tableau par trimestre
df_train['Quarter'] = df_train['Date'].dt.quarter
quarterly_stats = (
    df_train.groupby('Quarter', observed=True)['Weekly_Sales']
    .agg(['mean', 'std'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'std': 'Ecart_type'})
)
quarterly_stats['Trimestre'] = quarterly_stats['Quarter'].map(
    {1: 'Q1 (Jan-Mar)', 2: 'Q2 (Avr-Jun)', 3: 'Q3 (Jul-Sep)', 4: 'Q4 (Oct-Déc)'}
)
quarterly_stats['CV (%)'] = (quarterly_stats['Ecart_type'] / quarterly_stats['Moyenne'] * 100).round(1)

print(f"\n► Ventes moyennes par trimestre :")
display(
    quarterly_stats[['Trimestre', 'Moyenne', 'Ecart_type', 'CV (%)']].style
    .set_caption('Profil trimestriel des ventes – Dataset TRAIN')
    .format({'Moyenne': '{:,.0f} $', 'Ecart_type': '{:,.0f}', 'CV (%)': '{:.1f}%'})
    .background_gradient(subset=['Moyenne'], cmap='Blues')
)

# ── Graphique : boxplots par mois (coloré par IsHoliday) ─────────────
fig_box_month = px.box(
    df_train,
    x='Month', y='Weekly_Sales',
    color='IsHoliday',
    color_discrete_map={True: '#F39C12', False: '#2E86C1'},
    title='Distribution des ventes par mois (coloré par statut férié)',
    labels={'Weekly_Sales': 'Ventes ($)', 'Month': 'Mois', 'IsHoliday': 'Férié'},
    category_orders={'Month': list(range(1, 13))}
)
fig_box_month.update_layout(
    xaxis=dict(tickvals=list(range(1, 13)), ticktext=month_labels),
    height=500, width=1000,
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_box_month.show()
```

### <span id="151" style="color:#2E86C1;">1.5.1 Analyse de l'effet des promotions</span>

Les variables **MarkDown1 à MarkDown5** représentent les **réductions de prix temporaires** appliquées par Walmart lors de campagnes promotionnelles. Ces données ne sont disponibles qu'à partir de **novembre 2011**, ce qui signifie qu'une partie significative des observations ne dispose d'aucune information promotionnelle.

L'objectif de cette sous-section est d'évaluer l'**impact réel des promotions sur les ventes hebdomadaires**, en répondant à plusieurs questions clés :

L'analyse se déroule en **six étapes** :

1. **Vue d'ensemble des variables MarkDown** — disponibilité, statistiques descriptives et création d'un indicateur de promotion active.
2. **Impact global des promotions** — comparaison des ventes avec et sans promotion, quantification du lift.
3. **Évolution temporelle des promotions** — chronologie des campagnes et corrélation avec les ventes.
4. **Impact par type de MarkDown** — analyse différenciée du lift pour chaque type de promotion (MarkDown1 à MarkDown5).
5. **Sensibilité par département et type de magasin** — identification des départements et types de magasins les plus réactifs aux promotions.
6. **Corrélation et synthèse** — relation entre l'intensité promotionnelle et les ventes, bilan global.

<strong>Vue d'ensemble des variables MarkDown</strong>

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

# ══════════════════════════════════════════════════════════════════════════
#  1) VUE D'ENSEMBLE DES VARIABLES MARKDOWN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) VUE D'ENSEMBLE DES VARIABLES MARKDOWN")
print("=" * 70)

md_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

# Statistiques descriptives
md_stats = pd.DataFrame({
    'Variable': md_cols,
    'Non-null': [df_train[c].notna().sum() for c in md_cols],
    'Disponibilité (%)': [df_train[c].notna().mean() * 100 for c in md_cols],
    'Moyenne ($)': [df_train[c].mean() for c in md_cols],
    'Médiane ($)': [df_train[c].median() for c in md_cols],
    'Écart-type ($)': [df_train[c].std() for c in md_cols],
    'Min ($)': [df_train[c].min() for c in md_cols],
    'Max ($)': [df_train[c].max() for c in md_cols],
    'Valeurs ≤ 0': [(df_train[c] <= 0).sum() for c in md_cols],
})
print("\n► Statistiques descriptives des variables MarkDown :")
display(md_stats.style.format({
    'Disponibilité (%)': '{:.1f}', 'Moyenne ($)': '{:,.0f}', 'Médiane ($)': '{:,.0f}',
    'Écart-type ($)': '{:,.0f}', 'Min ($)': '{:,.0f}', 'Max ($)': '{:,.0f}'
}).set_caption("Statistiques descriptives des 5 variables MarkDown"))

# Indicateur de promotion active (au moins un MarkDown > 0)
df_train['HasPromo'] = df_train[md_cols].gt(0).any(axis=1)
df_train['Total_MarkDown'] = df_train[md_cols].clip(lower=0).sum(axis=1)

n_promo = df_train['HasPromo'].sum()
n_total = len(df_train)

print(f"\n► Indicateur de promotion active (HasPromo) :")
print(f"  - Semaines avec au moins une promotion : {n_promo:,} / {n_total:,} ({n_promo/n_total*100:.1f} %)")
print(f"  - Semaines sans promotion : {n_total - n_promo:,} ({(n_total - n_promo)/n_total*100:.1f} %)")

# Période de disponibilité des MarkDown
md_dispo = df_train[df_train[md_cols].notna().any(axis=1)]['Date']
print(f"\n► Période de disponibilité des MarkDown :")
print(f"  - Première date avec MarkDown : {md_dispo.min().strftime('%d/%m/%Y')}")
print(f"  - Dernière date avec MarkDown : {md_dispo.max().strftime('%d/%m/%Y')}")
```

L'examen des cinq variables MarkDown met en lumière plusieurs caractéristiques essentielles :

- **Valeurs hétérogènes** : les ordres de grandeur varient fortement d'un MarkDown à l'autre — certains atteignent des montants très élevés tandis que d'autres restent modestes, reflétant des types de campagnes promotionnelles différents.
- **Indicateur synthétique** : la variable `HasPromo` (au moins un MarkDown > 0) permet de distinguer les semaines promotionnelles des semaines normales. Le `Total_MarkDown` quantifie l'investissement promotionnel total.

> 💡 **Point clé** : Les MarkDown offrent un signal promotionnel exploitable, mais leur caractère incomplet et asymétrique nécessitera un traitement adapté lors de l'ingénierie de features (imputation par 0, clipping des valeurs négatives).

```python
# ══════════════════════════════════════════════════════════════════════════
#  2) IMPACT GLOBAL DES PROMOTIONS SUR LES VENTES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  2) IMPACT GLOBAL DES PROMOTIONS SUR LES VENTES")
print("=" * 70)

# Comparaison avec / sans promotion
promo_comp = df_train.groupby('HasPromo', observed=True)['Weekly_Sales'].agg(
    ['mean', 'median', 'std', 'count']
).rename(index={True: 'Avec promotion', False: 'Sans promotion'})
promo_comp.columns = ['Moyenne ($)', 'Médiane ($)', 'Écart-type ($)', 'Nb observations']

print("\n► Comparaison des ventes avec vs sans promotion :")
display(promo_comp.style.format({
    'Moyenne ($)': '{:,.0f}', 'Médiane ($)': '{:,.0f}',
    'Écart-type ($)': '{:,.0f}', 'Nb observations': '{:,}'
}).set_caption("Ventes hebdomadaires : Avec vs Sans promotion"))

lift_mean = (promo_comp.loc['Avec promotion', 'Moyenne ($)'] / promo_comp.loc['Sans promotion', 'Moyenne ($)'] - 1) * 100
lift_med = (promo_comp.loc['Avec promotion', 'Médiane ($)'] / promo_comp.loc['Sans promotion', 'Médiane ($)'] - 1) * 100
print(f"\n► Lift promotionnel (moyenne) : {lift_mean:+.1f} %")
print(f"► Lift promotionnel (médiane) : {lift_med:+.1f} %")

# Distribution des ventes avec vs sans promo
fig_promo_dist = go.Figure()
for label, flag in [('Sans promotion', False), ('Avec promotion', True)]:
    subset = df_train[df_train['HasPromo'] == flag]['Weekly_Sales']
    fig_promo_dist.add_trace(go.Histogram(
        x=subset.clip(upper=subset.quantile(0.99)),
        name=label, opacity=0.65, nbinsx=80
    ))

fig_promo_dist.update_layout(
    title="Distribution des ventes hebdomadaires – Avec vs Sans promotion",
    xaxis_title="Ventes hebdomadaires ($)", yaxis_title="Fréquence",
    barmode='overlay', height=450, width=1000,
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_promo_dist.show()
```

La comparaison directe entre semaines avec et sans promotion confirme l'existence d'un **effet promotionnel positif** :

- **Lift en moyenne** : les semaines avec au moins une promotion active affichent des ventes moyennes supérieures à celles sans promotion, traduisant un effet d'attraction des campagnes MarkDown.
- **Lift en médiane** : l'écart se maintient sur la médiane, ce qui prouve que l'effet n'est pas uniquement tiré par quelques valeurs extrêmes — il s'agit bien d'un effet de masse.
- **Distribution plus étalée** : la distribution des ventes en période de promotion est plus dispersée, ce qui suggère que l'effet des promotions n'est pas uniforme — certains magasins/départements en profitent davantage que d'autres.
- **Attention au biais temporel** : les promotions étant concentrées à partir de fin 2011 (incluant les périodes de fêtes), une partie du lift observé pourrait être confondue avec l'effet saisonnier.

> 💡 **Point clé** : L'effet promotionnel est réel mais doit être interprété avec prudence en raison du biais temporel (promotions absentes avant novembre 2011).

```python
# ══════════════════════════════════════════════════════════════════════════
#  4) IMPACT PAR TYPE DE MARKDOWN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  4) IMPACT PAR TYPE DE MARKDOWN")
print("=" * 70)

md_impact = []
for md in md_cols:
    mask = df_train[md].gt(0)
    ventes_avec = df_train.loc[mask, 'Weekly_Sales'].mean()
    ventes_sans = df_train.loc[~mask & df_train[md].notna(), 'Weekly_Sales'].mean()
    lift_md = (ventes_avec / ventes_sans - 1) * 100 if ventes_sans else 0
    md_impact.append({
        'Variable': md,
        'Ventes avec ($)': ventes_avec,
        'Ventes sans ($)': ventes_sans,
        'Lift (%)': lift_md,
        'Nb obs (avec)': mask.sum(),
        'Investissement moyen ($)': df_train.loc[mask, md].mean(),
    })

md_impact_df = pd.DataFrame(md_impact).sort_values('Lift (%)', ascending=False)
print("\n► Lift des ventes par type de MarkDown :")
display(md_impact_df.style.format({
    'Ventes avec ($)': '{:,.0f}', 'Ventes sans ($)': '{:,.0f}',
    'Lift (%)': '{:+.1f}', 'Nb obs (avec)': '{:,}', 'Investissement moyen ($)': '{:,.0f}'
}).set_caption("Impact différencié des 5 types de MarkDown"))

# Graphique : lift par MarkDown
colors_lift = ['#27AE60' if v > 0 else '#E74C3C' for v in md_impact_df['Lift (%)']]
fig_md_lift = go.Figure(go.Bar(
    x=md_impact_df['Variable'], y=md_impact_df['Lift (%)'],
    marker_color=colors_lift,
    text=[f"{v:+.1f} %" for v in md_impact_df['Lift (%)']],
    textposition='outside'
))
fig_md_lift.update_layout(
    title='Lift des ventes par type de MarkDown',
    xaxis_title='Variable MarkDown', yaxis_title='Lift (%)',
    height=420, width=800, template='plotly_white'
)
fig_md_lift.show()
```

**📝 Commentaire – Impact par type de MarkDown**

L'analyse différenciée par type de MarkDown révèle des profils d'impact contrastés :

- **MarkDown hétérogènes** : chaque type de MarkDown a un effet distinct sur les ventes. Certains MarkDown sont associés à un lift significatif, tandis que d'autres présentent un effet marginal, voire négatif.
- **Niveaux d'investissement variables** : la valeur moyenne et la fréquence d'utilisation diffèrent fortement entre les cinq types, suggérant des stratégies promotionnelles distinctes selon les catégories de produits.
- **Pouvoir prédictif différencié** : les MarkDown les plus impactants devraient recevoir un poids plus important dans les modèles prédictifs.

> 💡 **Point clé** : Il est pertinent de conserver les 5 variables MarkDown séparément (plutôt que de les agréger systématiquement), car chacune capture un type de promotion différent avec un effet spécifique sur les ventes.

<strong>Sensibilité par département et type de magasin</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  5) SENSIBILITÉ PAR DÉPARTEMENT ET TYPE DE MAGASIN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  5) SENSIBILITÉ PAR DÉPARTEMENT ET TYPE DE MAGASIN")
print("=" * 70)

# ── A) Lift par département ───────────────────────────────────────────
dept_promo = (
    df_train.groupby(['Dept', 'HasPromo'], observed=True)['Weekly_Sales']
    .mean().unstack(fill_value=0)
)
dept_promo.columns = ['Sans promo', 'Avec promo']
dept_promo['Lift (%)'] = (dept_promo['Avec promo'] / dept_promo['Sans promo'] - 1) * 100
dept_promo = dept_promo.sort_values('Lift (%)', ascending=False)

print("\n► Top 15 départements les plus sensibles aux promotions :")
display(dept_promo.head(15).style.format({
    'Sans promo': '{:,.0f}', 'Avec promo': '{:,.0f}', 'Lift (%)': '{:+.1f}'
}))

print("\n► 5 départements les moins sensibles (ou effet négatif) :")
display(dept_promo.tail(5).style.format({
    'Sans promo': '{:,.0f}', 'Avec promo': '{:,.0f}', 'Lift (%)': '{:+.1f}'
}))

# Graphique : top 20 départements par lift
top20_lift = dept_promo.head(20).reset_index()
fig_dept_lift = go.Figure(go.Bar(
    y=top20_lift['Dept'].astype(str),
    x=top20_lift['Lift (%)'],
    orientation='h',
    marker_color='#2E86C1',
    text=[f"{v:+.1f} %" for v in top20_lift['Lift (%)']],
    textposition='outside'
))
fig_dept_lift.update_layout(
    title='Top 20 départements – Lift promotionnel (ventes moyennes)',
    xaxis_title='Lift (%)', yaxis_title='Département',
    height=550, width=900, template='plotly_white',
    yaxis=dict(autorange='reversed')
)
fig_dept_lift.show()

# ── B) Lift par type de magasin ───────────────────────────────────────
print("\n► Lift par type de magasin :")
for t in sorted(df_train['Type'].unique()):
    avec = df_train[(df_train['Type'] == t) & (df_train['HasPromo'] == True)]['Weekly_Sales'].mean()
    sans = df_train[(df_train['Type'] == t) & (df_train['HasPromo'] == False)]['Weekly_Sales'].mean()
    lift_t = (avec / sans - 1) * 100
    print(f"  Type {t} : avec promo = {avec:,.0f} $ | sans promo = {sans:,.0f} $ | lift = {lift_t:+.1f} %")

# Box plot par type et promotion
fig_type_promo = px.box(
    df_train[df_train['Weekly_Sales'] <= df_train['Weekly_Sales'].quantile(0.99)],
    x='Type', y='Weekly_Sales', color='HasPromo',
    labels={'Weekly_Sales': 'Ventes hebdomadaires ($)', 'Type': 'Type de magasin', 'HasPromo': 'Promotion'},
    title='Distribution des ventes par type de magasin et statut promotionnel',
    color_discrete_map={True: '#E67E22', False: '#2E86C1'},
    category_orders={'Type': ['A', 'B', 'C']}
)
fig_type_promo.update_layout(
    height=500, width=900, template='plotly_white',
    legend=dict(title='Promotion active', orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_type_promo.show()
```

- **Départements les plus sensibles** : les départements à forte composante saisonnière ou événementielle tendent à mieux répondre aux promotions.
- **Effet par type de magasin** : les magasins de type A (grande taille) tendent à mieux capitaliser sur les promotions, probablement grâce à un assortiment plus large et un trafic plus important. Les types B et C montrent un effet plus modéré.

### <span id="152" style="color:#2E86C1;">1.5.2 Analyse de l'impact des périodes de fêtes</span>

La section 1.4.3 a mis en évidence l'existence d'un **effet significatif des périodes de fêtes** sur les ventes hebdomadaires, avec un lift positif pour les semaines marquées `IsHoliday = True`. Toutefois, cette première analyse restait globale et ne distinguait pas les différents événements commerciaux ni leurs mécanismes d'influence.

L'objectif de cette sous-section est d'**approfondir l'analyse de l'impact des fêtes** en répondant à plusieurs questions clés :

- Quel est l'**effet spécifique** de chaque fête (Super Bowl, Thanksgiving, Noël, Fête du Travail) sur les ventes ?
- L'effet des fêtes est-il **uniforme** ou varie-t-il selon le **type de magasin** et le **département** ?
- Les fêtes génèrent-elles un **pic ponctuel** ou un **effet étalé** sur plusieurs semaines (avant/après) ?
- Quels sont les **départements les plus sensibles** aux périodes de fêtes ?

L'analyse se déroule en **cinq étapes** :

1. **Effet global des fêtes** — rappel chiffré et test statistique de la différence entre semaines fériées et normales.
2. **Impact par fête spécifique** — classement des quatre fêtes par lift moyen et visualisation comparative.
3. **Effet temporel élargi (avant/après fête)** — analyse de la dynamique des ventes dans les semaines encadrant chaque fête.
4. **Sensibilité par type de magasin** — interaction entre le statut férié et le type de magasin (A, B, C).
5. **Départements les plus sensibles aux fêtes** — identification des départements qui bénéficient le plus des périodes fériées.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Effet global des fêtes et test statistique</strong>
</div>

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
from scipy import stats

# ══════════════════════════════════════════════════════════════════════════
#  1) EFFET GLOBAL DES FÊTES ET TEST STATISTIQUE
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) EFFET GLOBAL DES FÊTES ET TEST STATISTIQUE")
print("=" * 70)

# Rappel chiffré : Holiday vs Normal
holiday_recap = (
    df_train.groupby('IsHoliday', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'std', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Médiane', 'std': 'Écart-type', 'count': 'Nb_obs'})
)
holiday_recap['Période'] = holiday_recap['IsHoliday'].map({True: 'Semaine fériée', False: 'Semaine normale'})

print("\n► Rappel — Ventes moyennes : Fériées vs Normales :")
display(
    holiday_recap[['Période', 'Moyenne', 'Médiane', 'Écart-type', 'Nb_obs']].style
    .set_caption('Comparaison des ventes — Semaines fériées vs normales')
    .format({'Moyenne': '{:,.0f} $', 'Médiane': '{:,.0f} $', 'Écart-type': '{:,.0f}', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='Blues')
)

moy_h = holiday_recap.loc[holiday_recap['IsHoliday'] == True, 'Moyenne'].values[0]
moy_n = holiday_recap.loc[holiday_recap['IsHoliday'] == False, 'Moyenne'].values[0]
lift_h = (moy_h / moy_n - 1) * 100
print(f"\n► Lift férié global : {lift_h:+.1f} % (ventes moyennes)")

# Test de Welch (variances inégales)
ventes_hol = df_train.loc[df_train['IsHoliday'], 'Weekly_Sales']
ventes_norm = df_train.loc[~df_train['IsHoliday'], 'Weekly_Sales']
t_stat, p_value = stats.ttest_ind(ventes_hol, ventes_norm, equal_var=False)
print(f"\n► Test de Welch (H₀ : μ_férié = μ_normal) :")
print(f"  - t-statistique : {t_stat:.2f}")
print(f"  - p-value       : {p_value:.2e}")
print(f"  - Conclusion    : {'Différence significative (p < 0.05) ✅' if p_value < 0.05 else 'Pas de différence significative ❌'}")

# Graphique : distribution Holiday vs Normal (violin)
fig_violin = go.Figure()
for label, mask, color in [('Semaine normale', ~df_train['IsHoliday'], '#2E86C1'),
                            ('Semaine fériée', df_train['IsHoliday'], '#F39C12')]:
    subset = df_train.loc[mask, 'Weekly_Sales'].clip(upper=df_train['Weekly_Sales'].quantile(0.99))
    fig_violin.add_trace(go.Violin(
        y=subset, name=label,
        box_visible=True, meanline_visible=True,
        fillcolor=color, opacity=0.6,
        line_color=color
    ))

fig_violin.update_layout(
    title='Distribution des ventes — Semaines fériées vs normales',
    yaxis_title='Ventes hebdomadaires ($)',
    height=500, width=800, template='plotly_white',
    showlegend=False
)
fig_violin.show()
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Effet global des fêtes et test statistique**

L'analyse statistique confirme les observations exploratoires de la section 1.4.3 :

- **Lift positif confirmé** : les semaines fériées affichent des ventes moyennes supérieures aux semaines normales, avec un lift de l'ordre de quelques pourcents.
- **Significativité statistique** : le test de Welch rejette l'hypothèse nulle d'égalité des moyennes avec une p-value extrêmement faible, confirmant que la différence n'est pas due au hasard.
- **Distributions chevauchantes** : malgré la différence de moyenne, les violons montrent un large chevauchement entre les deux distributions, ce qui signifie que l'effet férié n'est pas universel — il dépend du magasin, du département et de la fête concernée.

> 💡 **Point clé** : L'indicateur `IsHoliday` est statistiquement significatif, mais son effet global modéré cache des disparités importantes qu'il convient de décomposer par fête et par contexte.

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Impact par fête spécifique</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  2) IMPACT PAR FÊTE SPÉCIFIQUE
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  2) IMPACT PAR FÊTE SPÉCIFIQUE")
print("=" * 70)

# Classification des fêtes par mois
def classify_holiday(date):
    m = date.month
    if m == 2: return 'Super Bowl'
    elif m == 9: return 'Labour Day'
    elif m == 11: return 'Thanksgiving'
    elif m == 12: return 'Christmas'
    else: return 'Autre'

holiday_rows = df_train[df_train['IsHoliday']].copy()
holiday_rows['Fete'] = holiday_rows['Date'].apply(classify_holiday)

# Moyenne normale (référence)
moy_normale = df_train.loc[~df_train['IsHoliday'], 'Weekly_Sales'].mean()

# Statistiques par fête
fete_detail = (
    holiday_rows.groupby('Fete', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'std', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Médiane', 'std': 'Écart-type', 'count': 'Nb_obs'})
)
fete_detail['Lift vs Normal (%)'] = ((fete_detail['Moyenne'] / moy_normale) - 1) * 100
fete_detail = fete_detail.sort_values('Lift vs Normal (%)', ascending=False)

print(f"\n► Référence : ventes moyennes en semaine normale = {moy_normale:,.0f} $")
print(f"\n► Ventes moyennes et lift par fête :")
display(
    fete_detail.style
    .set_caption('Impact différencié des 4 fêtes majeures sur les ventes')
    .format({
        'Moyenne': '{:,.0f} $', 'Médiane': '{:,.0f} $', 'Écart-type': '{:,.0f}',
        'Nb_obs': '{:,}', 'Lift vs Normal (%)': '{:+.1f}'
    })
    .background_gradient(subset=['Lift vs Normal (%)'], cmap='RdYlGn')
)

# Graphique : barres horizontales par fête
ordre_fetes = fete_detail['Fete'].tolist()
colors_fetes = {'Thanksgiving': '#E67E22', 'Super Bowl': '#3498DB',
                'Christmas': '#E74C3C', 'Labour Day': '#27AE60', 'Autre': '#95A5A6'}

fig_fete = go.Figure()
fig_fete.add_trace(go.Bar(
    y=fete_detail['Fete'], x=fete_detail['Lift vs Normal (%)'],
    orientation='h',
    marker_color=[colors_fetes.get(f, '#95A5A6') for f in fete_detail['Fete']],
    text=[f"{v:+.1f} %" for v in fete_detail['Lift vs Normal (%)']],
    textposition='outside'
))
fig_fete.add_vline(x=0, line_dash='dash', line_color='gray', line_width=1)
fig_fete.update_layout(
    title='Lift des ventes par fête (vs semaines normales)',
    xaxis_title='Lift (%)', yaxis_title='',
    height=400, width=800, template='plotly_white',
    yaxis=dict(autorange='reversed')
)
fig_fete.show()

# Boxplot comparatif : distribution par fête
holiday_rows_plot = holiday_rows.copy()
holiday_rows_plot = holiday_rows_plot[holiday_rows_plot['Weekly_Sales'] <= holiday_rows_plot['Weekly_Sales'].quantile(0.99)]

fig_box_fete = px.box(
    holiday_rows_plot,
    x='Fete', y='Weekly_Sales',
    color='Fete',
    color_discrete_map=colors_fetes,
    category_orders={'Fete': ordre_fetes},
    title='Distribution des ventes par fête',
    labels={'Weekly_Sales': 'Ventes hebdomadaires ($)', 'Fete': 'Événement'}
)
fig_box_fete.update_layout(
    height=500, width=900, template='plotly_white',
    showlegend=False
)
fig_box_fete.show()
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Impact par fête spécifique**

L'analyse différenciée par fête révèle des profils d'impact très contrastés :

- **Thanksgiving** domine nettement avec le lift le plus élevé. Cet événement, suivi du Black Friday, constitue le pic commercial majeur de l'année pour Walmart.
- **Super Bowl** génère également un effet positif notable, lié aux achats alimentaires et de divertissement pour les soirées de visionnage.
- **Christmas** présente un profil paradoxal : bien que décembre soit un mois de forte consommation, la semaine de Noël elle-même (semaine 52) peut afficher des ventes plus faibles car les achats sont déjà réalisés en amont.
- **Labour Day** a un impact plus modeste, cohérent avec un événement de moindre envergure commerciale.

> 💡 **Point clé** : Un indicateur binaire `IsHoliday` unique est insuffisant pour capturer la diversité des effets. L'ingénierie de features devra créer des **indicateurs spécifiques par fête** (one-hot encoding) et potentiellement des **variables de proximité** (semaines avant/après la fête).

<strong>Effet temporel élargi (semaines avant/après fête)</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  3) EFFET TEMPOREL ÉLARGI — SEMAINES AVANT / APRÈS FÊTE
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  3) EFFET TEMPOREL ÉLARGI — SEMAINES AVANT / APRÈS FÊTE")
print("=" * 70)

# Ventes moyennes par semaine (référence)
weekly_avg = (
    df_train.groupby('Date', observed=True)
    .agg(Ventes_Moy=('Weekly_Sales', 'mean'), Holiday=('IsHoliday', 'first'))
    .reset_index().sort_values('Date')
)
weekly_avg['WeekOfYear'] = weekly_avg['Date'].dt.isocalendar().week.astype(int)

# Identifier les semaines de fête
fete_weeks = weekly_avg[weekly_avg['Holiday']]['Date'].unique()

# Pour chaque fête, calculer les ventes sur une fenêtre [-3, +2] semaines
window_data = []
for fete_date in fete_weeks:
    fete_date = pd.Timestamp(fete_date)
    for delta in range(-3, 3):
        target_date = fete_date + pd.Timedelta(weeks=delta)
        row = weekly_avg[weekly_avg['Date'] == target_date]
        if not row.empty:
            window_data.append({
                'Fete_Date': fete_date,
                'Delta_Semaine': delta,
                'Ventes_Moy': row['Ventes_Moy'].values[0],
                'Fete': classify_holiday(fete_date)
            })

window_df = pd.DataFrame(window_data)

# Moyenne par delta et par fête
window_agg = (
    window_df.groupby(['Fete', 'Delta_Semaine'], observed=True)['Ventes_Moy']
    .mean().reset_index()
)

print(f"\n► Fenêtre d'analyse : 3 semaines avant → 2 semaines après chaque fête")
print(f"► Nombre de dates de fêtes analysées : {len(fete_weeks)}")

# Graphique : profil temporel par fête
fig_window = px.line(
    window_agg,
    x='Delta_Semaine', y='Ventes_Moy',
    color='Fete',
    color_discrete_map=colors_fetes,
    markers=True,
    title='Profil temporel des ventes autour des fêtes (fenêtre ±3 semaines)',
    labels={'Delta_Semaine': 'Semaines relatives à la fête (0 = semaine fériée)',
            'Ventes_Moy': 'Ventes moyennes ($)', 'Fete': 'Événement'}
)
fig_window.add_vline(x=0, line_dash='dash', line_color='rgba(255,200,50,0.5)', line_width=2,
                     annotation_text='Semaine fériée', annotation_position='top right')
fig_window.update_layout(
    height=500, width=1000, template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_window.show()

# Tableau : ventes moyennes par delta (toutes fêtes confondues)
global_window = (
    window_df.groupby('Delta_Semaine')['Ventes_Moy']
    .agg(['mean', 'std']).reset_index()
    .rename(columns={'mean': 'Ventes_Moy', 'std': 'Écart-type'})
)
global_window['Label'] = global_window['Delta_Semaine'].apply(
    lambda d: f"{'Semaine fériée' if d == 0 else f'{d:+d} sem.'}"
)
print("\n► Ventes moyennes par position relative (toutes fêtes) :")
display(
    global_window[['Label', 'Ventes_Moy', 'Écart-type']].style
    .set_caption('Profil temporel moyen autour des fêtes')
    .format({'Ventes_Moy': '{:,.0f} $', 'Écart-type': '{:,.0f}'})
    .background_gradient(subset=['Ventes_Moy'], cmap='YlOrRd')
)
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Effet temporel élargi (semaines avant/après fête)**

L'analyse de la dynamique temporelle autour des fêtes apporte des enseignements pratiques :

- **Effet d'anticipation** : pour certaines fêtes (notamment Thanksgiving et Noël), les ventes commencent à augmenter **1 à 2 semaines avant** l'événement, reflétant les achats préparatoires des consommateurs.
- **Pic à la semaine fériée** : la semaine fériée elle-même (delta = 0) concentre généralement le pic de ventes, particulièrement pour Thanksgiving.
- **Effet de retombée** : la semaine suivant la fête (delta = +1) montre souvent une **chute brutale** des ventes, caractéristique d'un effet de report de consommation.
- **Profils différenciés** : chaque fête a un profil temporel spécifique — Thanksgiving a un pic aigu, tandis que Noël a un profil plus étalé sur les semaines précédentes.

> 💡 **Point clé** : Pour la modélisation, créer des features de **distance temporelle à la prochaine fête** (nombre de semaines avant/après) permettra de capturer cet effet d'anticipation et de retombée, plutôt que de se limiter à un simple indicateur binaire.

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Sensibilité par type de magasin</strong>
</div>

```python
# ══════════════════════════════════════════════════════════════════════════
#  4) SENSIBILITÉ PAR TYPE DE MAGASIN
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  4) SENSIBILITÉ PAR TYPE DE MAGASIN")
print("=" * 70)

# Interaction Type × IsHoliday × Fête
holiday_rows_type = holiday_rows.merge(
    df_train[['Store', 'Type']].drop_duplicates(), on='Store', how='left',
    suffixes=('', '_dup')
)
# Si 'Type' est déjà dans holiday_rows, utiliser directement
if 'Type_dup' in holiday_rows_type.columns:
    holiday_rows_type.drop(columns=['Type_dup'], inplace=True)

# Lift par type de magasin (toutes fêtes)
type_holiday = (
    df_train.groupby(['Type', 'IsHoliday'], observed=True)['Weekly_Sales']
    .mean().unstack()
)
type_holiday.columns = ['Semaine normale', 'Semaine fériée']
type_holiday['Lift (%)'] = ((type_holiday['Semaine fériée'] / type_holiday['Semaine normale']) - 1) * 100

print("\n► Lift férié par type de magasin :")
display(
    type_holiday.style
    .set_caption('Ventes moyennes et lift — Par type de magasin')
    .format({'Semaine normale': '{:,.0f} $', 'Semaine fériée': '{:,.0f} $', 'Lift (%)': '{:+.1f}'})
    .background_gradient(subset=['Lift (%)'], cmap='RdYlGn')
)

# Interaction Type × Fête
type_fete = holiday_rows.groupby(['Type', 'Fete'], observed=True)['Weekly_Sales'].mean().reset_index()
normal_by_type = df_train[~df_train['IsHoliday']].groupby('Type', observed=True)['Weekly_Sales'].mean()
type_fete = type_fete.merge(normal_by_type.rename('Moy_Normale').reset_index(), on='Type')
type_fete['Lift (%)'] = ((type_fete['Weekly_Sales'] / type_fete['Moy_Normale']) - 1) * 100

print("\n► Lift par type de magasin et par fête :")
pivot_type_fete = type_fete.pivot_table(index='Type', columns='Fete', values='Lift (%)')
display(
    pivot_type_fete.style
    .set_caption('Lift (%) par type de magasin et par fête')
    .format('{:+.1f}')
    .background_gradient(cmap='RdYlGn', axis=None)
)

# Graphique : Grouped bar — Lift par Type × Fête
fig_type_fete = px.bar(
    type_fete,
    x='Fete', y='Lift (%)',
    color='Type',
    barmode='group',
    color_discrete_map={'A': '#2E86C1', 'B': '#E67E22', 'C': '#27AE60'},
    title='Lift des ventes par fête et par type de magasin',
    labels={'Lift (%)': 'Lift vs normal (%)', 'Fete': 'Événement'},
    text_auto='.1f',
    category_orders={'Fete': ['Thanksgiving', 'Super Bowl', 'Christmas', 'Labour Day']}
)
fig_type_fete.update_layout(
    height=500, width=900, template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_type_fete.show()
```

L'interaction entre le type de magasin et les périodes de fêtes révèle des dynamiques différenciées :

- **Réactivité variable** : les trois types de magasins (A, B, C) ne réagissent pas de la même manière aux fêtes. Le lift diffère significativement selon le format du magasin.
- **Effet Thanksgiving dominant** : quelle que soit la catégorie de magasin, Thanksgiving génère le lift le plus élevé, ce qui confirme son statut d'événement commercial incontournable.
- **Type × Fête** : certaines combinaisons (type de magasin × fête) produisent des effets amplifiés ou atténués, suggérant des différences de positionnement commercial et de clientèle.

> 💡 **Point clé** : L'interaction **Type × IsHoliday** et plus finement **Type × Fête** constitue une feature potentiellement puissante pour la modélisation. Un modèle qui ignore cette interaction risque de sous-estimer les ventes des magasins les plus réactifs aux fêtes.

### <span id="153" style="color:#2E86C1;">1.5.3 Analyse des variables économiques et saisonnières</span>

Les sections précédentes ont mis en évidence l'impact des promotions (1.5.1) et des périodes de fêtes (1.5.2) sur les ventes hebdomadaires. Au-delà de ces facteurs internes et événementiels, les ventes en grande distribution sont également influencées par des **variables contextuelles externes** — économiques et climatiques — que le dataset Walmart capture à travers quatre indicateurs : la **température**, le **prix du carburant**, l'**indice des prix à la consommation (CPI)** et le **taux de chômage**.

L'objectif de cette sous-section est d'évaluer l'**influence de chaque variable économique et saisonnière sur les ventes**, en répondant à plusieurs questions clés :

- La **température** influence-t-elle les ventes de manière linéaire ou non linéaire (seuils, saisons) ?
- Le **prix du carburant** affecte-t-il la fréquentation et les montants d'achat ?
- Le **CPI** reflète-t-il un comportement d'achat différencié selon le pouvoir d'achat ?
- Le **taux de chômage** est-il associé à une baisse de consommation, et cet effet varie-t-il selon le type de magasin ?

L'analyse se déroule en **six étapes** :

1. **Vue d'ensemble des variables économiques** — statistiques descriptives, distributions et corrélations initiales.
2. **Température et ventes** — profil saisonnier, relation non linéaire et effet par type de magasin.
3. **Prix du carburant et ventes** — corrélation, évolution temporelle et impact régional.
4. **CPI et ventes** — segmentation par niveau de CPI et effet différencié par type de magasin.
5. **Chômage et ventes** — relation inversée attendue, effet par zone géographique et type.
6. **Interactions entre variables économiques et synthèse** — corrélations croisées et bilan global.

<strong>Vue d'ensemble des variables économiques</strong>

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

# ══════════════════════════════════════════════════════════════════════════
#  1) VUE D'ENSEMBLE DES VARIABLES ÉCONOMIQUES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) VUE D'ENSEMBLE DES VARIABLES ÉCONOMIQUES")
print("=" * 70)

eco_cols = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

# Statistiques descriptives
eco_stats = pd.DataFrame({
    'Variable': eco_cols,
    'Non-null': [df_train[c].notna().sum() for c in eco_cols],
    'Manquants': [df_train[c].isna().sum() for c in eco_cols],
    'Moyenne': [df_train[c].mean() for c in eco_cols],
    'Médiane': [df_train[c].median() for c in eco_cols],
    'Écart-type': [df_train[c].std() for c in eco_cols],
    'Min': [df_train[c].min() for c in eco_cols],
    'Max': [df_train[c].max() for c in eco_cols],
    'CV (%)': [(df_train[c].std() / df_train[c].mean() * 100) for c in eco_cols],
})

print("\n► Statistiques descriptives des variables économiques :")
display(eco_stats.style.format({
    'Moyenne': '{:,.2f}', 'Médiane': '{:,.2f}', 'Écart-type': '{:,.2f}',
    'Min': '{:,.2f}', 'Max': '{:,.2f}', 'CV (%)': '{:.1f}',
    'Non-null': '{:,}', 'Manquants': '{:,}'
}).set_caption("Statistiques descriptives des 4 variables économiques"))

# Corrélations initiales avec Weekly_Sales
print("\n► Corrélations linéaires (Pearson) avec Weekly_Sales :")
for col in eco_cols:
    corr = df_train[col].corr(df_train['Weekly_Sales'])
    print(f"  • {col:15s} ↔ Weekly_Sales : r = {corr:+.4f}")

# Graphique : distributions (histogrammes)
fig_eco_dist = make_subplots(rows=2, cols=2,
    subplot_titles=['Température (°F)', 'Prix du carburant ($)', 'CPI', 'Taux de chômage (%)'])

colors_eco = ['#2E86C1', '#E67E22', '#27AE60', '#E74C3C']
for i, (col, color) in enumerate(zip(eco_cols, colors_eco)):
    row, col_idx = divmod(i, 2)
    fig_eco_dist.add_trace(
        go.Histogram(x=df_train[col], nbinsx=50, marker_color=color, opacity=0.7, name=col),
        row=row+1, col=col_idx+1
    )

fig_eco_dist.update_layout(
    title='Distribution des variables économiques et saisonnières',
    height=500, width=1000, template='plotly_white', showlegend=False
)
fig_eco_dist.show()
```

**📝 Commentaire – Vue d'ensemble des variables économiques**

L'examen des quatre variables économiques révèle des caractéristiques distinctes :

- **Température** : forte variabilité (large amplitude entre minimum et maximum), reflétant la diversité climatique des 45 magasins Walmart répartis dans différentes régions des États-Unis. La distribution est relativement symétrique.
- **Fuel_Price** : amplitude modérée, avec une tendance haussière probable au fil du temps. Le coefficient de variation est faible, indiquant une relative homogénéité.
- **CPI** : distribution bimodale probable, reflétant des zones géographiques à niveaux de prix très différents. La dispersion est significative.
- **Unemployment** : distribution étalée vers la droite, avec des zones à chômage élevé (effet de la crise 2008-2010 encore visible dans la période d'étude).
- **Corrélations faibles** : les corrélations linéaires brutes avec les ventes sont généralement faibles, ce qui ne signifie pas absence de relation — des effets non linéaires ou conditionnels peuvent exister.

> 💡 **Point clé** : Les corrélations linéaires seules ne suffisent pas à capturer les relations entre variables économiques et ventes. Une analyse par segments (bins, types de magasin) sera nécessaire.

```python
# ══════════════════════════════════════════════════════════════════════════
#  2) TEMPÉRATURE ET VENTES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  2) TEMPÉRATURE ET VENTES")
print("=" * 70)

# Discrétisation de la température en bins
df_train['Temp_Bin'] = pd.cut(df_train['Temperature'], bins=10, precision=0)
temp_sales = (
    df_train.groupby('Temp_Bin', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Médiane', 'count': 'Nb_obs'})
)

print("\n► Ventes moyennes par tranche de température :")
display(
    temp_sales.style
    .set_caption('Relation température → ventes (intervalles)')
    .format({'Moyenne': '{:,.0f} $', 'Médiane': '{:,.0f} $', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='coolwarm')
)

# Graphique : ventes moyennes par bin de température
fig_temp = go.Figure()
fig_temp.add_trace(go.Bar(
    x=temp_sales['Temp_Bin'].astype(str), y=temp_sales['Moyenne'],
    marker_color=px.colors.sample_colorscale('RdYlBu_r',
        [i / (len(temp_sales)-1) for i in range(len(temp_sales))]),
    text=[f"{v:,.0f} $" for v in temp_sales['Moyenne']],
    textposition='outside'
))
fig_temp.update_layout(
    title='Ventes moyennes par tranche de température',
    xaxis_title='Température (°F)', yaxis_title='Ventes moyennes ($)',
    height=450, width=1000, template='plotly_white'
)
fig_temp.show()

# Relation par type de magasin
temp_type = (
    df_train.groupby(['Temp_Bin', 'Type'], observed=True)['Weekly_Sales']
    .mean()
    .reset_index()
)

temp_type['Temp_Bin'] = temp_type['Temp_Bin'].astype(str)

fig_temp_type = px.line(
    temp_type,
    x='Temp_Bin',
    y='Weekly_Sales',
    color='Type',
    color_discrete_map={'A': '#2E86C1', 'B': '#E67E22', 'C': '#27AE60'},
    markers=True,
    title='Relation température → ventes par type de magasin',
    labels={'Weekly_Sales': 'Ventes moyennes ($)', 'Temp_Bin': 'Température (°F)'}
)

fig_temp_type.update_layout(
    height=450,
    width=1000,
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)

fig_temp_type.show()

# Corrélation par mois
temp_corr_month = (
    df_train.groupby('Month', observed=True)
    .apply(lambda g: g['Temperature'].corr(g['Weekly_Sales']), include_groups=False)
    .reset_index()
    .rename(columns={0: 'Corrélation'})
)
print("\n► Corrélation température ↔ ventes par mois :")
for _, row in temp_corr_month.iterrows():
    print(f"  • Mois {int(row['Month']):2d} : r = {row['Corrélation']:+.3f}")
```

**📝 Commentaire – Température et ventes**

L'analyse de la relation entre température et ventes met en évidence un profil complexe :

- **Relation non linéaire** : les ventes ne suivent pas une tendance monotone avec la température. On observe souvent un profil en U inversé ou un plateau, avec des ventes plus élevées aux températures modérées.
- **Effet saisonnier confondant** : la température est fortement corrélée au mois de l'année, ce qui signifie qu'une partie de la relation observée reflète la saisonnalité des ventes (pic de fin d'année = températures basses) plutôt qu'un effet causal de la température.
- **Différence par type de magasin** : les trois types de magasins réagissent différemment aux variations de température, ce qui reflète probablement des différences géographiques (localisation dans des zones climatiques distinctes).
- **Corrélation mensuelle variable** : la corrélation température ↔ ventes change de signe selon le mois, confirmant l'absence de relation linéaire simple.

> 💡 **Point clé** : La température est un proxy de la saisonnalité plus qu'un facteur causal direct. Pour la modélisation, elle reste utile comme feature contextuelle, mais les features temporelles (mois, semaine) captureront mieux l'effet saisonnier.

```python
# ══════════════════════════════════════════════════════════════════════════
#  3) PRIX DU CARBURANT ET VENTES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  3) PRIX DU CARBURANT ET VENTES")
print("=" * 70)

# Discrétisation du prix du carburant
df_train['Fuel_Bin'] = pd.cut(df_train['Fuel_Price'], bins=8, precision=2)
fuel_sales = (
    df_train.groupby('Fuel_Bin', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Médiane', 'count': 'Nb_obs'})
)

print("\n► Ventes moyennes par tranche de prix du carburant :")
display(
    fuel_sales.style
    .set_caption('Relation prix du carburant → ventes')
    .format({'Moyenne': '{:,.0f} $', 'Médiane': '{:,.0f} $', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='YlOrRd')
)

# Évolution temporelle croisée : Fuel_Price vs Ventes moyennes
weekly_eco = (
    df_train.groupby('Date', observed=True)
    .agg(Ventes_Moy=('Weekly_Sales', 'mean'),
         Fuel_Moy=('Fuel_Price', 'mean'))
    .reset_index().sort_values('Date')
)

fig_fuel = make_subplots(specs=[[{"secondary_y": True}]])
fig_fuel.add_trace(
    go.Scatter(x=weekly_eco['Date'], y=weekly_eco['Ventes_Moy'],
               mode='lines', name='Ventes moyennes ($)',
               line=dict(color='#2E86C1', width=1.5)),
    secondary_y=False
)
fig_fuel.add_trace(
    go.Scatter(x=weekly_eco['Date'], y=weekly_eco['Fuel_Moy'],
               mode='lines', name='Prix carburant ($)',
               line=dict(color='#E74C3C', width=2, dash='dot')),
    secondary_y=True
)
fig_fuel.update_layout(
    title='Évolution parallèle : ventes et prix du carburant',
    height=450, width=1000, template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.04, xanchor='right', x=1)
)
fig_fuel.update_yaxes(title_text='Ventes moyennes ($)', secondary_y=False)
fig_fuel.update_yaxes(title_text='Prix carburant ($)', secondary_y=True)
fig_fuel.show()

# Corrélation par type de magasin
print("\n► Corrélation Fuel_Price ↔ Weekly_Sales par type de magasin :")
for t in sorted(df_train['Type'].unique()):
    subset = df_train[df_train['Type'] == t]
    corr = subset['Fuel_Price'].corr(subset['Weekly_Sales'])
    print(f"  • Type {t} : r = {corr:+.4f}")
```

- **Variable de contexte** : le prix du carburant est davantage un indicateur macroéconomique qu'un facteur direct de variation des ventes.

> 💡 **Point clé** : Le prix du carburant a un pouvoir prédictif limité en soi, mais peut contribuer comme variable de contexte économique dans un modèle multivarié, notamment en interaction avec la localisation géographique.

<strong>CPI et ventes</strong>

```python
# ══════════════════════════════════════════════════════════════════════════
#  4) CPI (INDICE DES PRIX À LA CONSOMMATION) ET VENTES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  4) CPI ET VENTES")
print("=" * 70)

# Discrétisation du CPI en quintiles
df_train['CPI_Quintile'] = pd.qcut(df_train['CPI'], q=5, precision=1, duplicates='drop')
cpi_sales = (
    df_train.groupby('CPI_Quintile', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'std', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Médiane', 'std': 'Écart-type', 'count': 'Nb_obs'})
)

print("\n► Ventes moyennes par quintile de CPI :")
display(
    cpi_sales.style
    .set_caption('Relation CPI → ventes (quintiles)')
    .format({'Moyenne': '{:,.0f} $', 'Médiane': '{:,.0f} $', 'Écart-type': '{:,.0f}', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='Purples')
)

# Scatter : CPI vs Ventes moyennes par magasin
store_eco = (
    df_train.groupby('Store', observed=True)
    .agg(Ventes_Moy=('Weekly_Sales', 'mean'),
         CPI_Moy=('CPI', 'mean'),
         Unemp_Moy=('Unemployment', 'mean'),
         Type=('Type', 'first'))
    .reset_index()
)

fig_cpi = px.scatter(
    store_eco,
    x='CPI_Moy', y='Ventes_Moy',
    color='Type',
    color_discrete_map={'A': '#2E86C1', 'B': '#E67E22', 'C': '#27AE60'},
    size='Ventes_Moy', size_max=18,
    hover_data=['Store'],
    title='CPI moyen vs Ventes moyennes par magasin (coloré par type)',
    labels={'CPI_Moy': 'CPI moyen', 'Ventes_Moy': 'Ventes moyennes ($)'},
    trendline='ols'
)
fig_cpi.update_layout(
    height=500, width=900, template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)
fig_cpi.show()

# CPI par type de magasin
cpi_type = (
    df_train.groupby(['CPI_Quintile', 'Type'], observed=True)['Weekly_Sales']
    .mean()
    .reset_index()
)

cpi_type['CPI_Quintile'] = cpi_type['CPI_Quintile'].astype(str)

fig_cpi_type = px.bar(
    cpi_type,
    x='CPI_Quintile',
    y='Weekly_Sales',
    color='Type',
    barmode='group',
    color_discrete_map={'A': '#2E86C1', 'B': '#E67E22', 'C': '#27AE60'},
    title='Ventes moyennes par quintile de CPI et type de magasin',
    labels={'Weekly_Sales': 'Ventes moyennes ($)', 'CPI_Quintile': 'Quintile CPI'},
    text_auto='.0f'
)

fig_cpi_type.update_layout(
    height=450,
    width=1000,
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)

fig_cpi_type.show()
```

**📝 Commentaire – CPI et ventes**

L'analyse de la relation entre l'indice des prix à la consommation et les ventes apporte des nuances importantes :

- **Relation non monotone** : les ventes ne suivent pas une tendance linéaire simple avec le CPI. Certains quintiles de CPI affichent des ventes moyennes supérieures, sans que ce soit systématiquement les CPI les plus bas ou les plus élevés.
- **Effet de localisation** : le scatter par magasin montre que le CPI est fortement lié à la localisation géographique du magasin (chaque magasin a un CPI propre à sa zone). Les magasins de type A (grands) se concentrent dans certaines zones de CPI.
- **Interaction Type × CPI** : l'effet du CPI sur les ventes dépend du type de magasin, ce qui suggère que le CPI capture en partie un effet de zone géographique (pouvoir d'achat local) plutôt qu'un simple effet prix.
- **Confondeur géographique** : le CPI étant spécifique à chaque zone géographique, il agit comme un proxy de la localisation — une information potentiellement redondante avec l'identifiant du magasin.

> 💡 **Point clé** : Le CPI est utile comme feature de contexte géographique et économique. En modélisation, il peut capter l'effet du pouvoir d'achat local, mais attention à la multicolinéarité avec les identifiants de magasin.

```python
# ══════════════════════════════════════════════════════════════════════════
#  5) TAUX DE CHÔMAGE ET VENTES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  5) TAUX DE CHÔMAGE ET VENTES")
print("=" * 70)

# Discrétisation du chômage en quintiles
df_train['Unemp_Quintile'] = pd.qcut(df_train['Unemployment'], q=5, precision=1, duplicates='drop')
unemp_sales = (
    df_train.groupby('Unemp_Quintile', observed=True)['Weekly_Sales']
    .agg(['mean', 'median', 'count'])
    .reset_index()
    .rename(columns={'mean': 'Moyenne', 'median': 'Médiane', 'count': 'Nb_obs'})
)

print("\n► Ventes moyennes par quintile de chômage :")
display(
    unemp_sales.style
    .set_caption('Relation chômage → ventes (quintiles)')
    .format({'Moyenne': '{:,.0f} $', 'Médiane': '{:,.0f} $', 'Nb_obs': '{:,}'})
    .background_gradient(subset=['Moyenne'], cmap='RdYlGn_r')
)



# Évolution temporelle croisée : Chômage vs Ventes
weekly_unemp = (
    df_train.groupby('Date', observed=True)
    .agg(Ventes_Moy=('Weekly_Sales', 'mean'),
         Unemp_Moy=('Unemployment', 'mean'))
    .reset_index().sort_values('Date')
)

fig_unemp_ts = make_subplots(specs=[[{"secondary_y": True}]])
fig_unemp_ts.add_trace(
    go.Scatter(x=weekly_unemp['Date'], y=weekly_unemp['Ventes_Moy'],
               mode='lines', name='Ventes moyennes ($)',
               line=dict(color='#2E86C1', width=1.5)),
    secondary_y=False
)
fig_unemp_ts.add_trace(
    go.Scatter(x=weekly_unemp['Date'], y=weekly_unemp['Unemp_Moy'],
               mode='lines', name='Taux de chômage (%)',
               line=dict(color='#E74C3C', width=2, dash='dot')),
    secondary_y=True
)
fig_unemp_ts.update_layout(
    title='Évolution parallèle : ventes et taux de chômage',
    height=450, width=1000, template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.04, xanchor='right', x=1)
)
fig_unemp_ts.update_yaxes(title_text='Ventes moyennes ($)', secondary_y=False)
fig_unemp_ts.update_yaxes(title_text='Chômage (%)', secondary_y=True)
fig_unemp_ts.show()

# Corrélation par type
print("\n► Corrélation Unemployment ↔ Weekly_Sales par type :")
for t in sorted(df_train['Type'].unique()):
    subset = df_train[df_train['Type'] == t]
    corr = subset['Unemployment'].corr(subset['Weekly_Sales'])
    print(f"  • Type {t} : r = {corr:+.4f}")
```

> 💡 **Point clé** : Le taux de chômage est un indicateur structurel qui distingue les zones géographiques plutôt qu'un facteur de variation temporelle des ventes. Il sera utile comme feature contextuelle, en interaction avec le type de magasin.

### <span id="154" style="color:#2E86C1;">1.5.4 Étude des corrélations entre variables et ventes</span>

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

Les sections précédentes ont exploré séparément l'influence des promotions (MarkDowns), des jours fériés et des variables économiques/saisonnières sur les ventes hebdomadaires (`Weekly_Sales`). Cette section adopte une **vision transversale** : elle mesure et compare simultanément la force de liaison entre **toutes les variables disponibles** et la cible.

L'objectif est triple :
1. **Classer** les variables par force de corrélation avec `Weekly_Sales` (Pearson et Spearman).
2. **Comparer** les profils de corrélation selon le **type de magasin** (A, B, C) et les **départements**.
3. **Détecter** d'éventuelles relations **non linéaires** (écarts Pearson – Spearman) et identifier les variables les plus prometteuses pour la modélisation.

> Les résultats guideront directement le choix des features et des transformations pour les modèles prédictifs de la Phase 2.

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Panorama des corrélations globales</strong>
</div>

```python
# ══════════════════════════════════════════════════════════════════════════
#  1) PANORAMA DES CORRÉLATIONS GLOBALES
# ══════════════════════════════════════════════════════════════════════════
print("=" * 70)
print("  1) PANORAMA DES CORRÉLATIONS GLOBALES")
print("=" * 70)

# Identifier toutes les colonnes numériques (exclure identifiants)
num_cols_corr = [c for c in df_train.select_dtypes(include='number').columns
                 if c not in ('Store', 'Dept', 'Weekly_Sales')]

# Corrélation de Pearson avec Weekly_Sales
pearson_all = (df_train[num_cols_corr + ['Weekly_Sales']]
               .corr()['Weekly_Sales']
               .drop('Weekly_Sales')
               .sort_values(key=abs, ascending=False))

print(f"\n► {len(pearson_all)} variables numériques analysées")
print("► Corrélations de Pearson avec Weekly_Sales :\n")

# Tableau stylisé
pearson_df = pearson_all.reset_index()
pearson_df.columns = ['Variable', 'Pearson_r']
pearson_df['|r|'] = pearson_df['Pearson_r'].abs()
pearson_df['Force'] = pd.cut(
    pearson_df['|r|'],
    bins=[0, 0.1, 0.3, 0.5, 0.7, 1.0],
    labels=['Négligeable', 'Faible', 'Modérée', 'Forte', 'Très forte']
)

display(
    pearson_df.style
    .set_caption('Corrélations de Pearson — toutes variables vs Weekly_Sales')
    .format({'Pearson_r': '{:+.4f}', '|r|': '{:.4f}'})
    .background_gradient(subset=['Pearson_r'], cmap='RdBu_r', vmin=-0.5, vmax=0.5)
    .background_gradient(subset=['|r|'], cmap='YlOrRd', vmin=0, vmax=0.5)
)

# Bar chart horizontal trié par |r|
colors_corr = ['#e74c3c' if v < 0 else '#2ecc71' for v in pearson_all.values]

fig_rank = go.Figure(go.Bar(
    x=pearson_all.values,
    y=pearson_all.index,
    orientation='h',
    marker_color=colors_corr,
    text=[f'{v:+.4f}' for v in pearson_all.values],
    textposition='outside'
))
fig_rank.update_layout(
    title='Classement des variables par corrélation de Pearson avec Weekly_Sales',
    xaxis_title='Coefficient de Pearson (r)',
    yaxis=dict(autorange='reversed'),
    height=max(400, len(pearson_all) * 30),
    template='plotly_white',
    margin=dict(l=140)
)
fig_rank.show()
```

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:12px 16px; border-radius:6px; margin-top:12px; color:#cdd6f4;">

**📝 Commentaire – Panorama des corrélations globales**

Le classement des variables par corrélation de Pearson avec `Weekly_Sales` révèle la hiérarchie suivante :

- **Size** domine clairement : la taille du magasin est le meilleur prédicteur linéaire des ventes, confirmant que le volume de ventes est structurellement lié à la surface de vente.
- **Variables économiques** (CPI, Unemployment, Temperature, Fuel_Price) : corrélations généralement **faibles** (|r| < 0.2), cohérent avec les conclusions de la section 1.5.3.
- **MarkDowns** : corrélations variables — certains MarkDowns capturent un effet promotionnel modéré mais avec beaucoup de valeurs manquantes.
- **IsHoliday** : corrélation très faible, ce qui ne contredit pas l'effet mesuré en 1.5.2 — l'impact des fêtes est concentré et asymétrique.
- La colonne **Force** classe chaque variable : Négligeable (|r| < 0.1), Faible (0.1–0.3), Modérée (0.3–0.5), Forte (0.5–0.7), Très forte (> 0.7).

> Ce classement Pearson mesure uniquement les relations **linéaires**. Les étapes suivantes exploreront les différences par segment et les non-linéarités.

<div style="background:#2a2a3e; border-left:4px solid #f9e2af; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0; font-size:25px;">
<strong>🎄 4. Impact des périodes de fêtes</strong>
</div>

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

**Résultats du test statistique (test de Welch) :**

> H₀ : les ventes moyennes des semaines fériées = semaines normales

- **Résultat** : H₀ rejetée avec une p-value extrêmement faible → différence **statistiquement significative** ✅
- **Lift global** : les semaines fériées affichent des ventes moyennes supérieures aux semaines normales

**Hiérarchie des fêtes par impact :**

| Fête | Impact |
|---|---|
| **Thanksgiving** (semaine 47) | Le plus fort — lift pouvant doubler les ventes moyennes |
| **Christmas** | Pic de décembre soutenu |
| **Labor Day** | Effet modéré |
| **Super Bowl** | Effet limité |

**Nuances importantes :**
- L'effet des fêtes **varie fortement selon le département** (saisonniers vs stables)
- L'effet est plus marqué dans les **magasins de type A** (plus fort trafic)
- Quelques départements présentent un **lift négatif** (report ou substitution d'achats)

> 💡 **Conclusion** : `IsHoliday` est une feature utile mais insuffisante seule — des **indicateurs spécifiques par fête** (one-hot encoding : `Is_Thanksgiving`, `Is_Christmas`, `Is_SuperBowl`, `Is_LaborDay`) et une variable de **distance temporelle à la prochaine fête** enrichiront significativement les modèles.

</div>

<div style="background:#2a2a3e; border-left:4px solid #fab387; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0; font-size:25px;">
<strong>🏷️ 5. Effet des promotions (MarkDowns)</strong>
</div>


<div style="background:#1e1e2e; border-left:4px solid #fab387; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

**Caractéristiques des données promotionnelles :**
- Disponibles uniquement à partir de **novembre 2011** (imputation à 0 pour les périodes antérieures)
- **~30–50 % des observations** présentent au moins une promotion active (`HasPromo = True`)

**Résultats :**
- **Lift promotionnel global positif** (ventes moyennes supérieures avec au moins un MarkDown actif)
- Le lift est confirmé sur la **médiane** (effet de masse, pas uniquement dû à des valeurs extrêmes)
- Chaque type de MarkDown (1 à 5) présente un **impact différencié** — certains génèrent un fort lift, d'autres un effet marginal
- L'impact promotionnel est **fortement hétérogène selon les départements** : certains voient leurs ventes bondir de 50 %+ avec une promotion, d'autres sont peu ou négativement affectés
- Les **magasins de type A** captalisent mieux les promotions que les types B et C

> ⚠️ **Attention** : les MarkDowns étant concentrés en fin d'année (périodes de fêtes), une partie du lift observé peut être confondue avec l'effet saisonnier. Une modélisation rigoureuse devra démêler ces deux effets.

> 💡 **Conclusion** : Conserver les **5 variables MarkDown séparément** (ne pas les agréger) est recommandé, car chacune capture un type de promotion distinct. La variable synthétique `HasPromo` et `Total_MarkDown` restent utiles comme features complémentaires.

</div>

<div style="background:#2a2a3e; border-left:4px solid #f38ba8; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0; font-size:25px;">

<strong>🤖 8. Implications pour la modélisation</strong>
</div>

<div style="background:#1e1e2e; border-left:4px solid #74c7ec; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

### 📌 Features prioritaires recommandées

| Catégorie | Feature(s) | Justification |
|---|---|---|
| **Identité** | `Store`, `Dept`, `Type`, `Size` | Déterminants structurels du volume de base |
| **Temporel** | `Year`, `Month`, `Week`, `Quarter` | Composante saisonnière dominante |
| **Fêtes** | `IsHoliday`, `Is_Thanksgiving`, `Is_Christmas`, `Is_LaborDay`, `Is_SuperBowl` | Pics ponctuels statistiquement significatifs |
| **Proximité temporelle** | `Weeks_to_next_holiday`, `Weeks_since_last_holiday` | Anticiper et prolonger les effets des fêtes |
| **Promotions** | `MarkDown1`, `MarkDown2`, `MarkDown3`, `MarkDown4`, `MarkDown5`, `HasPromo`, `Total_MarkDown` | Lift variable mais réel par département |
| **Économique** | `CPI`, `Unemployment`, `Temperature`, `Fuel_Price` | Contexte géographique et saisonnier résiduel |
| **Lag / historique** | `Sales_lag_1`, `Sales_lag_4`, `Sales_lag_52` | Persistance temporelle des ventes |

---

### 🔗 Interactions à capturer

- **`Store × Dept`** — chaque combinaison a un comportement propre (3 645 paires) ; l'encodage par `Store` et `Dept` séparément ne suffit pas
- **`Dept × IsHoliday`** — l'effet des fêtes n'est pas uniforme : certains rayons profitent massivement des fêtes, d'autres pas
- **`Type × MarkDown`** — le lift des promotions varie selon la taille et le type de magasin
- **`Month × Type`** — la saisonnalité est proportionnelle à la taille, pas identique

---

### 🏆 Algorithmes recommandés

| Algorithme | Intérêt | Point d'attention |
|---|---|---|
| **Random Forest** | Robuste aux outliers, capte les non-linéarités, peu d'hyperparamètres critiques | Peut sur-lisser les pics de Thanksgiving |
| **XGBoost / LightGBM** | Meilleure performance sur données tabulaires hétérogènes, gère les features manquantes | Risque d'overfitting sans régularisation |
| **Gradient Boosting (sklearn)** | Bonne baseline solide | Plus lent que LightGBM sur grands datasets |
| **Modèles par segment** | Un modèle entraîné par type (`A`, `B`, `C`) ou par département à fort CA | Complexité de déploiement accrue |
| **Modèle baseline (médiane historique)** | Référence simple pour mesurer le gain réel des modèles ML | À calculer en premier (benchmark) |

---

### ⚠️ Points de vigilance

- **Risque de data leakage** : les MarkDowns futurs ne sont pas connus à l'avance → ne pas inclure des features calculées sur des données post-date de prédiction
- **Découpage temporel strict** : le split train/validation doit respecter l'ordre chronologique (pas de split aléatoire standard) — utiliser les semaines antérieures à une date pivot
- **Gestion des MarkDowns manquants** : les valeurs imputées à 0 avant novembre 2011 peuvent biaiser l'apprentissage si non signalées explicitement — envisager une variable booléenne `MarkDown_available`
- **Asymétrie de la distribution** : la métrique d'évaluation recommandée est le **WMAE** (Weighted Mean Absolute Error, avec poids 5× pour les semaines fériées) — conforme au schéma de notation Kaggle Walmart M5
- **Hétérogénéité Store × Dept** : un modèle global risque d'être dominé par les grands magasins de type A — envisager une normalisation par magasin ou une stratification

> ✅ **Bilan final** : Les analyses de la Phase 1 fournissent une base solide et documentée pour construire un pipeline ML efficace. Les features identifiées, les interactions à modéliser et les précautions mentionnées permettront d'atteindre l'objectif de précision fixé (WMAE minimal) en Phase 2.

<div style="background:#1e1e2e; border-left:4px solid #e8a838; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

Cette section traduit ces enseignements analytiques en **recommandations concrètes et opérationnelles** à destination des décideurs : directeurs de magasin, responsables marketing, équipes logistiques et direction régionale.

<div style="background:#2a2a3e; border-left:4px solid #a6e3a1; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0; font-size:25px;">
<strong>🏪 R1 — Gestion des stocks : anticiper par type de magasin et département</strong>
</div>

<div style="background:#1e1e2e; border-left:4px solid #a6e3a1; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

**Constat analytique :**
- Les ventes hebdomadaires sont **fortement hétérogènes** entre magasins (CV > 100 %) et entre départements (loi de Pareto : ~20 % des départements = ~80 % du CA)
- Les magasins de **Type A** concentrent la majorité du CA et présentent les plus fortes amplitudes de ventes

**Recommandations :**

| Priorité | Action | Bénéfice attendu |
|---|---|---|
| 🔴 **Haute** | Calibrer les niveaux de stock par **type de magasin** (A, B, C) — ne pas appliquer des règles uniformes | Réduction des surstocks dans les Types B et C |
| 🔴 **Haute** | Concentrer les efforts de planification sur les **top 20 départements** (qui génèrent ~80 % du CA) | Optimisation des ressources logistiques |
| 🟡 **Moyenne** | Mettre en place un **stock de sécurité différencié** selon la variabilité historique de chaque combinaison Store × Département | Réduction des ruptures de stock de l'ordre de 20 % |
| 🟡 **Moyenne** | En cas de **stock limité**, prioriser l'approvisionnement des magasins de **Type A** et des **départements à forte rotation** | Minimisation des pertes de ventes |

> ✅ **Objectif métier** : Réduire les ruptures de stock de 20 % grâce à une planification différenciée par segment, alignée sur les prévisions de ventes hebdomadaires.

</div>

<div style="background:#2a2a3e; border-left:4px solid #f9e2af; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0; font-size:25px;">
<strong>🎄 R2 — Préparation des périodes de fêtes : anticiper les pics avec précision</strong>
</div>

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

**Constat analytique :**
- Les **semaines fériées** génèrent des ventes moyennes significativement supérieures (différence confirmée par test de Welch, p-value < 0,001)
- **Thanksgiving** (semaine 47) est de loin la fête la plus impactante — les ventes peuvent **doubler** la moyenne hebdomadaire
- L'effet des fêtes est **asymétrique selon les départements** : certains voient un lift > 50 %, d'autres restent stables

**Recommandations :**

| Priorité | Action | Bénéfice attendu |
|---|---|---|
| 🔴 **Haute** | Constituer un **stock d'anticipation 3 à 4 semaines avant Thanksgiving** sur les départements à forte saisonnalité (alimentation festive, jouets, électronique, décoration) | Éviter les ruptures critiques en semaine 47 |
| 🔴 **Haute** | Préparer un **calendrier annuel des pics prévisibles** (S47 Thanksgiving, S52-S1 Noël, S6 Super Bowl, S36 Labor Day) et aligner les commandes fournisseurs en conséquence | Sécuriser les approvisionnements aux moments critiques |
| 🟡 **Moyenne** | Identifier les **départements insensibles aux fêtes** et éviter de sur-approvisionner ces rayons pendant les périodes festives | Réduction des surstocks et démarques |
| 🟢 **Standard** | Mettre en place une **alerte automatique** 2 semaines avant chaque fête pour les départements à lift historiquement élevé | Réactivité opérationnelle accrue |

> ✅ **Objectif métier** : Réduire les ruptures de stock pendant les périodes clés de 30 % en ciblant les bons magasins, les bons départements et les bonnes fenêtres temporelles.

<strong>🏷️ R3 — Stratégie promotionnelle : cibler, temporiser et mesurer</strong>
</div>

<div style="background:#1e1e2e; border-left:4px solid #fab387; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

**Constat analytique :**
- Les promotions (MarkDowns) génèrent un **lift positif global** sur les ventes moyennes, confirmé sur la médiane
- L'impact est **fortement hétérogène** : certains départements voient leurs ventes augmenter de > 50 %, d'autres sont peu ou négativement affectés
- Les **magasins de Type A** capitalisent mieux sur les promotions que les Types B et C
- Les MarkDowns sont concentrés sur la **période novembre 2011 – octobre 2012** — leur effet est parfois confondu avec la saisonnalité

**Recommandations :**

| Priorité | Action | Bénéfice attendu |
|---|---|---|
| 🔴 **Haute** | **Cibler les promotions sur les départements à fort lift historique** — éviter les actions promotionnelles sur les rayons naturellement peu réactifs | Amélioration du ROI promotionnel de ~15 % |
| 🔴 **Haute** | **Éviter les promotions pendant les semaines de forte demande naturelle** (Thanksgiving, pic de décembre) — l'investissement promotionnel y a peu de valeur ajoutée | Préservation de la marge, économie de frais promotionnels |
| 🟡 **Moyenne** | **Concentrer les MarkDowns sur les magasins de Type A** pour maximiser le rendement (volume de trafic supérieur) | Maximisation du chiffre d'affaires additionnel |
| 🟡 **Moyenne** | **Mesurer systématiquement le lift net** en comparant les ventes observées à une baseline prévisionnelle — distinguer l'effet promo de l'effet saisonnier | Évaluation objective de l'efficacité promotionnelle |
| 🟢 **Standard** | **Planifier les 5 types de MarkDown différemment** selon leur profil d'impact historique — ne pas les traiter comme équivalents | Optimisation de l'allocations budgétaire promotionnelle |

> ✅ **Objectif métier** : Augmenter le ROI des promotions de 15 % en ciblant les bons départements, les bons magasins et les bonnes fenêtres temporelles — en évitant les promotions inutiles lors des pics de demande naturels.

- Le profil saisonnier est **remarquablement stable d'une année à l'autre** — le cycle annuel est prévisible et reproductible
- Le **Q4 (octobre-décembre)** concentre le CA le plus élevé dans tous les types de magasins
- Certains départements ont des **profils saisonniers très marqués** (jardinage, jouets, électronique) tandis que d'autres sont quasi-stables (produits de première nécessité)

**Recommandations :**

| Priorité | Action | Bénéfice attendu |
|---|---|---|

| 🔴 **Haute** | **Réduire les commandes en janvier** sur les catégories non essentielles — le creux post-fêtes est prévisible et systématique | Réduction du capital immobilisé en stock mort |
| 🟡 **Moyenne** | Établir des **contrats d'approvisionnement flexibles avec les fournisseurs** permettant des ajustements de volume ±20 % selon les prévisions | Réactivité face aux variations de demande |
| 🟡 **Moyenne** | **Différencier les politiques d'achat par type de magasin** : les Types B et C nécessitent des volumes plus modérés avec moins de variabilité | Éviter les surstocks dans les petits magasins |
| 🟢 **Standard** | **Utiliser les prévisions hebdomadaires** (modèles Phase 2) comme base de calcul des ordres de réapprovisionnement sur un horizon de 4 à 12 semaines | Planification proactive des achats |

> ✅ **Objectif métier** : Réduire les coûts de stockage de 10 à 15 % et les ruptures d'approvisionnement en alignant précisément les commandes sur le cycle saisonnier historique.

</div>

<strong>🌍 R5 — Adaptation locale : tenir compte du contexte économique régional</strong>
</div>

<div style="background:#1e1e2e; border-left:4px solid #f38ba8; padding:14px 18px; border-radius:8px; color:#cdd6f4;">

**Constat analytique :**
- Le **CPI** (indice des prix à la consommation) présente une distribution bimodale révélant deux groupes géographiques distincts parmi les 45 magasins — zones à fort et faible coût de la vie
- Le **taux de chômage** varie significativement selon les zones (3,7 % à 14,3 %) et est corrélé négativement avec les ventes — les zones à fort chômage ont des ventes plus faibles
- Ces effets sont **structurels** (lents) et varient peu d'une semaine à l'autre, mais différencient fortement les magasins entre eux

**Recommandations :**

| Priorité | Action | Bénéfice attendu |
|---|---|---|
| 🟡 **Moyenne** | **Segmenter les magasins par profil économique local** (CPI bas / CPI élevé, chômage fort / faible) et adapter les stratégies prix et promotions à chaque segment | Pertinence commerciale accrue par zone |
| 🟡 **Moyenne** | **Dans les zones à fort chômage**, privilégier les promotions sur les produits essentiels et les articles à faible coût unitaire plutôt que sur le haut de gamme | Adaptation au pouvoir d'achat local |
| 🟡 **Moyenne** | **Dans les zones à CPI élevé**, monitorer les arbitrages de consommation pour détecter un éventuel report vers les MDD (marques de distributeur) | Anticipation des changements de mix produit |
| 🟢 **Standard** | Intégrer les indicateurs économiques locaux (CPI, chômage) dans les **tableaux de bord de suivi de performance** régionale pour alerter les directeurs en cas de détérioration | Suivi proactif du contexte économique |

> ✅ **Objectif métier** : Personnaliser les stratégies commerciales selon le contexte économique local pour améliorer la pertinence des actions et la satisfaction client dans chaque zone de chalandise.

</div>
