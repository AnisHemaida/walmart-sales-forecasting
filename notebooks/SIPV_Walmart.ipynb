## <span style="color:#2E86C1;">Introduction</span>

La **grande distribution** est un secteur hautement concurrentiel où la capacité à **anticiper la demande** constitue un avantage stratégique majeur. Walmart, en tant que leader mondial du commerce de détail, génère chaque semaine des volumes considérables de données transactionnelles à travers ses magasins répartis sur différentes régions.

Ce projet, intitulé **« Système Intelligent de Prévision des Ventes en Grande Distribution »**, vise à concevoir et déployer un pipeline complet de science des données pour **prédire les ventes hebdomadaires** de Walmart. En exploitant des données historiques enrichies de variables contextuelles — telles que la température, le prix du carburant, le taux de chômage, l'indice des prix à la consommation (CPI) et les périodes promotionnelles (*markdown*) — nous chercherons à construire des modèles prédictifs robustes et interprétables.

### Objectifs principaux

- **Comprendre** le contexte métier de la grande distribution et les facteurs qui influencent les ventes.
- **Explorer et préparer** les données en identifiant les anomalies, les valeurs manquantes et les relations entre variables.
- **Modéliser** les ventes à l'aide d'algorithmes de Machine Learning (régression, arbres de décision, forêts aléatoires, gradient boosting, etc.).
- **Évaluer** la performance des modèles à travers des métriques adaptées (MAE, RMSE, WMAE).
- **Exploiter** les résultats pour formuler des recommandations stratégiques à destination des décideurs.

### Jeux de données utilisés

| Fichier | Description |
|---|---|
| `train.csv` | Données d'entraînement : ventes hebdomadaires par magasin et département (2010-02 à 2012-10) |
| `test.csv` | Données de test pour la prédiction (2012-11 à 2013-07) |
| `stores.csv` | Informations sur les magasins (type et superficie) |
| `features.csv` | Variables contextuelles (température, carburant, CPI, chômage, promotions) |

### Démarche adoptée

Le projet est structuré en **trois grandes parties** :
1. **Compréhension et Préparation des Données** — Exploration, nettoyage et ingénierie des variables.
2. **Modélisation et Évaluation Prédictive** — Entraînement, optimisation et comparaison des modèles.
3. **Intelligence Décisionnelle et Exploitation** — Interprétation des résultats et recommandations stratégiques.

## <span id="11" style="color:#2E86C1;">1.1 Comprendre le contexte retail Walmart</span>

Le secteur du **retail (commerce de détail)** repose sur la vente directe de produits aux consommateurs à travers des magasins physiques ou des plateformes en ligne. Dans ce contexte, les entreprises génèrent de grandes quantités de données liées aux ventes, aux clients et aux opérations. L’analyse de ces données permet d’améliorer la prise de décision, d’optimiser les stocks et d’anticiper la demande.

**Walmart**, l’un des plus grands détaillants au monde, produit chaque jour un volume massif de données grâce à ses milliers de magasins et ses millions de transactions. Ces données constituent un contexte pertinent pour appliquer des méthodes de **science des données et de prévision des ventes**.

Dans ce projet, l’objectif est d’analyser les **données historiques de ventes** afin de comprendre les facteurs influençant la demande et de développer des **modèles capables de prédire les ventes futures**, permettant ainsi d’améliorer la gestion des stocks et la planification commerciale.

### <span id="111" style="color:#2E86C1;">1.1.1 Analyse du fonctionnement de la grande distribution</span>

La grande distribution constitue l'un des secteurs les plus complexes à piloter sur le plan analytique, en raison de la multiplicité des variables qui influencent la performance commerciale. Walmart, en tant que leader mondial de la distribution avec plus de 10 000 points de vente dans le monde, illustre parfaitement les défis inhérents à ce modèle économique.
### Modèle économique et logique de volume
Le modèle économique de la grande distribution repose sur un principe fondamental : générer de la rentabilité par la rotation rapide des stocks plutôt que par des marges unitaires élevées. Walmart applique cette logique à travers sa stratégie historique "Everyday Low Prices" (EDLP), qui privilégie des prix bas permanents à des promotions ponctuelles. Dans ce contexte, la capacité à anticiper précisément la demande devient un avantage concurrentiel majeur : une rupture de stock signifie une vente perdue et une dégradation de l'expérience client, tandis qu'un surstock génère des coûts de stockage, des démarques et potentiellement des destructions de marchandise.
### Organisation par départements et hétérogénéité des ventes
Au sein de chaque magasin, les ventes sont enregistrées à la granularité Store × Département × Semaine, soit une combinaison de 45 magasins et 81 départements couvrant l'ensemble des rayons — épicerie, électronique, habillement, jardinage, automobile, pharmacie, entre autres. Cette granularité fine révèle une forte hétérogénéité interne : certains départements, tels que le département 92 (épicerie générale) ou le département 95 (jardinage), concentrent une part disproportionnée du chiffre d'affaires total, tandis que d'autres affichent des volumes marginaux mais des profils saisonniers très marqués. Cette disparité impose de traiter chaque combinaison Store × Département comme une entité à part entière dans la modélisation.
### Les leviers opérationnels de la grande distribution
La performance d'un distributeur comme Walmart repose sur la maîtrise de trois leviers opérationnels interdépendants, qui constituent précisément les axes de décision que ce projet vise à soutenir.
### 1. Gestion des stocks et réapprovisionnement
La planification des stocks s'appuie sur l'anticipation de la demande future. Une prévision inexacte se traduit immédiatement par un impact financier mesurable : en cas de sous-estimation, la rupture de stock prive le magasin de ventes et dégrade la satisfaction client ; en cas de surestimation, le surstock mobilise des ressources logistiques, génère des coûts de stockage et conduit à des démarques souvent subies. Dans ce projet, l'objectif est de réduire les ruptures de stock de 20% grâce à une meilleure anticipation des volumes hebdomadaires.
### 2. Politique promotionnelle et gestion des MarkDowns
Walmart recourt à des remises temporaires, appelées MarkDowns, pour stimuler les ventes sur certains rayons, écouler des stocks excédentaires ou répondre à des dynamiques concurrentielles. Ces promotions génèrent des pics de demande dont l'intensité varie selon le type de magasin, le département concerné et la période de l'année. Anticiper l'effet d'un MarkDown sur les ventes hebdomadaires est donc un enjeu majeur pour dimensionner correctement les stocks en amont et maximiser le retour sur investissement promotionnel.
### 3. Planification des achats et gestion des délais fournisseurs
Les décisions d'achat sont structurellement prises plusieurs semaines avant la mise en rayon, en raison des délais logistiques et des contraintes fournisseurs. Elles doivent intégrer à la fois les événements prévisibles — fêtes nationales, variations saisonnières — et le contexte économique local propre à chaque zone de chalandise. Une planification des achats fondée sur des prévisions fiables permet de sécuriser les approvisionnements aux moments critiques tout en limitant les immobilisations de trésorerie.
### 4. Les facteurs externes structurant la demande
Au-delà des décisions internes, les ventes en grande distribution sont étroitement corrélées à des variables contextuelles que le dataset Walmart capture de manière explicite. Les événements de fêtes nationales, représentés par la variable IsHoliday, provoquent des pics de ventes très marqués pouvant doubler la moyenne hebdomadaire, comme c'est le cas lors de Thanksgiving en semaine 47. L'indice des prix à la consommation (CPI) influence les arbitrages d'achat des ménages, notamment sur les catégories non-essentielles. Le taux de chômage local (Unemployment) accroît la sensibilité au prix dans les zones économiquement fragiles. Le prix du carburant (Fuel_Price) affecte quant à lui la fréquence et le périmètre des déplacements en magasin. Enfin, la température (Temperature) conditionne la saisonnalité de certains rayons comme le jardinage, les articles de saison ou l'habillement.
La prise en compte de l'ensemble de ces facteurs dans le modèle prédictif est indispensable pour capturer la variabilité réelle des ventes et produire des prévisions robustes.

### <span id="112" style="color:#2E86C1;">1.1.2 Identification des décisions à soutenir (stocks, promotions...)</span>

### 1) Décisions “Stocks” à soutenir (réassort & allocation)

### A. Réassort (quand et combien commander)

- Question : “Combien commander pour Store X / Dept Y la semaine prochaine (ou 2–4 semaines) ?”

- Sortie attendue : prévision des ventes + quantité recommandée (réassort).

- KPI : taux de rupture (stockout), taux de service, surstock, coût de stockage.

>Les décisions de réassort consistent à déterminer quand et combien de produits commander pour chaque magasin et département. Grâce aux prévisions de ventes, il est possible d’estimer les quantités nécessaires pour les semaines à venir. L’objectif est d’éviter les ruptures de stock tout en limitant le surstock et les coûts de stockage.

### B. Stock de sécurité (safety stock)

- Question : “Quel stock minimum garder pour éviter les ruptures pendant pics saisonniers / holidays ?”

- Sortie : niveau de stock de sécurité recommandé par dept/magasin (selon volatilité).

- KPI : ruptures évitées, stabilité de disponibilité.

>Le stock de sécurité correspond à la quantité minimale de produits à garder pour éviter les ruptures de stock, surtout pendant les périodes de forte demande comme les fêtes ou les événements saisonniers. En analysant la variabilité des ventes, on peut recommander un niveau de stock de sécurité adapté pour chaque magasin et département. L’objectif est de maintenir une bonne disponibilité des produits pour les clients.

### C. Allocation inter-magasins (prioriser où envoyer le stock)

- Question : “Si stock limité, quels magasins doivent être servis en priorité ?”

- Sortie : classement des magasins/depts selon demande prévue + risque de rupture.

- KPI : pertes de ventes évitées, disponibilité globale.

>Lorsque le stock est limité, il est important de décider quels magasins doivent être approvisionnés en priorité. En utilisant les prévisions de ventes, on peut identifier les magasins et départements où la demande sera la plus forte et où le risque de rupture est élevé. Cela permet d’envoyer les produits là où ils seront le plus nécessaires et de réduire les pertes de ventes.

### D. Préparation des pics (événements & périodes de fêtes)

- Question : “Quelles semaines vont exploser (Noël, Thanksgiving, etc.) et sur quels depts ?”

- Sortie : calendrier de pics + “top depts à risque”.

- KPI : précision des pics, réduction des ruptures en périodes critiques.

==> Les prévisions permettent aussi d’identifier les périodes où les ventes vont fortement augmenter, comme pendant les fêtes ou certains événements saisonniers. En repérant ces semaines à l’avance et les départements concernés, Walmart peut mieux préparer ses stocks. Cela aide à éviter les ruptures de produits pendant les périodes de forte demande.

### 2) Décisions “Promotions” à soutenir (planification & évaluation)

### E. Planifier les promos (quand lancer / sur quels produits)

- Question : “Quelle promo faire, à quelle semaine, dans quel magasin/département ?”

- Sortie : recommandation de périodes “propices” (là où la demande est sensible / saisonnière).

- KPI : uplift ventes (l’augmentation des ventes grâce à la promotion), marge, ROI promo(le retour sur investissement de la promotion).

>Les prévisions de ventes peuvent aussi aider à planifier les promotions. En analysant la saisonnalité et la sensibilité de la demande, il devient possible d’identifier les semaines et les départements où une promotion serait la plus efficace. Cela permet d’augmenter les ventes tout en optimisant la marge et le retour sur investissement des campagnes promotionnelles.

### F. Mesurer l’impact des promotions (uplift)

- Question : “La promo a-t-elle réellement augmenté les ventes ou juste déplacé la demande ?”

- Sortie : comparaison ventes attendues (baseline) vs ventes observées pendant promo.

- KPI : uplift net(l’augmentation réelle des ventes grâce à la promotion), cannibalisation (quand une promotion fait baisser les ventes d’un autre produit ou département), efficacité par dept/magasin.

>Cette analyse permet de mesurer si une promotion a réellement augmenté les ventes ou si elle a seulement déplacé la demande dans le temps. Pour cela, on compare les ventes prévues sans promotion avec les ventes réellement observées pendant la période promotionnelle. Cela permet d’évaluer l’efficacité de la promotion pour chaque magasin et département.

### G. Éviter les promos “inutiles” (mauvais timing)

- Question : “Est-ce qu’on a besoin d’une promo si la demande est déjà naturellement haute ?”

### <span id="121" style="color:#2E86C1;">1.2.1 Définir objectif 1: Prévision ventes par magasin et département</span>


---

### Objectif en une phrase

---

### 📦 Ce que disent nos données (train.csv)

Avant de définir l'objectif, on a regardé ce qu'il y a dans les données :

| Fait | Chiffre réel |
|---|---|
| Nombre de magasins | **45** (22 grands de type A, 17 moyens B, 6 petits C) |
| Nombre de départements | **81** (alimentaire, électronique, vêtements…) |
| Période couverte | **2010-02-05 → 2012-10-26** (≈ 143 semaines) |

> 💡 **Ce que ça veut dire :** On a 45 × 81 = ~3 600 combinaisons Store/Dept possibles, sur 143 semaines. Chaque ligne = une case de ce tableau que le modèle doit apprendre à prédire.

---

### 🤖 Comment on formule ça pour la machine

C'est un problème de **régression** : on donne des chiffres (historique + contexte), la machine prédit un autre chiffre (les ventes futures).

```
Entrées connues                         Sortie à prédire
────────────────────────────────────    ─────────────────
Store, Dept, Date (semaine)         →   Weekly_Sales ($)
Type de magasin (A/B/C), Taille
Fête ou pas (IsHoliday)
Température, Prix carburant
Promotions (MarkDown1 à 5)
CPI (inflation), Chômage
```

---

### 📏 Comment savoir si notre modèle est bon ?

On mesure l'écart entre ce que le modèle prédit et la réalité.  
Le concours Kaggle utilise le **WMAE** (Weighted Mean Absolute Error — Erreur Absolue Moyenne Pondérée) :

$$\text{WMAE} = \frac{\sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|}{\sum_{i=1}^{n} w_i}$$

En simple : *c'est la moyenne des erreurs (en $), mais les erreurs faites pendant les semaines de fêtes comptent 5× plus.*

| Semaine | Poids $w_i$ | Pourquoi |
|---|---|---|
| Semaine normale | **1×** | Standard |
| Semaine de fête | **5×** | Ces semaines sont stratégiques pour Walmart |

### <span id="133" style="color:#2E86C1;">1.3.3 Identification des valeurs incohérences</span>

````python

# ── 1.3.3 Identification des valeurs incohérentes ─────────────────────────────

problemes = {}

# 1) Ventes négatives (ne devrait pas exister)
neg = df_train[df_train['Weekly_Sales'] < 0]
problemes['Ventes négatives (Weekly_Sales < 0)'] = len(neg)
if len(neg) > 0:
    print(f"⚠️  {len(neg)} lignes avec ventes négatives :")
    display(neg[['Store', 'Dept', 'Date', 'Weekly_Sales']].head(10))

# 2) Ventes = 0 (dept inactif ou fermé ?)
zero = df_train[df_train['Weekly_Sales'] == 0]
problemes['Ventes exactement à 0'] = len(zero)

# 3) Doublons exacts
dupes = df_train.duplicated(subset=['Store', 'Dept', 'Date'])
problemes['Doublons (Store, Dept, Date)'] = dupes.sum()

# 4) Incohérence IsHoliday (conflit entre train et features)
# Vérification que la colonne IsHoliday est cohérente
holiday_check = df_train[['Store', 'Date', 'IsHoliday']].drop_duplicates()
holiday_conflict = holiday_check.groupby('Date')['IsHoliday'].nunique()
conflits_holiday = holiday_conflict[holiday_conflict > 1]
problemes['Conflits IsHoliday sur même date'] = len(conflits_holiday)

# 5) Taille de magasin (Size) : valeurs nulles ou négatives
problemes['Size manquant ou ≤ 0'] = (df_train['Size'].isnull() | (df_train['Size'] <= 0)).sum()

# ── Synthèse ──────────────────────────────────────────────────────────────────
print("=" * 55)
print("SYNTHÈSE DES INCOHÉRENCES DÉTECTÉES")
print("=" * 55)
for desc, count in problemes.items():
    statut = "✅" if count == 0 else "⚠️ "
    print(f"  {statut}  {desc:45s} → {count:,}")

print(f"\n  ℹ️  Les ventes négatives sont connues dans ce dataset.")
print(f"     Elles correspondent à des retours/corrections de caisse.")
print(f"     Elles seront conservées car représentent une réalité métier.")

# ── Distribution des ventes négatives ─────────────────────────────────────────
if len(neg) > 0:
    print(f"\n  Stats des ventes négatives :")
    print(neg['Weekly_Sales'].describe().round(2).to_string())
````

### <span id="134" style="color:#2E86C1;">1.3.4 Traiter les valeurs manquantes</span>

````python

# ── 1.3.4 Traitement des valeurs manquantes ───────────────────────────────────

print("── Avant traitement ──")
print(df_train[['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5',
                'CPI','Unemployment']].isnull().sum().to_string())

# ── MarkDown1 à MarkDown5 ─────────────────────────────────────────────────────
# NaN = aucune promotion cette semaine-là → on remplace par 0
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
df_train[markdown_cols] = df_train[markdown_cols].fillna(0)
df_test[markdown_cols]  = df_test[markdown_cols].fillna(0)

# ── CPI et Unemployment ───────────────────────────────────────────────────────
# Ces variables évoluent lentement → interpolation linéaire dans chaque magasin
# puis forward-fill et backward-fill pour les extrémités
for col in ['CPI', 'Unemployment']:
    df_train[col] = (df_train
                     .sort_values('Date')
                     .groupby('Store')[col]
                     .transform(lambda x: x.interpolate(method='linear')
                                           .ffill()
                                           .bfill()))
    df_test[col]  = (df_test
                     .sort_values('Date')
                     .groupby('Store')[col]
                     .transform(lambda x: x.interpolate(method='linear')
                                           .ffill()
                                           .bfill()))

# ── Fuel_Price — vérification (normalement complet) ──────────────────────────
for col in ['Fuel_Price', 'Temperature']:
    remaining = df_train[col].isnull().sum()
    if remaining > 0:
        df_train[col] = (df_train
                         .sort_values('Date')
                         .groupby('Store')[col]
                         .transform(lambda x: x.interpolate().ffill().bfill()))

print("\n── Après traitement ──")
missing_apres = df_train.isnull().sum()
missing_apres = missing_apres[missing_apres > 0]
if len(missing_apres) == 0:
    print("  ✅ Aucune valeur manquante restante dans df_train")
else:
    print(missing_apres.to_string())

missing_test = df_test.isnull().sum()
missing_test = missing_test[missing_test > 0]
if len(missing_test) == 0:
    print("  ✅ Aucune valeur manquante restante dans df_test")
else:
    print("  Restant dans df_test :")
    print(missing_test.to_string())

print(f"\n  Stratégie appliquée :")
print(f"  • MarkDown1-5  → remplacement par 0 (absence de promo = 0)")
print(f"  • CPI, Unemployment → interpolation linéaire par magasin + ffill/bfill")
````

**📋 Interprétation — Traitement des valeurs manquantes**

Deux stratégies ont été appliquées selon la nature de chaque variable :

**MarkDown1 à MarkDown5 → remplacement par `0`**
- Un `NaN` dans ces colonnes signifie littéralement « aucune promotion cette semaine-là ».
- Remplacer par `0` est donc **sémantiquement correct** : 0 $ de remise = pas de promotion.
- Après traitement : **0 valeur manquante** sur les 5 colonnes MarkDown.

**CPI et Unemployment → interpolation linéaire par magasin**
- Ces variables évoluent lentement dans le temps (données économiques mensuelles/trimestrielles).
- L'interpolation linéaire entre deux valeurs connues est donc la méthode la plus fidèle à la réalité.
- Un `ffill()` puis `bfill()` est appliqué pour couvrir les éventuelles extrémités de série.
- Le groupement par `Store` garantit qu'on ne mélange pas les données économiques de zones différentes.

> ✅ Après traitement : **aucune valeur manquante** dans `df_train` ni dans `df_test`.

````python
# ── 1.4.1 Structure et variables du dataset ───────────────────────────────────
import pandas as pd

print("=" * 55)
print("STRUCTURE DU DATASET df_train")
print("=" * 55)
print(f"  Lignes        : {df_train.shape[0]:,}")
print(f"  Colonnes      : {df_train.shape[1]}")
print(f"  Magasins      : {df_train['Store'].nunique()}")
print(f"  Départements  : {df_train['Dept'].nunique()}")
print(f"  Semaines      : {df_train['Date'].nunique()}")
print(f"  Période       : {df_train['Date'].min().date()} → {df_train['Date'].max().date()}")

# ── Types de variables ────────────────────────────────────────────────────────
print("\n── Types de variables et valeurs uniques ──")
overview = pd.DataFrame({
    'Dtype'           : df_train.dtypes,
    'Valeurs uniques' : df_train.nunique(),
    'Exemple'         : [df_train[c].dropna().iloc[0] for c in df_train.columns]
})
display(overview)

# ── Répartition des types de magasins ────────────────────────────────────────
type_labels = {0: 'C (petit)', 1: 'B (moyen)', 2: 'A (grand)'}
store_types = (df_train.drop_duplicates('Store')[['Store', 'Type']]
               .assign(Libelle=lambda d: d['Type'].map(type_labels))
               .groupby('Libelle').size()
               .rename('Nb magasins')
               .to_frame())
print("\n── Répartition des types de magasins ──")
display(store_types)

# ── Stats descriptives ────────────────────────────────────────────────────────
print("\n── Statistiques descriptives (Weekly_Sales) ──")
display(df_train[['Weekly_Sales']].describe().round(2))

# ── Identification des semaines de fêtes (IsHoliday == True) ──────────────────
# SOURCE : documentation officielle de la compétition Kaggle
#          "Walmart Recruiting – Store Sales Forecasting" (2013)
#
# Walmart indexe 4 événements comme jours fériés (IsHoliday = True),
# pondérés ×5 dans la WMAE pour pénaliser les erreurs sur ces semaines :
#
#   Événement      Date de référence              Règle de détection
#   ─────────────  ──────────────────────────────  ──────────────────────────────
#   Super Bowl     vendredi ~ 12 février           mois == 2
#   Labor Day      vendredi du 1er lundi sept.     mois == 9
#   Thanksgiving   vendredi du 4e jeudi nov.       mois == 11
#   Christmas      vendredi ~ 25 décembre          mois == 12
#
# La date dans le CSV est le vendredi de début de semaine (convention Walmart).
# Comme chaque événement tombé dans un mois différent, le mois seul suffit
# à identifier la fête de façon univoque sur l'ensemble du dataset.

def nom_fete(date):
    """Identifie la fête d'après le mois (règle issue de la doc. Kaggle)."""
    m = pd.Timestamp(date).month
    if m == 2:  return 'Super Bowl'
    if m == 9:  return 'Labor Day'
    if m == 11: return 'Thanksgiving'
    return 'Christmas'  # mois 12 (ou janvier si chevauchement d'année)

holiday_dates = df_train[df_train['IsHoliday'] == 1]['Date'].sort_values().unique()

print("\n── Identification des semaines de fêtes ──")
print(f"  {'Date':<15} {'Événement':<18} {'Règle (source : Kaggle)'}")
print("  " + "─" * 60)
for d in holiday_dates:
    ts = pd.Timestamp(d)
    event = nom_fete(d)
    if ts.month == 2:
        rule = "Vendredi ~12 fév → semaine du Super Bowl"
    elif ts.month == 9:
        rule = "1er lundi de septembre → Labor Day"
    elif ts.month == 11:
        rule = "4e jeudi de novembre → Thanksgiving"
    else:
        rule = "Vendredi ~25 déc → semaine de Noël"
    print(f"  {str(ts.date()):<15} {event:<18} {rule}")

print(f"\n  → {len(holiday_dates)} semaines sur {df_train['Date'].nunique()} "
      f"({len(holiday_dates)/df_train['Date'].nunique()*100:.0f}%) "
      f"sont pondérées ×5 dans la WMAE")
print("  ⚠  Thanksgiving et Noël 2012 absents car les données s'arrêtent en octobre 2012")
````

**🗓️ Comment identifier les semaines de fêtes ?**

La colonne `IsHoliday` est fournie directement dans les fichiers `train.csv` et `features.csv` du Kaggle. La documentation officielle de la compétition précise que **4 événements** sont marqués comme jours fériés, chacun tombant dans un mois distinct :

| Événement | Date de référence | Comment nous le savons |
|---|---|---|
| **Super Bowl** | ~12 février | Décrit dans la page Kaggle ; les dates (12 fév 2010, 11 fév 2011, 10 fév 2012) correspondent au vendredi le plus proche du Super Bowl |
| **Labor Day** | 1er lundi de septembre | Fête nationale américaine fixe ; les dates (10 sept 2010, 9 sept 2011, 7 sept 2012) confirment le premier lundi de septembre |
| **Thanksgiving** | 4e jeudi de novembre | Fête nationale américaine fixe ; les dates (26 nov 2010, 25 nov 2011) = 4e jeudi de novembre |
| **Christmas** | ~25 décembre | Décrit dans la page Kaggle ; les dates (31 déc 2010, 30 déc 2011) = vendredi précédant ou suivant le 25 décembre |

**Règle de détection :** comme chaque événement tombe dans un **mois différent** (2, 9, 11, 12), le mois seul suffit à identifier la fête de façon univoque. La fonction `nom_fete()` exploite cette propriété.

````python
# ── 1.4.4 Distribution temporelle des valeurs négatives (temporaire) ─────────

neg = df_train[df_train['Weekly_Sales'] < 0].copy()
neg['Date'] = pd.to_datetime(neg['Date'])
df_dates = df_train.copy()
df_dates['Date'] = pd.to_datetime(df_dates['Date'])
neg['Year'] = neg['Date'].dt.year

neg_weekly_count = neg.groupby('Date').size()
neg_weekly_sum = neg.groupby('Date')['Weekly_Sales'].sum()
neg_monthly_count = neg.set_index('Date').resample('ME').size()
neg_yearly_count = neg.groupby('Year').size().sort_index()
neg_yearly_share = (neg_yearly_count / df_dates.groupby(df_dates['Date'].dt.year).size() * 100)
neg_yearly_amount = neg.groupby('Year')['Weekly_Sales'].sum().sort_index()

# Base absolue (montants negatifs en valeur positive)
neg_yearly_amount_abs = neg_yearly_amount.abs()

# Part des montants negatifs sur le total annuel des ventes
annual_total_sales = df_dates.groupby(df_dates['Date'].dt.year)['Weekly_Sales'].sum().sort_index()
neg_amount_share_of_total_pct = (neg_yearly_amount_abs / annual_total_sales * 100)

# Evolution annuelle classique (YoY) -> 2010 n'a pas de valeur precedente
neg_amount_evol_yoy_pct = neg_yearly_amount_abs.pct_change() * 100

# Evolution par rapport a 2010 (base 2010 = 0%) -> inclut 2010
neg_amount_evol_base2010_pct = (neg_yearly_amount_abs / neg_yearly_amount_abs.iloc[0] - 1) * 100

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

axes[0, 0].plot(neg_weekly_count.index, neg_weekly_count.values, color='firebrick', linewidth=2)
axes[0, 0].set_title("Nombre de ventes negatives par semaine", fontsize=12)
axes[0, 0].set_xlabel("Date")
axes[0, 0].set_ylabel("Nombre de lignes negatives")
axes[0, 0].grid(alpha=0.3)

axes[0, 1].fill_between(neg_weekly_sum.index, neg_weekly_sum.values, color='indianred', alpha=0.65)
axes[0, 1].axhline(0, color='black', linewidth=1)
axes[0, 1].set_title("Montant total negatif par semaine", fontsize=12)
axes[0, 1].set_xlabel("Date")
axes[0, 1].set_ylabel("Somme Weekly_Sales < 0 ($)")
axes[0, 1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}$"))
axes[0, 1].grid(alpha=0.3)

axes[1, 0].bar(neg_monthly_count.index, neg_monthly_count.values, width=25, color='peru')
axes[1, 0].set_title("Frequence mensuelle des valeurs negatives", fontsize=12)
axes[1, 0].set_xlabel("Mois")
axes[1, 0].set_ylabel("Nombre de lignes negatives")
axes[1, 0].grid(axis='y', alpha=0.3)

bars = axes[1, 1].bar(neg_yearly_share.index.astype(str), neg_yearly_share.values, color=['#C0392B', '#D35400', '#A93226'])
axes[1, 1].set_title("Part des valeurs negatives par annee", fontsize=12)
axes[1, 1].set_xlabel("Annee")
axes[1, 1].set_ylabel("Part des lignes negatives (%)")
axes[1, 1].grid(axis='y', alpha=0.3)
for bar, val in zip(bars, neg_yearly_share.values):
    axes[1, 1].text(bar.get_x() + bar.get_width() / 2, val + 0.01, f"{val:.2f}%",
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# ── Visuel demande: evolution (%) du montant negatif avec 2010 visible ───────
fig, ax = plt.subplots(figsize=(10, 5))

evol_plot = neg_amount_evol_base2010_pct
colors = ['slategray' if i == 0 else ('crimson' if v > 0 else 'seagreen')
          for i, v in enumerate(evol_plot.values)]
bars = ax.bar(evol_plot.index.astype(str), evol_plot.values, color=colors)
ax.axhline(0, color='black', linewidth=1)
ax.set_title("Evolution (%) du montant negatif - base 2010", fontsize=13)
ax.set_xlabel("Annee")
ax.set_ylabel("Variation vs 2010 (%)")
ax.grid(axis='y', alpha=0.3)

for bar, val in zip(bars, evol_plot.values):
    ax.text(bar.get_x() + bar.get_width() / 2,
            val + (0.8 if val >= 0 else -0.8),
            f"{val:+.1f}%",
            ha='center',
            va='bottom' if val >= 0 else 'top',
            fontsize=10,
            fontweight='bold')

ax.text(0.99, 0.02,
        "2010 = annee de reference (0%)",
        transform=ax.transAxes, ha='right', va='bottom',
        fontsize=9, color='gray', style='italic')

plt.tight_layout()
plt.show()

# ── Nouveau visuel: part des montants negatifs dans le total annuel ──────────
fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(
    neg_amount_share_of_total_pct.index.astype(str),
    neg_amount_share_of_total_pct.values,
    color=['#7D3C98', '#884EA0', '#A569BD']
)
ax.set_title("Part du montant negatif dans le total annuel des ventes", fontsize=13)
ax.set_xlabel("Annee")
ax.set_ylabel("Part du montant negatif (%)")
ax.grid(axis='y', alpha=0.3)
for bar, val in zip(bars, neg_amount_share_of_total_pct.values):
    ax.text(bar.get_x() + bar.get_width() / 2,
            val + 0.0005,
            f"{val:.4f}%",
            ha='center', va='bottom', fontsize=10, fontweight='bold')

ax.text(0.99, 0.02,
        "Base: |montants negatifs| / ventes annuelles totales",
        transform=ax.transAxes, ha='right', va='bottom',
        fontsize=9, color='gray', style='italic')

plt.tight_layout()
plt.show()

print("── Synthese des ventes negatives ──")
print(f"Nombre total de lignes negatives : {len(neg):,}")
print(f"Part du dataset                  : {len(neg) / len(df_train) * 100:.2f}%")
print(f"Nombre de semaines touchees      : {neg['Date'].nunique()} sur {df_dates['Date'].nunique()}")
print(f"Montant negatif cumule           : {neg['Weekly_Sales'].sum():,.2f}$")

print("\nEvolution annuelle des valeurs negatives :")
display(
    pd.DataFrame({
        'Nb_valeurs_negatives': neg_yearly_count,
        'Part_lignes_negatives_%': neg_yearly_share.round(2),
        'Montant_negatif_total_$': neg_yearly_amount.round(2),
        'Montant_negatif_abs_$': neg_yearly_amount_abs.round(2),
        'Part_montant_negatif_sur_total_annuel_%': neg_amount_share_of_total_pct.round(4),
        'Evol_montant_negatif_YoY_%': neg_amount_evol_yoy_pct.round(2),
        'Evol_montant_negatif_vs_2010_%': neg_amount_evol_base2010_pct.round(2)
    })
)

print("\nTop 10 semaines les plus touchees :")
display(
    neg.groupby('Date')['Weekly_Sales']
       .agg(Nb_valeurs_negatives='size', Montant_negatif_total='sum')
       .sort_values('Nb_valeurs_negatives', ascending=False)
       .head(10)
       .round(2)
)
````

Les résultats montrent que les valeurs négatives sont **faibles en volume global** mais **présentes tout au long de la période**. Les résultats montrent que 35,9% des semaines comportent au moins une promotion, contre 64,1% sans promotion.

 Elle indique que les promotions ne sont pas systématiques, mais qu’elles concernent tout de même une part importante des observations.

On peut donc considérer que les promotions constituent un levier commercial significatif dans le dataset, tout en restant ponctuelles. Cette variable pourra être utilisée pour comparer les niveaux de ventes entre semaines promotionnelles et non promotionnelles.


````python
# ── Comparaison des ventes avec/sans promotion ──────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))


# Boxplot des ventes
sns.boxplot(data=df_train, x='Has_Promotion', y='Weekly_Sales', ax=ax1, palette=['#E74C3C', '#27AE60'])
ax1.set_title('Distribution des ventes selon la présence de promotions', fontsize=13, fontweight='bold')
ax1.set_xlabel('Promotion active')
ax1.set_ylabel('Ventes hebdomadaires ($)')
ax1.set_xticklabels(['Sans promotion', 'Avec promotion'])

# Violin plot pour voir la distribution
sns.violinplot(data=df_train, x='Has_Promotion', y='Weekly_Sales', ax=ax2, palette=['#E74C3C', '#27AE60'])
ax2.set_title('Distribution détaillée des ventes', fontsize=13, fontweight='bold')
ax2.set_xlabel('Promotion active')
ax2.set_ylabel('Ventes hebdomadaires ($)')
ax2.set_xticklabels(['Sans promotion', 'Avec promotion'])

plt.tight_layout()
plt.savefig('../images/1.5.1_markdowns.png', dpi=300, bbox_inches='tight')
plt.show()
````

📌 **Interprétation :**

Les graphiques comparent la distribution des ventes hebdomadaires selon la présence ou non d’une promotion active. On observe que les semaines avec promotion présentent une médiane légèrement plus élevée ainsi qu’une dispersion un peu plus importante que les semaines sans promotion.

Les deux distributions restent toutefois largement superposées, avec de nombreux outliers dans chaque groupe. Cela montre que la présence d’une promotion n’explique pas à elle seule les différences de ventes.

Ainsi, les promotions semblent avoir un effet positif potentiel sur les ventes, mais cet effet demeure hétérogène et probablement dépendant d’autres facteurs comme le type de magasin, le département, la saisonnalité ou les périodes de fêtes.


````python
# ── Analyse de l'intensité des promotions ───────────────────────────────────
# Puisque toutes les semaines ont des promotions, analysons l'impact de l'intensité promotionnelle

# Calcul du montant total des promotions par semaine
df_train['Total_Promotion'] = df_train[markdown_cols].sum(axis=1, skipna=True)

# Analyse de la corrélation entre montant des promotions et ventes
correlation = df_train['Total_Promotion'].corr(df_train['Weekly_Sales'])

print(f"\nAnalyse de l'intensité des promotions :")
print(f"  Montant moyen des promotions par semaine : ${df_train['Total_Promotion'].mean():,.0f}")
print(f"  Corrélation entre montant des promotions et ventes : {correlation:.3f}")

# Visualisation de la relation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_train, x='Total_Promotion', y='Weekly_Sales', alpha=0.6, color='#2E86C1')
plt.title('Relation entre montant des promotions et ventes hebdomadaires', fontsize=13, fontweight='bold')
plt.xlabel('Montant total des promotions ($)')
plt.ylabel('Ventes hebdomadaires ($)')
plt.tight_layout()
plt.savefig('../images/1.5.1_promotion_intensity.png', dpi=300, bbox_inches='tight')
plt.show()
````


📌 **Interprétation :**

Le montant moyen des promotions s’élève à environ 6 684 $ par semaine, ce qui montre que les actions promotionnelles représentent un levier commercial important dans le dataset.

Cependant, la corrélation entre le montant total des promotions et les ventes hebdomadaires est de seulement 0,065. Cette valeur indique une relation linéaire très faible, bien que légèrement positive.

Ainsi, une hausse du montant promotionnel semble associée à une légère augmentation des ventes, mais cet effet reste limité lorsqu’on considère l’ensemble des observations. Le nuage de points met en évidence une forte dispersion, ce qui suggère que l’impact des promotions dépend aussi d’autres facteurs comme le type de magasin, le département, la saisonnalité ou les fêtes.

En conclusion, l’intensité des promotions ne suffit pas à expliquer seule les variations de ventes et doit être interprétée en combinaison avec d’autres variables métier.

### <span id="211" style="color:#2E86C1;">2.1.1 Encodage des variables catégorielles</span>

````python
# ── 2.1.1 Encodage des variables catégorielles & features promo (MarkDown) ─────

import pandas as pd
import numpy as np

print("Taille de df_train avant préparation:", df_train.shape)
if 'df_test' in globals():
    print("Taille de df_test avant préparation:", df_test.shape)

# Hypothèse de travail (décision projet) :
# Dans ce projet, on choisit d'interpréter les NaN des MarkDown comme 0 (pas de promo).
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
promo_flag_cols = [f'{c}_has_promo' for c in markdown_cols]

# Assurer Date au bon format
df_train['Date'] = pd.to_datetime(df_train['Date'])
if 'df_test' in globals():
    df_test['Date'] = pd.to_datetime(df_test['Date'])

# Nettoyage (idempotence) : supprimer d'éventuels anciens flags *_is_missing / *_has_promo
cols_to_remove_train = [c for c in df_train.columns if ('_is_missing' in c) or ('_has_promo' in c) or (c == 'n_markdowns_active')]
if cols_to_remove_train:
    df_train = df_train.drop(columns=cols_to_remove_train)

if 'df_test' in globals():
    cols_to_remove_test = [c for c in df_test.columns if ('_is_missing' in c) or ('_has_promo' in c) or (c == 'n_markdowns_active')]
    if cols_to_remove_test:
        df_test = df_test.drop(columns=cols_to_remove_test)

# 1) MarkDown : NaN -> 0 (pas de promo)
df_train[markdown_cols] = df_train[markdown_cols].fillna(0.0)
if 'df_test' in globals():
    df_test[markdown_cols] = df_test[markdown_cols].fillna(0.0)

# 2) Indicateurs promo : 1 si MarkDown > 0 (promo active), sinon 0
for col in markdown_cols:
    df_train[f'{col}_has_promo'] = (df_train[col] > 0).astype('int8')
    if 'df_test' in globals():
        df_test[f'{col}_has_promo'] = (df_test[col] > 0).astype('int8')

# 3) Synthèse : nombre de MarkDown actifs la semaine (0..5)
df_train['n_markdowns_active'] = df_train[promo_flag_cols].sum(axis=1).astype('int8')
if 'df_test' in globals():
    df_test['n_markdowns_active'] = df_test[promo_flag_cols].sum(axis=1).astype('int8')

# 4) Encodage booléen pour IsHoliday (True/False -> 1/0)
df_train['IsHoliday'] = df_train['IsHoliday'].astype('int8')
if 'df_test' in globals():
    df_test['IsHoliday'] = df_test['IsHoliday'].astype('int8')

print("\nDtypes (contrôle rapide) :")
cols_to_check = ['IsHoliday', 'Type'] + promo_flag_cols + ['n_markdowns_active']
print(df_train[cols_to_check].dtypes)

print("\nAperçu (train) :")
display(df_train[['Date','Store','Dept'] + markdown_cols + promo_flag_cols + ['n_markdowns_active']].head())
````

#### Interprétation — 2.1.1 (MarkDown & variables catégorielles)

- L'effet attendu sur les ventes dépend surtout de la présence d'une promotion non nulle (signal “promo active”).
- Les variables binaires `*_has_promo` capturent ce signal métier (1 = promo, 0 = pas de promo).

**Variable de synthèse `n_markdowns_active`**
- Résume l’intensité de la promotion (0 à 5 MarkDowns actifs la même semaine).

### <span id="212" style="color:#2E86C1;">2.1.2 Création de variables temporelles et contextuelles</span>

````python
# ── 2.1.2 Création de variables temporelles et contextuelles ───────────────────

def extract_date_features(df):
    """
    Extrait les composantes de la date et ajoute des flags spécifiques pour 
    les périodes de fêtes, cruciaux pour un modèle ensembliste.
    """
    # 1. Extraction des composantes classiques (Saisonnalité générale)
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
    
    # 2. Ajout des drapeaux pour les semaines de vacances spécifiques
    # Basé sur l'analyse de la partie 1.5.2 et les règles Kaggle
    df['Is_SuperBowl'] = df['Date'].isin(pd.to_datetime(['2010-02-12', '2011-02-11', '2012-02-10', '2013-02-08'])).astype(int)
    df['Is_LabourDay'] = df['Date'].isin(pd.to_datetime(['2010-09-10', '2011-09-09', '2012-09-07', '2013-09-06'])).astype(int)
    df['Is_Thanksgiving'] = df['Date'].isin(pd.to_datetime(['2010-11-26', '2011-11-25', '2012-11-23', '2013-11-29'])).astype(int)
    df['Is_Christmas'] = df['Date'].isin(pd.to_datetime(['2010-12-31', '2011-12-30', '2012-12-28', '2013-12-27'])).astype(int)
    
    return df

print("Processus en cours sur df_train...")
df_train = extract_date_features(df_train)

if 'df_test' in globals():
    print("Processus en cours sur df_test...")
    df_test = extract_date_features(df_test)

print("\nAperçu des nouvelles features temporelles (Exemple aléatoire df_train) :")
display(df_train[['Date', 'Year', 'Month', 'Week', 'Is_Thanksgiving', 'IsHoliday']].sample(5, random_state=42))
````

### <span id="213" style="color:#2E86C1;">2.1.3 Séparation des jeux d'entraînement et de test</span>

````python
# ── 2.1.3 Séparation des jeux d'entraînement et de test (*Time Series Split*) ──

# Lors d'un problème métier temporel, on ne split jamais "au hasard"
# On doit simuler les conditions du réel : prédire le futur à partir du passé

# 1. Tri chronologique stricte indispensable (par précaution)
df_train = df_train.sort_values(by=['Date', 'Store', 'Dept']).reset_index(drop=True)

# 2. Définition de la ligne de coupure temporelle (objectif ~80% train / ~20% valid)
# Ici, on choisit une DATE de coupure telle qu'environ 80% des lignes (observations)
# soient dans le passé (train) et ~20% dans le futur (validation).
target_train_frac = 0.80

counts_by_date = df_train.groupby('Date').size().sort_index()
cum_frac = counts_by_date.cumsum() / counts_by_date.sum()

# Première date où l'on atteint (ou dépasse) 80% des lignes
split_date = cum_frac[cum_frac >= target_train_frac].index[0]

# Création des masques (train = passé, valid = futur)
mask_train = df_train['Date'] <= split_date
mask_valid = df_train['Date'] > split_date

# 3. Exclusions des colonnes (La Date ne rentre pas purement dans l'algo XGBoost, on a Year/Month/Week pour ça)
cols_to_drop = ['Weekly_Sales', 'Date']

# Train set interne
X_train_local = df_train[mask_train].drop(columns=cols_to_drop)
y_train_local = df_train[mask_train]['Weekly_Sales']

# Validation set interne
X_valid_local = df_train[mask_valid].drop(columns=cols_to_drop)
y_valid_local = df_train[mask_valid]['Weekly_Sales']

# Affichage des volumes
print(f"── Date de coupure (split_date) : {pd.to_datetime(split_date).date()}")
print(f"── Période d'Entraînement : du {df_train[mask_train]['Date'].min().date()} au {df_train[mask_train]['Date'].max().date()}")
print(f"── Période de Validation  : du {df_train[mask_valid]['Date'].min().date()} au {df_train[mask_valid]['Date'].max().date()}")
print("-" * 50)
print(f"Dimensions de X_train_local : {X_train_local.shape} (soit {len(X_train_local)/len(df_train):.1%} des données globales)")
print(f"Dimensions de X_valid_local : {X_valid_local.shape} (soit {len(X_valid_local)/len(df_train):.1%} des données globales)")
````

````python
# Visuels 2.1.3 — validation temporelle (split)

import matplotlib.pyplot as plt

# Série temporelle agrégée (toutes ventes hebdo) pour visualiser la coupure
weekly_total = df_train.groupby('Date', as_index=False)['Weekly_Sales'].sum().sort_values('Date')

plt.figure(figsize=(10, 4))
plt.plot(weekly_total['Date'], weekly_total['Weekly_Sales'], linewidth=1)
plt.axvline(split_date, color='red', linestyle='--', linewidth=2, label=f"Split: {split_date.date()}")
plt.title("Ventes hebdomadaires totales (df_train) et date de split")
plt.xlabel("Date")
plt.ylabel("Somme Weekly_Sales")
plt.legend()
plt.tight_layout()
plt.show()

# Comparaison des tailles train/valid (contrôle)
sizes = [len(X_train_local), len(X_valid_local)]
labels = ['Train (< split)', 'Validation (>= split)']
plt.figure(figsize=(6, 4))
plt.bar(labels, sizes, color=['#4C72B0', '#DD8452'])
plt.title("Nombre d'observations par sous-ensemble")
plt.ylabel("Nombre de lignes")
for i, v in enumerate(sizes):
    plt.text(i, v, f"{v:,}", ha='center', va='bottom')
plt.tight_layout()
plt.show()
````

#### Interprétation — 2.1.3 (split temporel)

**Pourquoi un split temporel ?**
- On veut simuler le cas réel : entraîner sur le passé et évaluer sur du futur.
- Un split aléatoire mélangerait des semaines futures dans l'entraînement (*data leakage*) et donnerait un score trop optimiste.

**Ce que montrent les visuels**
- La ligne rouge matérialise la date de coupure entre entraînement et validation.
- La date de coupure est choisie pour obtenir environ **80%** des observations en entraînement et **20%** en validation (au plus proche possible, car on coupe par semaine/date).
- La comparaison des tailles confirme que la validation représente une portion limitée mais suffisante pour tester la généralisation sur une période récente.

### 2.1.3bis — Validation croisée temporelle (*TimeSeriesSplit*, CV)

> Objectif : rendre l'évaluation plus **robuste** qu'un seul split, en répétant l'entraînement/validation sur plusieurs fenêtres temporelles.

- On respecte l'ordre du temps (pas de mélange passé/futur) → évite le *data leakage*.
- On obtient **5 scores** (un par fold) puis une **moyenne** (et un écart-type) des métriques.
- Ici, on split **par dates (semaines)** : chaque fold valide sur des semaines entières, cohérent avec le problème.

````python
# ── 2.1.3bis — TimeSeriesSplit (CV) : évaluation robuste ──

# On calcule des métriques sur 5 folds temporels (train = passé, valid = futur).
# NB: Comme les observations sont hebdomadaires, on split par Date (semaine) pour garder des semaines entières.

import numpy as np
import pandas as pd

from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import HistGradientBoostingRegressor

# Sécurité : ordre temporel
df_cv = df_train.sort_values(by=['Date', 'Store', 'Dept']).reset_index(drop=True)

# Features/target
cols_to_drop_cv = ['Weekly_Sales', 'Date']
X_all = df_cv.drop(columns=cols_to_drop_cv)
y_all = df_cv['Weekly_Sales']

# Encodage minimal si jamais il reste des colonnes non numériques (ex.: Type)
cat_cols = X_all.select_dtypes(include=['object', 'category']).columns.tolist()
if len(cat_cols) > 0:
    X_all = pd.get_dummies(X_all, columns=cat_cols, drop_first=False)

# Split par dates (semaines)
unique_dates = np.array(sorted(df_cv['Date'].unique()))
tscv = TimeSeriesSplit(n_splits=5)

# Modèle de base (rapide, sans dépendances externes)
model = HistGradientBoostingRegressor(
    random_state=42,
    max_depth=6,
    learning_rate=0.1,
    max_iter=100,
 )

fold_rows = []
for fold, (train_date_idx, valid_date_idx) in enumerate(tscv.split(unique_dates), start=1):
    train_dates = unique_dates[train_date_idx]
    valid_dates = unique_dates[valid_date_idx]

    mask_tr = df_cv['Date'].isin(train_dates)
    mask_va = df_cv['Date'].isin(valid_dates)

    X_tr, y_tr = X_all.loc[mask_tr], y_all.loc[mask_tr]
    X_va, y_va = X_all.loc[mask_va], y_all.loc[mask_va]

    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_va)

    mae = mean_absolute_error(y_va, y_pred)
    rmse = float(np.sqrt(mean_squared_error(y_va, y_pred)))

    # WMAE (Kaggle Walmart): poids 5 si IsHoliday==True sinon 1 (si dispo)
    if 'IsHoliday' in df_cv.columns:
        w = np.where(df_cv.loc[mask_va, 'IsHoliday'].astype(bool).values, 5.0, 1.0)
        wmae = float(np.sum(w * np.abs(y_va.values - y_pred)) / np.sum(w))
    else:
        wmae = np.nan

    fold_rows.append({
        'fold': fold,
        'train_dates': len(train_dates),
        'valid_dates': len(valid_dates),
        'train_rows': int(mask_tr.sum()),
        'valid_rows': int(mask_va.sum()),
        'train_start': pd.to_datetime(train_dates.min()).date(),
        'train_end': pd.to_datetime(train_dates.max()).date(),
        'valid_start': pd.to_datetime(valid_dates.min()).date(),
        'valid_end': pd.to_datetime(valid_dates.max()).date(),
        'MAE': mae,
        'RMSE': rmse,
        'WMAE': wmae,
    })

cv_results = pd.DataFrame(fold_rows)
display(cv_results)

summary = cv_results[['MAE', 'RMSE', 'WMAE']].agg(['mean', 'std']).T
summary.columns = ['moyenne', 'ecart_type']
display(summary)
````

````python
# Visuels — 2.1.3bis : ce que fait TimeSeriesSplit, visuellement

import matplotlib.dates as mdates
import matplotlib.pyplot as plt

if 'cv_results' not in globals():
    raise RuntimeError("Exécute d'abord la cellule 2.1.3bis (TimeSeriesSplit) pour créer cv_results.")

cv_viz = cv_results.copy()
for c in ['train_start', 'train_end', 'valid_start', 'valid_end']:
    cv_viz[c] = pd.to_datetime(cv_viz[c])

# 1) Fenêtres temporelles (train vs validation) pour chaque fold
fig, ax = plt.subplots(figsize=(11, 3.8))
for _, r in cv_viz.sort_values('fold').iterrows():
    y = int(r['fold'])
    ax.hlines(y, r['train_start'], r['train_end'], color='#4C72B0', lw=8, label='Train' if y == 1 else None)
    ax.hlines(y, r['valid_start'], r['valid_end'], color='#DD8452', lw=8, label='Validation' if y == 1 else None)

ax.set_title("TimeSeriesSplit (5 folds) — fenêtres temporelles par fold")
ax.set_xlabel("Date")
ax.set_ylabel("Fold")
ax.set_yticks(cv_viz['fold'].tolist())
ax.legend(loc='upper left')
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.grid(True, axis='x', linestyle=':', alpha=0.35)
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

# 2) Variabilité des métriques selon le fold (plus robuste qu'un seul split)
metrics = ['MAE', 'RMSE', 'WMAE']
fig, axes = plt.subplots(1, 3, figsize=(12, 3.2), sharex=True)
for ax, metric in zip(axes, metrics):
    ax.bar(cv_viz['fold'], cv_viz[metric], color='#4C72B0')
    ax.set_title(metric)
    ax.set_xlabel('Fold')
    ax.grid(True, axis='y', linestyle=':', alpha=0.35)
axes[0].set_ylabel('Valeur')
plt.suptitle('Métriques par fold (TimeSeriesSplit)', y=1.05)
plt.tight_layout()
plt.show()
````

#### Interprétation — 2.1.3bis (TimeSeriesSplit / validation croisée temporelle)

**Ce que montrent les visuels (folds)**
- Chaque ligne correspond à un *fold* : la partie **bleue** est la période utilisée pour entraîner, la partie **orange** est la période utilisée pour valider.
- Les folds avancent dans le temps : on entraîne toujours sur le passé et on valide sur une fenêtre plus récente → pas de mélange passé/futur.

**Lecture des métriques**
- On obtient 5 valeurs (une par fold) pour **MAE / RMSE / WMAE** : cela donne une estimation plus robuste qu'un seul split, car on voit la variabilité selon les périodes.
- Un écart‑type élevé indique que la difficulté de prédiction varie selon la période (saisonnalité, événements, changements de distribution, etc.).

**Important : le modèle ici est volontairement "démo"**

<h2 id="annexeA" style="color:#2E86C1;">📚 Annexe A — Comprendre les métriques d'évaluation</h2>

> Cette annexe explique pas-à-pas les métriques utilisées pour mesurer la qualité d'un modèle de prévision.  
> Pas de prérequis mathématiques — chaque notion est illustrée avec des chiffres concrets.

---

### A.1 — Le problème de base : mesurer une erreur

Imagine qu'on prédit les ventes d'un département pour 5 semaines :

| Semaine | Ventes réelles ($) | Ventes prédites ($) | Erreur ($) |
|---|---|---|---|
| S1 | 20 000 | 18 000 | −2 000 |
| S2 | 15 000 | 17 000 | +2 000 |
| S3 | 30 000 | 28 000 | −2 000 |
| S4 |  5 000 |  6 000 | +1 000 |
| S5 | 10 000 |  9 000 | −1 000 |

**Problème :** Si on fait juste la moyenne des erreurs → $\frac{(-2000)+(2000)+(-2000)+(1000)+(-1000)}{5} = -400$  
Les erreurs positives et négatives s'annulent → ce n'est pas utile !

---

### A.2 — MAE (Mean Absolute Error) — Erreur Absolue Moyenne

**Idée :** On prend la valeur absolue de chaque erreur (on ignore le signe), puis on fait la moyenne.

$$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

Avec notre exemple :

| Semaine | \|Erreur\| ($) |
|---|---|
| S1 | 2 000 |
| S2 | 2 000 |
| S3 | 2 000 |
| S4 | 1 000 |
| S5 | 1 000 |

$$\text{MAE} = \frac{2000 + 2000 + 2000 + 1000 + 1000}{5} = \frac{8000}{5} = \textbf{1 600 \$}$$

**Lecture :** En moyenne, notre modèle se trompe de **1 600 $** par semaine.  
**Avantage :** Simple, dans la même unité que la variable (ici des $).  
**Limite :** Traite les grandes et petites erreurs de façon égale.

---

### A.3 — RMSE (Root Mean Square Error) — Racine de l'Erreur Quadratique Moyenne

**Idée :** On met les erreurs au carré (pour pénaliser les grandes erreurs davantage), on fait la moyenne, puis on prend la racine carrée.

$$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

Avec notre exemple :

| Semaine | Erreur² |
|---|---|
| S1 | 2 000² = 4 000 000 |
| S2 | 2 000² = 4 000 000 |
| S3 | 2 000² = 4 000 000 |
| S4 | 1 000² = 1 000 000 |
| S5 | 1 000² = 1 000 000 |

$$\text{RMSE} = \sqrt{\frac{14\,000\,000}{5}} = \sqrt{2\,800\,000} \approx \textbf{1 673 \$}$$

**Lecture :** Similaire au MAE, mais légèrement plus élevé car les grandes erreurs sont pénalisées.  
**Avantage :** Sensible aux erreurs importantes → utile pour détecter les grosses bavures.  
**Différence avec MAE :** Si RMSE >> MAE, il y a probablement quelques semaines avec de très grosses erreurs.

---

### A.4 — WMAE (Weighted MAE) — La métrique officielle Kaggle Walmart 🏆

**Idée :** Même chose que le MAE, mais certaines semaines comptent plus que d'autres.  
Walmart a décidé que les **semaines de fêtes** sont 5× plus importantes à bien prédire (Thanksgiving, Noël, Super Bowl, Pâques).

$$\text{WMAE} = \frac{\sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|}{\sum_{i=1}^{n} w_i}$$

Avec $w_i = 5$ si la semaine est fériée, $w_i = 1$ sinon.

**Exemple concret** (S2 = semaine de Thanksgiving) :

| Semaine | Fête ? | Poids $w_i$ | \|Erreur\| ($) | $w_i \times$ \|Erreur\| |
|---|---|---|---|---|
| S1 | Non | 1 | 2 000 | 2 000 |
| S2 | **Oui** | **5** | 2 000 | **10 000** |
| S3 | Non | 1 | 2 000 | 2 000 |
| S4 | Non | 1 | 1 000 | 1 000 |
| S5 | Non | 1 | 1 000 | 1 000 |

$$\text{WMAE} = \frac{2000 + 10000 + 2000 + 1000 + 1000}{1+5+1+1+1} = \frac{16\,000}{9} \approx \textbf{1 778 \$}$$

> Même erreur de 2 000 $ en S2 (fête) qu'en S1 (normale), mais son impact dans le score est **5× plus grand**.

**Pourquoi ?** Les fêtes représentent seulement 7% des semaines dans les données, mais génèrent un volume de ventes disproportionné. Rater Thanksgiving coûte beaucoup plus cher à Walmart que rater une semaine ordinaire.

### A.5 — Résumé comparatif

| Métrique | Formule (simplifiée) | Sensible aux grosses erreurs ? | Pénalise les fêtes ? | Usage dans ce projet |
|---|---|---|---|---|
| **MAE** | Moyenne des \|erreurs\| | Non | Non | Référence simple, facile à lire |
| **RMSE** | √ Moyenne des erreurs² | ✅ Oui | Non | Détection de prédictions très mauvaises |
| **WMAE** | MAE pondéré × poids fêtes | Non | ✅ Oui (×5) | **Score officiel Kaggle — à optimiser** |

> 🎯 **Conclusion :** On optimise le **WMAE**. On utilise MAE et RMSE comme métriques secondaires pour diagnostiquer le modèle, mais c'est le WMAE qui détermine si on est bon ou pas.


````python

# ── ANNEXE A : Démonstration interactive des métriques ───────────────────────
import numpy as np
import pandas as pd

# Exemple fictif : 5 semaines de prévisions
exemple = pd.DataFrame({
    'Semaine'   : ['S1', 'S2 (Thanksgiving)', 'S3', 'S4', 'S5'],
    'Réel ($)'  : [20000, 15000, 30000, 5000, 10000],
    'Prédit ($)': [18000, 17000, 28000, 6000,  9000],
    'IsHoliday' : [False, True, False, False, False]
})

y_true  = np.array(exemple['Réel ($)'])
y_pred  = np.array(exemple['Prédit ($)'])
weights = np.where(exemple['IsHoliday'], 5, 1)

erreurs_abs = np.abs(y_true - y_pred)

MAE  = np.mean(erreurs_abs)
RMSE = np.sqrt(np.mean((y_true - y_pred)**2))
WMAE = np.sum(weights * erreurs_abs) / np.sum(weights)

exemple['|Erreur| ($)'] = erreurs_abs
exemple['Poids (w)']    = weights
exemple['w × |Erreur|'] = weights * erreurs_abs

print("=" * 60)
print("  TABLEAU DES ERREURS PAR SEMAINE")
print("=" * 60)
print(exemple[['Semaine','Réel ($)','Prédit ($)','IsHoliday',
               '|Erreur| ($)','Poids (w)','w × |Erreur|']].to_string(index=False))

print("\n" + "=" * 60)
print(f"  MAE  (moyenne |erreurs|)        = {MAE:>8,.0f} $")
print(f"  RMSE (racine erreurs carrées)   = {RMSE:>8,.0f} $")
print(f"  WMAE (avec poids fêtes ×5)      = {WMAE:>8,.0f} $")
print("=" * 60)
print(f"\n→ WMAE ({WMAE:,.0f}$) > MAE ({MAE:,.0f}$) à cause du poids ×5 de S2 (Thanksgiving)")
print(f"  Même erreur de {erreurs_abs[1]:,.0f}$ mais poids {weights[1]}× = impact {weights[1]*erreurs_abs[1]:,.0f}$ dans le score")
````
