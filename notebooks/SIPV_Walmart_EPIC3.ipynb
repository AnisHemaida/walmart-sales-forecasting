## <span id="31" style="color:#2E86C1;">3.1 Déploiement du modèle de prévision</span>

Cette section couvre le **déploiement opérationnel** du modèle de prévision des ventes sélectionné lors de l'EPIC 2.

Le modèle champion retenu est **XGBoost (tuné)**, avec une WMAE de **1 300 $** et un R² de **0.9852**, entraîné sur l'historique des ventes Walmart (2010–2012) pour 45 magasins et jusqu'à 81 départements.

Le déploiement repose sur la méthode **pickle**, recommandée pour sa simplicité et sa compatibilité avec Flask. Elle permet de sauvegarder le modèle entraîné ainsi que tous les objets de prétraitement nécessaires, puis de les recharger en production sans ré-entraîner le modèle.

**Objets sauvegardés :**

| Fichier | Contenu |
|---|---|
| `model.pkl` | Modèle XGBoost (tuné) prêt à prédire |
| `columns.pkl` | Liste ordonnée des features (ordre d'entrée exact) |
| `scaler.pkl` | StandardScaler ajusté sur les données d'entraînement |
| `store_mapping.pkl` | Dictionnaire Target Encoding des magasins |
| `dept_mapping.pkl` | Dictionnaire Target Encoding des départements |

> **Note :** Le scaler est conservé pour garantir la cohérence du prétraitement en production, même si XGBoost est invariant à l'échelle. Il sera utilisé par l'API Flask pour transformer les nouvelles données avec les mêmes paramètres (μ, σ) que lors de l'entraînement.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Reconstruction du pipeline & Sauvegarde du modèle (pickle)</strong>
</div>

````python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import pickle
from pathlib import Path
from datetime import timedelta
from sklearn.preprocessing import StandardScaler
import xgboost as xgb

print("=" * 60)
print("  DÉPLOIEMENT — SIPV WALMART")
print("  Reconstruction du pipeline et sauvegarde du modèle")
print("=" * 60)

# ── 1) Chargement des données ──────────────────────────────────
DATA_PATH = Path('../Data')

df_train_raw = pd.read_csv(DATA_PATH / 'train.csv',    parse_dates=['Date'])
df_features  = pd.read_csv(DATA_PATH / 'features.csv', parse_dates=['Date'])
df_stores    = pd.read_csv(DATA_PATH / 'stores.csv')

df = df_train_raw.merge(df_features, on=['Store', 'Date', 'IsHoliday'], how='left')
df = df.merge(df_stores, on='Store', how='left')

print(f"\n  ► Dataset brut : {df.shape[0]:,} lignes × {df.shape[1]} colonnes")

# ── 2) Nettoyage ───────────────────────────────────────────────
for col in ['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']:
    df[col] = df[col].fillna(0)

df = df.sort_values(['Store', 'Date'])
df[['CPI', 'Unemployment']] = df.groupby('Store')[['CPI', 'Unemployment']].transform(
    lambda x: x.interpolate(method='linear').ffill().bfill()
)
df = df[df['Weekly_Sales'] >= 0].copy()

# Conserver les IDs originaux (nécessaires pour les lags et le groupby)
df['Store_orig'] = df['Store']
df['Dept_orig']  = df['Dept']

# ── 3) Features de lag & moyennes mobiles ─────────────────────
df = df.sort_values(['Store_orig', 'Dept_orig', 'Date'])

# Lags : t-1, t-4, t-12, t-52 semaines — par groupe Store × Dept
for lag in [1, 4, 12, 52]:
    df[f'Lag_{lag}'] = (
        df.groupby(['Store_orig', 'Dept_orig'])['Weekly_Sales']
        .shift(lag)
    )
print("  ✅ Lag_1, Lag_4, Lag_12, Lag_52 créés")

# Moyennes mobiles décalées d'1 semaine (évite la fuite de données)
for w in [4, 12]:
    df[f'MA_{w}'] = (
        df.groupby(['Store_orig', 'Dept_orig'])['Weekly_Sales']
        .transform(lambda x: x.rolling(window=w, min_periods=1).mean().shift(1))
    )
print("  ✅ MA_4, MA_12 créées")

# Supprimer les lignes avec NaN (premières semaines sans historique)
lag_cols = ['Lag_1', 'Lag_4', 'Lag_12', 'Lag_52', 'MA_4', 'MA_12']
df = df.dropna(subset=lag_cols).reset_index(drop=True)
print(f"  ✅ Après dropna : {df.shape[0]:,} lignes")

# ── 4) Feature Engineering ─────────────────────────────────────
df['Year']      = df['Date'].dt.year
df['Month']     = df['Date'].dt.month
df['Week']      = df['Date'].dt.isocalendar().week.astype(int)
df['Quarter']   = df['Date'].dt.quarter
df['DayOfYear'] = df['Date'].dt.dayofyear

df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
df['Week_sin']  = np.sin(2 * np.pi * df['Week']  / 52)
df['Week_cos']  = np.cos(2 * np.pi * df['Week']  / 52)

type_map        = {'A': 3, 'B': 2, 'C': 1}
df['Type_ord']  = df['Type'].map(type_map)
df['IsHoliday'] = df['IsHoliday'].astype(int)

df['MarkDown_total'] = df[['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']].sum(axis=1)
df['HasMarkDown']    = (df['MarkDown_total'] > 0).astype(int)

# ── 5) Target Encoding Store & Dept ───────────────────────────
global_mean = df['Weekly_Sales'].mean()
smoothing   = 100

store_stats        = df.groupby('Store')['Weekly_Sales'].agg(['mean', 'count'])
store_stats['enc'] = (store_stats['mean'] * store_stats['count'] + global_mean * smoothing) / \
                     (store_stats['count'] + smoothing)
store_mapping      = store_stats['enc'].to_dict()
df['Store']        = df['Store_orig'].map(store_mapping)

dept_stats        = df.groupby('Dept')['Weekly_Sales'].agg(['mean', 'count'])
dept_stats['enc'] = (dept_stats['mean'] * dept_stats['count'] + global_mean * smoothing) / \
                    (dept_stats['count'] + smoothing)
dept_mapping      = dept_stats['enc'].to_dict()
df['Dept']        = df['Dept_orig'].map(dept_mapping)

print("  ✅ Feature engineering + Target Encoding terminés")

# ── 6) Préparation des features ────────────────────────────────
target       = 'Weekly_Sales'
cols_to_drop = ['Date', 'Store_orig', 'Dept_orig', 'Type']
feature_cols = [c for c in df.columns if c != target and c not in cols_to_drop]

cols_to_scale = [c for c in ['Temperature','Fuel_Price','CPI','Unemployment',
                              'MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5',
                              'MarkDown_total','Size'] if c in feature_cols]
scaler            = StandardScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

print(f"  ► {len(feature_cols)} features")

# ── 7) Split temporel — 12 dernières semaines en test ─────────
date_max   = df['Date'].max()
split_date = date_max - timedelta(weeks=12)

mask_train = df['Date'] <= split_date
mask_test  = df['Date'] >  split_date

X = df[feature_cols]
y = df[target]

X_train, y_train = X[mask_train], y[mask_train]
X_test,  y_test  = X[mask_test],  y[mask_test]

print(f"\n  ► Date de coupure : {split_date.strftime('%Y-%m-%d')}")
print(f"  ► Train : {X_train.shape[0]:,} lignes | Test : {X_test.shape[0]:,} lignes")

# ── 8) Entraînement — XGBoost (tuné) ──────────────────────────
print("\n  ► Entraînement : XGBoost (tuné) ...")

champion_model = xgb.XGBRegressor(
    n_estimators=500, learning_rate=0.05, max_depth=8,
    min_child_weight=5, subsample=0.8, colsample_bytree=0.8,
    reg_alpha=0.1, reg_lambda=1.0,
    random_state=42, n_jobs=-1, verbosity=0
)
champion_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)

y_pred          = champion_model.predict(X_test)
is_holiday_test = df.loc[mask_test, 'IsHoliday'].values
weights         = np.where(is_holiday_test == 1, 5, 1)
wmae = np.sum(weights * np.abs(y_test.values - y_pred)) / np.sum(weights)
r2   = 1 - np.sum((y_test.values - y_pred)**2) / np.sum((y_test.values - y_test.mean())**2)

print(f"  ✅ Champion XGBoost — WMAE : {wmae:,.0f} $ | R² : {r2:.4f}")

# ── 9) Sauvegarde en pickle ────────────────────────────────────
MODELS_PATH = Path('../models')
MODELS_PATH.mkdir(exist_ok=True)

pickle.dump(champion_model, open(MODELS_PATH / 'model.pkl',         'wb'))
pickle.dump(feature_cols,   open(MODELS_PATH / 'columns.pkl',       'wb'))
pickle.dump(scaler,         open(MODELS_PATH / 'scaler.pkl',         'wb'))
pickle.dump(store_mapping,  open(MODELS_PATH / 'store_mapping.pkl',  'wb'))
pickle.dump(dept_mapping,   open(MODELS_PATH / 'dept_mapping.pkl',   'wb'))

print("\n  ✅ Fichiers pickle sauvegardés dans models/")
print("     • model.pkl         → Modèle XGBoost (tuné)")
print("     • columns.pkl       → Liste des features (ordre exact)")
print("     • scaler.pkl        → StandardScaler (paramètres μ, σ)")
print("     • store_mapping.pkl → Target Encoding Store")
print("     • dept_mapping.pkl  → Target Encoding Dept")

df_full = df.copy()
print("\n  ✅ Pipeline complet — prêt")
print("=" * 60)
````

### <span id="311" style="color:#2E86C1;">3.1.1 Choix du mode de restitution (rapport, dashboard, fichier)</span>

Une fois le modèle entraîné et sauvegardé, il faut choisir **comment rendre les prévisions accessibles** aux utilisateurs finaux (gestionnaires de stocks, responsables d'achat, direction).

Trois modes de restitution complémentaires sont retenus pour le SIPV Walmart :

| Mode | Cible | Usage |
|---|---|---|
| **API Flask** | Systèmes d'information (ERP, WMS) | Prédiction à la demande en temps réel |
| **Dashboard Plotly** | Managers & gestionnaires | Visualisation interactive des KPIs |
| **Export CSV** | Analystes & reporting | Archivage et analyse externe (Excel, BI) |

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Présentation des 3 modes de restitution retenus</strong>
</div>

### Mode 1 — API Flask *(Production)*

L'API Flask charge le modèle `model.pkl` une seule fois au démarrage du serveur, puis répond à chaque requête HTTP POST en appliquant le pipeline de prétraitement et en retournant la prévision en temps réel.

**Avantages :**
- Intégration directe avec les systèmes d'information Walmart (ERP, WMS)
- Prédiction en temps réel (< 1 seconde par requête)
- Scalable : peut gérer plusieurs requêtes simultanées

**Fichiers requis :** `model.pkl` · `columns.pkl` · `scaler.pkl` · `store_mapping.pkl` · `dept_mapping.pkl`

---

### Mode 2 — Dashboard Plotly *(Opérationnel)*

Une interface web interactive qui permet aux managers de visualiser les prévisions hebdomadaires, les KPIs, les alertes de rupture de stock et de simuler l'impact des promotions — sans aucune compétence technique requise.

**Avantages :**
- Accessible à tous les profils (achats, logistique, direction)
- Filtres dynamiques par magasin, département et période
- Alertes visuelles colorées selon le niveau de risque (CRITIQUE / ÉLEVÉ / OK)

**Fichiers requis :** `dashboard_walmart.py` · `model.pkl`

---

### Mode 3 — Export CSV *(Analytique)*

Les prévisions sont exportées sous forme de fichiers CSV agrégés par magasin et par département, utilisables directement dans Excel, Power BI ou Tableau.

**Avantages :**
- Compatible avec tous les outils de reporting existants

### <span id="313" style="color:#2E86C1;">3.1.3 Documentation du fonctionnement du modèle</span>

Cette section documente le **fonctionnement interne du modèle de prévision** afin de permettre à tout utilisateur technique de comprendre, reproduire et maintenir le système.

Le modèle retenu est **XGBoost (Extreme Gradient Boosting)**, un algorithme d'apprentissage automatique basé sur l'ensemble de plusieurs arbres de décision entraînés séquentiellement, chacun corrigeant les erreurs du précédent.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Documentation du pipeline et des paramètres du modèle</strong>
</div>

### Pipeline de prévision — Vue d'ensemble

```
Nouvelles données brutes
        ↓
1. Nettoyage (valeurs manquantes, aberrantes)
        ↓
2. Création des features de lag (Lag_1, Lag_4, Lag_12, Lag_52)
   et moyennes mobiles (MA_4, MA_12)
        ↓
3. Feature Engineering (Year, Month, Week, sin/cos, IsHoliday, MarkDown_total...)
        ↓
4. Target Encoding — Store & Dept (store_mapping.pkl / dept_mapping.pkl)
        ↓
5. Normalisation — StandardScaler (scaler.pkl)
        ↓
6. Prédiction — XGBoost champion (model.pkl)
        ↓
Prévision Weekly_Sales ($)
```

---

### Hyperparamètres du modèle XGBoost

| Paramètre | Valeur | Rôle |
|---|---|---|
| `n_estimators` | 500 | Nombre d'arbres de décision |
| `learning_rate` | 0.05 | Vitesse d'apprentissage (faible = plus précis) |
| `max_depth` | 8 | Profondeur maximale de chaque arbre |
| `min_child_weight` | 5 | Nombre minimum d'observations par feuille |
| `subsample` | 0.8 | Proportion des données utilisées par arbre |
| `colsample_bytree` | 0.8 | Proportion des features utilisées par arbre |
| `reg_alpha` | 0.1 | Régularisation L1 (réduit les features inutiles) |
| `reg_lambda` | 1.0 | Régularisation L2 (évite le surapprentissage) |

La documentation du modèle ne se limite pas à ses paramètres techniques. Pour garantir la **confiance des utilisateurs** et répondre aux exigences d'explicabilité, une analyse **SHAP (SHapley Additive exPlanations)** est réalisée.

SHAP permet de mesurer la contribution de chaque feature à chaque prévision individuelle, en s'appuyant sur la théorie des jeux coopératifs. Contrairement à la simple importance des features (qui donne une vue globale), SHAP fournit une explication **locale** (pour chaque prédiction) et **globale** (pour l'ensemble du modèle).

| Visualisation | Type | Description |
|---|---|---|
| **Summary Plot** | Global | Distribution des valeurs SHAP pour chaque feature |
| **Bar Plot** | Global | Importance moyenne absolue de chaque feature |
| **Waterfall Plot** | Local | Explication détaillée d'une prévision individuelle |

### <span id="314" style="color:#2E86C1;">3.1.4 Définition des limites et hypothèses du modèle</span>

Tout modèle de machine learning repose sur des **hypothèses** et présente des **limites** qu'il est essentiel de connaître avant toute utilisation en production. Cette section définit clairement le cadre d'utilisation du modèle XGBoost déployé dans le SIPV Walmart.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Limites et hypothèses du modèle XGBoost</strong>
</div>

### Hypothèses du modèle

| # | Hypothèse | Implication |
|---|---|---|
| 1 | Les patterns historiques (2010–2012) se répètent dans le futur | Le modèle peut perdre en précision lors de changements structurels du marché |
| 2 | Les features de lag sont disponibles au moment de la prédiction | Pour prédire la semaine N, les ventes des semaines N-1, N-4, N-12, N-52 doivent être connues |
| 3 | La relation entre les features et les ventes est stable dans le temps | Une évolution des comportements d'achat peut nécessiter un ré-entraînement |
| 4 | Les données d'entrée respectent le même format que les données d'entraînement | Toute nouvelle donnée doit passer par le même pipeline de prétraitement |
| 5 | Les 45 magasins et leurs départements sont représentatifs | Le modèle n'est pas conçu pour des magasins hors du périmètre d'entraînement |

---

### Limites du modèle

**Limites temporelles**
- Le modèle est entraîné sur des données de **2010 à 2012** uniquement. Il ne tient pas compte des évolutions récentes (inflation, COVID, nouvelles tendances de consommation).
- L'horizon de prévision recommandé est de **1 à 12 semaines**. Au-delà, l'incertitude augmente significativement.

**Limites structurelles**
- Le modèle **ne prédit pas** les ventes de nouveaux magasins ou de nouveaux départements absents des données d'entraînement.
- Les **événements exceptionnels** non représentés dans les données (catastrophes naturelles, pandémies, grèves) ne peuvent pas être anticipés.
- Le modèle **ne gère pas** les ruptures de stock — il prédit les ventes potentielles, pas les ventes contraintes par l'inventaire.
- Le modèle traite chaque **Store × Dept indépendamment** — les effets de corrélation entre magasins ou entre départements ne sont pas capturés.

**Limites techniques**
- Les features de lag nécessitent un **historique minimum de 52 semaines** pour être calculées correctement.
- Le modèle doit être **ré-entraîné périodiquement** (recommandé : tous les 6 mois) pour intégrer les nouvelles données et maintenir sa performance.
- En cas de **dérive des données** (data drift), les performances se dégradent progressivement sans alerte automatique.

**Limites liées aux données**
- Pas de données sur les **prix des concurrents**, les réseaux sociaux ou la météo locale précise — des facteurs pourtant influents sur les ventes.
- Certains magasins ou départements disposent de **moins de données historiques**, ce qui peut réduire la précision des prévisions pour ces entités.

**Limites d'interprétabilité**
- XGBoost est une **"boîte noire"** — ses décisions sont difficiles à expliquer sans outils XAI. Une analyse SHAP (prévue en section 3.1.3) est nécessaire pour interpréter les prévisions individuelles.

---

> **Recommandation :** Ce modèle est conçu comme un **outil d'aide à la décision**, non comme un système de décision automatique. Les prévisions doivent être interprétées par des gestionnaires qui connaissent le contexte métier.

### <span id="322" style="color:#2E86C1;">3.2.2 Générer prévisions par département</span>

Cette sous-section produit des **prévisions détaillées au niveau Store × Département**, enrichies de scores de fiabilité et d'un classement des combinaisons Store × Dept les plus performantes.

Contrairement à la vue agrégée par magasin (3.2.1), ce niveau de granularité permet aux gestionnaires d'identifier précisément **quels départements** au sein de **quels magasins** génèrent les meilleures ventes prévues, et avec quelle fiabilité.

**Score de fiabilité :**

Le score de fiabilité mesure la **précision du modèle** pour chaque combinaison Store × Département. Il est calculé comme suit :

```
Fiabilité (%) = 100 - Erreur relative absolue moyenne (%)
```

où l'erreur relative = |Prédit − Réel| / Réel × 100

- Un score de **100%** signifie que la prévision est parfaite
- Un score de **80%** signifie une erreur moyenne de 20% par rapport aux ventes réelles
- Un score de **0%** signifie que le modèle a très mal prédit — soit parce que les ventes réelles sont très faibles (erreur relative amplifiée), soit parce que l'écart absolu est très grand

> **Note :** Les combinaisons Store × Dept avec des ventes réelles très faibles tendent à avoir un score de fiabilité bas, car une petite erreur absolue se traduit par une grande erreur relative.

| Indicateur | Description |

| `Weekly_Sales_Predicted` | Prévision des ventes par Store × Dept × Semaine |
| `Fiabilite` | Score de fiabilité basé sur l'erreur relative (%) |
| `Rang` | Classement Store × Dept par total des ventes prévues |

````python
import pandas as pd
import numpy as np

print("=" * 60)
print("  3.2.2 — PRÉVISIONS PAR DÉPARTEMENT (Store × Dept)")
print("=" * 60)

# ── 1) Prévisions détaillées Store × Dept ─────────────────────
df_dept = df_previsions.copy()

# Score de fiabilité : 100% - erreur relative absolue moyenne
df_dept['Erreur_relative'] = (
    np.abs(df_dept['Weekly_Sales_Predicted'] - df_dept['Weekly_Sales']) /
    df_dept['Weekly_Sales'].replace(0, np.nan) * 100
).fillna(0).round(2)

# ── 2) Agrégation par Store × Dept ────────────────────────────
df_dept_agg = df_dept.groupby(['Store', 'Dept']).agg(
    Total_Prevu    = ('Weekly_Sales_Predicted', 'sum'),
    Total_Reel     = ('Weekly_Sales',           'sum'),
    Fiabilite_moy  = ('Erreur_relative',        'mean')
).reset_index()

df_dept_agg['Fiabilite']      = (100 - df_dept_agg['Fiabilite_moy']).clip(0, 100).round(1)
df_dept_agg['Total_Prevu']    = df_dept_agg['Total_Prevu'].round(2)
df_dept_agg['Total_Reel']     = df_dept_agg['Total_Reel'].round(2)
df_dept_agg['Difference']     = (df_dept_agg['Total_Prevu'] - df_dept_agg['Total_Reel']).round(2)
df_dept_agg                   = df_dept_agg.sort_values('Total_Prevu', ascending=False).reset_index(drop=True)
df_dept_agg['Rang']           = df_dept_agg.index + 1

print(f"\n  ► {df_dept_agg.shape[0]:,} combinaisons Store × Dept")
print(f"  ► Fiabilité moyenne : {df_dept_agg['Fiabilite'].mean():.1f}%")

# ── 3) Aperçu — Top 15 ────────────────────────────────────────
cols = ['Rang', 'Store', 'Dept', 'Total_Reel', 'Total_Prevu', 'Difference', 'Fiabilite']
fmt  = {'Total_Reel': '{:,.2f}', 'Total_Prevu': '{:,.2f}', 'Difference': '{:,.2f}', 'Fiabilite': '{:.1f}%'}

print("\n  ► Top 15 combinaisons Store × Dept par ventes prévues :")
df_dept_agg[cols].head(15).style.format(fmt)
````

### <span id="331" style="color:#2E86C1;">3.3.1 Calculer niveaux de stock optimaux</span>

Cette section traduit les prévisions hebdomadaires en **décisions concrètes de réapprovisionnement** grâce à deux outils classiques de gestion des stocks :

- **EOQ (Economic Order Quantity)** : quantité optimale à commander pour minimiser les coûts totaux (commande + stockage)
- **Stock de sécurité** : tampon contre la variabilité de la demande pour éviter les ruptures
- **Stock cible** : niveau de stock à maintenir en magasin avant de déclencher une commande

**Hypothèses utilisées :**

| Paramètre | Valeur | Description | Explication |
|-----------|--------|-------------|-------------|
| **S** — coût de commande | 50 $ | Coût fixe par commande passée | Chaque fois qu'un magasin passe une commande à un fournisseur, il paie 50$ indépendamment de la quantité commandée (frais administratifs, transport, traitement). Plus S est élevé, plus l'EOQ est grand → on commande en grosses quantités pour limiter le nombre de commandes. |
| **H** — taux de possession | 20 % | Pourcentage annuel de la valeur du stock | Garder du stock coûte de l'argent : espace, assurance, risque d'invendu. Ici, H = 20% signifie que stocker 1 000$ de marchandise coûte 200$ par an. Plus H est élevé, plus l'EOQ est petit → on préfère commander souvent en petites quantités. |
| **L** — délai de livraison | 1 semaine | Temps entre commande et réception | Quand on passe une commande, il faut 1 semaine avant de recevoir la marchandise. Pendant ce délai, on doit avoir assez de stock pour répondre à la demande. L intervient dans le calcul du stock de sécurité et du stock cible. |
| **Z** — niveau de service | 1,65 | Facteur statistique pour 95% de probabilité sans rupture | Z est tiré de la loi normale. Z=1,65 signifie qu'on accepte un risque de rupture de 5% : dans 95 cas sur 100, le stock sera suffisant. Un Z plus élevé (ex. 2,33 → 99%) donne plus de sécurité mais nécessite plus de stock. |

**Formules :**

**EOQ ($)** = racine( 2 × D × S / H ) — D = demande annuelle estimée (52 × ventes hebdomadaires moyennes prévues)

**Stock de sécurité ($)** = Z × σ_demande × racine(L) — σ_demande = écart-type des ventes hebdomadaires prédites par Store × Dept

**Stock cible ($)** = Stock de sécurité + ( Ventes moyennes hebdomadaires × L ) — représente le niveau de stock à avoir en magasin au moment de passer une commande. Il couvre à la fois les ventes attendues pendant le délai de livraison (L semaine) et le stock tampon en cas de pic de demande.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Calcul EOQ et stock de sécurité par Store × Département</strong>
</div>

````python
import pandas as pd
import numpy as np

print("=" * 60)
print("  3.3.1 — NIVEAUX DE STOCK OPTIMAUX (EOQ + STOCK DE SÉCURITÉ)")
print("=" * 60)

# ── Paramètres ─────────────────────────────────────────────────
S      = 50     # Coût fixe par commande ($)
H_rate = 0.20   # Taux de possession annuel (20%)
L      = 1      # Délai de livraison (semaines)
Z      = 1.65   # Facteur de service 95%

# ── 1) Agrégation des prévisions par Store × Dept ─────────────
df_stock = df_previsions.groupby(['Store', 'Dept']).agg(
    Ventes_moy_hebdo   = ('Weekly_Sales_Predicted', 'mean'),
    Ventes_std_hebdo   = ('Weekly_Sales_Predicted', 'std'),
    Ventes_total_prevu = ('Weekly_Sales_Predicted', 'sum')
).reset_index()

df_stock['Ventes_std_hebdo'] = df_stock['Ventes_std_hebdo'].fillna(0)

# ── 2) Calcul EOQ ──────────────────────────────────────────────
# D = demande annuelle en $ (52 semaines × ventes hebdo moyennes)
# H = H_rate (taux de possession, en % de la valeur — ne dépend pas de D)
# EOQ ($) = sqrt(2 × D × S / H_rate)
df_stock['D_annuel'] = df_stock['Ventes_moy_hebdo'] * 52

df_stock['EOQ'] = np.where(
    df_stock['Ventes_moy_hebdo'] > 0,
    np.sqrt(2 * df_stock['D_annuel'] * S / H_rate),
    0
).round(2)

# ── 3) Calcul Stock de sécurité ────────────────────────────────
# Stock sécurité = Z × σ_demande × sqrt(L)
df_stock['Stock_securite'] = (Z * df_stock['Ventes_std_hebdo'] * np.sqrt(L)).round(2)

# ── 4) Stock cible = Stock sécurité + demande pendant délai ───
df_stock['Stock_cible'] = (
    df_stock['Stock_securite'] + df_stock['Ventes_moy_hebdo'] * L
).round(2)

# ── 5) Mise en forme ───────────────────────────────────────────
df_stock = df_stock.sort_values('Ventes_total_prevu', ascending=False).reset_index(drop=True)

cols_affich = ['Store', 'Dept', 'Ventes_moy_hebdo', 'Ventes_std_hebdo',
               'EOQ', 'Stock_securite', 'Stock_cible']
fmt = {c: '{:,.2f}' for c in cols_affich if c not in ['Store', 'Dept']}

print(f"\n  ► {len(df_stock):,} combinaisons Store × Dept traitées")
print(f"  ► EOQ moyen            : {df_stock['EOQ'].mean():,.0f} $")
print(f"  ► Stock sécurité moyen : {df_stock['Stock_securite'].mean():,.0f} $")
print(f"  ► Stock cible moyen    : {df_stock['Stock_cible'].mean():,.0f} $")

print("\n  ► Aperçu — Top 15 Store × Dept par ventes prévues :")
df_stock[cols_affich].head(15).style.format(fmt)
````

### <span id="332" style="color:#2E86C1;">3.3.2 Générer alertes de rupture</span>

Cette section génère des **alertes automatiques de rupture de stock** pour chaque combinaison Store × Département, en estimant la probabilité que la demande sur les 2 prochaines semaines dépasse le stock disponible.

**Logique de détection :**

On suppose que chaque magasin démarre avec un stock équivalent au **Stock cible** calculé en 3.3.1. La demande sur 2 semaines suit une loi normale dont les paramètres sont dérivés des prévisions du modèle XGBoost :

- Demande moyenne sur 2 semaines = Ventes_moy_hebdo × 2
- Écart-type sur 2 semaines = Ventes_std_hebdo × racine(2)

La probabilité de rupture est calculée comme :

**P(rupture)** = 1 - Phi( (Stock_cible - mu_2) / sigma_2 ) -- où Phi est la fonction de répartition de la loi normale standard.

**Niveaux d alerte :**

| Niveau | Condition | Action recommandée |
|--------|-----------|-------------------|
| CRITIQUE | P(rupture) > 50 % | Commander immédiatement -- rupture très probable |
| HAUTE | 30 % < P(rupture) <= 50 % | Préparer une commande dans les 3 jours |
| MODEREE | 20 % < P(rupture) <= 30 % | Surveiller -- stock sous pression |
| OK | P(rupture) <= 20 % | Stock suffisant -- aucune action requise |

<div style="background:#2a2a3e; border-left:4px solid #f38ba8; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Génération des alertes automatiques de rupture — horizon 2 semaines</strong>
</div>

Pour chaque paire Store × Département, la probabilité de rupture est estimée par une **loi normale** :

- **Demande sur 2 semaines** ~ N(2·μ, √2·σ) — somme de deux semaines indépendantes
- **Stock disponible** ≈ *Stock\_cible* (niveau cible supposé atteint)
- **P(rupture)** = P(Demande\_2sem > Stock\_cible) calculée via la fonction de répartition normale

| Niveau urgence | Seuil | Action suggérée |
|----------------|-------|-----------------|
| 🔴 **CRITIQUE** | P > 50 % | Commande urgente immédiate |
| 🟠 **HAUTE**    | P > 30 % | Commande planifiée sous 48 h |
| 🟡 **MODÉRÉE**  | P > 20 % | Surveillance renforcée |


````python
import numpy as np
import pandas as pd

print("=" * 60)
print("  3.3.2 - ALERTES DE RUPTURE (horizon 2 semaines)")
print("=" * 60)

SEUIL_ALERTE = 0.15

df_alertes = df_stock.copy()

# CV = std / mean
df_alertes["CV"] = (
    df_alertes["Ventes_std_hebdo"] /
    df_alertes["Ventes_moy_hebdo"].replace(0, float("nan"))
).fillna(0).round(4)

df_alertes["Score_risque"] = (df_alertes["CV"] * df_alertes["Ventes_moy_hebdo"]).round(2)

def urgence(cv):
    if cv > 0.40: return "CRITIQUE"
    if cv > 0.25: return "HAUTE"
    if cv > 0.15: return "MODEREE"
    return "OK"

df_alertes["Urgence"] = df_alertes["CV"].apply(urgence)
df_alertes["Label"]   = df_alertes.apply(lambda r: f"S{int(r.Store)}-D{int(r.Dept)}", axis=1)

df_alerte_active = (
    df_alertes[df_alertes["CV"] > SEUIL_ALERTE]
    .sort_values("Score_risque", ascending=False)
    .reset_index(drop=True)
)

n_critique = (df_alerte_active["Urgence"] == "CRITIQUE").sum()
n_haute    = (df_alerte_active["Urgence"] == "HAUTE").sum()
n_moderee  = (df_alerte_active["Urgence"] == "MODEREE").sum()

print("")
print("  Items analyses   :", len(df_alertes))
print("  Alertes actives  :", len(df_alerte_active),
      " (", round(len(df_alerte_active)/len(df_alertes)*100, 1), "% )")
print("  CRITIQUE (CV > 40 %) :", n_critique)
print("  HAUTE    (CV > 25 %) :", n_haute)
print("  MODEREE  (CV > 15 %) :", n_moderee)

top20 = df_alerte_active.head(20)[["Label","Ventes_moy_hebdo","Ventes_std_hebdo","CV","Score_risque","Urgence"]].copy()
top20["CV (%)"]             = (top20["CV"] * 100).round(1).astype(str) + " %"
top20["Ventes moy/sem ($)"] = top20["Ventes_moy_hebdo"].round(0).astype(int)
top20["Ventes std ($)"]     = top20["Ventes_std_hebdo"].round(0).astype(int)

print("")
print("  Top 20 alertes par score de risque :")
print(top20[["Label","Ventes moy/sem ($)","Ventes std ($)","CV (%)","Score_risque","Urgence"]]
      .rename(columns={"Label": "StorexDept"})
      .to_string(index=False))
````

### <span id="343" style="color:#2E86C1;">3.4.3 Estimer ROI et comparer scénarios</span>

Cette sous-section evalue la **rentabilite de la promotion** en calculant le ROI (Return on Investment) et en comparant plusieurs scenarios de montants de MarkDown.

**Formule du ROI :**

**ROI (%) = (Gain de ventes - Cout de la promotion) / Cout de la promotion x 100**

- Si ROI > 0 % : la promotion est **rentable** — le gain de ventes depasse le cout
- Si ROI = 0 % : la promotion est **neutre** — le gain compense exactement le cout
- Si ROI < 0 % : la promotion est **non rentable** — le cout depasse le gain genere

Plusieurs scenarios sont simules automatiquement pour identifier le **montant de promotion optimal** — celui qui maximise le ROI.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Estimation du ROI et comparaison de scenarios de promotion</strong>
</div>

L'interface répond à ce besoin en offrant un point d'accès unique à toutes les informations nécessaires à la prise de décision : prévisions de ventes, alertes de rupture, recommandations de stock et simulation de promotions.
