### <span id="211" style="color:#2E86C1;">2.1.1 Encodage des variables catГ©gorielles</span>

Le dataset issu de l'EPIC1 contient **deux variables catГ©gorielles Г  haute cardinalitГ©** (`Store` : 45, `Dept` : 81) et **deux variables Г  faible cardinalitГ©** (`Type` : 3, `IsHoliday` : 2). Chaque type nГ©cessite une stratГ©gie d'encodage adaptГ©e :

| Variable | Type actuel | ModalitГ©s | StratГ©gie | Justification |
|---|---|---|---|---|
| `Store` | int64 | 45 | **Target Encoding** (lissГ©) | Encode le niveau moyen de ventes par magasin вЂ” Г©vite 45 colonnes One-Hot |
| `Dept` | int64 | 81 | **Target Encoding** (lissГ©) | Encode le niveau moyen de ventes par dГ©partement вЂ” Г©vite 81 colonnes One-Hot |
| `Type` | category | A, B, C | **Ordinal Encoding** (A=1, B=2, C=3) | Variable ordinale liГ©e Г  `Size` (A = grands magasins в†’ C = petits) |
| `IsHoliday` | bool | True / False | Conversion `int` (0/1) | DГ©jГ  binaire |

**Pourquoi le Target Encoding pour Store et Dept ?**
- Le One-Hot crГ©erait **45 + 81 = 126 colonnes** supplГ©mentaires, alourdissant le modГЁle
- Le Target Encoding rГ©sume chaque catГ©gorie par sa **moyenne de `Weekly_Sales`**, capturant le niveau de ventes propre Г  chaque magasin/dГ©partement
- Un **lissage (smoothing)** est appliquГ© pour Г©viter la fuite de donnГ©es (*target leakage*) : les moyennes bruitГ©es des petits groupes sont rГ©gularisГ©es vers la moyenne globale.



````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.1.1 вЂ” Encodage des variables catГ©gorielles
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
import numpy as np

print("=" * 70)
print("  ENCODAGE DES VARIABLES CATГ‰GORIELLES")
print("=" * 70)

# в”Ђв”Ђ 1) Target Encoding вЂ” Store (lissГ© avec smoothing) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 1) Target Encoding : Store в”Ђв”Ђ")

global_mean = df_train['Weekly_Sales'].mean()
smoothing = 100  # facteur de lissage

store_stats = df_train.groupby('Store')['Weekly_Sales'].agg(['mean', 'count'])
store_stats['Store_encoded'] = (
    (store_stats['count'] * store_stats['mean'] + smoothing * global_mean)
    / (store_stats['count'] + smoothing)
)
store_mapping = store_stats['Store_encoded'].to_dict()

# Garder Store original pour rГ©fГ©rence
df_train['Store_orig'] = df_train['Store']
df_train['Store'] = df_train['Store_orig'].map(store_mapping)

print(f"  вњ… Store в†’ Target Encoding (smoothing={smoothing})")
print(f"     вЂў Moyenne globale : {global_mean:,.0f} $")
print(f"     вЂў Plage encodГ©e  : [{df_train['Store'].min():,.0f} вЂ” {df_train['Store'].max():,.0f}] $")

# AperГ§u top 5 / bottom 5
top5 = store_stats.nlargest(5, 'Store_encoded')[['mean', 'count', 'Store_encoded']]
bot5 = store_stats.nsmallest(5, 'Store_encoded')[['mean', 'count', 'Store_encoded']]
print("\n     Top 5 magasins (ventes moyennes les plus Г©levГ©es) :")
for idx, row in top5.iterrows():
    print(f"       Store {idx:>2} в†’ {row['Store_encoded']:>10,.0f} $ (moy brute: {row['mean']:>10,.0f} $, n={row['count']:>5,})")
print("     Bottom 5 magasins :")
for idx, row in bot5.iterrows():
    print(f"       Store {idx:>2} в†’ {row['Store_encoded']:>10,.0f} $ (moy brute: {row['mean']:>10,.0f} $, n={row['count']:>5,})")

# в”Ђв”Ђ 2) Target Encoding вЂ” Dept (lissГ© avec smoothing) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 2) Target Encoding : Dept в”Ђв”Ђ")

dept_stats = df_train.groupby('Dept')['Weekly_Sales'].agg(['mean', 'count'])
dept_stats['Dept_encoded'] = (
    (dept_stats['count'] * dept_stats['mean'] + smoothing * global_mean)
    / (dept_stats['count'] + smoothing)
)
dept_mapping = dept_stats['Dept_encoded'].to_dict()

df_train['Dept_orig'] = df_train['Dept']
df_train['Dept'] = df_train['Dept_orig'].map(dept_mapping)

print(f"  вњ… Dept в†’ Target Encoding (smoothing={smoothing})")
print(f"     вЂў Plage encodГ©e : [{df_train['Dept'].min():,.0f} вЂ” {df_train['Dept'].max():,.0f}] $")
print(f"     вЂў Nb dГ©partements : {len(dept_mapping)}")

# в”Ђв”Ђ 3) Ordinal Encoding вЂ” Type (A, B, C) в†’ liГ© Г  Size в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 3) Ordinal Encoding : Type в”Ђв”Ђ")

type_mapping = {'A': 1, 'B': 2, 'C': 3}
df_train['Type'] = df_train['Type'].map(type_mapping)

print(f"  вњ… Type в†’ Ordinal Encoding (A=1, B=2, C=3)")
print(f"     вЂў Justification : Type est liГ© Г  Size (A = grands magasins в†’ C = petits)")
for label, code in type_mapping.items():
    n = (df_train['Type'] == code).sum()
    print(f"     вЂў Type {label} (={code}) : {n:,} lignes ({n/len(df_train)*100:.1f}%)")

# в”Ђв”Ђ 4) IsHoliday вЂ” conversion bool в†’ int (0/1) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 4) IsHoliday в”Ђв”Ђ")
df_train['IsHoliday'] = df_train['IsHoliday'].astype(int)
n_holiday = df_train['IsHoliday'].sum()
print(f"  вњ… IsHoliday в†’ int (Trueв†’1, Falseв†’0)")
print(f"     вЂў Type aprГЁs conversion : {df_train['IsHoliday'].dtype}")
print(f"     вЂў Semaines normales : {len(df_train) - n_holiday:,}")
print(f"     вЂў Semaines de fГЄte  : {n_holiday:,} ({n_holiday/len(df_train)*100:.1f}%)")

# в”Ђв”Ђ 5) VГ©rification finale в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\n" + "=" * 70)
print("  VГ‰RIFICATION APRГ€S ENCODAGE")
print("=" * 70)

print(f"\n  в–є Shape : {df_train.shape[0]:,} lignes Г— {df_train.shape[1]} colonnes")
print(f"  в–є Colonnes catГ©gorielles restantes : {list(df_train.select_dtypes(include='category').columns)}")

df_types_after = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values],
    "Exemple": [str(df_train[c].iloc[0]) for c in df_train.columns]
})

display(
    df_types_after.style
    .set_properties(**{"text-align": "left", "padding": "8px", "border": "1px solid #ddd"})
    .set_caption("Types des variables aprГЁs encodage")

)

df_train.head()

````

Les 4 transformations appliquГ©es ont converti l'ensemble des variables catГ©gorielles en valeurs numГ©riques exploitables par les algorithmes de machine learning :

Le lissage rГ©gularise les moyennes des groupes Г  faible effectif vers la moyenne globale, limitant ainsi le risque de *target leakage*. - **IsHoliday** a Г©tГ© converti de `bool` en `int64` (0/1), format attendu par les modГЁles.

>  **RГ©sultat** : le dataset ne contient plus aucune colonne catГ©gorielle (`category`). Toutes les variables sont dГ©sormais numГ©riques (`float64`, `int64` ou `int32`), prГЄtes pour les Г©tapes de feature engineering et de modГ©lisation.

### <span id="212" style="color:#2E86C1;">2.1.2 CrГ©ation de variables temporelles et contextuelles</span>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Transformation de la variable Date</strong>
</div>

La colonne `Date` (datetime64) ne peut pas ГЄtre utilisГ©e telle quelle par les algorithmes de machine learning. Elle doit ГЄtre **dГ©composГ©e en composantes temporelles** qui capturent les cycles et tendances des ventes.

| Variable crГ©Г©e | Source | RГґle |
|---|---|---|
| `Year` | `Date.year` | Capturer la tendance annuelle (croissance/dГ©clin) |
| `Month` | `Date.month` | SaisonnalitГ© mensuelle |
| `Week` | `Date.isocalendar().week` | NumГ©ro de semaine ISO вЂ” granularitГ© principale des ventes |
| `Month_sin` / `Month_cos` | sin/cos(2ПЂ Г— Month/12) | Encodage cyclique du mois (dГ©cembre в†” janvier sont proches) |
| `Week_sin` / `Week_cos` | sin/cos(2ПЂ Г— Week/52) | Encodage cyclique de la semaine (sem. 52 в†” sem. 1 sont proches) |

> рџ’Ў **Pourquoi l'encodage cyclique ?** Le mois 12 (dГ©cembre) et le mois 1 (janvier) sont sГ©parГ©s de 11 unitГ©s avec un encodage linГ©aire, alors qu'ils sont **voisins dans le calendrier**. L'encodage sin/cos projette le temps sur un cercle, prГ©servant cette proximitГ© naturelle.

````python

print("=" * 70)
print("  TRANSFORMATION DE LA VARIABLE TEMPORELLE (Date)")
print("=" * 70)

# в”Ђв”Ђ 1) Variables temporelles de base в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 1) Variables temporelles de base в”Ђв”Ђ")

df_train['Year']  = df_train['Date'].dt.year
df_train['Month'] = df_train['Date'].dt.month
df_train['Week']  = df_train['Date'].dt.isocalendar().week.astype(int)

print(f"  вњ… Year  в†’ {df_train['Year'].unique()}")
print(f"  вњ… Month в†’ {sorted(df_train['Month'].unique())}")
print(f"  вњ… Week  в†’ [{df_train['Week'].min()} вЂ” {df_train['Week'].max()}]")

# в”Ђв”Ђ 2) Encodage cyclique (sin/cos) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 2) Encodage cyclique (sin/cos) в”Ђв”Ђ")

df_train['Month_sin'] = np.sin(2 * np.pi * df_train['Month'] / 12)
df_train['Month_cos'] = np.cos(2 * np.pi * df_train['Month'] / 12)
df_train['Week_sin']  = np.sin(2 * np.pi * df_train['Week'] / 52)
df_train['Week_cos']  = np.cos(2 * np.pi * df_train['Week'] / 52)

print(f"  вњ… Month_sin / Month_cos crГ©Г©s")
print(f"  вњ… Week_sin  / Week_cos  crГ©Г©s")
print(f"     вЂў Exemple (semaine 1)  : sin={np.sin(2*np.pi*1/52):.4f}, cos={np.cos(2*np.pi*1/52):.4f}")
print(f"     вЂў Exemple (semaine 52) : sin={np.sin(2*np.pi*52/52):.4f}, cos={np.cos(2*np.pi*52/52):.4f}")
print(f"     в†’ Les semaines 1 et 52 sont proches sur le cercle вњ“")


# в”Ђв”Ђ 3) VГ©rification finale в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ


df_types_fe = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values],
    "Exemple": [str(df_train[c].iloc[0]) for c in df_train.columns]
})

display(
    df_types_fe.style
    .set_properties(**{"text-align": "left", "padding": "8px", "border": "1px solid #ddd"})
    .set_caption("Variables aprГЁs transformation de la variable Date")
)

df_train.head()
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire - RГ©sultats de la transformation de la variable Date :**

La colonne `Date` a Г©tГ© dГ©composГ©e en **7 nouvelles variables** numГ©riques :

- **`Year`**, **`Month`**, **`Week`** : composantes calendaires directes, capturant respectivement la tendance annuelle, la saisonnalitГ© mensuelle et la granularitГ© hebdomadaire des ventes.

- **`Month_sin`/`Month_cos`** et **`Week_sin`/`Week_cos`** : encodage cyclique via sin/cos. Cette transformation projette le temps sur un cercle unitaire, permettant aux modГЁles de comprendre que **dГ©cembre (12) et janvier (1) sont voisins**, et que **la semaine 52 et la semaine 1 sont consГ©cutives**.


</div>



<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Transformation des variables Markdown</strong>
</div>

Les variables `MarkDown1` Г  `MarkDown5` reprГ©sentent les montants de rГ©ductions promotionnelles par type. PlutГґt que de les utiliser uniquement de maniГЁre individuelle, nous crГ©ons **4 variables agrГ©gГ©es** qui synthГ©tisent l'activitГ© promotionnelle globale :

| Variable crГ©Г©e | Calcul | RГґle |
|---|---|---|
| `MarkDown_Total` | Somme de MarkDown1 Г  MarkDown5 | Pression promotionnelle totale (en $) |
| `MarkDown_Density` | MarkDown_Total / Size | Pression promotionnelle **relative** Г  la taille du magasin ($/sqft) |
| `Has_MarkDown` | 1 si MarkDown_Total > 0, sinon 0 | Signal binaire : promo active ou non |
| `Nb_MarkDowns` | Nombre de MarkDowns > 0 | IntensitГ© promotionnelle (combien de types de promos simultanГ©es) |

> рџ’Ў `MarkDown_Density` normalise le budget promo par la surface du magasin : un mГЄme montant de 50 000 $ a un impact trГЁs diffГ©rent dans un grand magasin (Type A) vs un petit (Type C). Les colonnes individuelles `MarkDown1`вЂ“`MarkDown5` sont **conservГ©es** pour distinguer l'effet de chaque type de promotion.

````python

print("=" * 70)
print("  TRANSFORMATION DES VARIABLES MARKDOWN")
print("=" * 70)

md_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

# в”Ђв”Ђ 1) MarkDown_Total вЂ” pression promotionnelle totale в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 1) MarkDown_Total в”Ђв”Ђ")

df_train['MarkDown_Total'] = df_train[md_cols].sum(axis=1)

print(f"  вњ… MarkDown_Total = somme des 5 MarkDowns")
print(f"     вЂў Moyenne : {df_train['MarkDown_Total'].mean():,.0f} $")
print(f"     вЂў MГ©diane : {df_train['MarkDown_Total'].median():,.0f} $")
print(f"     вЂў Max     : {df_train['MarkDown_Total'].max():,.0f} $")
print(f"     вЂў Lignes Г  0 : {(df_train['MarkDown_Total'] == 0).sum():,} ({(df_train['MarkDown_Total'] == 0).mean()*100:.1f}%)")

# в”Ђв”Ђ 2) MarkDown_Density вЂ” pression promo relative Г  la taille du magasin в”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 2) MarkDown_Density в”Ђв”Ђ")

df_train['MarkDown_Density'] = df_train['MarkDown_Total'] / df_train['Size']

print(f"  вњ… MarkDown_Density = MarkDown_Total / Size ($/sqft)")
print(f"     вЂў Moyenne : {df_train['MarkDown_Density'].mean():.6f}")
print(f"     вЂў MГ©diane : {df_train['MarkDown_Density'].median():.6f}")
print(f"     вЂў Max     : {df_train['MarkDown_Density'].max():.6f}")
print(f"     вЂў Lignes Г  0 : {(df_train['MarkDown_Density'] == 0).sum():,} ({(df_train['MarkDown_Density'] == 0).mean()*100:.1f}%)")

# в”Ђв”Ђ 3) Has_MarkDown вЂ” signal binaire promo active в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 3) Has_MarkDown в”Ђв”Ђ")

df_train['Has_MarkDown'] = (df_train['MarkDown_Total'] > 0).astype(int)

n_promo = df_train['Has_MarkDown'].sum()
print(f"  вњ… Has_MarkDown = 1 si au moins un MarkDown > 0")
print(f"     вЂў Semaines avec promo  : {n_promo:,} ({n_promo/len(df_train)*100:.1f}%)")
print(f"     вЂў Semaines sans promo  : {len(df_train) - n_promo:,} ({(len(df_train) - n_promo)/len(df_train)*100:.1f}%)")

# в”Ђв”Ђ 4) Nb_MarkDowns вЂ” nombre de promos actives simultanГ©ment в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 4) Nb_MarkDowns в”Ђв”Ђ")

df_train['Nb_MarkDowns'] = (df_train[md_cols] > 0).sum(axis=1)

print(f"  вњ… Nb_MarkDowns = nombre de MarkDowns > 0 par ligne")
print(f"     вЂў Distribution :")
for val in sorted(df_train['Nb_MarkDowns'].unique()):
    n = (df_train['Nb_MarkDowns'] == val).sum()
    print(f"       {val} promo(s) active(s) : {n:,} lignes ({n/len(df_train)*100:.1f}%)")

# в”Ђв”Ђ 5) VГ©rification в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\n" + "=" * 70)
print("  VГ‰RIFICATION")
print("=" * 70)

new_md_cols = ['MarkDown_Total', 'MarkDown_Density', 'Has_MarkDown', 'Nb_MarkDowns']
print(f"\n  в–є Nouvelles colonnes : {new_md_cols}")
print(f"  в–є MarkDown1-5 individuels : conservГ©s вњ“")
print(f"  в–є Shape : {df_train.shape[0]:,} lignes Г— {df_train.shape[1]} colonnes")

# AperГ§u des lignes avec promo
display(
    df_train[df_train['Has_MarkDown'] == 1][md_cols + new_md_cols].head(5)
    .style
    .set_properties(**{"text-align": "right", "padding": "6px"})
    .set_caption("Exemples de lignes avec MarkDowns actifs")
)
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire вЂ” RГ©sultats de la transformation des MarkDowns :**

Les 5 variables promotionnelles individuelles (`MarkDown1`вЂ“`MarkDown5`) ont Г©tГ© conservГ©es et complГ©tГ©es par **4 variables agrГ©gГ©es** :

- **`MarkDown_Total`** : somme des 5 MarkDowns вЂ” mesure la pression promotionnelle totale en dollars pour chaque semaine/magasin/dГ©partement.

- **`MarkDown_Density`** : ratio `MarkDown_Total / Size` вЂ” normalise le budget promo par la surface du magasin ($/sqft). Un mГЄme investissement promo a un impact plus fort dans un petit magasin que dans un grand. Cette variable est particuliГЁrement utile pour Ridge, qui ne peut pas apprendre seul les interactions entre variables.

- **`Has_MarkDown`** : indicateur binaire (0/1) вЂ” signale si au moins une promotion est active. Permet aux modГЁles de distinguer rapidement les semaines avec/sans promo.

- **`Nb_MarkDowns`** : nombre de types de promos simultanГ©ment actifs (0 Г  5) вЂ” capture l'intensitГ© de la stratГ©gie promotionnelle.

> **RГ©sultat** : le dataset passe de MarkDown1-5 (5 colonnes brutes) Г  **9 colonnes promotionnelles** au total (5 individuelles + 4 agrГ©gГ©es), offrant aux modГЁles Г  la fois le dГ©tail par type de promo et une vue synthГ©tique de l'activitГ© promotionnelle.

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>CrГ©ation des variables de dГ©calage temporel (Lags)</strong>
</div>

Les variables de **dГ©calage temporel (Lags)** permettent au modГЁle d'utiliser les ventes passГ©es comme prГ©dicteurs. L'idГ©e : les ventes d'un magasin/dГ©partement la semaine derniГЁre sont fortement prГ©dictives des ventes de cette semaine.

| Variable crГ©Г©e | DГ©calage | RГґle |
|---|---|---|
| `Lag_1` | t в€’ 1 semaine | Signal le plus rГ©cent вЂ” rГ©action rapide aux changements |
| `Lag_4` | t в€’ 4 semaines | Tendance Г  moyen terme (~1 mois) |
| `Lag_12` | t в€’ 12 semaines | Tendance trimestrielle (~3 mois) |
| `Lag_52` | t в€’ 52 semaines | SaisonnalitГ© annuelle (mГЄme semaine l'an dernier) |

> вљ пёЏ **Calcul par groupe** `(Store_orig, Dept_orig)` : chaque lag est calculГ© au sein de la sГ©rie temporelle d'un magasin/dГ©partement donnГ©, et non sur l'ensemble du dataset. Les premiГЁres lignes de chaque groupe contiendront des `NaN` (pas d'historique disponible) вЂ” ils seront traitГ©s lors du split train/test.

````python

print("=" * 70)
print("  CRГ‰ATION DES VARIABLES DE DГ‰CALAGE TEMPOREL (LAGS)")
print("=" * 70)

# в”Ђв”Ђ Tri chronologique par groupe (indispensable pour les lags) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
df_train = df_train.sort_values(['Store_orig', 'Dept_orig', 'Date']).reset_index(drop=True)
print("\n  вњ… DonnГ©es triГ©es par (Store_orig, Dept_orig, Date)")

# в”Ђв”Ђ DГ©finition des lags Г  crГ©er в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
lag_periods = [1, 4, 12, 52]

print(f"\nв”Ђв”Ђ CrГ©ation des Lags : {lag_periods} в”Ђв”Ђ")

for lag in lag_periods:
    col_name = f'Lag_{lag}'
    df_train[col_name] = (
        df_train
        .groupby(['Store_orig', 'Dept_orig'])['Weekly_Sales']
        .shift(lag)
    )
    n_nan = df_train[col_name].isna().sum()
    n_valid = df_train[col_name].notna().sum()
    print(f"\n  вњ… {col_name} (t в€’ {lag} semaine{'s' if lag > 1 else ''})")
    print(f"     вЂў Valeurs valides : {n_valid:,} ({n_valid/len(df_train)*100:.1f}%)")
    print(f"     вЂў NaN (dГ©but de sГ©rie) : {n_nan:,} ({n_nan/len(df_train)*100:.1f}%)")
    if n_valid > 0:
        print(f"     вЂў Moyenne : {df_train[col_name].mean():,.0f} $")

# в”Ђв”Ђ VГ©rification в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\n" + "=" * 70)
print("  VГ‰RIFICATION")
print("=" * 70)

lag_cols = [f'Lag_{lag}' for lag in lag_periods]
print(f"\n  в–є Nouvelles colonnes : {lag_cols}")
print(f"  в–є Shape : {df_train.shape[0]:,} lignes Г— {df_train.shape[1]} colonnes")

# AperГ§u pour un magasin/dГ©partement
sample = df_train[
    (df_train['Store_orig'] == 1) & (df_train['Dept_orig'] == 1)
][['Store_orig', 'Dept_orig', 'Date', 'Weekly_Sales'] + lag_cols].head(8)

print("\n  в–є Exemple : Store 1, Dept 1 (premiГЁres semaines)")
display(
    sample.style
    .set_properties(**{"text-align": "right", "padding": "6px"})
    .format({col: "{:,.0f}" for col in ['Weekly_Sales'] + lag_cols}, na_rep="NaN")
    .set_caption("Lags вЂ” Store 1, Dept 1")
)
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire вЂ” RГ©sultats de la crГ©ation des Lags :**

Quatre variables de dГ©calage temporel ont Г©tГ© crГ©Г©es, calculГ©es **par groupe** `(Store_orig, Dept_orig)` pour Г©viter toute contamination entre sГ©ries :

- **`Lag_1`** (t в€’ 1 sem.) : signal le plus rГ©cent вЂ” capture les variations hebdomadaires immГ©diates. TrГЁs peu de NaN (1 ligne par groupe).

- **`Lag_4`** (t в€’ 4 sem.) : tendance mensuelle вЂ” lisse les fluctuations hebdomadaires et rГ©vГЁle la dynamique Г  court terme.

- **`Lag_12`** (t в€’ 12 sem.) : tendance trimestrielle вЂ” capture les cycles promotionnels et les variations saisonniГЁres intra-annuelles.

- **`Lag_52`** (t в€’ 52 sem.) : saisonnalitГ© annuelle вЂ” les ventes de la mГЄme semaine l'annГ©e prГ©cГ©dente sont un prГ©dicteur trГЁs puissant en retail (ex: Black Friday, rentrГ©e scolaire).

> **NaN attendus** : les premiГЁres semaines de chaque sГ©rie (Store Г— Dept) n'ont pas d'historique suffisant, gГ©nГ©rant des NaN. Le `Lag_52` produit le plus de NaN (~52 lignes par groupe Г— 3 645 groupes). Ces valeurs manquantes seront gГ©rГ©es lors du split train/test (les lignes sans Lag_52 correspondent Г  la premiГЁre annГ©e de donnГ©es).

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>CrГ©ation des moyennes mobiles (Moving Averages)</strong>
</div>

Les moyennes mobiles capturent la **tendance locale** des ventes en lissant les fluctuations hebdomadaires. Contrairement aux lags (valeurs ponctuelles), elles rГ©sument le comportement rГ©cent sur une fenГЄtre glissante.

> вљ пёЏ Pour Г©viter toute fuite de donnГ©es, chaque moyenne mobile est **dГ©calГ©e d'une semaine** (`shift(1)`) вЂ” la semaine courante n'entre jamais dans le calcul.

| Variable | Description |
|----------|-------------|
| `MA_4` | Moyenne mobile sur **4 semaines** вЂ” tendance court terme |
| `MA_12` | Moyenne mobile sur **12 semaines** вЂ” tendance moyen terme |

Calcul par groupe `(Store Г— Dept)` pour respecter la granularitГ© des sГ©ries.

````python
# в”Ђв”Ђ Moyennes Mobiles (Moving Averages) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
ma_windows = [4, 12]
group = df_train.groupby(['Store_orig', 'Dept_orig'])['Weekly_Sales']

ma_cols = []
for w in ma_windows:
    col = f'MA_{w}'
    # rolling sur la fenГЄtre w, shift(1) pour exclure la semaine courante
    df_train[col] = group.transform(
        lambda x: x.rolling(window=w, min_periods=1).mean().shift(1)
    )
    ma_cols.append(col)

print(f"Colonnes crГ©Г©es : {ma_cols}")
print(f"Shape : {df_train.shape}")
print(f"NaN introduits :\n{df_train[ma_cols].isna().sum()}\n")

# AperГ§u pour Store 1, Dept 1
sample = df_train[
    (df_train['Store_orig'] == 1) & (df_train['Dept_orig'] == 1)
][['Date', 'Weekly_Sales', 'Lag_1'] + ma_cols].head(15)
sample
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**RГ©sultat вЂ” Moyennes Mobiles**

| Г‰lГ©ment | DГ©tail |
|---------|--------|
| **Colonnes crГ©Г©es** | `MA_4` (4 semaines), `MA_12` (12 semaines) |
| **Shape** | 418 727 Г— 35 |
| **NaN introduits** | 3 323 par colonne (1ГЁre semaine de chaque groupe Store Г— Dept) |
| **Protection anti-fuite** | `shift(1)` вЂ” la semaine courante est exclue du calcul |

**ComplГ©mentaritГ© avec les lags :**
- Les **lags** capturent des valeurs ponctuelles passГ©es (t-1, t-4, t-12, t-52)
- Les **moyennes mobiles** rГ©sument la tendance sur une fenГЄtre в†’ plus robuste au bruit

Les NaN seront traitГ©s lors de la prГ©paration finale avant l'entraГ®nement.

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Imputation des valeurs manquantes (Moyenne du groupe)</strong>
</div>

Les Г©tapes prГ©cГ©dentes (Lags, Moyennes Mobiles) ont introduit des **valeurs manquantes (NaN)**, principalement pour les premiГЁres semaines de chaque groupe `Store Г— Dept` oГ№ l'historique est insuffisant. Les variables `MarkDown1вЂ“5` contiennent aussi des NaN pour les pГ©riodes sans promotion.

**StratГ©gie : Moyenne du groupe (`Store_orig Г— Dept_orig`)**

Pour chaque colonne contenant des NaN, on remplace les valeurs manquantes par la **moyenne du groupe** (mГЄme magasin, mГЄme dГ©partement). Cette approche respecte la hiГ©rarchie naturelle des donnГ©es :

$$\text{NaN}_{i,j} \leftarrow \overline{x}_{(\text{Store}=i,\, \text{Dept}=j)}$$

| Avantage | DГ©tail |
|----------|--------|
| **CohГ©rence locale** | Un magasin haut de gamme (Type A) n'est pas diluГ© par les petits magasins |
| **Conservation des lignes** | On Г©vite de supprimer ~160k lignes d'entraГ®nement (38% du dataset) |
| **Robustesse** | Si un groupe n'a aucune valeur valide, on recourt Г  la **moyenne globale** |



````python
# в”Ђв”Ђ Imputation des valeurs manquantes par la moyenne du groupe в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("=" * 70)
print("  IMPUTATION DES VALEURS MANQUANTES вЂ” Moyenne du groupe (Store Г— Dept)")
print("=" * 70)

# Identifier les colonnes avec NaN
cols_with_nan = df_train.columns[df_train.isna().any()].tolist()
print(f"\nв–є Colonnes avec NaN : {cols_with_nan}")
print(f"в–є Nombre total de NaN avant imputation : {df_train[cols_with_nan].isna().sum().sum():,}")

# DГ©tail des NaN par colonne
print("\nв”Ђв”Ђ DГ©tail des NaN par colonne в”Ђв”Ђ")
for col in cols_with_nan:
    n_nan = df_train[col].isna().sum()
    pct = n_nan / len(df_train) * 100
    print(f"  {col:20s} : {n_nan:>7,} NaN ({pct:.1f}%)")

# Imputation par la moyenne du groupe (Store_orig Г— Dept_orig)
print("\nв”Ђв”Ђ Imputation par groupe (Store_orig Г— Dept_orig) в”Ђв”Ђ")
nan_before_total = df_train[cols_with_nan].isna().sum().sum()

for col in cols_with_nan:
    # Moyenne par groupe
    group_mean = df_train.groupby(['Store_orig', 'Dept_orig'])[col].transform('mean')
    # Imputer les NaN avec la moyenne du groupe
    df_train[col] = df_train[col].fillna(group_mean)

# Fallback : si un groupe entier est NaN, on utilise la moyenne globale
nan_after_group = df_train[cols_with_nan].isna().sum().sum()
if nan_after_group > 0:
    print(f"  вљ пёЏ {nan_after_group:,} NaN restants aprГЁs imputation par groupe в†’ fallback moyenne globale")
    for col in cols_with_nan:
        if df_train[col].isna().any():
            global_col_mean = df_train[col].mean()
            n_remaining = df_train[col].isna().sum()
            df_train[col] = df_train[col].fillna(global_col_mean)
            print(f"    {col:20s} : {n_remaining:,} NaN в†’ remplacГ©s par moyenne globale = {global_col_mean:.2f}")

# VГ©rification finale
nan_after_total = df_train[cols_with_nan].isna().sum().sum()
print(f"\nв”Ђв”Ђ RГ©sultat в”Ђв”Ђ")
print(f"  вњ… NaN avant  : {nan_before_total:,}")
print(f"  вњ… NaN aprГЁs  : {nan_after_total}")
print(f"  вњ… Lignes conservГ©es : {len(df_train):,} (aucune suppression)")
print(f"  вњ… Shape : {df_train.shape}")
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**RГ©sultat вЂ” Imputation des valeurs manquantes**

| Г‰lГ©ment | DГ©tail |
|---------|--------|
| **Colonnes imputГ©es** | `Lag_1`, `Lag_4`, `Lag_12`, `Lag_52`, `MA_4`, `MA_12` |
| **NaN avant** | 221 411 (rГ©partis sur 6 colonnes) |
| **NaN aprГЁs** | 0 вњ… |
| **MГ©thode principale** | Moyenne du groupe `Store_orig Г— Dept_orig` |
| **Fallback** | 6 174 NaN в†’ moyenne globale (groupes entiГЁrement manquants) |
| **Lignes conservГ©es** | 418 727 (aucune suppression) |

**Observations :**
- `Lag_52` concentrait **72% des NaN** (159 934) вЂ” les 52 premiГЁres semaines de chaque groupe n'ont pas d'historique Г  1 an
- Les `MarkDown1вЂ“5` n'avaient **aucun NaN** (dГ©jГ  traitГ©s lors de la prГ©paration EPIC1)
- 6 174 NaN (2.8%) ont nГ©cessitГ© le fallback vers la moyenne globale, pour des groupes trГЁs courts sans aucune valeur valide

</div>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Normalisation des variables continues (StandardScaler)</strong>
</div>

Les variables continues du dataset prГ©sentent des **Г©chelles trГЁs diffГ©rentes** (ex: `Size` ~34 000вЂ“220 000 vs `Fuel_Price` ~2.5вЂ“4.5). Cette hГ©tГ©rogГ©nГ©itГ© pose problГЁme pour **Ridge**, qui pГ©nalise les coefficients proportionnellement Г  l'Г©chelle des variables.

**StratГ©gie : `StandardScaler`** (z-score) вЂ” chaque variable est recentrГ©e Г  moyenne 0 et Г©cart-type 1 :

$$z = \frac{x - \mu}{\sigma}$$

| Colonnes normalisГ©es | Raison |
|---|---|
| `Store`, `Dept` | Target-encoded (~1kвЂ“50k $) |
| `Temperature`, `Fuel_Price`, `CPI`, `Unemployment` | Г‰chelles trГЁs diffГ©rentes |
| `Size` | TrГЁs grande Г©chelle (~34kвЂ“220k sqft) |
| `MarkDown1вЂ“5`, `MarkDown_Total`, `MarkDown_Density` | Promos en $ et $/sqft |
| `Month`, `Week`, `Year` | Composantes calendaires |
| `Lag_1/4/12/52`, `MA_4/12` | MГЄme Г©chelle que Weekly_Sales (~-4k Г  +700k) |

| Colonnes **exclues** | Raison |
|---|---|
| `Weekly_Sales` | Variable cible вЂ” ne pas transformer |
| `IsHoliday`, `Has_MarkDown` | Binaires (0/1) вЂ” dГ©jГ  normalisГ©es |
| `Type` | Ordinal (1/2/3) вЂ” plage rГ©duite |
| `Nb_MarkDowns` | Compteur (0вЂ“5) вЂ” plage rГ©duite |
| `Month_sin/cos`, `Week_sin/cos` | Cycliques вЂ” dГ©jГ  dans [-1, 1] |
| `Store_orig`, `Dept_orig`, `Date` | Colonnes de rГ©fГ©rence вЂ” pas des features |

> вљ пёЏ **Note** : les modГЁles tree-based (RF, LightGBM, XGBoost) sont **invariants Г  l'Г©chelle** et n'ont pas besoin de normalisation. Cette Г©tape bГ©nГ©ficie principalement Г  **Ridge** (baseline). On normalise ici sur l'ensemble des donnГ©es avant le split ; en production, le scaler serait ajustГ© uniquement sur le jeu d'entraГ®nement.

````python
# в”Ђв”Ђ Normalisation des variables continues (StandardScaler) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
from sklearn.preprocessing import StandardScaler

print("=" * 70)
print("  NORMALISATION DES VARIABLES CONTINUES")
print("=" * 70)

# в”Ђв”Ђ Colonnes Г  normaliser в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
cols_to_scale = [
    # Target-encoded
    'Store', 'Dept',
    # Variables Г©conomiques et environnementales
    'Temperature', 'Fuel_Price', 'CPI', 'Unemployment',
    # Taille du magasin
    'Size',
    # MarkDowns individuels et agrГ©gГ©s
    'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5',
    'MarkDown_Total', 'MarkDown_Density',
    # Composantes calendaires
    'Year', 'Month', 'Week',
    # Lags et Moyennes Mobiles
    'Lag_1', 'Lag_4', 'Lag_12', 'Lag_52',
    'MA_4', 'MA_12',
]

# в”Ђв”Ђ Colonnes exclues в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
cols_excluded = [
    'Weekly_Sales',                          # cible
    'IsHoliday', 'Has_MarkDown',             # binaires
    'Type', 'Nb_MarkDowns',                  # ordinal / compteur Г  plage rГ©duite
    'Month_sin', 'Month_cos',                # cycliques [-1, 1]
    'Week_sin', 'Week_cos',
    'Store_orig', 'Dept_orig', 'Date',       # rГ©fГ©rence
]

print(f"\n  в–є Colonnes Г  normaliser : {len(cols_to_scale)}")
print(f"  в–є Colonnes exclues      : {len(cols_excluded)}")

# в”Ђв”Ђ AperГ§u AVANT normalisation в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ Statistiques AVANT normalisation (Г©chantillon) в”Ђв”Ђ")
before_stats = df_train[cols_to_scale].describe().loc[['mean', 'std', 'min', 'max']]
display(
    before_stats[['Store', 'Temperature', 'Fuel_Price', 'Size', 'CPI', 'Lag_1']].style
    .format("{:,.2f}")
    .set_caption("Avant normalisation вЂ” Г©chelles trГЁs diffГ©rentes")
)

# в”Ђв”Ђ Application du StandardScaler в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
scaler = StandardScaler()
df_train[cols_to_scale] = scaler.fit_transform(df_train[cols_to_scale])

print("\n  вњ… StandardScaler appliquГ© sur les colonnes continues")
print(f"     вЂў Moyenne aprГЁs normalisation  : ~0")
print(f"     вЂў Г‰cart-type aprГЁs normalisation : ~1")

# в”Ђв”Ђ AperГ§u APRГ€S normalisation в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ Statistiques APRГ€S normalisation (Г©chantillon) в”Ђв”Ђ")
after_stats = df_train[cols_to_scale].describe().loc[['mean', 'std', 'min', 'max']]
display(
    after_stats[['Store', 'Temperature', 'Fuel_Price', 'Size', 'CPI', 'Lag_1']].style
    .format("{:,.4f}")
    .set_caption("AprГЁs normalisation вЂ” toutes les variables sur la mГЄme Г©chelle")
)

# в”Ђв”Ђ VГ©rification finale в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\n" + "=" * 70)
print("  VГ‰RIFICATION")
print("=" * 70)
print(f"\n  в–є Shape : {df_train.shape[0]:,} lignes Г— {df_train.shape[1]} colonnes")
print(f"  в–є Colonnes normalisГ©es : {len(cols_to_scale)}")
print(f"  в–є Scaler sauvegardГ© en mГ©moire pour le dГ©ploiement (section 3.1)")

df_train.head()
````

````python
# в”Ђв”Ђ SГ©lection des colonnes pour la modГ©lisation в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

print("=" * 70)
print("  SГ‰LECTION DES COLONNES POUR LA MODГ‰LISATION")
print("=" * 70)

# в”Ђв”Ђ Variable cible в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
target = 'Weekly_Sales'

# в”Ђв”Ђ Colonnes Г  exclure (rГ©fГ©rence / non-prГ©dictives) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
cols_to_drop = ['Date', 'Store_orig', 'Dept_orig']

# в”Ђв”Ђ Features = toutes les colonnes sauf cible et exclues в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
feature_cols = [c for c in df_train.columns if c != target and c not in cols_to_drop]

print(f"\n  в–є Variable cible     : {target}")
print(f"  в–є Colonnes exclues   : {cols_to_drop}")
print(f"  в–є Nombre de features : {len(feature_cols)}")

# в”Ђв”Ђ Affichage organisГ© par groupe в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
groups = {
    'IdentitГ© magasin': ['Store', 'Dept', 'Type', 'Size'],
    'Г‰conomiques': ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment'],
    'Promotionnelles (individuelles)': ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'],
    'Promotionnelles (agrГ©gГ©es)': ['MarkDown_Total', 'MarkDown_Density', 'Has_MarkDown', 'Nb_MarkDowns'],
    'Temporelles': ['Year', 'Month', 'Week', 'Month_sin', 'Month_cos', 'Week_sin', 'Week_cos'],
    'FГЄte': ['IsHoliday'],
    'Lags': ['Lag_1', 'Lag_4', 'Lag_12', 'Lag_52'],
    'Moyennes mobiles': ['MA_4', 'MA_12'],
}

print(f"\nв”Ђв”Ђ DГ©tail des {len(feature_cols)} features в”Ђв”Ђ\n")
total = 0
for group_name, cols in groups.items():
    present = [c for c in cols if c in feature_cols]
    total += len(present)
    print(f"  {group_name} ({len(present)}) :")
    print(f"    в†’ {present}")

print(f"\n  вњ… Total vГ©rifiГ© : {total} features")

# в”Ђв”Ђ Suppression des colonnes auxiliaires в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\nв”Ђв”Ђ Suppression des colonnes auxiliaires в”Ђв”Ђ")
df_model = df_train.drop(columns=cols_to_drop)
print(f"  вњ… Colonnes supprimГ©es : {cols_to_drop}")
print(f"  в–є Shape df_model : {df_model.shape[0]:,} lignes Г— {df_model.shape[1]} colonnes")
print(f"    ({len(feature_cols)} features + 1 cible)")

df_model.head()
````

### <span id="213" style="color:#2E86C1;">2.1.3 SГ©paration des jeux d'entraГ®nement et de test</span>

La sГ©paration des donnГ©es doit respecter la **chronologie des ventes** pour simuler un scГ©nario rГ©aliste : le modГЁle est entraГ®nГ© sur le passГ© et Г©valuГ© sur le futur.

| Г‰lГ©ment | DГ©tail |
|---|---|
| **MГ©thode** | Split temporel (pas alГ©atoire) |
| **Plage des donnГ©es** | FГ©v. 2010 в†’ Oct. 2012 (~143 semaines) |
| **Seuil de coupure** | DerniГЁres **12 semaines** (~3 mois) rГ©servГ©es au test |
| **Ratio approximatif** | ~92% train / ~8% test |

> вљ пёЏ **Pourquoi un split temporel ?** Un split alГ©atoire mГ©langerait des semaines de 2010 dans le test et des semaines de 2012 dans le train. Le modГЁle "verrait" le futur pendant l'entraГ®nement в†’ **data leakage**. Le split temporel garantit que le train contient uniquement des donnГ©es **antГ©rieures** au test.

````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.1.3 вЂ” SГ©paration temporelle Train / Test
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ

print("=" * 70)
print("  SГ‰PARATION TEMPORELLE TRAIN / TEST")
print("=" * 70)

# в”Ђв”Ђ 1) Identifier la plage temporelle в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 1) Plage temporelle des donnГ©es в”Ђв”Ђ")

date_min = df_train['Date'].min()
date_max = df_train['Date'].max()
n_weeks = df_train['Date'].nunique()

print(f"  в–є DГ©but  : {date_min.strftime('%Y-%m-%d')}")
print(f"  в–є Fin    : {date_max.strftime('%Y-%m-%d')}")
print(f"  в–є Semaines uniques : {n_weeks}")

# в”Ђв”Ђ 2) DГ©finir le seuil de coupure в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 2) Seuil de coupure в”Ђв”Ђ")

# DerniГЁres 12 semaines (~3 mois) pour le test
from datetime import timedelta
split_date = date_max - timedelta(weeks=12)

print(f"  в–є Date de coupure : {split_date.strftime('%Y-%m-%d')}")
print(f"  в–є Train : [{date_min.strftime('%Y-%m-%d')} в†’ {split_date.strftime('%Y-%m-%d')}]")
print(f"  в–є Test  : ({split_date.strftime('%Y-%m-%d')} в†’ {date_max.strftime('%Y-%m-%d')}]")

# в”Ђв”Ђ 3) CrГ©er le masque temporel в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ 3) Application du split в”Ђв”Ђ")

mask_train = df_train['Date'] <= split_date
mask_test  = df_train['Date'] > split_date

# Appliquer sur df_model (mГЄme index que df_train)
X_train = df_model.loc[mask_train, feature_cols]
y_train = df_model.loc[mask_train, target]
X_test  = df_model.loc[mask_test, feature_cols]
y_test  = df_model.loc[mask_test, target]

print(f"  в–є Train : {X_train.shape[0]:,} lignes")
print(f"  в–є Test  : {X_test.shape[0]:,} lignes")


# в”Ђв”Ђ 4) RГ©sumГ© final в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\n" + "=" * 70)
print("  RГ‰SUMГ‰ DU SPLIT")
print("=" * 70)

total = X_train.shape[0] + X_test.shape[0]
print(f"\n  в–є X_train : {X_train.shape[0]:,} lignes Г— {X_train.shape[1]} features ({X_train.shape[0]/total*100:.1f}%)")
print(f"  в–є X_test  : {X_test.shape[0]:,} lignes Г— {X_test.shape[1]} features ({X_test.shape[0]/total*100:.1f}%)")
print(f"  в–є y_train : {y_train.shape[0]:,} valeurs")
print(f"  в–є y_test  : {y_test.shape[0]:,} valeurs")
print(f"  в–є Total   : {total:,} lignes (100% conservГ©)")

# Distribution de la cible
print(f"\n  в–є Weekly_Sales (train) : moy={y_train.mean():,.0f} $ | med={y_train.median():,.0f} $")
print(f"  в–є Weekly_Sales (test)  : moy={y_test.mean():,.0f} $ | med={y_test.median():,.0f} $")
````

## <span id="22" style="color:#2E86C1;">2.2 DГ©veloppement et entraГ®nement des modГЁles de prГ©vision</span>

Cette section porte sur le **dГ©veloppement et l'entraГ®nement des modГЁles de prГ©vision** des ventes hebdomadaires. ГЂ partir des donnГ©es prГ©parГ©es en section 2.1 (`X_train` : 383 872 lignes Г— 31 features), nous entraГ®nons progressivement **4 modГЁles** selon une approche du plus simple au plus complexe :

| # | ModГЁle | Type | RГґle |
|---|--------|------|------|
| 1 | **Ridge** | RГ©gression linГ©aire pГ©nalisГ©e (L2) | **Baseline** вЂ” rГ©fГ©rence minimale Г  battre |
| 2 | **Random Forest** | Ensemble d'arbres (bagging) | PremiГЁre approche non-linГ©aire |
| 3 | **LightGBM** | Gradient Boosting (histograms) | ModГЁle principal вЂ” rapide et performant |
| 4 | **XGBoost** | Gradient Boosting (exact/approx) | Validation croisГ©e вЂ” robustesse |

**MГ©trique d'Г©valuation principale : WMAE** (Weighted Mean Absolute Error)

$$\text{WMAE} = \frac{1}{\sum w_i} \sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|$$

oГ№ $w_i = 5$ si la semaine est un **jour fГ©riГ©** (IsHoliday = 1), sinon $w_i = 1$.

> *В« se tromper pendant une semaine fГ©riГ©e est **5 fois plus grave** que se tromper une semaine normale. В»* 

Cette pondГ©ration reflГЁte l'importance business des pГ©riodes de forte affluence (Thanksgiving, NoГ«l, Super Bowl, Labour Day). Une erreur de prГ©vision durant ces semaines coГ»te beaucoup plus cher : sous-estimation в†’ rupture de stock et ventes perdues, surestimation в†’ surplus et gaspillage.

**DГ©marche pour chaque modГЁle :**
1. EntraГ®nement sur `X_train` / `y_train`
2. PrГ©diction sur `X_test`
3. Calcul de la WMAE et des mГ©triques complГ©mentaires (MAE, RMSE, RВІ)
4. Ajustement des hyperparamГЁtres si nГ©cessaire

### <span id="221" style="color:#2E86C1;">2.2.1 SГ©lection des algorithmes de prГ©vision adaptГ©s</span>

Le choix des algorithmes repose sur une **stratГ©gie progressive** : partir d'un modГЁle simple (baseline) puis monter en complexitГ© pour mesurer le gain rГ©el de chaque approche.

#### 1. Ridge Regression (Baseline)

RГ©gression linГ©aire avec pГ©nalisation **L2** ($\lambda \sum \beta_j^2$). Elle suppose une relation linГ©aire entre les features et les ventes. C'est la **rГ©fГ©rence minimale** : si un modГЁle avancГ© ne bat pas Ridge, il n'apporte aucune valeur.

| Avantage | Limite |
|----------|--------|
| TrГЁs rapide Г  entraГ®ner | Ne capture pas les interactions non-linГ©aires |
| InterprГ©table (coefficients) | Sensible Г  l'Г©chelle в†’ nГ©cessite la normalisation |
| Robuste au surapprentissage | Performance plafonnГ©e sur donnГ©es complexes |

#### 2. Random Forest

Ensemble de **centaines d'arbres de dГ©cision** entraГ®nГ©s sur des sous-Г©chantillons alГ©atoires (bagging). Chaque arbre vote, et la prГ©diction finale est la **moyenne des votes**.

| Avantage | Limite |
|----------|--------|
| Capture les non-linГ©aritГ©s et interactions | Plus lent que Ridge |
| Robuste aux outliers | Moins performant que le boosting en gГ©nГ©ral |
| Pas besoin de normalisation | Peut sur-apprendre si trop d'arbres peu profonds |

#### 3. LightGBM (ModГЁle principal)

Gradient Boosting basГ© sur des **histogrammes** : les arbres sont construits **sГ©quentiellement**, chacun corrigeant les erreurs du prГ©cГ©dent. L'approche par histogrammes rend LightGBM **trГЁs rapide** mГЄme sur de gros volumes.

| Avantage | Limite |
|----------|--------|
| TrГЁs performant sur donnГ©es tabulaires | Risque de surapprentissage si mal rГ©glГ© |
| Rapide (histogrammes + croissance par feuille) | Plus d'hyperparamГЁtres Г  ajuster |
| GГЁre nativement les NaN et les catГ©gorielles | Moins interprГ©table que Ridge |

#### 4. XGBoost (Validation)

Autre implГ©mentation du Gradient Boosting, avec une **rГ©gularisation L1+L2** intГ©grГ©e. UtilisГ© pour **valider** les rГ©sultats de LightGBM et confirmer la robustesse.

| Avantage | Limite |
|----------|--------|
| RГ©gularisation native (L1 + L2) | Plus lent que LightGBM |
| TrГЁs utilisГ© en compГ©titions Kaggle | Consomme plus de mГ©moire |
| ComplГ©mentaire Г  LightGBM | HyperparamГЁtres similaires |

#### RГ©sumГ© de la stratГ©gie

```
Ridge (baseline)  в†’  Random Forest  в†’  LightGBM  в†’  XGBoost
   linГ©aire           non-linГ©aire      boosting      validation
   simple              intermГ©diaire     principal     robustesse
```

> Cette approche permet de **quantifier le gain** Г  chaque Г©tape : si Ridge donne une WMAE de X et LightGBM donne Y, on sait exactement combien la complexitГ© supplГ©mentaire apporte.

````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.2.1 вЂ” SГ©lection des algorithmes & DГ©finition de la mГ©trique WMAE
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("=" * 70)
print("  SГ‰LECTION DES ALGORITHMES & MГ‰TRIQUE WMAE")
print("=" * 70)

# в”Ђв”Ђ 1) Fonction WMAE в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
def wmae(y_true, y_pred, is_holiday):
    """
    Weighted Mean Absolute Error (mГ©trique Walmart/Kaggle).
    Poids : 5 pour les semaines fГ©riГ©es, 1 sinon.
    """
    weights = np.where(is_holiday == 1, 5, 1)
    return np.sum(weights * np.abs(y_true - y_pred)) / np.sum(weights)

print("\nвњ… Fonction wmae(y_true, y_pred, is_holiday) dГ©finie")

# в”Ђв”Ђ 2) RГ©cupГ©rer IsHoliday pour le calcul des poids в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# IsHoliday est dans X_train et X_test (c'est une feature)
is_holiday_train = X_train['IsHoliday'].values
is_holiday_test  = X_test['IsHoliday'].values

n_holiday_train = (is_holiday_train == 1).sum()
n_holiday_test  = (is_holiday_test == 1).sum()

print(f"\nв”Ђв”Ђ IsHoliday dans les donnГ©es в”Ђв”Ђ")
print(f"  в–є Train : {n_holiday_train:,} semaines fГ©riГ©es ({n_holiday_train/len(X_train)*100:.1f}%)")
print(f"  в–є Test  : {n_holiday_test:,} semaines fГ©riГ©es ({n_holiday_test/len(X_test)*100:.1f}%)")

# в”Ђв”Ђ 3) Fonction d'Г©valuation complГЁte в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
def evaluate_model(name, y_true, y_pred, is_holiday):
    """Affiche toutes les mГ©triques pour un modГЁle donnГ©."""
    w = wmae(y_true, y_pred, is_holiday)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    
    print(f"\n{'в”Ђ' * 50}")
    print(f"  {name}")
    print(f"{'в”Ђ' * 50}")
    print(f"  WMAE  : {w:,.0f} $")
    print(f"  MAE   : {mae:,.0f} $")
    print(f"  RMSE  : {rmse:,.0f} $")
    print(f"  RВІ    : {r2:.4f}")
    
    return {'model': name, 'WMAE': w, 'MAE': mae, 'RMSE': rmse, 'R2': r2}

print("вњ… Fonction evaluate_model(name, y_true, y_pred, is_holiday) dГ©finie")

# в”Ђв”Ђ 4) Liste des modГЁles Г  entraГ®ner в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\nв”Ђв”Ђ ModГЁles sГ©lectionnГ©s в”Ђв”Ђ")
models = {
    'Ridge': Ridge(alpha=1.0),
    'Random Forest': RandomForestRegressor(
        n_estimators=200, max_depth=20, min_samples_leaf=5,
        n_jobs=-1, random_state=42
    ),
}

# в”Ђв”Ђ DГ©tection CUDA (XGBoost uniquement) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import subprocess as _sp
try:
    _r = _sp.run(['nvidia-smi'], capture_output=True, text=True, timeout=5)
    _cuda = _r.returncode == 0
except Exception:
    _cuda = False
if _cuda:
    print("  вњ… GPU NVIDIA dГ©tectГ© в†’ CUDA activГ© pour XGBoost")
else:
    print("  в„№пёЏ  Pas de GPU в†’ entraГ®nement sur CPU")

# LightGBM et XGBoost вЂ” import conditionnel
try:
    import lightgbm as lgb
    models['LightGBM'] = lgb.LGBMRegressor(
        n_estimators=500, max_depth=8, learning_rate=0.05,
        num_leaves=63, min_child_samples=50,
        subsample=0.8, colsample_bytree=0.8,
        n_jobs=-1, random_state=42, verbose=-1
    )
    print("  вњ… LightGBM disponible")
except ImportError:
    print("  вљ пёЏ LightGBM non installГ© вЂ” pip install lightgbm")

try:
    import xgboost as xgb
    _xgb_gpu = {'device': 'cuda', 'tree_method': 'hist'} if _cuda else {}
    models['XGBoost'] = xgb.XGBRegressor(
        n_estimators=500, max_depth=8, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        reg_alpha=0.1, reg_lambda=1.0,
        n_jobs=-1, random_state=42, verbosity=0,
        **_xgb_gpu
    )
    print("  вњ… XGBoost disponible")
except ImportError:
    print("  вљ пёЏ XGBoost non installГ© вЂ” pip install xgboost")

print(f"\nв–є {len(models)} modГЁles prГЄts : {list(models.keys())}")

# Stockage des rГ©sultats
results = []
trained_models = {}
````

**Commentaire вЂ” SГ©lection des algorithmes et mГ©trique WMAE :**

Cette cellule met en place l'infrastructure d'Г©valuation et les 4 modГЁles retenus pour la prГ©vision des ventes :

- **MГ©trique WMAE** (*Weighted Mean Absolute Error*) : Les erreurs commises durant les **semaines fГ©riГ©es** sont pГ©nalisГ©es **5 fois plus** que les semaines normales, reflГ©tant l'impact financier disproportionnГ© d'une mauvaise prГ©vision pendant ces pГ©riodes de forte affluence. La formule est :
$$WMAE = \frac{\sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|}{\sum_{i=1}^{n} w_i} \quad \text{avec } w_i = \begin{cases} 5 & \text{si semaine fГ©riГ©e} \\ 1 & \text{sinon} \end{cases}$$

- **4 modГЁles sГ©lectionnГ©s**, du plus simple au plus complexe :
  1. **Ridge** (rГ©gression linГ©aire rГ©gularisГ©e L2) вЂ” sert de *baseline* pour mesurer l'apport des modГЁles non linГ©aires.
  2. **Random Forest** (200 arbres, profondeur max 20) вЂ” modГЁle d'ensemble robuste au bruit et aux outliers.
  3. **LightGBM** (500 itГ©rations, lr=0.05) вЂ” gradient boosting optimisГ© pour la vitesse, idГ©al pour les grands volumes de donnГ©es.
  4. **XGBoost** (500 itГ©rations, lr=0.05, rГ©gularisation L1/L2) вЂ” gradient boosting avec rГ©gularisation intГ©grГ©e pour limiter le surapprentissage.

- **Fonction `evaluate_model()`** : Г©value chaque modГЁle avec 4 mГ©triques complГ©mentaires вЂ” WMAE (mГ©trique principale), MAE, RMSE et RВІ.

> **RГ©sultat** : les 4 modГЁles sont initialisГ©s et prГЄts pour l'entraГ®nement.

### <span id="222" style="color:#2E86C1;">2.2.2 EntraГ®nement des modГЁles sur les donnГ©es historiques</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-bottom:10px; color:#D5D8DC;">

**Objectif** вЂ” EntraГ®ner les 4 modГЁles sГ©lectionnГ©s (Ridge, Random Forest, LightGBM, XGBoost) sur le jeu d'entraГ®nement (`X_train`, `y_train`), puis gГ©nГ©rer les prГ©dictions sur le jeu de test (`X_test`) et Г©valuer chaque modГЁle avec l'ensemble des mГ©triques (WMAE, MAE, RMSE, RВІ).

**Approche :**
- Chaque modГЁle est entraГ®nГ© sur les **383 872 observations** du jeu d'entraГ®nement (avant le 2012-08-03).
- Les prГ©dictions sont ensuite gГ©nГ©rГ©es sur les **34 855 observations** du jeu de test (12 derniГЁres semaines).
- Le temps d'entraГ®nement et de prГ©diction est mesurГ© pour chaque modГЁle afin de comparer leur efficacitГ© computationnelle.
- Un **tableau comparatif** est produit, triГ© par WMAE croissant, pour identifier le modГЁle le plus performant.

</div>

````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.2.2 вЂ” EntraГ®nement des modГЁles & Г‰valuation initiale
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
import time
import pandas as pd

print("=" * 70)
print("  ENTRAГЋNEMENT DES MODГ€LES & Г‰VALUATION INITIALE")
print("=" * 70)

# в”Ђв”Ђ Conversion de Type (category в†’ int) pour XGBoost в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
if X_train['Type'].dtype.name == 'category':
    X_train['Type'] = X_train['Type'].astype(int)
    X_test['Type']  = X_test['Type'].astype(int)
    print("  в„№пёЏ  Colonne 'Type' convertie de category в†’ int\n")

results = []
trained_models = {}

for name, model in models.items():
    print(f"\n{'в”Ѓ' * 60}")
    print(f"  в–є EntraГ®nement : {name}")
    print(f"{'в”Ѓ' * 60}")
    
    # в”Ђв”Ђ EntraГ®nement в”Ђв”Ђ
    t0 = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - t0
    print(f"  вњ… EntraГ®nГ© en {train_time:.1f} s")
    
    # в”Ђв”Ђ PrГ©diction в”Ђв”Ђ
    t0 = time.time()
    y_pred = model.predict(X_test)
    pred_time = time.time() - t0
    print(f"  вњ… PrГ©diction en {pred_time:.3f} s")
    
    # в”Ђв”Ђ Г‰valuation в”Ђв”Ђ
    res = evaluate_model(name, y_test.values, y_pred, is_holiday_test)
    res['Train_Time_s'] = round(train_time, 1)
    res['Pred_Time_s'] = round(pred_time, 3)
    results.append(res)
    
    # Stocker le modГЁle entraГ®nГ©
    trained_models[name] = model

# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# Tableau comparatif
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
print(f"\n\n{'=' * 70}")
print("  TABLEAU COMPARATIF DES MODГ€LES")
print(f"{'=' * 70}\n")

df_results = pd.DataFrame(results).sort_values('WMAE')
df_results.index = range(1, len(df_results) + 1)
df_results.index.name = 'Rang'

# Formater l'affichage
df_display = df_results.copy()
df_display['WMAE'] = df_display['WMAE'].apply(lambda x: f"{x:,.0f} $")
df_display['MAE']  = df_display['MAE'].apply(lambda x: f"{x:,.0f} $")
df_display['RMSE'] = df_display['RMSE'].apply(lambda x: f"{x:,.0f} $")
df_display['R2']   = df_display['R2'].apply(lambda x: f"{x:.4f}")
df_display['Train_Time_s'] = df_display['Train_Time_s'].apply(lambda x: f"{x:.1f} s")
df_display['Pred_Time_s']  = df_display['Pred_Time_s'].apply(lambda x: f"{x:.3f} s")

df_display.rename(columns={
    'model': 'ModГЁle', 'Train_Time_s': 'Temps entr.',
    'Pred_Time_s': 'Temps prГ©d.'
}, inplace=True)

display(df_display)

# в”Ђв”Ђ Meilleur modГЁle в”Ђв”Ђ
best = df_results.iloc[0]
print(f"\nрџЏ† Meilleur modГЁle (WMAE le plus bas) : {best['model']}")
print(f"   WMAE = {best['WMAE']:,.0f} $  |  RВІ = {best['R2']:.4f}")
````

### <span id="223" style="color:#2E86C1;">2.2.3 Ajustement des hyperparamГЁtres</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-bottom:10px; color:#D5D8DC;">

**Objectif** вЂ” Affiner les hyperparamГЁtres des **2 meilleurs modГЁles** (XGBoost et LightGBM) afin de rГ©duire davantage le WMAE par rapport aux rГ©sultats initiaux (1 343 $ et 1 378 $).

**Approche вЂ” `RandomizedSearchCV`** :
- **Pourquoi RandomizedSearchCV ?** Avec 383 872 lignes et de nombreux hyperparamГЁtres, un `GridSearchCV` exhaustif serait prohibitif. `RandomizedSearchCV` Г©chantillonne alГ©atoirement `n_iter` combinaisons dans l'espace de recherche, offrant un bon compromis entre exploration et temps de calcul.
- **Validation croisГ©e** : 3 folds (compromis entre robustesse statistique et temps d'exГ©cution).
- **MГ©trique d'optimisation** : `neg_mean_absolute_error` (proxy de la WMAE вЂ” la pondГ©ration par IsHoliday sera vГ©rifiГ©e aprГЁs avec la vraie WMAE sur le jeu de test).
- **Espace de recherche** : les hyperparamГЁtres les plus influents pour chaque algorithme (profondeur, taux d'apprentissage, rГ©gularisation, sous-Г©chantillonnage).

</div>

````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.2.3 вЂ” Ajustement des hyperparamГЁtres (XGBoost & LightGBM)
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
import time, warnings, subprocess
warnings.filterwarnings('ignore')

print("=" * 70)
print("  AJUSTEMENT DES HYPERPARAMГ€TRES вЂ” RandomizedSearchCV + TimeSeriesSplit")
print("=" * 70)

# в”Ђв”Ђ DГ©tection CUDA (XGBoost uniquement) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
try:
    _res = subprocess.run(['nvidia-smi'], capture_output=True, text=True, timeout=5)
    cuda_available = _res.returncode == 0
except Exception:
    cuda_available = False

if cuda_available:
    print("  вњ… GPU NVIDIA dГ©tectГ© в†’ CUDA activГ© pour XGBoost")
    xgb_device_params = {'device': 'cuda', 'tree_method': 'hist'}
else:
    print("  в„№пёЏ  Aucun GPU dГ©tectГ© в†’ entraГ®nement sur CPU")
    xgb_device_params = {}

print("  в„№пёЏ  LightGBM в†’ CPU (version pip sans support GPU)")

# в”Ђв”Ђ RГ©sultats avant tuning (section 2.2.2) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
baseline_wmae = {
    'XGBoost':  df_results.loc[df_results['model'] == 'XGBoost', 'WMAE'].values[0],
    'LightGBM': df_results.loc[df_results['model'] == 'LightGBM', 'WMAE'].values[0],
}

# в”Ђв”Ђ Tri de X_train par Date pour TimeSeriesSplit в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# Copie locale uniquement вЂ” X_train global inchangГ© pour les cellules suivantes
_sort_idx  = df_train.loc[X_train.index, 'Date'].sort_values().index
X_train_ts = X_train.loc[_sort_idx]
y_train_ts = y_train.loc[_sort_idx]
print(f"\n  вњ… X_train_ts triГ© par Date ({len(X_train_ts):,} lignes)")

tscv = TimeSeriesSplit(n_splits=3)
print(f"  вњ… TimeSeriesSplit(n_splits=3)\n")

# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 1) XGBoost вЂ” RandomizedSearchCV + TimeSeriesSplit
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
print(f"{'в”Ѓ' * 60}")
print(f"  в–є Tuning : XGBoost  (20 tirages Г— 3 folds)")
print(f"{'в”Ѓ' * 60}")

import xgboost as xgb

xgb_param_dist = {
    'n_estimators':     [300, 500, 700, 1000],
    'max_depth':        [6, 8, 10, 12],
    'learning_rate':    [0.01, 0.03, 0.05, 0.08, 0.1],
    'subsample':        [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'reg_alpha':        [0, 0.01, 0.1, 0.5],
    'reg_lambda':       [0.5, 1.0, 2.0, 5.0],
    'min_child_weight': [3, 5, 10, 20],
}

xgb_base = xgb.XGBRegressor(
    n_jobs=-1, random_state=42, verbosity=0,
    **xgb_device_params
)

t0 = time.time()
xgb_search = RandomizedSearchCV(
    xgb_base,
    param_distributions=xgb_param_dist,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=tscv,
    random_state=42,
    n_jobs=1,
    verbose=0,
)
xgb_search.fit(X_train_ts, y_train_ts)
xgb_time = time.time() - t0

print(f"  вњ… TerminГ© en {xgb_time/60:.1f} min ({int(xgb_time)} s)")
print(f"  в–є Meilleur score CV (neg_MAE) : {xgb_search.best_score_:,.0f}")
print(f"\n  Meilleurs hyperparamГЁtres :")
for k, v in xgb_search.best_params_.items():
    print(f"    {k:20s} = {v}")

y_pred_xgb = xgb_search.best_estimator_.predict(X_test)
res_xgb = evaluate_model('XGBoost (tunГ©)', y_test.values, y_pred_xgb, is_holiday_test)

# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2) LightGBM вЂ” RandomizedSearchCV + TimeSeriesSplit   (CPU uniquement)
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
print(f"\n{'в”Ѓ' * 60}")
print(f"  в–є Tuning : LightGBM  (20 tirages Г— 3 folds, CPU)")
print(f"{'в”Ѓ' * 60}")

import lightgbm as lgb

lgb_param_dist = {
    'n_estimators':      [300, 500, 700, 1000],
    'max_depth':         [6, 8, 10, 12, -1],
    'learning_rate':     [0.01, 0.03, 0.05, 0.08, 0.1],
    'num_leaves':        [31, 63, 127, 255],
    'min_child_samples': [20, 50, 100],
    'subsample':         [0.7, 0.8, 0.9],
    'colsample_bytree':  [0.7, 0.8, 0.9],
    'reg_alpha':         [0, 0.01, 0.1, 0.5],
    'reg_lambda':        [0, 0.1, 1.0, 5.0],
}

lgb_base = lgb.LGBMRegressor(
    n_jobs=-1, random_state=42, verbose=-1
)

t0 = time.time()
lgb_search = RandomizedSearchCV(
    lgb_base,
    param_distributions=lgb_param_dist,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=tscv,
    random_state=42,
    n_jobs=1,
    verbose=0,
)
lgb_search.fit(X_train_ts, y_train_ts)
lgb_time = time.time() - t0

print(f"  вњ… TerminГ© en {lgb_time/60:.1f} min ({int(lgb_time)} s)")
print(f"  в–є Meilleur score CV (neg_MAE) : {lgb_search.best_score_:,.0f}")
print(f"\n  Meilleurs hyperparamГЁtres :")
for k, v in lgb_search.best_params_.items():
    print(f"    {k:20s} = {v}")

y_pred_lgb = lgb_search.best_estimator_.predict(X_test)
res_lgb = evaluate_model('LightGBM (tunГ©)', y_test.values, y_pred_lgb, is_holiday_test)

# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 3) Tableau comparatif : Avant vs AprГЁs tuning
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
print(f"\n\n{'=' * 70}")
print("  COMPARAISON AVANT / APRГ€S TUNING")
print(f"{'=' * 70}\n")

comparison = []
for name_key, res_tuned, y_pred_tuned in [
    ('XGBoost', res_xgb, y_pred_xgb),
    ('LightGBM', res_lgb, y_pred_lgb),
]:
    before = baseline_wmae[name_key]
    after  = res_tuned['WMAE']
    delta  = after - before
    pct_change = (delta / before) * 100
    comparison.append({
        'ModГЁle': name_key,
        'WMAE avant': f"{before:,.0f} $",
        'WMAE aprГЁs': f"{after:,.0f} $",
        'О” WMAE': f"{delta:+,.0f} $",
        'Variation': f"{pct_change:+.1f} %",
        'RВІ aprГЁs': f"{res_tuned['R2']:.4f}",
    })

df_comparison = pd.DataFrame(comparison)
df_comparison.index = range(1, len(df_comparison) + 1)
display(df_comparison)

# в”Ђв”Ђ Mettre Г  jour les modГЁles entraГ®nГ©s в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
trained_models['XGBoost']  = xgb_search.best_estimator_
trained_models['LightGBM'] = lgb_search.best_estimator_

best_tuned = min(
    [('XGBoost', res_xgb), ('LightGBM', res_lgb)],
    key=lambda x: x[1]['WMAE']
)
print(f"\nрџЏ† Champion aprГЁs tuning : {best_tuned[0]}")
print(f"   WMAE = {best_tuned[1]['WMAE']:,.0f} $  |  RВІ = {best_tuned[1]['R2']:.4f}")
````

````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.2.4 вЂ” GГ©nГ©ration des prГ©visions (modГЁle champion : XGBoost tunГ©)
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
import numpy as np
import pandas as pd

print("=" * 70)
print("  GГ‰NГ‰RATION DES PRГ‰VISIONS вЂ” XGBoost (tunГ©)")
print("=" * 70)

# в”Ђв”Ђ 1) ModГЁle champion в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
champion_name = best_tuned[0]
champion_model = trained_models[champion_name]
print(f"\n  в–є ModГЁle champion : {champion_name}")
print(f"  в–є WMAE (test)     : {best_tuned[1]['WMAE']:,.0f} $")
print(f"  в–є RВІ (test)       : {best_tuned[1]['R2']:.4f}")

# в”Ђв”Ђ 2) GГ©nГ©rer les prГ©dictions finales в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\nв”Ђв”Ђ PrГ©dictions sur l'ensemble de test ({X_test.shape[0]:,} lignes) в”Ђв”Ђ")

y_pred_final = champion_model.predict(X_test)
print(f"  вњ… {len(y_pred_final):,} prГ©visions gГ©nГ©rГ©es")

# в”Ђв”Ђ 3) Construire le DataFrame de prГ©visions в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print("\nв”Ђв”Ђ Construction du DataFrame de prГ©visions в”Ђв”Ђ")

df_predictions = pd.DataFrame({
    'Store':          df_train.loc[mask_test, 'Store_orig'].values,
    'Dept':           df_train.loc[mask_test, 'Dept_orig'].values,
    'Date':           df_train.loc[mask_test, 'Date'].values,
    'IsHoliday':      is_holiday_test,
    'Reel':           y_test.values,
    'Prevu':          y_pred_final,
})

# Colonnes dГ©rivГ©es
df_predictions['Erreur']     = df_predictions['Prevu'] - df_predictions['Reel']
df_predictions['Erreur_Abs'] = df_predictions['Erreur'].abs()
df_predictions['Erreur_Pct'] = (df_predictions['Erreur_Abs'] / df_predictions['Reel'].abs().replace(0, np.nan) * 100).round(2)

print(f"  вњ… DataFrame : {df_predictions.shape[0]:,} lignes Г— {df_predictions.shape[1]} colonnes")
print(f"  в–є Colonnes   : {list(df_predictions.columns)}")

# в”Ђв”Ђ 4) Statistiques descriptives des prГ©visions в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\n{'=' * 70}")
print("  STATISTIQUES DES PRГ‰VISIONS")
print(f"{'=' * 70}\n")

print(f"  в–є Ventes rГ©elles (test) :")
print(f"    Moyenne      : {df_predictions['Reel'].mean():>12,.0f} $")
print(f"    MГ©diane      : {df_predictions['Reel'].median():>12,.0f} $")
print(f"    Min          : {df_predictions['Reel'].min():>12,.0f} $")
print(f"    Max          : {df_predictions['Reel'].max():>12,.0f} $")

print(f"\n  в–є Ventes prГ©vues :")
print(f"    Moyenne      : {df_predictions['Prevu'].mean():>12,.0f} $")
print(f"    MГ©diane      : {df_predictions['Prevu'].median():>12,.0f} $")
print(f"    Min          : {df_predictions['Prevu'].min():>12,.0f} $")
print(f"    Max          : {df_predictions['Prevu'].max():>12,.0f} $")

print(f"\n  в–є Erreurs absolues :")
print(f"    Moyenne (MAE): {df_predictions['Erreur_Abs'].mean():>12,.0f} $")
print(f"    MГ©diane      : {df_predictions['Erreur_Abs'].median():>12,.0f} $")
print(f"    90e centile  : {df_predictions['Erreur_Abs'].quantile(0.90):>12,.0f} $")
print(f"    95e centile  : {df_predictions['Erreur_Abs'].quantile(0.95):>12,.0f} $")
print(f"    Max          : {df_predictions['Erreur_Abs'].max():>12,.0f} $")

# Erreur mГ©diane en %
median_pct = df_predictions['Erreur_Pct'].median()
print(f"\n  в–є Erreur % mГ©diane : {median_pct:.1f} %")

# в”Ђв”Ђ 5) RГ©partition des erreurs в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\n{'=' * 70}")
print("  RГ‰PARTITION DES ERREURS")
print(f"{'=' * 70}\n")

seuils = [500, 1000, 2000, 5000]
for s in seuils:
    n = (df_predictions['Erreur_Abs'] <= s).sum()
    pct = n / len(df_predictions) * 100
    print(f"  в–є Erreur в‰¤ {s:>5,} $ : {n:>6,} lignes ({pct:5.1f} %)")

# в”Ђв”Ђ 6) PrГ©visions fГ©riГ©s vs normaux в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\n{'=' * 70}")
print("  PRГ‰VISIONS : JOURS FГ‰RIГ‰S vs NORMAUX")
print(f"{'=' * 70}\n")

for label, mask_h in [("Normaux (IsHoliday=0)", df_predictions['IsHoliday'] == 0),
                       ("FГ©riГ©s  (IsHoliday=1)", df_predictions['IsHoliday'] == 1)]:
    sub = df_predictions[mask_h]
    if len(sub) > 0:
        mae_h = sub['Erreur_Abs'].mean()
        med_h = sub['Erreur_Abs'].median()
        print(f"  в–є {label} : {len(sub):,} lignes")
        print(f"    MAE     = {mae_h:,.0f} $")
        print(f"    MГ©diane = {med_h:,.0f} $\n")

# в”Ђв”Ђ 7) AperГ§u : meilleures et pires prГ©visions в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\n{'=' * 70}")
print("  TOP 10 вЂ” MEILLEURES PRГ‰VISIONS (erreur la plus faible)")
print(f"{'=' * 70}\n")

top10 = df_predictions.nsmallest(10, 'Erreur_Abs')[
    ['Store', 'Dept', 'Date', 'Reel', 'Prevu', 'Erreur', 'Erreur_Pct']
].copy()
top10['Reel']   = top10['Reel'].apply(lambda x: f"{x:,.0f} $")
top10['Prevu']  = top10['Prevu'].apply(lambda x: f"{x:,.0f} $")
top10['Erreur'] = top10['Erreur'].apply(lambda x: f"{x:+,.0f} $")
top10['Erreur_Pct'] = top10['Erreur_Pct'].apply(lambda x: f"{x:.2f} %")
top10.index = range(1, 11)
display(top10)

print(f"\n{'=' * 70}")
print("  TOP 10 вЂ” PIRES PRГ‰VISIONS (erreur la plus Г©levГ©e)")
print(f"{'=' * 70}\n")

worst10 = df_predictions.nlargest(10, 'Erreur_Abs')[
    ['Store', 'Dept', 'Date', 'Reel', 'Prevu', 'Erreur', 'Erreur_Pct']
].copy()
worst10['Reel']   = worst10['Reel'].apply(lambda x: f"{x:,.0f} $")
worst10['Prevu']  = worst10['Prevu'].apply(lambda x: f"{x:,.0f} $")
worst10['Erreur'] = worst10['Erreur'].apply(lambda x: f"{x:+,.0f} $")
worst10['Erreur_Pct'] = worst10['Erreur_Pct'].apply(lambda x: f"{x:.2f} %")
worst10.index = range(1, 11)
display(worst10)

print(f"\n{'=' * 70}")
print(f"  вњ… DataFrame 'df_predictions' prГЄt ({df_predictions.shape[0]:,} lignes)")
print(f"{'=' * 70}")
````

### <span id="231" style="color:#2E86C1;">2.3.1 Choix des mГ©triques d'Г©valuation</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-top:12px; color:#D5D8DC;">

**Objectif :** DГ©finir et justifier les mГ©triques utilisГ©es pour Г©valuer les modГЁles de prГ©vision des ventes. Le choix de la **WMAE** comme mГ©trique principale est dictГ© par le concours Kaggle Walmart, mais nous calculons Г©galement des mГ©triques complГ©mentaires pour une vision complГЁte.

**MГ©triques retenues :**
- **WMAE** (Weighted MAE) вЂ” mГ©trique principale, pondГЁre Г—5 les semaines fГ©riГ©es
- **MAE** (Mean Absolute Error) вЂ” erreur moyenne en dollars, interprГ©table directement
- **RMSE** (Root Mean Squared Error) вЂ” pГ©nalise les grosses erreurs
- **RВІ** (Coefficient de dГ©termination) вЂ” variance expliquГ©e par le modГЁle
- **MAPE** (Mean Absolute Percentage Error) вЂ” erreur relative en %

</div>

````python
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
# 2.3.1 вЂ” Choix des mГ©triques d'Г©valuation
# в•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђв•ђ
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("=" * 70)
print("  MГ‰TRIQUES D'Г‰VALUATION RETENUES")
print("=" * 70)

# в”Ђв”Ђ 1) Rappel des formules в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

# в”Ђв”Ђ 2) Calcul complet pour le champion (XGBoost tunГ©) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"в”Ђв”Ђ MГ©triques du champion : {champion_name} (tunГ©) в”Ђв”Ђ\n")

y_true = y_test.values
y_pred_ch = y_pred_final

# WMAE
w_test = np.where(is_holiday_test == 1, 5, 1)
val_wmae = np.sum(w_test * np.abs(y_true - y_pred_ch)) / np.sum(w_test)

# MAE
val_mae = mean_absolute_error(y_true, y_pred_ch)

# RMSE
val_rmse = np.sqrt(mean_squared_error(y_true, y_pred_ch))

# RВІ
val_r2 = r2_score(y_true, y_pred_ch)

# MAPE (excluant les ventes nulles/nГ©gatives pour Г©viter division par 0)
mask_valid = np.abs(y_true) > 0
val_mape = np.mean(np.abs(y_true[mask_valid] - y_pred_ch[mask_valid]) / np.abs(y_true[mask_valid])) * 100
n_excluded = (~mask_valid).sum()

print(f"  в–є WMAE  = {val_wmae:>10,.0f} $   в†ђ mГ©trique principale (Kaggle)")
print(f"  в–є MAE   = {val_mae:>10,.0f} $   в†ђ erreur absolue moyenne")
print(f"  в–є RMSE  = {val_rmse:>10,.0f} $   в†ђ sensible aux outliers")
print(f"  в–є RВІ    = {val_r2:>10.4f}     в†ђ variance expliquГ©e")
print(f"  в–є MAPE  = {val_mape:>10.2f} %   в†ђ erreur relative moyenne")
if n_excluded > 0:
    print(f"    (MAPE calculГ© sur {mask_valid.sum():,} lignes, {n_excluded:,} exclues car ventes = 0)")

# в”Ђв”Ђ 3) Tableau rГ©capitulatif formatГ© в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
print(f"\n{'=' * 70}")
print("  TABLEAU RГ‰CAPITULATIF вЂ” CHAMPION")
print(f"{'=' * 70}\n")

df_metrics = pd.DataFrame([{
    'MГ©trique': 'WMAE (pondГ©rГ©e)',
    'Valeur': f"{val_wmae:,.0f} $",
    'RГґle': 'MГ©trique principale вЂ” pondГЁre x5 les fГ©riГ©s',
}, {
    'MГ©trique': 'MAE',
    'Valeur': f"{val_mae:,.0f} $",
    'RГґle': 'Erreur moyenne en dollars',
}, {
    'MГ©trique': 'RMSE',
    'Valeur': f"{val_rmse:,.0f} $",
    'RГґle': 'PГ©nalise les grandes erreurs',
}, {
    'MГ©trique': 'RВІ',
    'Valeur': f"{val_r2:.4f}",
    'RГґle': f'{val_r2*100:.2f}% de la variance expliquГ©e',
}, {
    'MГ©trique': 'MAPE',
    'Valeur': f"{val_mape:.2f} %",
    'RГґle': 'Erreur relative вЂ” interprГ©table par le mГ©tier',
}])
df_metrics.index = range(1, len(df_metrics) + 1)
display(df_metrics)

````

**Commentaire вЂ” MГ©triques d'Г©valuation :**

| MГ©trique | Valeur | InterprГ©tation |
|:---|:---|:---|
| **WMAE** | **1 332 $** | En moyenne, l'erreur pondГ©rГ©e est de 1 332 $ par magasin/dГ©partement/semaine |
| **MAE** | **1 263 $** | Sans pondГ©ration fГ©riГ©e, l'erreur descend вЂ” les fГ©riГ©s sont plus difficiles Г  prГ©dire |
| **RMSE** | **2 801 $** | Le ratio RMSE/MAE в‰€ 2.2 indique la prГ©sence de quelques erreurs importantes |
| **RВІ** | **0.9837** | Le modГЁle explique **98.4%** de la variance des ventes вЂ” excellente qualitГ© |
| **MAPE** | **477 %** | вљ пёЏ Valeur gonflГ©e par les dГ©partements Г  ventes quasi-nulles (division par ~0) |

**Observations :**
- Le **MAPE est non pertinent** ici en raison des ventes proches de 0 dans certains dГ©partements (ex: rayons saisonniers). C'est un phГ©nomГЁne classique вЂ” la MAPE n'est pas adaptГ©e aux datasets avec des valeurs cibles proches de zГ©ro.
