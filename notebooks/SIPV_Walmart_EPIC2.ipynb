## <span id="21" style="color:#2E86C1;">2.1 Préparation des données pour la modélisation</span>

Avant d'entraîner tout modèle de machine learning, les données brutes issues de l'étape `Compréhension et Préparation des Données` doivent être **transformées et structurées** de façon à être exploitables par les algorithmes de prévision. Cette phase de préparation est déterminante : la qualité des features construites conditionne directement la précision des prédictions.

Cette section couvre trois étapes structurantes :

| Étape | Contenu |
|---|---|
| **2.1.1 Encodage** | Transformation des variables catégorielles (`Store`, `Dept`, `Type`, `IsHoliday`) en valeurs numériques exploitables |
| **2.1.2 Feature engineering** | Extraction de variables temporelles (`Year`, `Month`, `Week`), indicateurs de saisonnalité et variables de contexte promotionnel |
| **2.1.3 Split train/test** | Séparation temporelle des données en jeu d'entraînement et jeu de validation, en respectant la chronologie des ventes |


> ⚠️ **Principe fondamental** : le split train/test doit être **temporel** et non aléatoire. Utiliser un split aléatoire entraînerait une **fuite de données** (*data leakage*) — le modèle verrait des semaines futures lors de l'entraînement, ce qui fausserait l'évaluation des performances.


````python
# ──────────────────────────────────────────────────────────────────
# IMPORT — Chargement de df_train_model depuis EPIC1
# ──────────────────────────────────────────────────────────────────
import pandas as pd

df_train = pd.read_parquet("../Data/prepared/df_train.parquet")

print(f"✅ df_train chargé depuis EPIC1")

df_types = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values]
})

display(
    df_types.style
    .set_properties(**{
        "text-align": "left",
        "padding": "8px",
        "border": "1px solid #ddd"
    })
    .set_caption("Types des variables du dataset")
)

df_train.head()





````

### <span id="211" style="color:#2E86C1;">2.1.1 Encodage des variables catégorielles</span>

Le dataset issu de l'EPIC1 contient **deux variables catégorielles à haute cardinalité** (`Store` : 45, `Dept` : 81) et **deux variables à faible cardinalité** (`Type` : 3, `IsHoliday` : 2). Chaque type nécessite une stratégie d'encodage adaptée :

| Variable | Type actuel | Modalités | Stratégie | Justification |
|---|---|---|---|---|
| `Store` | int64 | 45 | **Target Encoding** (lissé) | Encode le niveau moyen de ventes par magasin — évite 45 colonnes One-Hot |
| `Dept` | int64 | 81 | **Target Encoding** (lissé) | Encode le niveau moyen de ventes par département — évite 81 colonnes One-Hot |
| `Type` | category | A, B, C | **Ordinal Encoding** (A=1, B=2, C=3) | Variable ordinale liée à `Size` (A = grands magasins → C = petits) |
| `IsHoliday` | bool | True / False | Conversion `int` (0/1) | Déjà binaire |

**Pourquoi le Target Encoding pour Store et Dept ?**
- Le One-Hot créerait **45 + 81 = 126 colonnes** supplémentaires, alourdissant le modèle
- Le Target Encoding résume chaque catégorie par sa **moyenne de `Weekly_Sales`**, capturant le niveau de ventes propre à chaque magasin/département
- Un **lissage (smoothing)** est appliqué pour éviter la fuite de données (*target leakage*) : les moyennes bruitées des petits groupes sont régularisées vers la moyenne globale.



````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.1.1 — Encodage des variables catégorielles
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np

print("=" * 70)
print("  ENCODAGE DES VARIABLES CATÉGORIELLES")
print("=" * 70)

# ── 1) Target Encoding — Store (lissé avec smoothing) ────────────────────────
print("\n── 1) Target Encoding : Store ──")

global_mean = df_train['Weekly_Sales'].mean()
smoothing = 100  # facteur de lissage

store_stats = df_train.groupby('Store')['Weekly_Sales'].agg(['mean', 'count'])
store_stats['Store_encoded'] = (
    (store_stats['count'] * store_stats['mean'] + smoothing * global_mean)
    / (store_stats['count'] + smoothing)
)
store_mapping = store_stats['Store_encoded'].to_dict()

# Garder Store original pour référence
df_train['Store_orig'] = df_train['Store']
df_train['Store'] = df_train['Store_orig'].map(store_mapping)

print(f"  ✅ Store → Target Encoding (smoothing={smoothing})")
print(f"     • Moyenne globale : {global_mean:,.0f} $")
print(f"     • Plage encodée  : [{df_train['Store'].min():,.0f} — {df_train['Store'].max():,.0f}] $")

# Aperçu top 5 / bottom 5
top5 = store_stats.nlargest(5, 'Store_encoded')[['mean', 'count', 'Store_encoded']]
bot5 = store_stats.nsmallest(5, 'Store_encoded')[['mean', 'count', 'Store_encoded']]
print("\n     Top 5 magasins (ventes moyennes les plus élevées) :")
for idx, row in top5.iterrows():
    print(f"       Store {idx:>2} → {row['Store_encoded']:>10,.0f} $ (moy brute: {row['mean']:>10,.0f} $, n={row['count']:>5,})")
print("     Bottom 5 magasins :")
for idx, row in bot5.iterrows():
    print(f"       Store {idx:>2} → {row['Store_encoded']:>10,.0f} $ (moy brute: {row['mean']:>10,.0f} $, n={row['count']:>5,})")

# ── 2) Target Encoding — Dept (lissé avec smoothing) ─────────────────────────
print("\n── 2) Target Encoding : Dept ──")

dept_stats = df_train.groupby('Dept')['Weekly_Sales'].agg(['mean', 'count'])
dept_stats['Dept_encoded'] = (
    (dept_stats['count'] * dept_stats['mean'] + smoothing * global_mean)
    / (dept_stats['count'] + smoothing)
)
dept_mapping = dept_stats['Dept_encoded'].to_dict()

df_train['Dept_orig'] = df_train['Dept']
df_train['Dept'] = df_train['Dept_orig'].map(dept_mapping)

print(f"  ✅ Dept → Target Encoding (smoothing={smoothing})")
print(f"     • Plage encodée : [{df_train['Dept'].min():,.0f} — {df_train['Dept'].max():,.0f}] $")
print(f"     • Nb départements : {len(dept_mapping)}")

# ── 3) Ordinal Encoding — Type (A, B, C) → lié à Size ────────────────────────
print("\n── 3) Ordinal Encoding : Type ──")

type_mapping = {'A': 1, 'B': 2, 'C': 3}
df_train['Type'] = df_train['Type'].map(type_mapping)

print(f"  ✅ Type → Ordinal Encoding (A=1, B=2, C=3)")
print(f"     • Justification : Type est lié à Size (A = grands magasins → C = petits)")
for label, code in type_mapping.items():
    n = (df_train['Type'] == code).sum()
    print(f"     • Type {label} (={code}) : {n:,} lignes ({n/len(df_train)*100:.1f}%)")

# ── 4) IsHoliday — conversion bool → int (0/1) ──────────────────────────────
print("\n── 4) IsHoliday ──")
df_train['IsHoliday'] = df_train['IsHoliday'].astype(int)
n_holiday = df_train['IsHoliday'].sum()
print(f"  ✅ IsHoliday → int (True→1, False→0)")
print(f"     • Type après conversion : {df_train['IsHoliday'].dtype}")
print(f"     • Semaines normales : {len(df_train) - n_holiday:,}")
print(f"     • Semaines de fête  : {n_holiday:,} ({n_holiday/len(df_train)*100:.1f}%)")

# ── 5) Vérification finale ───────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  VÉRIFICATION APRÈS ENCODAGE")
print("=" * 70)

print(f"\n  ► Shape : {df_train.shape[0]:,} lignes × {df_train.shape[1]} colonnes")
print(f"  ► Colonnes catégorielles restantes : {list(df_train.select_dtypes(include='category').columns)}")

df_types_after = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values],
    "Exemple": [str(df_train[c].iloc[0]) for c in df_train.columns]
})

display(
    df_types_after.style
    .set_properties(**{"text-align": "left", "padding": "8px", "border": "1px solid #ddd"})
    .set_caption("Types des variables après encodage")

)

df_train.head()

````

**Commentaire — Résultats de l'encodage :**

Les 4 transformations appliquées ont converti l'ensemble des variables catégorielles en valeurs numériques exploitables par les algorithmes de machine learning :
- **Store** et **Dept** ont été encodés via un **Target Encoding lissé** (smoothing = 100). Le lissage régularise les moyennes des groupes à faible effectif vers la moyenne globale, limitant ainsi le risque de *target leakage*. Les colonnes `Store_orig` et `Dept_orig` sont conservées pour référence.
- **Type** a été converti en **Ordinal Encoding** (A=1, B=2, C=3), reflétant le lien ordinal avec la taille des magasins (`Size`) : les magasins de Type A sont les plus grands, ceux de Type C les plus petits.

- **IsHoliday** a été converti de `bool` en `int64` (0/1), format attendu par les modèles.

>  **Résultat** : le dataset ne contient plus aucune colonne catégorielle (`category`). Toutes les variables sont désormais numériques (`float64`, `int64` ou `int32`), prêtes pour les étapes de feature engineering et de modélisation.

### <span id="212" style="color:#2E86C1;">2.1.2 Création de variables temporelles et contextuelles</span>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Transformation de la variable Date</strong>
</div>

La colonne `Date` (datetime64) ne peut pas être utilisée telle quelle par les algorithmes de machine learning. Elle doit être **décomposée en composantes temporelles** qui capturent les cycles et tendances des ventes.

| Variable créée | Source | Rôle |
|---|---|---|
| `Year` | `Date.year` | Capturer la tendance annuelle (croissance/déclin) |
| `Month` | `Date.month` | Saisonnalité mensuelle |
| `Week` | `Date.isocalendar().week` | Numéro de semaine ISO — granularité principale des ventes |
| `Month_sin` / `Month_cos` | sin/cos(2π × Month/12) | Encodage cyclique du mois (décembre ↔ janvier sont proches) |
| `Week_sin` / `Week_cos` | sin/cos(2π × Week/52) | Encodage cyclique de la semaine (sem. 52 ↔ sem. 1 sont proches) |

> 💡 **Pourquoi l'encodage cyclique ?** Le mois 12 (décembre) et le mois 1 (janvier) sont séparés de 11 unités avec un encodage linéaire, alors qu'ils sont **voisins dans le calendrier**. L'encodage sin/cos projette le temps sur un cercle, préservant cette proximité naturelle.

````python

print("=" * 70)
print("  TRANSFORMATION DE LA VARIABLE TEMPORELLE (Date)")
print("=" * 70)

# ── 1) Variables temporelles de base ─────────────────────────────────────────
print("\n── 1) Variables temporelles de base ──")

df_train['Year']  = df_train['Date'].dt.year
df_train['Month'] = df_train['Date'].dt.month
df_train['Week']  = df_train['Date'].dt.isocalendar().week.astype(int)

print(f"  ✅ Year  → {df_train['Year'].unique()}")
print(f"  ✅ Month → {sorted(df_train['Month'].unique())}")
print(f"  ✅ Week  → [{df_train['Week'].min()} — {df_train['Week'].max()}]")

# ── 2) Encodage cyclique (sin/cos) ──────────────────────────────────────────
print("\n── 2) Encodage cyclique (sin/cos) ──")

df_train['Month_sin'] = np.sin(2 * np.pi * df_train['Month'] / 12)
df_train['Month_cos'] = np.cos(2 * np.pi * df_train['Month'] / 12)
df_train['Week_sin']  = np.sin(2 * np.pi * df_train['Week'] / 52)
df_train['Week_cos']  = np.cos(2 * np.pi * df_train['Week'] / 52)

print(f"  ✅ Month_sin / Month_cos créés")
print(f"  ✅ Week_sin  / Week_cos  créés")
print(f"     • Exemple (semaine 1)  : sin={np.sin(2*np.pi*1/52):.4f}, cos={np.cos(2*np.pi*1/52):.4f}")
print(f"     • Exemple (semaine 52) : sin={np.sin(2*np.pi*52/52):.4f}, cos={np.cos(2*np.pi*52/52):.4f}")
print(f"     → Les semaines 1 et 52 sont proches sur le cercle ✓")


# ── 3) Vérification finale ───────────────────────────────────────────────────


df_types_fe = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values],
    "Exemple": [str(df_train[c].iloc[0]) for c in df_train.columns]
})

display(
    df_types_fe.style
    .set_properties(**{"text-align": "left", "padding": "8px", "border": "1px solid #ddd"})
    .set_caption("Variables après transformation de la variable Date")
)

df_train.head()
````

**Commentaire - Résultats de la transformation de la variable Date :**

La colonne `Date` a été décomposée en **7 nouvelles variables** numériques :

- **`Year`**, **`Month`**, **`Week`** : composantes calendaires directes, capturant respectivement la tendance annuelle, la saisonnalité mensuelle et la granularité hebdomadaire des ventes.

- **`Month_sin`/`Month_cos`** et **`Week_sin`/`Week_cos`** : encodage cyclique via sin/cos. Cette transformation projette le temps sur un cercle unitaire, permettant aux modèles de comprendre que **décembre (12) et janvier (1) sont voisins**, et que **la semaine 52 et la semaine 1 sont consécutives**.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Transformation des variables Markdown</strong>
</div>

Les variables `MarkDown1` à `MarkDown5` représentent les montants de réductions promotionnelles par type. Plutôt que de les utiliser uniquement de manière individuelle, nous créons **4 variables agrégées** qui synthétisent l'activité promotionnelle globale :

| Variable créée | Calcul | Rôle |
|---|---|---|
| `MarkDown_Total` | Somme de MarkDown1 à MarkDown5 | Pression promotionnelle totale (en $) |
| `MarkDown_Density` | MarkDown_Total / Size | Pression promotionnelle **relative** à la taille du magasin ($/sqft) |
| `Has_MarkDown` | 1 si MarkDown_Total > 0, sinon 0 | Signal binaire : promo active ou non |
| `Nb_MarkDowns` | Nombre de MarkDowns > 0 | Intensité promotionnelle (combien de types de promos simultanées) |

> 💡 `MarkDown_Density` normalise le budget promo par la surface du magasin : un même montant de 50 000 $ a un impact très différent dans un grand magasin (Type A) vs un petit (Type C). Les colonnes individuelles `MarkDown1`–`MarkDown5` sont **conservées** pour distinguer l'effet de chaque type de promotion.

````python

print("=" * 70)
print("  TRANSFORMATION DES VARIABLES MARKDOWN")
print("=" * 70)

md_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

# ── 1) MarkDown_Total — pression promotionnelle totale ───────────────────────
print("\n── 1) MarkDown_Total ──")

df_train['MarkDown_Total'] = df_train[md_cols].sum(axis=1)

print(f"  ✅ MarkDown_Total = somme des 5 MarkDowns")
print(f"     • Moyenne : {df_train['MarkDown_Total'].mean():,.0f} $")
print(f"     • Médiane : {df_train['MarkDown_Total'].median():,.0f} $")
print(f"     • Max     : {df_train['MarkDown_Total'].max():,.0f} $")
print(f"     • Lignes à 0 : {(df_train['MarkDown_Total'] == 0).sum():,} ({(df_train['MarkDown_Total'] == 0).mean()*100:.1f}%)")

# ── 2) MarkDown_Density — pression promo relative à la taille du magasin ─────
print("\n── 2) MarkDown_Density ──")

df_train['MarkDown_Density'] = df_train['MarkDown_Total'] / df_train['Size']

print(f"  ✅ MarkDown_Density = MarkDown_Total / Size ($/sqft)")
print(f"     • Moyenne : {df_train['MarkDown_Density'].mean():.6f}")
print(f"     • Médiane : {df_train['MarkDown_Density'].median():.6f}")
print(f"     • Max     : {df_train['MarkDown_Density'].max():.6f}")
print(f"     • Lignes à 0 : {(df_train['MarkDown_Density'] == 0).sum():,} ({(df_train['MarkDown_Density'] == 0).mean()*100:.1f}%)")

# ── 3) Has_MarkDown — signal binaire promo active ───────────────────────────
print("\n── 3) Has_MarkDown ──")

df_train['Has_MarkDown'] = (df_train['MarkDown_Total'] > 0).astype(int)

n_promo = df_train['Has_MarkDown'].sum()
print(f"  ✅ Has_MarkDown = 1 si au moins un MarkDown > 0")
print(f"     • Semaines avec promo  : {n_promo:,} ({n_promo/len(df_train)*100:.1f}%)")
print(f"     • Semaines sans promo  : {len(df_train) - n_promo:,} ({(len(df_train) - n_promo)/len(df_train)*100:.1f}%)")

# ── 4) Nb_MarkDowns — nombre de promos actives simultanément ─────────────────
print("\n── 4) Nb_MarkDowns ──")

df_train['Nb_MarkDowns'] = (df_train[md_cols] > 0).sum(axis=1)

print(f"  ✅ Nb_MarkDowns = nombre de MarkDowns > 0 par ligne")
print(f"     • Distribution :")
for val in sorted(df_train['Nb_MarkDowns'].unique()):
    n = (df_train['Nb_MarkDowns'] == val).sum()
    print(f"       {val} promo(s) active(s) : {n:,} lignes ({n/len(df_train)*100:.1f}%)")

# ── 5) Vérification ──────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  VÉRIFICATION")
print("=" * 70)

new_md_cols = ['MarkDown_Total', 'MarkDown_Density', 'Has_MarkDown', 'Nb_MarkDowns']
print(f"\n  ► Nouvelles colonnes : {new_md_cols}")
print(f"  ► MarkDown1-5 individuels : conservés ✓")
print(f"  ► Shape : {df_train.shape[0]:,} lignes × {df_train.shape[1]} colonnes")

# Aperçu des lignes avec promo
display(
    df_train[df_train['Has_MarkDown'] == 1][md_cols + new_md_cols].head(5)
    .style
    .set_properties(**{"text-align": "right", "padding": "6px"})
    .set_caption("Exemples de lignes avec MarkDowns actifs")
)
````

**Commentaire — Résultats de la transformation des MarkDowns :**

Les 5 variables promotionnelles individuelles (`MarkDown1`–`MarkDown5`) ont été conservées et complétées par **4 variables agrégées** :

- **`MarkDown_Total`** : somme des 5 MarkDowns — mesure la pression promotionnelle totale en dollars pour chaque semaine/magasin/département.

- **`MarkDown_Density`** : ratio `MarkDown_Total / Size` — normalise le budget promo par la surface du magasin ($/sqft). Un même investissement promo a un impact plus fort dans un petit magasin que dans un grand. Cette variable est particulièrement utile pour Ridge, qui ne peut pas apprendre seul les interactions entre variables.

- **`Has_MarkDown`** : indicateur binaire (0/1) — signale si au moins une promotion est active. Permet aux modèles de distinguer rapidement les semaines avec/sans promo.

- **`Nb_MarkDowns`** : nombre de types de promos simultanément actifs (0 à 5) — capture l'intensité de la stratégie promotionnelle.

> **Résultat** : le dataset passe de MarkDown1-5 (5 colonnes brutes) à **9 colonnes promotionnelles** au total (5 individuelles + 4 agrégées), offrant aux modèles à la fois le détail par type de promo et une vue synthétique de l'activité promotionnelle.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Création des variables de décalage temporel (Lags)</strong>
</div>

Les variables de **décalage temporel (Lags)** permettent au modèle d'utiliser les ventes passées comme prédicteurs. L'idée : les ventes d'un magasin/département la semaine dernière sont fortement prédictives des ventes de cette semaine.

| Variable créée | Décalage | Rôle |
|---|---|---|
| `Lag_1` | t − 1 semaine | Signal le plus récent — réaction rapide aux changements |
| `Lag_4` | t − 4 semaines | Tendance à moyen terme (~1 mois) |
| `Lag_12` | t − 12 semaines | Tendance trimestrielle (~3 mois) |
| `Lag_52` | t − 52 semaines | Saisonnalité annuelle (même semaine l'an dernier) |

> ⚠️ **Calcul par groupe** `(Store_orig, Dept_orig)` : chaque lag est calculé au sein de la série temporelle d'un magasin/département donné, et non sur l'ensemble du dataset. Les premières lignes de chaque groupe contiendront des `NaN` (pas d'historique disponible) — ils seront traités lors du split train/test.

````python

print("=" * 70)
print("  CRÉATION DES VARIABLES DE DÉCALAGE TEMPOREL (LAGS)")
print("=" * 70)

# ── Tri chronologique par groupe (indispensable pour les lags) ───────────────
df_train = df_train.sort_values(['Store_orig', 'Dept_orig', 'Date']).reset_index(drop=True)
print("\n  ✅ Données triées par (Store_orig, Dept_orig, Date)")

# ── Définition des lags à créer ─────────────────────────────────────────────
lag_periods = [1, 4, 12, 52]

print(f"\n── Création des Lags : {lag_periods} ──")

for lag in lag_periods:
    col_name = f'Lag_{lag}'
    df_train[col_name] = (
        df_train
        .groupby(['Store_orig', 'Dept_orig'])['Weekly_Sales']
        .shift(lag)
    )
    n_nan = df_train[col_name].isna().sum()
    n_valid = df_train[col_name].notna().sum()
    print(f"\n  ✅ {col_name} (t − {lag} semaine{'s' if lag > 1 else ''})")
    print(f"     • Valeurs valides : {n_valid:,} ({n_valid/len(df_train)*100:.1f}%)")
    print(f"     • NaN (début de série) : {n_nan:,} ({n_nan/len(df_train)*100:.1f}%)")
    if n_valid > 0:
        print(f"     • Moyenne : {df_train[col_name].mean():,.0f} $")

# ── Vérification ─────────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  VÉRIFICATION")
print("=" * 70)

lag_cols = [f'Lag_{lag}' for lag in lag_periods]
print(f"\n  ► Nouvelles colonnes : {lag_cols}")
print(f"  ► Shape : {df_train.shape[0]:,} lignes × {df_train.shape[1]} colonnes")

# Aperçu pour un magasin/département
sample = df_train[
    (df_train['Store_orig'] == 1) & (df_train['Dept_orig'] == 1)
][['Store_orig', 'Dept_orig', 'Date', 'Weekly_Sales'] + lag_cols].head(8)

print("\n  ► Exemple : Store 1, Dept 1 (premières semaines)")
display(
    sample.style
    .set_properties(**{"text-align": "right", "padding": "6px"})
    .format({col: "{:,.0f}" for col in ['Weekly_Sales'] + lag_cols}, na_rep="NaN")
    .set_caption("Lags — Store 1, Dept 1")
)
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire — Résultats de la création des Lags :**

Quatre variables de décalage temporel ont été créées, calculées **par groupe** `(Store_orig, Dept_orig)` pour éviter toute contamination entre séries :

- **`Lag_1`** (t − 1 sem.) : signal le plus récent — capture les variations hebdomadaires immédiates. Très peu de NaN (1 ligne par groupe).

- **`Lag_4`** (t − 4 sem.) : tendance mensuelle — lisse les fluctuations hebdomadaires et révèle la dynamique à court terme.

- **`Lag_12`** (t − 12 sem.) : tendance trimestrielle — capture les cycles promotionnels et les variations saisonnières intra-annuelles.

- **`Lag_52`** (t − 52 sem.) : saisonnalité annuelle — les ventes de la même semaine l'année précédente sont un prédicteur très puissant en retail (ex: Black Friday, rentrée scolaire).

> **NaN attendus** : les premières semaines de chaque série (Store × Dept) n'ont pas d'historique suffisant, générant des NaN. Le `Lag_52` produit le plus de NaN (~52 lignes par groupe × 3 645 groupes). Ces valeurs manquantes seront gérées lors du split train/test (les lignes sans Lag_52 correspondent à la première année de données).

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Création des moyennes mobiles (Moving Averages)</strong>
</div>

Les moyennes mobiles capturent la **tendance locale** des ventes en lissant les fluctuations hebdomadaires. Contrairement aux lags (valeurs ponctuelles), elles résument le comportement récent sur une fenêtre glissante.

> ⚠️ Pour éviter toute fuite de données, chaque moyenne mobile est **décalée d'une semaine** (`shift(1)`) — la semaine courante n'entre jamais dans le calcul.

| Variable | Description |
|----------|-------------|
| `MA_4` | Moyenne mobile sur **4 semaines** — tendance court terme |
| `MA_12` | Moyenne mobile sur **12 semaines** — tendance moyen terme |

Calcul par groupe `(Store × Dept)` pour respecter la granularité des séries.

````python
# ── Moyennes Mobiles (Moving Averages) ──────────────────────────────────────
ma_windows = [4, 12]
group = df_train.groupby(['Store_orig', 'Dept_orig'])['Weekly_Sales']

ma_cols = []
for w in ma_windows:
    col = f'MA_{w}'
    # rolling sur la fenêtre w, shift(1) pour exclure la semaine courante
    df_train[col] = group.transform(
        lambda x: x.rolling(window=w, min_periods=1).mean().shift(1)
    )
    ma_cols.append(col)

print(f"Colonnes créées : {ma_cols}")
print(f"Shape : {df_train.shape}")
print(f"NaN introduits :\n{df_train[ma_cols].isna().sum()}\n")

# Aperçu pour Store 1, Dept 1
sample = df_train[
    (df_train['Store_orig'] == 1) & (df_train['Dept_orig'] == 1)
][['Date', 'Weekly_Sales', 'Lag_1'] + ma_cols].head(15)
sample
````

**Résultat — Moyennes Mobiles**

| Élément | Détail |
|---------|--------|
| **Colonnes créées** | `MA_4` (4 semaines), `MA_12` (12 semaines) |
| **Shape** | 418 727 × 35 |
| **NaN introduits** | 3 323 par colonne (1ère semaine de chaque groupe Store × Dept) |
| **Protection anti-fuite** | `shift(1)` — la semaine courante est exclue du calcul |

**Complémentarité avec les lags :**
- Les **lags** capturent des valeurs ponctuelles passées (t-1, t-4, t-12, t-52)
- Les **moyennes mobiles** résument la tendance sur une fenêtre → plus robuste au bruit

Les NaN seront traités lors de la préparation finale avant l'entraînement.

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Imputation des valeurs manquantes (Moyenne du groupe)</strong>
</div>

Les étapes précédentes (Lags, Moyennes Mobiles) ont introduit des **valeurs manquantes (NaN)**, principalement pour les premières semaines de chaque groupe `Store × Dept` où l'historique est insuffisant. Les variables `MarkDown1–5` contiennent aussi des NaN pour les périodes sans promotion.

**Stratégie : Moyenne du groupe (`Store_orig × Dept_orig`)**

Pour chaque colonne contenant des NaN, on remplace les valeurs manquantes par la **moyenne du groupe** (même magasin, même département). Cette approche respecte la hiérarchie naturelle des données :

$$\text{NaN}_{i,j} \leftarrow \overline{x}_{(\text{Store}=i,\, \text{Dept}=j)}$$

| Avantage | Détail |
|----------|--------|
| **Cohérence locale** | Un magasin haut de gamme (Type A) n'est pas dilué par les petits magasins |
| **Conservation des lignes** | On évite de supprimer ~160k lignes d'entraînement (38% du dataset) |
| **Robustesse** | Si un groupe n'a aucune valeur valide, on recourt à la **moyenne globale** |



````python
# ── Imputation des valeurs manquantes par la moyenne du groupe ───────────────
print("=" * 70)
print("  IMPUTATION DES VALEURS MANQUANTES — Moyenne du groupe (Store × Dept)")
print("=" * 70)

# Identifier les colonnes avec NaN
cols_with_nan = df_train.columns[df_train.isna().any()].tolist()
print(f"\n► Colonnes avec NaN : {cols_with_nan}")
print(f"► Nombre total de NaN avant imputation : {df_train[cols_with_nan].isna().sum().sum():,}")

# Détail des NaN par colonne
print("\n── Détail des NaN par colonne ──")
for col in cols_with_nan:
    n_nan = df_train[col].isna().sum()
    pct = n_nan / len(df_train) * 100
    print(f"  {col:20s} : {n_nan:>7,} NaN ({pct:.1f}%)")

# Imputation par la moyenne du groupe (Store_orig × Dept_orig)
print("\n── Imputation par groupe (Store_orig × Dept_orig) ──")
nan_before_total = df_train[cols_with_nan].isna().sum().sum()

for col in cols_with_nan:
    # Moyenne par groupe
    group_mean = df_train.groupby(['Store_orig', 'Dept_orig'])[col].transform('mean')
    # Imputer les NaN avec la moyenne du groupe
    df_train[col] = df_train[col].fillna(group_mean)

# Fallback : si un groupe entier est NaN, on utilise la moyenne globale
nan_after_group = df_train[cols_with_nan].isna().sum().sum()
if nan_after_group > 0:
    print(f"  ⚠️ {nan_after_group:,} NaN restants après imputation par groupe → fallback moyenne globale")
    for col in cols_with_nan:
        if df_train[col].isna().any():
            global_col_mean = df_train[col].mean()
            n_remaining = df_train[col].isna().sum()
            df_train[col] = df_train[col].fillna(global_col_mean)
            print(f"    {col:20s} : {n_remaining:,} NaN → remplacés par moyenne globale = {global_col_mean:.2f}")

# Vérification finale
nan_after_total = df_train[cols_with_nan].isna().sum().sum()
print(f"\n── Résultat ──")
print(f"  ✅ NaN avant  : {nan_before_total:,}")
print(f"  ✅ NaN après  : {nan_after_total}")
print(f"  ✅ Lignes conservées : {len(df_train):,} (aucune suppression)")
print(f"  ✅ Shape : {df_train.shape}")
````

| Élément | Détail |
|---------|--------|
| **Colonnes imputées** | `Lag_1`, `Lag_4`, `Lag_12`, `Lag_52`, `MA_4`, `MA_12` |
| **NaN avant** | 221 411 (répartis sur 6 colonnes) |
| **NaN après** | 0 ✅ |
| **Méthode principale** | Moyenne du groupe `Store_orig × Dept_orig` |
| **Fallback** | 6 174 NaN → moyenne globale (groupes entièrement manquants) |
| **Lignes conservées** | 418 727 (aucune suppression) |

- Les `MarkDown1–5` n'avaient **aucun NaN** (déjà traités lors de la préparation EPIC1)
- 6 174 NaN (2.8%) ont nécessité le fallback vers la moyenne globale, pour des groupes très courts sans aucune valeur valide

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Normalisation des variables continues (StandardScaler)</strong>
</div>

Les variables continues du dataset présentent des **échelles très différentes** (ex: `Size` ~34 000–220 000 vs `Fuel_Price` ~2.5–4.5). Cette hétérogénéité pose problème pour **Ridge**, qui pénalise les coefficients proportionnellement à l'échelle des variables.

**Stratégie : `StandardScaler`** (z-score) — chaque variable est recentrée à moyenne 0 et écart-type 1 :

$$z = \frac{x - \mu}{\sigma}$$

| Colonnes normalisées | Raison |
|---|---|
| `Store`, `Dept` | Target-encoded (~1k–50k $) |
| `Temperature`, `Fuel_Price`, `CPI`, `Unemployment` | Échelles très différentes |
| `Size` | Très grande échelle (~34k–220k sqft) |
| `MarkDown1–5`, `MarkDown_Total`, `MarkDown_Density` | Promos en $ et $/sqft |
| `Month`, `Week`, `Year` | Composantes calendaires |
| `Lag_1/4/12/52`, `MA_4/12` | Même échelle que Weekly_Sales (~-4k à +700k) |

| Colonnes **exclues** | Raison |
|---|---|
| `Weekly_Sales` | Variable cible — ne pas transformer |
| `IsHoliday`, `Has_MarkDown` | Binaires (0/1) — déjà normalisées |
| `Type` | Ordinal (1/2/3) — plage réduite |
| `Nb_MarkDowns` | Compteur (0–5) — plage réduite |
| `Month_sin/cos`, `Week_sin/cos` | Cycliques — déjà dans [-1, 1] |
| `Store_orig`, `Dept_orig`, `Date` | Colonnes de référence — pas des features |

> ⚠️ **Note** : les modèles tree-based (RF, LightGBM, XGBoost) sont **invariants à l'échelle** et n'ont pas besoin de normalisation. Cette étape bénéficie principalement à **Ridge** (baseline). On normalise ici sur l'ensemble des données avant le split ; en production, le scaler serait ajusté uniquement sur le jeu d'entraînement.

### <span id="213" style="color:#2E86C1;">2.1.3 Séparation des jeux d'entraînement et de test</span>

La séparation des données doit respecter la **chronologie des ventes** pour simuler un scénario réaliste : le modèle est entraîné sur le passé et évalué sur le futur.

| Élément | Détail |
|---|---|
| **Méthode** | Split temporel (pas aléatoire) |
| **Plage des données** | Fév. 2010 → Oct. 2012 (~143 semaines) |
| **Seuil de coupure** | Dernières **12 semaines** (~3 mois) réservées au test |
| **Ratio approximatif** | ~92% train / ~8% test |

> ⚠️ **Pourquoi un split temporel ?** Un split aléatoire mélangerait des semaines de 2010 dans le test et des semaines de 2012 dans le train. Le modèle "verrait" le futur pendant l'entraînement → **data leakage**. Le split temporel garantit que le train contient uniquement des données **antérieures** au test.

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.1.3 — Séparation temporelle Train / Test
# ══════════════════════════════════════════════════════════════════════════════

print("=" * 70)
print("  SÉPARATION TEMPORELLE TRAIN / TEST")
print("=" * 70)

# ── 1) Identifier la plage temporelle ────────────────────────────────────────
print("\n── 1) Plage temporelle des données ──")

date_min = df_train['Date'].min()
date_max = df_train['Date'].max()
n_weeks = df_train['Date'].nunique()

print(f"  ► Début  : {date_min.strftime('%Y-%m-%d')}")
print(f"  ► Fin    : {date_max.strftime('%Y-%m-%d')}")
print(f"  ► Semaines uniques : {n_weeks}")

# ── 2) Définir le seuil de coupure ───────────────────────────────────────────
print("\n── 2) Seuil de coupure ──")

# Dernières 12 semaines (~3 mois) pour le test
from datetime import timedelta
split_date = date_max - timedelta(weeks=12)

print(f"  ► Date de coupure : {split_date.strftime('%Y-%m-%d')}")
print(f"  ► Train : [{date_min.strftime('%Y-%m-%d')} → {split_date.strftime('%Y-%m-%d')}]")
print(f"  ► Test  : ({split_date.strftime('%Y-%m-%d')} → {date_max.strftime('%Y-%m-%d')}]")

# ── 3) Créer le masque temporel ──────────────────────────────────────────────
print("\n── 3) Application du split ──")

mask_train = df_train['Date'] <= split_date
mask_test  = df_train['Date'] > split_date

# Appliquer sur df_model (même index que df_train)
X_train = df_model.loc[mask_train, feature_cols]
y_train = df_model.loc[mask_train, target]
X_test  = df_model.loc[mask_test, feature_cols]
y_test  = df_model.loc[mask_test, target]

print(f"  ► Train : {X_train.shape[0]:,} lignes")
print(f"  ► Test  : {X_test.shape[0]:,} lignes")


# ── 4) Résumé final ──────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  RÉSUMÉ DU SPLIT")
print("=" * 70)

total = X_train.shape[0] + X_test.shape[0]
print(f"\n  ► X_train : {X_train.shape[0]:,} lignes × {X_train.shape[1]} features ({X_train.shape[0]/total*100:.1f}%)")
print(f"  ► X_test  : {X_test.shape[0]:,} lignes × {X_test.shape[1]} features ({X_test.shape[0]/total*100:.1f}%)")
print(f"  ► y_train : {y_train.shape[0]:,} valeurs")
print(f"  ► y_test  : {y_test.shape[0]:,} valeurs")
print(f"  ► Total   : {total:,} lignes (100% conservé)")

# Distribution de la cible
print(f"\n  ► Weekly_Sales (train) : moy={y_train.mean():,.0f} $ | med={y_train.median():,.0f} $")
print(f"  ► Weekly_Sales (test)  : moy={y_test.mean():,.0f} $ | med={y_test.median():,.0f} $")
````

## <span id="22" style="color:#2E86C1;">2.2 Développement et entraînement des modèles de prévision</span>

Cette section porte sur le **développement et l'entraînement des modèles de prévision** des ventes hebdomadaires. À partir des données préparées en section 2.1 (`X_train` : 383 872 lignes × 31 features), nous entraînons progressivement **4 modèles** selon une approche du plus simple au plus complexe :

| # | Modèle | Type | Rôle |
|---|--------|------|------|
| 1 | **Ridge** | Régression linéaire pénalisée (L2) | **Baseline** — référence minimale à battre |
| 2 | **Random Forest** | Ensemble d'arbres (bagging) | Première approche non-linéaire |
| 3 | **LightGBM** | Gradient Boosting (histograms) | Modèle principal — rapide et performant |
| 4 | **XGBoost** | Gradient Boosting (exact/approx) | Validation croisée — robustesse |

**Métrique d'évaluation principale : WMAE** (Weighted Mean Absolute Error)

$$\text{WMAE} = \frac{1}{\sum w_i} \sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|$$

où $w_i = 5$ si la semaine est un **jour férié** (IsHoliday = 1), sinon $w_i = 1$.

> *« se tromper pendant une semaine fériée est **5 fois plus grave** que se tromper une semaine normale. »* 

Cette pondération reflète l'importance business des périodes de forte affluence (Thanksgiving, Noël, Super Bowl, Labour Day). Une erreur de prévision durant ces semaines coûte beaucoup plus cher : sous-estimation → rupture de stock et ventes perdues, surestimation → surplus et gaspillage.

**Démarche pour chaque modèle :**
1. Entraînement sur `X_train` / `y_train`
2. Prédiction sur `X_test`
3. Calcul de la WMAE et des métriques complémentaires (MAE, RMSE, R²)
4. Ajustement des hyperparamètres si nécessaire

### <span id="221" style="color:#2E86C1;">2.2.1 Sélection des algorithmes de prévision adaptés</span>

Le choix des algorithmes repose sur une **stratégie progressive** : partir d'un modèle simple (baseline) puis monter en complexité pour mesurer le gain réel de chaque approche.

#### 1. Ridge Regression (Baseline)

Régression linéaire avec pénalisation **L2** ($\lambda \sum \beta_j^2$). Elle suppose une relation linéaire entre les features et les ventes. C'est la **référence minimale** : si un modèle avancé ne bat pas Ridge, il n'apporte aucune valeur.

| Avantage | Limite |
|----------|--------|
| Très rapide à entraîner | Ne capture pas les interactions non-linéaires |
| Interprétable (coefficients) | Sensible à l'échelle → nécessite la normalisation |
| Robuste au surapprentissage | Performance plafonnée sur données complexes |

#### 2. Random Forest

Ensemble de **centaines d'arbres de décision** entraînés sur des sous-échantillons aléatoires (bagging). Chaque arbre vote, et la prédiction finale est la **moyenne des votes**.

| Avantage | Limite |
|----------|--------|
| Capture les non-linéarités et interactions | Plus lent que Ridge |
| Robuste aux outliers | Moins performant que le boosting en général |
| Pas besoin de normalisation | Peut sur-apprendre si trop d'arbres peu profonds |

#### 3. LightGBM (Modèle principal)

Gradient Boosting basé sur des **histogrammes** : les arbres sont construits **séquentiellement**, chacun corrigeant les erreurs du précédent. L'approche par histogrammes rend LightGBM **très rapide** même sur de gros volumes.

| Avantage | Limite |
|----------|--------|
| Très performant sur données tabulaires | Risque de surapprentissage si mal réglé |
| Rapide (histogrammes + croissance par feuille) | Plus d'hyperparamètres à ajuster |
| Gère nativement les NaN et les catégorielles | Moins interprétable que Ridge |

#### 4. XGBoost (Validation)

Autre implémentation du Gradient Boosting, avec une **régularisation L1+L2** intégrée. Utilisé pour **valider** les résultats de LightGBM et confirmer la robustesse.

| Avantage | Limite |
|----------|--------|
| Régularisation native (L1 + L2) | Plus lent que LightGBM |
| Très utilisé en compétitions Kaggle | Consomme plus de mémoire |
| Complémentaire à LightGBM | Hyperparamètres similaires |

#### Résumé de la stratégie

```
Ridge (baseline)  →  Random Forest  →  LightGBM  →  XGBoost
   linéaire           non-linéaire      boosting      validation
   simple              intermédiaire     principal     robustesse
```

> Cette approche permet de **quantifier le gain** à chaque étape : si Ridge donne une WMAE de X et LightGBM donne Y, on sait exactement combien la complexité supplémentaire apporte.

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.1 — Sélection des algorithmes & Définition de la métrique WMAE
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("=" * 70)
print("  SÉLECTION DES ALGORITHMES & MÉTRIQUE WMAE")
print("=" * 70)

# ── 1) Fonction WMAE ─────────────────────────────────────────────────────────
def wmae(y_true, y_pred, is_holiday):
    """
    Weighted Mean Absolute Error (métrique Walmart/Kaggle).
    Poids : 5 pour les semaines fériées, 1 sinon.
    """
    weights = np.where(is_holiday == 1, 5, 1)
    return np.sum(weights * np.abs(y_true - y_pred)) / np.sum(weights)

print("\n✅ Fonction wmae(y_true, y_pred, is_holiday) définie")

# ── 2) Récupérer IsHoliday pour le calcul des poids ─────────────────────────
# IsHoliday est dans X_train et X_test (c'est une feature)
is_holiday_train = X_train['IsHoliday'].values
is_holiday_test  = X_test['IsHoliday'].values

n_holiday_train = (is_holiday_train == 1).sum()
n_holiday_test  = (is_holiday_test == 1).sum()

print(f"\n── IsHoliday dans les données ──")
print(f"  ► Train : {n_holiday_train:,} semaines fériées ({n_holiday_train/len(X_train)*100:.1f}%)")
print(f"  ► Test  : {n_holiday_test:,} semaines fériées ({n_holiday_test/len(X_test)*100:.1f}%)")

# ── 3) Fonction d'évaluation complète ────────────────────────────────────────
def evaluate_model(name, y_true, y_pred, is_holiday):
    """Affiche toutes les métriques pour un modèle donné."""
    w = wmae(y_true, y_pred, is_holiday)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    
    print(f"\n{'─' * 50}")
    print(f"  {name}")
    print(f"{'─' * 50}")
    print(f"  WMAE  : {w:,.0f} $")
    print(f"  MAE   : {mae:,.0f} $")
    print(f"  RMSE  : {rmse:,.0f} $")
    print(f"  R²    : {r2:.4f}")
    
    return {'model': name, 'WMAE': w, 'MAE': mae, 'RMSE': rmse, 'R2': r2}

print("✅ Fonction evaluate_model(name, y_true, y_pred, is_holiday) définie")

# ── 4) Liste des modèles à entraîner ────────────────────────────────────────
print(f"\n── Modèles sélectionnés ──")
models = {
    'Ridge': Ridge(alpha=1.0),
    'Random Forest': RandomForestRegressor(
        n_estimators=200, max_depth=20, min_samples_leaf=5,
        n_jobs=-1, random_state=42
    ),
}

# ── Détection CUDA (XGBoost uniquement) ────────────────────────────────────────────
import subprocess as _sp
try:
    _r = _sp.run(['nvidia-smi'], capture_output=True, text=True, timeout=5)
    _cuda = _r.returncode == 0
except Exception:
    _cuda = False
if _cuda:
    print("  ✅ GPU NVIDIA détecté → CUDA activé pour XGBoost")
else:
    print("  ℹ️  Pas de GPU → entraînement sur CPU")

# LightGBM et XGBoost — import conditionnel
try:
    import lightgbm as lgb
    models['LightGBM'] = lgb.LGBMRegressor(
        n_estimators=500, max_depth=8, learning_rate=0.05,
        num_leaves=63, min_child_samples=50,
        subsample=0.8, colsample_bytree=0.8,
        n_jobs=-1, random_state=42, verbose=-1
    )
    print("  ✅ LightGBM disponible")
except ImportError:
    print("  ⚠️ LightGBM non installé — pip install lightgbm")

try:
    import xgboost as xgb
    _xgb_gpu = {'device': 'cuda', 'tree_method': 'hist'} if _cuda else {}
    models['XGBoost'] = xgb.XGBRegressor(
        n_estimators=500, max_depth=8, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        reg_alpha=0.1, reg_lambda=1.0,
        n_jobs=-1, random_state=42, verbosity=0,
        **_xgb_gpu
    )
    print("  ✅ XGBoost disponible")
except ImportError:
    print("  ⚠️ XGBoost non installé — pip install xgboost")

print(f"\n► {len(models)} modèles prêts : {list(models.keys())}")

# Stockage des résultats
results = []
trained_models = {}
````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire — Sélection des algorithmes et métrique WMAE :**

Cette cellule met en place l'infrastructure d'évaluation et les 4 modèles retenus pour la prévision des ventes :

- **Métrique WMAE** (*Weighted Mean Absolute Error*) : Les erreurs commises durant les **semaines fériées** sont pénalisées **5 fois plus** que les semaines normales, reflétant l'impact financier disproportionné d'une mauvaise prévision pendant ces périodes de forte affluence. La formule est :
$$WMAE = \frac{\sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|}{\sum_{i=1}^{n} w_i} \quad \text{avec } w_i = \begin{cases} 5 & \text{si semaine fériée} \\ 1 & \text{sinon} \end{cases}$$

- **4 modèles sélectionnés**, du plus simple au plus complexe :
  1. **Ridge** (régression linéaire régularisée L2) — sert de *baseline* pour mesurer l'apport des modèles non linéaires.
  2. **Random Forest** (200 arbres, profondeur max 20) — modèle d'ensemble robuste au bruit et aux outliers.
  3. **LightGBM** (500 itérations, lr=0.05) — gradient boosting optimisé pour la vitesse, idéal pour les grands volumes de données.
  4. **XGBoost** (500 itérations, lr=0.05, régularisation L1/L2) — gradient boosting avec régularisation intégrée pour limiter le surapprentissage.

- **Fonction `evaluate_model()`** : évalue chaque modèle avec 4 métriques complémentaires — WMAE (métrique principale), MAE, RMSE et R².

> **Résultat** : les 4 modèles sont initialisés et prêts pour l'entraînement.

### <span id="222" style="color:#2E86C1;">2.2.2 Entraînement des modèles sur les données historiques</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-bottom:10px; color:#D5D8DC;">

**Objectif** — Entraîner les 4 modèles sélectionnés (Ridge, Random Forest, LightGBM, XGBoost) sur le jeu d'entraînement (`X_train`, `y_train`), puis générer les prédictions sur le jeu de test (`X_test`) et évaluer chaque modèle avec l'ensemble des métriques (WMAE, MAE, RMSE, R²).

**Approche :**
- Chaque modèle est entraîné sur les **383 872 observations** du jeu d'entraînement (avant le 2012-08-03).
- Les prédictions sont ensuite générées sur les **34 855 observations** du jeu de test (12 dernières semaines).
- Le temps d'entraînement et de prédiction est mesuré pour chaque modèle afin de comparer leur efficacité computationnelle.
- Un **tableau comparatif** est produit, trié par WMAE croissant, pour identifier le modèle le plus performant.

</div>

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.2 — Entraînement des modèles & Évaluation initiale
# ══════════════════════════════════════════════════════════════════════════════
import time
import pandas as pd

print("=" * 70)
print("  ENTRAÎNEMENT DES MODÈLES & ÉVALUATION INITIALE")
print("=" * 70)

# ── Conversion de Type (category → int) pour XGBoost ────────────────────────
if X_train['Type'].dtype.name == 'category':
    X_train['Type'] = X_train['Type'].astype(int)
    X_test['Type']  = X_test['Type'].astype(int)
    print("  ℹ️  Colonne 'Type' convertie de category → int\n")

results = []
trained_models = {}

for name, model in models.items():
    print(f"\n{'━' * 60}")
    print(f"  ► Entraînement : {name}")
    print(f"{'━' * 60}")
    
    # ── Entraînement ──
    t0 = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - t0
    print(f"  ✅ Entraîné en {train_time:.1f} s")
    
    # ── Prédiction ──
    t0 = time.time()
    y_pred = model.predict(X_test)
    pred_time = time.time() - t0
    print(f"  ✅ Prédiction en {pred_time:.3f} s")
    
    # ── Évaluation ──
    res = evaluate_model(name, y_test.values, y_pred, is_holiday_test)
    res['Train_Time_s'] = round(train_time, 1)
    res['Pred_Time_s'] = round(pred_time, 3)
    results.append(res)
    
    # Stocker le modèle entraîné
    trained_models[name] = model

# ══════════════════════════════════════════════════════════════════════════════
# Tableau comparatif
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n\n{'=' * 70}")
print("  TABLEAU COMPARATIF DES MODÈLES")
print(f"{'=' * 70}\n")

df_results = pd.DataFrame(results).sort_values('WMAE')
df_results.index = range(1, len(df_results) + 1)
df_results.index.name = 'Rang'

# Formater l'affichage
df_display = df_results.copy()
df_display['WMAE'] = df_display['WMAE'].apply(lambda x: f"{x:,.0f} $")
df_display['MAE']  = df_display['MAE'].apply(lambda x: f"{x:,.0f} $")
df_display['RMSE'] = df_display['RMSE'].apply(lambda x: f"{x:,.0f} $")
df_display['R2']   = df_display['R2'].apply(lambda x: f"{x:.4f}")
df_display['Train_Time_s'] = df_display['Train_Time_s'].apply(lambda x: f"{x:.1f} s")
df_display['Pred_Time_s']  = df_display['Pred_Time_s'].apply(lambda x: f"{x:.3f} s")

df_display.rename(columns={
    'model': 'Modèle', 'Train_Time_s': 'Temps entr.',
    'Pred_Time_s': 'Temps préd.'
}, inplace=True)

display(df_display)

# ── Meilleur modèle ──
best = df_results.iloc[0]
print(f"\n🏆 Meilleur modèle (WMAE le plus bas) : {best['model']}")
print(f"   WMAE = {best['WMAE']:,.0f} $  |  R² = {best['R2']:.4f}")
````

### <span id="223" style="color:#2E86C1;">2.2.3 Ajustement des hyperparamètres</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-bottom:10px; color:#D5D8DC;">

**Objectif** — Affiner les hyperparamètres des **2 meilleurs modèles** (XGBoost et LightGBM) afin de réduire davantage le WMAE par rapport aux résultats initiaux (1 343 $ et 1 378 $).

**Approche — `RandomizedSearchCV`** :
- **Pourquoi RandomizedSearchCV ?** Avec 383 872 lignes et de nombreux hyperparamètres, un `GridSearchCV` exhaustif serait prohibitif. `RandomizedSearchCV` échantillonne aléatoirement `n_iter` combinaisons dans l'espace de recherche, offrant un bon compromis entre exploration et temps de calcul.
- **Validation croisée** : 3 folds (compromis entre robustesse statistique et temps d'exécution).
- **Métrique d'optimisation** : `neg_mean_absolute_error` (proxy de la WMAE — la pondération par IsHoliday sera vérifiée après avec la vraie WMAE sur le jeu de test).
- **Espace de recherche** : les hyperparamètres les plus influents pour chaque algorithme (profondeur, taux d'apprentissage, régularisation, sous-échantillonnage).

</div>

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.3 — Ajustement des hyperparamètres (XGBoost & LightGBM)
# ══════════════════════════════════════════════════════════════════════════════
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
import time, warnings, subprocess
warnings.filterwarnings('ignore')

print("=" * 70)
print("  AJUSTEMENT DES HYPERPARAMÈTRES — RandomizedSearchCV + TimeSeriesSplit")
print("=" * 70)

# ── Détection CUDA (XGBoost uniquement) ──────────────────────────────────────
try:
    _res = subprocess.run(['nvidia-smi'], capture_output=True, text=True, timeout=5)
    cuda_available = _res.returncode == 0
except Exception:
    cuda_available = False

if cuda_available:
    print("  ✅ GPU NVIDIA détecté → CUDA activé pour XGBoost")
    xgb_device_params = {'device': 'cuda', 'tree_method': 'hist'}
else:
    print("  ℹ️  Aucun GPU détecté → entraînement sur CPU")
    xgb_device_params = {}

print("  ℹ️  LightGBM → CPU (version pip sans support GPU)")

# ── Résultats avant tuning (section 2.2.2) ───────────────────────────────────
baseline_wmae = {
    'XGBoost':  df_results.loc[df_results['model'] == 'XGBoost', 'WMAE'].values[0],
    'LightGBM': df_results.loc[df_results['model'] == 'LightGBM', 'WMAE'].values[0],
}

# ── Tri de X_train par Date pour TimeSeriesSplit ──────────────────────────────
# Copie locale uniquement — X_train global inchangé pour les cellules suivantes
_sort_idx  = df_train.loc[X_train.index, 'Date'].sort_values().index
X_train_ts = X_train.loc[_sort_idx]
y_train_ts = y_train.loc[_sort_idx]
print(f"\n  ✅ X_train_ts trié par Date ({len(X_train_ts):,} lignes)")

tscv = TimeSeriesSplit(n_splits=3)
print(f"  ✅ TimeSeriesSplit(n_splits=3)\n")

# ══════════════════════════════════════════════════════════════════════════════
# 1) XGBoost — RandomizedSearchCV + TimeSeriesSplit
# ══════════════════════════════════════════════════════════════════════════════
print(f"{'━' * 60}")
print(f"  ► Tuning : XGBoost  (20 tirages × 3 folds)")
print(f"{'━' * 60}")

import xgboost as xgb

xgb_param_dist = {
    'n_estimators':     [300, 500, 700, 1000],
    'max_depth':        [6, 8, 10, 12],
    'learning_rate':    [0.01, 0.03, 0.05, 0.08, 0.1],
    'subsample':        [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'reg_alpha':        [0, 0.01, 0.1, 0.5],
    'reg_lambda':       [0.5, 1.0, 2.0, 5.0],
    'min_child_weight': [3, 5, 10, 20],
}

xgb_base = xgb.XGBRegressor(
    n_jobs=-1, random_state=42, verbosity=0,
    **xgb_device_params
)

t0 = time.time()
xgb_search = RandomizedSearchCV(
    xgb_base,
    param_distributions=xgb_param_dist,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=tscv,
    random_state=42,
    n_jobs=1,
    verbose=0,
)
xgb_search.fit(X_train_ts, y_train_ts)
xgb_time = time.time() - t0

print(f"  ✅ Terminé en {xgb_time/60:.1f} min ({int(xgb_time)} s)")
print(f"  ► Meilleur score CV (neg_MAE) : {xgb_search.best_score_:,.0f}")
print(f"\n  Meilleurs hyperparamètres :")
for k, v in xgb_search.best_params_.items():
    print(f"    {k:20s} = {v}")

y_pred_xgb = xgb_search.best_estimator_.predict(X_test)
res_xgb = evaluate_model('XGBoost (tuné)', y_test.values, y_pred_xgb, is_holiday_test)

# ══════════════════════════════════════════════════════════════════════════════
# 2) LightGBM — RandomizedSearchCV + TimeSeriesSplit   (CPU uniquement)
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n{'━' * 60}")
print(f"  ► Tuning : LightGBM  (20 tirages × 3 folds, CPU)")
print(f"{'━' * 60}")

import lightgbm as lgb

lgb_param_dist = {
    'n_estimators':      [300, 500, 700, 1000],
    'max_depth':         [6, 8, 10, 12, -1],
    'learning_rate':     [0.01, 0.03, 0.05, 0.08, 0.1],
    'num_leaves':        [31, 63, 127, 255],
    'min_child_samples': [20, 50, 100],
    'subsample':         [0.7, 0.8, 0.9],
    'colsample_bytree':  [0.7, 0.8, 0.9],
    'reg_alpha':         [0, 0.01, 0.1, 0.5],
    'reg_lambda':        [0, 0.1, 1.0, 5.0],
}

lgb_base = lgb.LGBMRegressor(
    n_jobs=-1, random_state=42, verbose=-1
)

t0 = time.time()
lgb_search = RandomizedSearchCV(
    lgb_base,
    param_distributions=lgb_param_dist,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=tscv,
    random_state=42,
    n_jobs=1,
    verbose=0,
)
lgb_search.fit(X_train_ts, y_train_ts)
lgb_time = time.time() - t0

print(f"  ✅ Terminé en {lgb_time/60:.1f} min ({int(lgb_time)} s)")
print(f"  ► Meilleur score CV (neg_MAE) : {lgb_search.best_score_:,.0f}")
print(f"\n  Meilleurs hyperparamètres :")
for k, v in lgb_search.best_params_.items():
    print(f"    {k:20s} = {v}")

y_pred_lgb = lgb_search.best_estimator_.predict(X_test)
res_lgb = evaluate_model('LightGBM (tuné)', y_test.values, y_pred_lgb, is_holiday_test)

# ══════════════════════════════════════════════════════════════════════════════
# 3) Tableau comparatif : Avant vs Après tuning
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n\n{'=' * 70}")
print("  COMPARAISON AVANT / APRÈS TUNING")
print(f"{'=' * 70}\n")

comparison = []
for name_key, res_tuned, y_pred_tuned in [
    ('XGBoost', res_xgb, y_pred_xgb),
    ('LightGBM', res_lgb, y_pred_lgb),
]:
    before = baseline_wmae[name_key]
    after  = res_tuned['WMAE']
    delta  = after - before
    pct_change = (delta / before) * 100
    comparison.append({
        'Modèle': name_key,
        'WMAE avant': f"{before:,.0f} $",
        'WMAE après': f"{after:,.0f} $",
        'Δ WMAE': f"{delta:+,.0f} $",
        'Variation': f"{pct_change:+.1f} %",
        'R² après': f"{res_tuned['R2']:.4f}",
    })

df_comparison = pd.DataFrame(comparison)
df_comparison.index = range(1, len(df_comparison) + 1)
display(df_comparison)

# ── Mettre à jour les modèles entraînés ──────────────────────────────────────
trained_models['XGBoost']  = xgb_search.best_estimator_
trained_models['LightGBM'] = lgb_search.best_estimator_

best_tuned = min(
    [('XGBoost', res_xgb), ('LightGBM', res_lgb)],
    key=lambda x: x[1]['WMAE']
)
print(f"\n🏆 Champion après tuning : {best_tuned[0]}")
print(f"   WMAE = {best_tuned[1]['WMAE']:,.0f} $  |  R² = {best_tuned[1]['R2']:.4f}")
````

**Commentaire — Résultats de l'ajustement des hyperparamètres :**

Le `RandomizedSearchCV` (20 combinaisons × 3 folds) a été appliqué aux deux meilleurs modèles. Voici la comparaison avant/après :

| Modèle | WMAE avant | WMAE après | Δ WMAE | Variation | R² après |
|--------|-----------|-----------|--------|-----------|----------|
| **XGBoost** | 1 343 $ | **1 332 $** | −11 $ | −0.8 % | 0.9837 |
| LightGBM | 1 378 $ | 1 357 $ | −21 $ | −1.5 % | 0.9839 |

**Observations clés :**

- **XGBoost reste le champion** avec un WMAE de 1 332$ (−11$ par rapport aux paramètres par défaut). LightGBM progresse davantage en valeur absolue (−21$) mais reste en seconde position.

- **Les gains sont modestes** (< 2 %) — ce qui est un signe positif : cela confirme que les hyperparamètres par défaut choisis en section 2.2.1 étaient déjà bien calibrés, et que **la qualité du feature engineering** (section 2.1) est le principal moteur de la performance.

- **Le R² est quasi identique** entre les deux modèles (~0.984), ce qui signifie qu'ils expliquent la même proportion de variance. La différence se joue sur les **semaines fériées** (pondération ×5 dans la WMAE), où XGBoost est légèrement plus précis.

- **Temps de tuning** : l'ensemble de la procédure (20 × 3 folds × 2 modèles) a pris environ 53 minutes, ce qui reste raisonnable pour un dataset de 384K lignes.

> **Modèle retenu** : **XGBoost (tuné)** — WMAE = 1 332 $, R² = 0.9837. Ce modèle sera utilisé pour les prévisions finales et le déploiement.

### <span id="224" style="color:#2E86C1;">2.2.4 Génération des prévisions</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-top:12px; color:#D5D8DC;">

**Objectif :** Utiliser le modèle champion (**XGBoost tuné**, WMAE = 1 332 $) pour générer les prévisions finales sur l'ensemble de test (12 dernières semaines). Cette étape produit un DataFrame de prévisions exploitable qui servira de base à l'évaluation détaillée (section 2.3) et à la validation business (section 2.4).

**Contenu :**

- Aperçu des résultats (Top/Bottom erreurs)

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.4 — Génération des prévisions (modèle champion : XGBoost tuné)
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np
import pandas as pd

print("=" * 70)
print("  GÉNÉRATION DES PRÉVISIONS — XGBoost (tuné)")
print("=" * 70)

# ── 1) Modèle champion ──────────────────────────────────────────────────────
champion_name = best_tuned[0]
champion_model = trained_models[champion_name]
print(f"\n  ► Modèle champion : {champion_name}")
print(f"  ► WMAE (test)     : {best_tuned[1]['WMAE']:,.0f} $")
print(f"  ► R² (test)       : {best_tuned[1]['R2']:.4f}")

# ── 2) Générer les prédictions finales ───────────────────────────────────────
print(f"\n── Prédictions sur l'ensemble de test ({X_test.shape[0]:,} lignes) ──")

y_pred_final = champion_model.predict(X_test)
print(f"  ✅ {len(y_pred_final):,} prévisions générées")

# ── 3) Construire le DataFrame de prévisions ────────────────────────────────
print("\n── Construction du DataFrame de prévisions ──")

df_predictions = pd.DataFrame({
    'Store':          df_train.loc[mask_test, 'Store_orig'].values,
    'Dept':           df_train.loc[mask_test, 'Dept_orig'].values,
    'Date':           df_train.loc[mask_test, 'Date'].values,
    'IsHoliday':      is_holiday_test,
    'Reel':           y_test.values,
    'Prevu':          y_pred_final,
})

# Colonnes dérivées
df_predictions['Erreur']     = df_predictions['Prevu'] - df_predictions['Reel']
df_predictions['Erreur_Abs'] = df_predictions['Erreur'].abs()
df_predictions['Erreur_Pct'] = (df_predictions['Erreur_Abs'] / df_predictions['Reel'].abs().replace(0, np.nan) * 100).round(2)

print(f"  ✅ DataFrame : {df_predictions.shape[0]:,} lignes × {df_predictions.shape[1]} colonnes")
print(f"  ► Colonnes   : {list(df_predictions.columns)}")

# ── 4) Statistiques descriptives des prévisions ─────────────────────────────
print(f"\n{'=' * 70}")
print("  STATISTIQUES DES PRÉVISIONS")
print(f"{'=' * 70}\n")

print(f"  ► Ventes réelles (test) :")
print(f"    Moyenne      : {df_predictions['Reel'].mean():>12,.0f} $")
print(f"    Médiane      : {df_predictions['Reel'].median():>12,.0f} $")
print(f"    Min          : {df_predictions['Reel'].min():>12,.0f} $")
print(f"    Max          : {df_predictions['Reel'].max():>12,.0f} $")

print(f"\n  ► Ventes prévues :")
print(f"    Moyenne      : {df_predictions['Prevu'].mean():>12,.0f} $")
print(f"    Médiane      : {df_predictions['Prevu'].median():>12,.0f} $")
print(f"    Min          : {df_predictions['Prevu'].min():>12,.0f} $")
print(f"    Max          : {df_predictions['Prevu'].max():>12,.0f} $")

print(f"\n  ► Erreurs absolues :")
print(f"    Moyenne (MAE): {df_predictions['Erreur_Abs'].mean():>12,.0f} $")
print(f"    Médiane      : {df_predictions['Erreur_Abs'].median():>12,.0f} $")
print(f"    90e centile  : {df_predictions['Erreur_Abs'].quantile(0.90):>12,.0f} $")
print(f"    95e centile  : {df_predictions['Erreur_Abs'].quantile(0.95):>12,.0f} $")
print(f"    Max          : {df_predictions['Erreur_Abs'].max():>12,.0f} $")

# Erreur médiane en %
median_pct = df_predictions['Erreur_Pct'].median()
print(f"\n  ► Erreur % médiane : {median_pct:.1f} %")

# ── 5) Répartition des erreurs ───────────────────────────────────────────────
print(f"\n{'=' * 70}")
print("  RÉPARTITION DES ERREURS")
print(f"{'=' * 70}\n")

seuils = [500, 1000, 2000, 5000]
for s in seuils:
    n = (df_predictions['Erreur_Abs'] <= s).sum()
    pct = n / len(df_predictions) * 100
    print(f"  ► Erreur ≤ {s:>5,} $ : {n:>6,} lignes ({pct:5.1f} %)")

# ── 6) Prévisions fériés vs normaux ─────────────────────────────────────────
print(f"\n{'=' * 70}")
print("  PRÉVISIONS : JOURS FÉRIÉS vs NORMAUX")
print(f"{'=' * 70}\n")

for label, mask_h in [("Normaux (IsHoliday=0)", df_predictions['IsHoliday'] == 0),
                       ("Fériés  (IsHoliday=1)", df_predictions['IsHoliday'] == 1)]:
    sub = df_predictions[mask_h]
    if len(sub) > 0:
        mae_h = sub['Erreur_Abs'].mean()
        med_h = sub['Erreur_Abs'].median()
        print(f"  ► {label} : {len(sub):,} lignes")
        print(f"    MAE     = {mae_h:,.0f} $")
        print(f"    Médiane = {med_h:,.0f} $\n")

# ── 7) Aperçu : meilleures et pires prévisions ──────────────────────────────
print(f"\n{'=' * 70}")
print("  TOP 10 — MEILLEURES PRÉVISIONS (erreur la plus faible)")
print(f"{'=' * 70}\n")

top10 = df_predictions.nsmallest(10, 'Erreur_Abs')[
    ['Store', 'Dept', 'Date', 'Reel', 'Prevu', 'Erreur', 'Erreur_Pct']
].copy()
top10['Reel']   = top10['Reel'].apply(lambda x: f"{x:,.0f} $")
top10['Prevu']  = top10['Prevu'].apply(lambda x: f"{x:,.0f} $")
top10['Erreur'] = top10['Erreur'].apply(lambda x: f"{x:+,.0f} $")
top10['Erreur_Pct'] = top10['Erreur_Pct'].apply(lambda x: f"{x:.2f} %")
top10.index = range(1, 11)
display(top10)

print(f"\n{'=' * 70}")
print("  TOP 10 — PIRES PRÉVISIONS (erreur la plus élevée)")
print(f"{'=' * 70}\n")

worst10 = df_predictions.nlargest(10, 'Erreur_Abs')[
    ['Store', 'Dept', 'Date', 'Reel', 'Prevu', 'Erreur', 'Erreur_Pct']
].copy()
worst10['Reel']   = worst10['Reel'].apply(lambda x: f"{x:,.0f} $")
worst10['Prevu']  = worst10['Prevu'].apply(lambda x: f"{x:,.0f} $")
worst10['Erreur'] = worst10['Erreur'].apply(lambda x: f"{x:+,.0f} $")
worst10['Erreur_Pct'] = worst10['Erreur_Pct'].apply(lambda x: f"{x:.2f} %")
worst10.index = range(1, 11)
display(worst10)

print(f"\n{'=' * 70}")
print(f"  ✅ DataFrame 'df_predictions' prêt ({df_predictions.shape[0]:,} lignes)")
print(f"{'=' * 70}")
````

### <span id="231" style="color:#2E86C1;">2.3.1 Choix des métriques d'évaluation</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-top:12px; color:#D5D8DC;">

**Objectif :** Définir et justifier les métriques utilisées pour évaluer les modèles de prévision des ventes. Le choix de la **WMAE** comme métrique principale est dicté par le concours Kaggle Walmart, mais nous calculons également des métriques complémentaires pour une vision complète.

**Métriques retenues :**
- **WMAE** (Weighted MAE) — métrique principale, pondère ×5 les semaines fériées
- **MAE** (Mean Absolute Error) — erreur moyenne en dollars, interprétable directement
- **RMSE** (Root Mean Squared Error) — pénalise les grosses erreurs
- **R²** (Coefficient de détermination) — variance expliquée par le modèle
- **MAPE** (Mean Absolute Percentage Error) — erreur relative en %

</div>

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.3.1 — Choix des métriques d'évaluation
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("=" * 70)
print("  MÉTRIQUES D'ÉVALUATION RETENUES")
print("=" * 70)

# ── 1) Rappel des formules ───────────────────────────────────────────────────

# ── 2) Calcul complet pour le champion (XGBoost tuné) ────────────────────────
print(f"── Métriques du champion : {champion_name} (tuné) ──\n")

y_true = y_test.values
y_pred_ch = y_pred_final

# WMAE
w_test = np.where(is_holiday_test == 1, 5, 1)
val_wmae = np.sum(w_test * np.abs(y_true - y_pred_ch)) / np.sum(w_test)

# MAE
val_mae = mean_absolute_error(y_true, y_pred_ch)

# RMSE
val_rmse = np.sqrt(mean_squared_error(y_true, y_pred_ch))

# R²
val_r2 = r2_score(y_true, y_pred_ch)

# MAPE (excluant les ventes nulles/négatives pour éviter division par 0)
mask_valid = np.abs(y_true) > 0
val_mape = np.mean(np.abs(y_true[mask_valid] - y_pred_ch[mask_valid]) / np.abs(y_true[mask_valid])) * 100
n_excluded = (~mask_valid).sum()

print(f"  ► WMAE  = {val_wmae:>10,.0f} $   ← métrique principale (Kaggle)")
print(f"  ► MAE   = {val_mae:>10,.0f} $   ← erreur absolue moyenne")
print(f"  ► RMSE  = {val_rmse:>10,.0f} $   ← sensible aux outliers")
print(f"  ► R²    = {val_r2:>10.4f}     ← variance expliquée")
print(f"  ► MAPE  = {val_mape:>10.2f} %   ← erreur relative moyenne")
if n_excluded > 0:
    print(f"    (MAPE calculé sur {mask_valid.sum():,} lignes, {n_excluded:,} exclues car ventes = 0)")

# ── 3) Tableau récapitulatif formaté ─────────────────────────────────────────
print(f"\n{'=' * 70}")
print("  TABLEAU RÉCAPITULATIF — CHAMPION")
print(f"{'=' * 70}\n")

df_metrics = pd.DataFrame([{
    'Métrique': 'WMAE (pondérée)',
    'Valeur': f"{val_wmae:,.0f} $",
    'Rôle': 'Métrique principale — pondère x5 les fériés',
}, {
    'Métrique': 'MAE',
    'Valeur': f"{val_mae:,.0f} $",
    'Rôle': 'Erreur moyenne en dollars',
}, {
    'Métrique': 'RMSE',
    'Valeur': f"{val_rmse:,.0f} $",
    'Rôle': 'Pénalise les grandes erreurs',
}, {
    'Métrique': 'R²',
    'Valeur': f"{val_r2:.4f}",
    'Rôle': f'{val_r2*100:.2f}% de la variance expliquée',
}, {
    'Métrique': 'MAPE',
    'Valeur': f"{val_mape:.2f} %",
    'Rôle': 'Erreur relative — interprétable par le métier',
}])
df_metrics.index = range(1, len(df_metrics) + 1)
display(df_metrics)

````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire — Métriques d'évaluation :**

| Métrique | Valeur | Interprétation |
|:---|:---|:---|
| **WMAE** | **1 332 $** | En moyenne, l'erreur pondérée est de 1 332 $ par magasin/département/semaine |
| **MAE** | **1 263 $** | Sans pondération fériée, l'erreur descend — les fériés sont plus difficiles à prédire |
| **RMSE** | **2 801 $** | Le ratio RMSE/MAE ≈ 2.2 indique la présence de quelques erreurs importantes |
| **R²** | **0.9837** | Le modèle explique **98.4%** de la variance des ventes — excellente qualité |
| **MAPE** | **477 %** | ⚠️ Valeur gonflée par les départements à ventes quasi-nulles (division par ~0) |

**Observations :**
- Le **MAPE est non pertinent** ici en raison des ventes proches de 0 dans certains départements (ex: rayons saisonniers). C'est un phénomène classique — la MAPE n'est pas adaptée aux datasets avec des valeurs cibles proches de zéro.
- La **WMAE reste la métrique de référence** pour ce projet : interprétable, business-oriented, et robuste.
- Le **R² de 0.9837** confirme que le modèle capture la quasi-totalité de la dynamique des ventes.
- L'écart **WMAE > MAE** (+69 $) montre que les semaines fériées concentrent des erreurs plus élevées, ce qui est attendu.

### <span id="233" style="color:#2E86C1;">2.3.3 Analyse des erreurs de prédiction</span>

**Objectif :** Analyser en profondeur les erreurs de prédiction du champion (XGBoost tuné) pour identifier les zones de force et de faiblesse du modèle. Cette analyse guide les améliorations futures et informe les utilisateurs métier des cas où le modèle est moins fiable.

**Visualisations :**
1. **Scatter plot** Prévu vs Réel — vérifier la linéarité
2. **Distribution des résidus** — vérifier la normalité et l'absence de biais
3. **Erreur par magasin** — identifier les magasins les plus difficiles à prédire

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.4.2 — Comparaison avec les ventes historiques
# ══════════════════════════════════════════════════════════════════════════════
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import pandas as pd

print('=' * 70)
print('  2.4.2 — COMPARAISON AVEC LES VENTES HISTORIQUES')
print('=' * 70)

# ── 1) Agrégation hebdomadaire ───────────────────────────────────────────────
df_predictions['Date'] = pd.to_datetime(df_predictions['Date'])
df_weekly = df_predictions.groupby('Date').agg(
    Reel_Total  = ('Reel',      'sum'),
    Prevu_Total = ('Prevu',     'sum'),
    IsHoliday   = ('IsHoliday', 'max')
).reset_index().sort_values('Date')
df_weekly['Erreur_Pct_Sem'] = (
    (df_weekly['Prevu_Total'] - df_weekly['Reel_Total']).abs()
    / df_weekly['Reel_Total'].abs() * 100
).round(2)

print(f'\n── Ventes agrégées ({len(df_weekly)} semaines de test) ──\n')
print(f'  ► Ventes réelles totales  : {df_weekly["Reel_Total"].sum():,.0f} $')
print(f'  ► Ventes prévues totales  : {df_weekly["Prevu_Total"].sum():,.0f} $')
ecart = df_weekly['Prevu_Total'].sum() - df_weekly['Reel_Total'].sum()
print(f'  ► Écart global            : {ecart:+,.0f} $ ({ecart/df_weekly["Reel_Total"].sum()*100:+.2f} %)')

# ── 2) Fériés vs normaux ─────────────────────────────────────────────────────
print(f'\n── Performance fériés vs normaux ──\n')
for label, mask_h in [('Normales', df_predictions['IsHoliday'] == 0),
                       ('Fériées ', df_predictions['IsHoliday'] == 1)]:
    sub = df_predictions[mask_h]
    if len(sub) > 0:
        mae_h  = sub['Erreur_Abs'].mean()
        mape_h = sub['Erreur_Pct'].median()
        print(f'  ► Semaines {label} ({len(sub):,} lignes) : MAE = {mae_h:,.0f} $  |  Erreur médiane = {mape_h:.1f} %')

# ── 3) Performance par magasin ───────────────────────────────────────────────
print(f'\n── Performance par magasin ──')
df_by_store = df_predictions.groupby('Store').agg(
    MAE          = ('Erreur_Abs', 'mean'),
    Erreur_Pct   = ('Erreur_Pct', 'median'),
    N            = ('Reel',       'count'),
    Ventes_Moy   = ('Reel',       'mean')
).round(1).sort_values('MAE')

print(f'\n  Top 5 magasins (meilleure précision) :')
display(df_by_store.head(5))
print(f'\n  Bottom 5 magasins (moins bonne précision) :')
display(df_by_store.tail(5))

# ── 4) Visualisations ───────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 1, figsize=(16, 12))
fig.suptitle('2.4.2 — Comparaison avec les ventes historiques', fontsize=14, fontweight='bold')

# 4a) Évolution temporelle
ax1 = axes[0]
ax1.plot(df_weekly['Date'], df_weekly['Reel_Total'] / 1e6,
         label='Ventes réelles', color='#2980b9', linewidth=2, marker='o', markersize=5)
ax1.plot(df_weekly['Date'], df_weekly['Prevu_Total'] / 1e6,
         label='Ventes prévues', color='#e74c3c', linewidth=2, linestyle='--', marker='s', markersize=5)
feries_mask = df_weekly['IsHoliday'] == 1
ax1.scatter(df_weekly.loc[feries_mask, 'Date'],
            df_weekly.loc[feries_mask, 'Reel_Total'] / 1e6,
            color='#f39c12', zorder=5, s=120, label='Semaine fériée', marker='*')
ax1.set_xlabel('Semaine', fontsize=11)
ax1.set_ylabel('Ventes totales (M$)', fontsize=11)
ax1.set_title('Évolution hebdomadaire — Réelles vs Prévues (ensemble de test)', fontsize=12, fontweight='bold')
ax1.legend(fontsize=10)
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.1f}M$'))
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=30, ha='right')

# 4b) MAE par magasin
ax2 = axes[1]
stores   = df_by_store.index.tolist()
maes     = df_by_store['MAE'].values
med_mae  = df_by_store['MAE'].median()
colors_bar = ['#27ae60' if m <= med_mae else '#e74c3c' for m in maes]
ax2.bar(range(len(stores)), maes, color=colors_bar, alpha=0.85)
ax2.axhline(df_by_store['MAE'].mean(), color='#2980b9', linestyle='--', linewidth=2,
            label=f"MAE moyenne = {df_by_store['MAE'].mean():,.0f} $")
ax2.set_xticks(range(len(stores)))
ax2.set_xticklabels([f'S{s}' for s in stores], rotation=45, ha='right', fontsize=8)
ax2.set_xlabel('Magasin', fontsize=11)
ax2.set_ylabel('MAE ($)', fontsize=11)
ax2.set_title('MAE par magasin (vert = sous la médiane, rouge = au-dessus)', fontsize=12, fontweight='bold')
ax2.legend(fontsize=10)
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1000:.0f}k$'))

plt.tight_layout()
plt.savefig('../images/242_comparaison_historique.png', dpi=150, bbox_inches='tight')
plt.show()
print('\n  ✅ Figure sauvegardée → images/242_comparaison_historique.png')
````

### <span id="243" style="color:#2E86C1;">2.4.3 Validation finale par les utilisateurs métier</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-top:12px; color:#D5D8DC;">

**Objectif** — Émettre un **verdict final GO/NO-GO** sur le déploiement du modèle, basé sur des seuils de tolérance métier standard dans le secteur retail :

| Seuil | Signification |
|---|---|
| Erreur ≤ 10% | Prévision **excellente** — planification optimale |
| Erreur ≤ 20% | Prévision **acceptable** — utilisation en production |
| Erreur > 30% | Prévision **insuffisante** — révision nécessaire |

Le **score de confiance** par magasin correspond au % de prévisions dans le seuil acceptable (≤ 20%). Un magasin avec un score ≥ 80% reçoit un verdict **GO**.

</div>

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.4.3 — Validation finale par les utilisateurs métier
# ══════════════════════════════════════════════════════════════════════════════
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print('=' * 70)
print('  2.4.3 — VALIDATION FINALE PAR LES UTILISATEURS MÉTIER')
print('=' * 70)

# ── Seuils de tolérance métier (standard retail) ────────────────────────────
SEUIL_EXCELLENT    = 10   # erreur ≤ 10% → excellent
SEUIL_ACCEPTABLE   = 20   # erreur ≤ 20% → acceptable pour la production
SEUIL_INACCEPTABLE = 30   # erreur > 30% → à améliorer

n_total = len(df_predictions)

print(f'\n── Seuils définis ──\n')
print(f'  ► Excellent    : erreur ≤ {SEUIL_EXCELLENT}%')
print(f'  ► Acceptable   : erreur ≤ {SEUIL_ACCEPTABLE}%')
print(f'  ► Insuffisant  : erreur > {SEUIL_INACCEPTABLE}%')

# ── Calcul des catégories ────────────────────────────────────────────────────
n_excellent    = (df_predictions['Erreur_Pct'] <= SEUIL_EXCELLENT).sum()
n_bon          = ((df_predictions['Erreur_Pct'] > SEUIL_EXCELLENT) &
                  (df_predictions['Erreur_Pct'] <= SEUIL_ACCEPTABLE)).sum()
n_limite       = ((df_predictions['Erreur_Pct'] > SEUIL_ACCEPTABLE) &
                  (df_predictions['Erreur_Pct'] <= SEUIL_INACCEPTABLE)).sum()
n_inacceptable = (df_predictions['Erreur_Pct'] > SEUIL_INACCEPTABLE).sum()
n_acceptable   = n_excellent + n_bon  # ≤ 20%

pct_excellent    = n_excellent    / n_total * 100
pct_bon          = n_bon          / n_total * 100
pct_limite       = n_limite       / n_total * 100
pct_inacceptable = n_inacceptable / n_total * 100
pct_acceptable   = n_acceptable   / n_total * 100

print(f'\n── Résultats ──\n')
print(f'  ► Excellent    (≤{SEUIL_EXCELLENT:>2}%) : {n_excellent:>6,}  ({pct_excellent:5.1f} %)')
print(f'  ► Bon          (≤{SEUIL_ACCEPTABLE:>2}%) : {n_bon:>6,}  ({pct_bon:5.1f} %)')
print(f'  ► Limite       (≤{SEUIL_INACCEPTABLE:>2}%) : {n_limite:>6,}  ({pct_limite:5.1f} %)')
print(f'  ► Insuffisant  (>{SEUIL_INACCEPTABLE:>2}%) : {n_inacceptable:>6,}  ({pct_inacceptable:5.1f} %)')
print(f'\n  ► TOTAL acceptable (≤{SEUIL_ACCEPTABLE}%) : {n_acceptable:,} / {n_total:,}  ({pct_acceptable:.1f} %)')

# ── Score de confiance par magasin ───────────────────────────────────────────
print(f'\n── Score de confiance par magasin ──\n')
df_score = df_predictions.groupby('Store').apply(
    lambda g: pd.Series({
        'Score_Confiance_%': round((g['Erreur_Pct'] <= SEUIL_ACCEPTABLE).mean() * 100, 1),
        'MAE_$':             round(g['Erreur_Abs'].mean(), 0),
        'N':                 len(g)
    })
).reset_index()

df_score['Verdict'] = df_score['Score_Confiance_%'].apply(
    lambda s: 'GO' if s >= 80 else ('CONDITIONNEL' if s >= 60 else 'NO-GO')
)
df_score = df_score.sort_values('Score_Confiance_%', ascending=False)
df_score.index = range(1, len(df_score) + 1)
display(df_score)

# ── Verdict global ───────────────────────────────────────────────────────────
n_go   = (df_score['Verdict'] == 'GO').sum()
n_cond = (df_score['Verdict'] == 'CONDITIONNEL').sum()
n_nogo = (df_score['Verdict'] == 'NO-GO').sum()

if pct_acceptable >= 75:
    verdict = 'GO — Modèle validé pour la mise en production'
    verdict_color = '#27ae60'
elif pct_acceptable >= 60:
    verdict = 'CONDITIONNEL — Utilisation possible avec surveillance'
    verdict_color = '#f39c12'
else:
    verdict = 'NO-GO — Modèle à améliorer avant déploiement'
    verdict_color = '#e74c3c'

print(f'\n{"═" * 70}')
print(f'  VERDICT GLOBAL : {verdict}')
print(f'  Score de couverture (≤{SEUIL_ACCEPTABLE}%) : {pct_acceptable:.1f} %')
print(f'  Magasins — GO : {n_go}  |  Conditionnels : {n_cond}  |  NO-GO : {n_nogo}')
print(f'{"═" * 70}')

# ── Visualisations ───────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('2.4.3 — Validation finale par les utilisateurs métier', fontsize=14, fontweight='bold')

# Camembert : répartition des prévisions par qualité
ax1 = axes[0]
categories = [f'Excellent\n(≤{SEUIL_EXCELLENT}%)',
              f'Bon\n({SEUIL_EXCELLENT}-{SEUIL_ACCEPTABLE}%)',
              f'Limite\n({SEUIL_ACCEPTABLE}-{SEUIL_INACCEPTABLE}%)',
              f'Insuffisant\n(>{SEUIL_INACCEPTABLE}%)']
values      = [n_excellent, n_bon, n_limite, n_inacceptable]
colors_pie  = ['#27ae60', '#2ecc71', '#f39c12', '#e74c3c']
explode     = (0.05, 0, 0, 0.05)
wedges, texts, autotexts = ax1.pie(
    values, labels=categories, colors=colors_pie,
    autopct='%1.1f%%', startangle=90, explode=explode,
    textprops={'fontsize': 10}
)
for at in autotexts:
    at.set_fontsize(9)
ax1.set_title('Répartition des prévisions par qualité', fontsize=12, fontweight='bold')

# Score de confiance par magasin (barplot horizontal)
ax2 = axes[1]
df_s = df_score.sort_values('Score_Confiance_%')
colors_bar = ['#27ae60' if s >= 80 else ('#f39c12' if s >= 60 else '#e74c3c')
              for s in df_s['Score_Confiance_%']]
ax2.barh(range(len(df_s)), df_s['Score_Confiance_%'], color=colors_bar, alpha=0.85)
ax2.axvline(80, color='#27ae60', linestyle='--', linewidth=1.8, label='Seuil GO (80%)')
ax2.axvline(60, color='#f39c12', linestyle='--', linewidth=1.8, label='Seuil Conditionnel (60%)')
ax2.set_yticks(range(len(df_s)))
ax2.set_yticklabels([f"Store {int(s)}" for s in df_s['Store']], fontsize=8)
ax2.set_xlabel('Score de confiance (%)', fontsize=11)
ax2.set_title('Score de confiance par magasin', fontsize=12, fontweight='bold')
ax2.legend(fontsize=9)
ax2.set_xlim(0, 108)
for i, (idx_r, row) in enumerate(df_s.iterrows()):
    ax2.text(row['Score_Confiance_%'] + 0.5, i, f"{row['Score_Confiance_%']:.0f}%",
             va='center', fontsize=7)

plt.tight_layout()
plt.savefig('../images/243_validation_metier.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\n  ✅ Figure sauvegardée → images/243_validation_metier.png')
print(f'\n{"═" * 70}')
print(f'  ✅ Section 2.4 complète — Validation métier terminée')
print(f'{"═" * 70}')
````
