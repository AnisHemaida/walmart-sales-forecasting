## <span id="21" style="color:#2E86C1;">2.1 Préparation des données pour la modélisation</span>

Avant d'entraîner tout modèle de machine learning, les données brutes issues de l'étape `Compréhension et Préparation des Données` doivent être **transformées et structurées** de façon à être exploitables par les algorithmes de prévision. Cette phase de préparation est déterminante : la qualité des features construites conditionne directement la précision des prédictions.

Cette section couvre trois étapes structurantes :

| Étape | Contenu |
|---|---|
| **2.1.1 Encodage** | Transformation des variables catégorielles (`Store`, `Dept`, `Type`, `IsHoliday`) en valeurs numériques exploitables |
| **2.1.2 Feature engineering** | Extraction de variables temporelles (`Year`, `Month`, `Week`), indicateurs de saisonnalité et variables de contexte promotionnel |
| **2.1.3 Split train/test** | Séparation temporelle des données en jeu d'entraînement et jeu de validation, en respectant la chronologie des ventes |


> ⚠️ **Principe fondamental** : le split train/test doit être **temporel** et non aléatoire. Utiliser un split aléatoire entraînerait une **fuite de données** (*data leakage*) — le modèle verrait des semaines futures lors de l'entraînement, ce qui fausserait l'évaluation des performances.


````python
# ──────────────────────────────────────────────────────────────────
# IMPORT — Chargement de df_train_model depuis EPIC1
# ──────────────────────────────────────────────────────────────────
import pandas as pd

df_train = pd.read_parquet("../Data/prepared/df_train.parquet")

print(f"✅ df_train chargé depuis EPIC1")

df_types = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values]
})

display(
    df_types.style
    .set_properties(**{
        "text-align": "left",
        "padding": "8px",
        "border": "1px solid #ddd"
    })
    .set_caption("Types des variables du dataset")
)

df_train.head()





````

### <span id="211" style="color:#2E86C1;">2.1.1 Encodage des variables catégorielles</span>

Le dataset issu de l'EPIC1 contient **deux variables catégorielles à haute cardinalité** (`Store` : 45, `Dept` : 81) et **deux variables à faible cardinalité** (`Type` : 3, `IsHoliday` : 2). Chaque type nécessite une stratégie d'encodage adaptée :

| Variable | Type actuel | Modalités | Stratégie | Justification |
|---|---|---|---|---|
| `Store` | int64 | 45 | **Target Encoding** (lissé) | Encode le niveau moyen de ventes par magasin — évite 45 colonnes One-Hot |
| `Dept` | int64 | 81 | **Target Encoding** (lissé) | Encode le niveau moyen de ventes par département — évite 81 colonnes One-Hot |
| `Type` | category | A, B, C | **Ordinal Encoding** (A=1, B=2, C=3) | Variable ordinale liée à `Size` (A = grands magasins → C = petits) |
| `IsHoliday` | bool | True / False | Conversion `int` (0/1) | Déjà binaire |

**Pourquoi le Target Encoding pour Store et Dept ?**
- Le One-Hot créerait **45 + 81 = 126 colonnes** supplémentaires, alourdissant le modèle
- Le Target Encoding résume chaque catégorie par sa **moyenne de `Weekly_Sales`**, capturant le niveau de ventes propre à chaque magasin/département
- Un **lissage (smoothing)** est appliqué pour éviter la fuite de données (*target leakage*) : les moyennes bruitées des petits groupes sont régularisées vers la moyenne globale.



````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.1.1 — Encodage des variables catégorielles
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np

print("=" * 70)
print("  ENCODAGE DES VARIABLES CATÉGORIELLES")
print("=" * 70)

# ── 1) Target Encoding — Store (lissé avec smoothing) ────────────────────────
print("\n── 1) Target Encoding : Store ──")

global_mean = df_train['Weekly_Sales'].mean()
smoothing = 100  # facteur de lissage

store_stats = df_train.groupby('Store')['Weekly_Sales'].agg(['mean', 'count'])
store_stats['Store_encoded'] = (
    (store_stats['count'] * store_stats['mean'] + smoothing * global_mean)
    / (store_stats['count'] + smoothing)
)
store_mapping = store_stats['Store_encoded'].to_dict()

# Garder Store original pour référence
df_train['Store_orig'] = df_train['Store']
df_train['Store'] = df_train['Store_orig'].map(store_mapping)

print(f"  ✅ Store → Target Encoding (smoothing={smoothing})")
print(f"     • Moyenne globale : {global_mean:,.0f} $")
print(f"     • Plage encodée  : [{df_train['Store'].min():,.0f} — {df_train['Store'].max():,.0f}] $")

# Aperçu top 5 / bottom 5
top5 = store_stats.nlargest(5, 'Store_encoded')[['mean', 'count', 'Store_encoded']]
bot5 = store_stats.nsmallest(5, 'Store_encoded')[['mean', 'count', 'Store_encoded']]
print("\n     Top 5 magasins (ventes moyennes les plus élevées) :")
for idx, row in top5.iterrows():
    print(f"       Store {idx:>2} → {row['Store_encoded']:>10,.0f} $ (moy brute: {row['mean']:>10,.0f} $, n={row['count']:>5,})")
print("     Bottom 5 magasins :")
for idx, row in bot5.iterrows():
    print(f"       Store {idx:>2} → {row['Store_encoded']:>10,.0f} $ (moy brute: {row['mean']:>10,.0f} $, n={row['count']:>5,})")

# ── 2) Target Encoding — Dept (lissé avec smoothing) ─────────────────────────
print("\n── 2) Target Encoding : Dept ──")

dept_stats = df_train.groupby('Dept')['Weekly_Sales'].agg(['mean', 'count'])
dept_stats['Dept_encoded'] = (
    (dept_stats['count'] * dept_stats['mean'] + smoothing * global_mean)
    / (dept_stats['count'] + smoothing)
)
dept_mapping = dept_stats['Dept_encoded'].to_dict()

df_train['Dept_orig'] = df_train['Dept']
df_train['Dept'] = df_train['Dept_orig'].map(dept_mapping)

print(f"  ✅ Dept → Target Encoding (smoothing={smoothing})")
print(f"     • Plage encodée : [{df_train['Dept'].min():,.0f} — {df_train['Dept'].max():,.0f}] $")
print(f"     • Nb départements : {len(dept_mapping)}")

# ── 3) Ordinal Encoding — Type (A, B, C) → lié à Size ────────────────────────
print("\n── 3) Ordinal Encoding : Type ──")

type_mapping = {'A': 1, 'B': 2, 'C': 3}
df_train['Type'] = df_train['Type'].map(type_mapping)

print(f"  ✅ Type → Ordinal Encoding (A=1, B=2, C=3)")
print(f"     • Justification : Type est lié à Size (A = grands magasins → C = petits)")
for label, code in type_mapping.items():
    n = (df_train['Type'] == code).sum()
    print(f"     • Type {label} (={code}) : {n:,} lignes ({n/len(df_train)*100:.1f}%)")

# ── 4) IsHoliday — conversion bool → int (0/1) ──────────────────────────────
print("\n── 4) IsHoliday ──")
df_train['IsHoliday'] = df_train['IsHoliday'].astype(int)
n_holiday = df_train['IsHoliday'].sum()
print(f"  ✅ IsHoliday → int (True→1, False→0)")
print(f"     • Type après conversion : {df_train['IsHoliday'].dtype}")
print(f"     • Semaines normales : {len(df_train) - n_holiday:,}")
print(f"     • Semaines de fête  : {n_holiday:,} ({n_holiday/len(df_train)*100:.1f}%)")

# ── 5) Vérification finale ───────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  VÉRIFICATION APRÈS ENCODAGE")
print("=" * 70)

print(f"\n  ► Shape : {df_train.shape[0]:,} lignes × {df_train.shape[1]} colonnes")
print(f"  ► Colonnes catégorielles restantes : {list(df_train.select_dtypes(include='category').columns)}")

df_types_after = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values],
    "Exemple": [str(df_train[c].iloc[0]) for c in df_train.columns]
})

display(
    df_types_after.style
    .set_properties(**{"text-align": "left", "padding": "8px", "border": "1px solid #ddd"})
    .set_caption("Types des variables après encodage")

)

df_train.head()

````

**Commentaire — Résultats de l'encodage :**

Les 4 transformations appliquées ont converti l'ensemble des variables catégorielles en valeurs numériques exploitables par les algorithmes de machine learning :
- **Store** et **Dept** ont été encodés via un **Target Encoding lissé** (smoothing = 100). Le lissage régularise les moyennes des groupes à faible effectif vers la moyenne globale, limitant ainsi le risque de *target leakage*. Les colonnes `Store_orig` et `Dept_orig` sont conservées pour référence.
- **Type** a été converti en **Ordinal Encoding** (A=1, B=2, C=3), reflétant le lien ordinal avec la taille des magasins (`Size`) : les magasins de Type A sont les plus grands, ceux de Type C les plus petits.

- **IsHoliday** a été converti de `bool` en `int64` (0/1), format attendu par les modèles.

>  **Résultat** : le dataset ne contient plus aucune colonne catégorielle (`category`). Toutes les variables sont désormais numériques (`float64`, `int64` ou `int32`), prêtes pour les étapes de feature engineering et de modélisation.

### <span id="212" style="color:#2E86C1;">2.1.2 Création de variables temporelles et contextuelles</span>

<div style="background:#2a2a3e; border-left:4px solid #89b4fa; padding:10px 14px; border-radius:5px; margin-top:16px; color:#FAD7A0;">
<strong>Transformation de la variable Date</strong>
</div>

La colonne `Date` (datetime64) ne peut pas être utilisée telle quelle par les algorithmes de machine learning. Elle doit être **décomposée en composantes temporelles** qui capturent les cycles et tendances des ventes.

| Variable créée | Source | Rôle |
|---|---|---|
| `Year` | `Date.year` | Capturer la tendance annuelle (croissance/déclin) |
| `Month` | `Date.month` | Saisonnalité mensuelle |
| `Week` | `Date.isocalendar().week` | Numéro de semaine ISO — granularité principale des ventes |
| `Month_sin` / `Month_cos` | sin/cos(2π × Month/12) | Encodage cyclique du mois (décembre ↔ janvier sont proches) |
| `Week_sin` / `Week_cos` | sin/cos(2π × Week/52) | Encodage cyclique de la semaine (sem. 52 ↔ sem. 1 sont proches) |

> 💡 **Pourquoi l'encodage cyclique ?** Le mois 12 (décembre) et le mois 1 (janvier) sont séparés de 11 unités avec un encodage linéaire, alors qu'ils sont **voisins dans le calendrier**. L'encodage sin/cos projette le temps sur un cercle, préservant cette proximité naturelle.

````python

print("=" * 70)
print("  TRANSFORMATION DE LA VARIABLE TEMPORELLE (Date)")
print("=" * 70)

# ── 1) Variables temporelles de base ─────────────────────────────────────────
print("\n── 1) Variables temporelles de base ──")

df_train['Year']  = df_train['Date'].dt.year
df_train['Month'] = df_train['Date'].dt.month
df_train['Week']  = df_train['Date'].dt.isocalendar().week.astype(int)

print(f"  ✅ Year  → {df_train['Year'].unique()}")
print(f"  ✅ Month → {sorted(df_train['Month'].unique())}")
print(f"  ✅ Week  → [{df_train['Week'].min()} — {df_train['Week'].max()}]")

# ── 2) Encodage cyclique (sin/cos) ──────────────────────────────────────────
print("\n── 2) Encodage cyclique (sin/cos) ──")

df_train['Month_sin'] = np.sin(2 * np.pi * df_train['Month'] / 12)
df_train['Month_cos'] = np.cos(2 * np.pi * df_train['Month'] / 12)
df_train['Week_sin']  = np.sin(2 * np.pi * df_train['Week'] / 52)
df_train['Week_cos']  = np.cos(2 * np.pi * df_train['Week'] / 52)

print(f"  ✅ Month_sin / Month_cos créés")
print(f"  ✅ Week_sin  / Week_cos  créés")
print(f"     • Exemple (semaine 1)  : sin={np.sin(2*np.pi*1/52):.4f}, cos={np.cos(2*np.pi*1/52):.4f}")
print(f"     • Exemple (semaine 52) : sin={np.sin(2*np.pi*52/52):.4f}, cos={np.cos(2*np.pi*52/52):.4f}")
print(f"     → Les semaines 1 et 52 sont proches sur le cercle ✓")


# ── 3) Vérification finale ───────────────────────────────────────────────────


df_types_fe = pd.DataFrame({
    "Variable": df_train.columns,
    "Type": [str(t) for t in df_train.dtypes.values],
    "Exemple": [str(df_train[c].iloc[0]) for c in df_train.columns]
})

display(
    df_types_fe.style
    .set_properties(**{"text-align": "left", "padding": "8px", "border": "1px solid #ddd"})
    .set_caption("Variables après transformation de la variable Date")
)

df_train.head()
````

## <span id="22" style="color:#2E86C1;">2.2 Développement et entraînement des modèles de prévision</span>

Cette section porte sur le **développement et l'entraînement des modèles de prévision** des ventes hebdomadaires. À partir des données préparées en section 2.1 (`X_train` : 383 872 lignes × 31 features), nous entraînons progressivement **4 modèles** selon une approche du plus simple au plus complexe :

| # | Modèle | Type | Rôle |
|---|--------|------|------|
| 1 | **Ridge** | Régression linéaire pénalisée (L2) | **Baseline** — référence minimale à battre |
| 2 | **Random Forest** | Ensemble d'arbres (bagging) | Première approche non-linéaire |
| 3 | **LightGBM** | Gradient Boosting (histograms) | Modèle principal — rapide et performant |
| 4 | **XGBoost** | Gradient Boosting (exact/approx) | Validation croisée — robustesse |

**Métrique d'évaluation principale : WMAE** (Weighted Mean Absolute Error)

$$\text{WMAE} = \frac{1}{\sum w_i} \sum_{i=1}^{n} w_i \cdot |y_i - \hat{y}_i|$$

où $w_i = 5$ si la semaine est un **jour férié** (IsHoliday = 1), sinon $w_i = 1$.

> *« se tromper pendant une semaine fériée est **5 fois plus grave** que se tromper une semaine normale. »* 

Cette pondération reflète l'importance business des périodes de forte affluence (Thanksgiving, Noël, Super Bowl, Labour Day). Une erreur de prévision durant ces semaines coûte beaucoup plus cher : sous-estimation → rupture de stock et ventes perdues, surestimation → surplus et gaspillage.

**Démarche pour chaque modèle :**
1. Entraînement sur `X_train` / `y_train`
2. Prédiction sur `X_test`
3. Calcul de la WMAE et des métriques complémentaires (MAE, RMSE, R²)
4. Ajustement des hyperparamètres si nécessaire

### <span id="221" style="color:#2E86C1;">2.2.1 Sélection des algorithmes de prévision adaptés</span>

Le choix des algorithmes repose sur une **stratégie progressive** : partir d'un modèle simple (baseline) puis monter en complexité pour mesurer le gain réel de chaque approche.

#### 1. Ridge Regression (Baseline)

Régression linéaire avec pénalisation **L2** ($\lambda \sum \beta_j^2$). Elle suppose une relation linéaire entre les features et les ventes. C'est la **référence minimale** : si un modèle avancé ne bat pas Ridge, il n'apporte aucune valeur.

| Avantage | Limite |
|----------|--------|
| Très rapide à entraîner | Ne capture pas les interactions non-linéaires |
| Interprétable (coefficients) | Sensible à l'échelle → nécessite la normalisation |
| Robuste au surapprentissage | Performance plafonnée sur données complexes |

#### 2. Random Forest

Ensemble de **centaines d'arbres de décision** entraînés sur des sous-échantillons aléatoires (bagging). Chaque arbre vote, et la prédiction finale est la **moyenne des votes**.

| Avantage | Limite |
|----------|--------|
| Capture les non-linéarités et interactions | Plus lent que Ridge |
| Robuste aux outliers | Moins performant que le boosting en général |
| Pas besoin de normalisation | Peut sur-apprendre si trop d'arbres peu profonds |

#### 3. LightGBM (Modèle principal)

Gradient Boosting basé sur des **histogrammes** : les arbres sont construits **séquentiellement**, chacun corrigeant les erreurs du précédent. L'approche par histogrammes rend LightGBM **très rapide** même sur de gros volumes.

| Avantage | Limite |
|----------|--------|
| Très performant sur données tabulaires | Risque de surapprentissage si mal réglé |
| Rapide (histogrammes + croissance par feuille) | Plus d'hyperparamètres à ajuster |
| Gère nativement les NaN et les catégorielles | Moins interprétable que Ridge |

#### 4. XGBoost (Validation)

Autre implémentation du Gradient Boosting, avec une **régularisation L1+L2** intégrée. Utilisé pour **valider** les résultats de LightGBM et confirmer la robustesse.

| Avantage | Limite |
|----------|--------|
| Régularisation native (L1 + L2) | Plus lent que LightGBM |
| Très utilisé en compétitions Kaggle | Consomme plus de mémoire |
| Complémentaire à LightGBM | Hyperparamètres similaires |

#### Résumé de la stratégie

```
Ridge (baseline)  →  Random Forest  →  LightGBM  →  XGBoost
   linéaire           non-linéaire      boosting      validation
   simple              intermédiaire     principal     robustesse
```

> Cette approche permet de **quantifier le gain** à chaque étape : si Ridge donne une WMAE de X et LightGBM donne Y, on sait exactement combien la complexité supplémentaire apporte.

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.1 — Sélection des algorithmes & Définition de la métrique WMAE
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("=" * 70)
print("  SÉLECTION DES ALGORITHMES & MÉTRIQUE WMAE")
print("=" * 70)

# ── 1) Fonction WMAE ─────────────────────────────────────────────────────────
def wmae(y_true, y_pred, is_holiday):
    """
    Weighted Mean Absolute Error (métrique Walmart/Kaggle).
    Poids : 5 pour les semaines fériées, 1 sinon.
    """
    weights = np.where(is_holiday == 1, 5, 1)
    return np.sum(weights * np.abs(y_true - y_pred)) / np.sum(weights)

print("\n✅ Fonction wmae(y_true, y_pred, is_holiday) définie")

# ── 2) Récupérer IsHoliday pour le calcul des poids ─────────────────────────
# IsHoliday est dans X_train et X_test (c'est une feature)
is_holiday_train = X_train['IsHoliday'].values
is_holiday_test  = X_test['IsHoliday'].values

n_holiday_train = (is_holiday_train == 1).sum()
n_holiday_test  = (is_holiday_test == 1).sum()

print(f"\n── IsHoliday dans les données ──")
print(f"  ► Train : {n_holiday_train:,} semaines fériées ({n_holiday_train/len(X_train)*100:.1f}%)")
print(f"  ► Test  : {n_holiday_test:,} semaines fériées ({n_holiday_test/len(X_test)*100:.1f}%)")

# ── 3) Fonction d'évaluation complète ────────────────────────────────────────
def evaluate_model(name, y_true, y_pred, is_holiday):
    """Affiche toutes les métriques pour un modèle donné."""
    w = wmae(y_true, y_pred, is_holiday)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    
    print(f"\n{'─' * 50}")
    print(f"  {name}")
    print(f"{'─' * 50}")
    print(f"  WMAE  : {w:,.0f} $")
    print(f"  MAE   : {mae:,.0f} $")
    print(f"  RMSE  : {rmse:,.0f} $")
    print(f"  R²    : {r2:.4f}")
    
    return {'model': name, 'WMAE': w, 'MAE': mae, 'RMSE': rmse, 'R2': r2}

print("✅ Fonction evaluate_model(name, y_true, y_pred, is_holiday) définie")

# ── 4) Liste des modèles à entraîner ────────────────────────────────────────
print(f"\n── Modèles sélectionnés ──")
models = {
    'Ridge': Ridge(alpha=1.0),
    'Random Forest': RandomForestRegressor(
        n_estimators=200, max_depth=20, min_samples_leaf=5,
        n_jobs=-1, random_state=42
    ),
}

# ── Détection CUDA (XGBoost uniquement) ────────────────────────────────────────────
import subprocess as _sp
try:
    _r = _sp.run(['nvidia-smi'], capture_output=True, text=True, timeout=5)
    _cuda = _r.returncode == 0
except Exception:
    _cuda = False
if _cuda:
    print("  ✅ GPU NVIDIA détecté → CUDA activé pour XGBoost")
else:
    print("  ℹ️  Pas de GPU → entraînement sur CPU")

# LightGBM et XGBoost — import conditionnel
try:
    import lightgbm as lgb
    models['LightGBM'] = lgb.LGBMRegressor(
        n_estimators=500, max_depth=8, learning_rate=0.05,
        num_leaves=63, min_child_samples=50,
        subsample=0.8, colsample_bytree=0.8,
        n_jobs=-1, random_state=42, verbose=-1
    )
    print("  ✅ LightGBM disponible")
except ImportError:
    print("  ⚠️ LightGBM non installé — pip install lightgbm")

try:
    import xgboost as xgb
    _xgb_gpu = {'device': 'cuda', 'tree_method': 'hist'} if _cuda else {}
    models['XGBoost'] = xgb.XGBRegressor(
        n_estimators=500, max_depth=8, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        reg_alpha=0.1, reg_lambda=1.0,
        n_jobs=-1, random_state=42, verbosity=0,
        **_xgb_gpu
    )
    print("  ✅ XGBoost disponible")
except ImportError:
    print("  ⚠️ XGBoost non installé — pip install xgboost")

print(f"\n► {len(models)} modèles prêts : {list(models.keys())}")

# Stockage des résultats
results = []
trained_models = {}
````

Cette cellule met en place l'infrastructure d'évaluation et les 4 modèles retenus pour la prévision des ventes :

### <span id="223" style="color:#2E86C1;">2.2.3 Ajustement des hyperparamètres</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-bottom:10px; color:#D5D8DC;">

**Objectif** — Affiner les hyperparamètres des **2 meilleurs modèles** (XGBoost et LightGBM) afin de réduire davantage le WMAE par rapport aux résultats initiaux (1 343 $ et 1 378 $).

**Approche — `RandomizedSearchCV`** :
- **Pourquoi RandomizedSearchCV ?** Avec 383 872 lignes et de nombreux hyperparamètres, un `GridSearchCV` exhaustif serait prohibitif. `RandomizedSearchCV` échantillonne aléatoirement `n_iter` combinaisons dans l'espace de recherche, offrant un bon compromis entre exploration et temps de calcul.
- **Validation croisée** : 3 folds (compromis entre robustesse statistique et temps d'exécution).
- **Métrique d'optimisation** : `neg_mean_absolute_error` (proxy de la WMAE — la pondération par IsHoliday sera vérifiée après avec la vraie WMAE sur le jeu de test).
- **Espace de recherche** : les hyperparamètres les plus influents pour chaque algorithme (profondeur, taux d'apprentissage, régularisation, sous-échantillonnage).

</div>

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.3 — Ajustement des hyperparamètres (XGBoost & LightGBM)
# ══════════════════════════════════════════════════════════════════════════════
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
import time, warnings, subprocess
warnings.filterwarnings('ignore')

print("=" * 70)
print("  AJUSTEMENT DES HYPERPARAMÈTRES — RandomizedSearchCV + TimeSeriesSplit")
print("=" * 70)

# ── Détection CUDA (XGBoost uniquement) ──────────────────────────────────────
try:
    _res = subprocess.run(['nvidia-smi'], capture_output=True, text=True, timeout=5)
    cuda_available = _res.returncode == 0
except Exception:
    cuda_available = False

if cuda_available:
    print("  ✅ GPU NVIDIA détecté → CUDA activé pour XGBoost")
    xgb_device_params = {'device': 'cuda', 'tree_method': 'hist'}
else:
    print("  ℹ️  Aucun GPU détecté → entraînement sur CPU")
    xgb_device_params = {}

print("  ℹ️  LightGBM → CPU (version pip sans support GPU)")

# ── Résultats avant tuning (section 2.2.2) ───────────────────────────────────
baseline_wmae = {
    'XGBoost':  df_results.loc[df_results['model'] == 'XGBoost', 'WMAE'].values[0],
    'LightGBM': df_results.loc[df_results['model'] == 'LightGBM', 'WMAE'].values[0],
}

# ── Tri de X_train par Date pour TimeSeriesSplit ──────────────────────────────
# Copie locale uniquement — X_train global inchangé pour les cellules suivantes
_sort_idx  = df_train.loc[X_train.index, 'Date'].sort_values().index
X_train_ts = X_train.loc[_sort_idx]
y_train_ts = y_train.loc[_sort_idx]
print(f"\n  ✅ X_train_ts trié par Date ({len(X_train_ts):,} lignes)")

tscv = TimeSeriesSplit(n_splits=3)
print(f"  ✅ TimeSeriesSplit(n_splits=3)\n")

# ══════════════════════════════════════════════════════════════════════════════
# 1) XGBoost — RandomizedSearchCV + TimeSeriesSplit
# ══════════════════════════════════════════════════════════════════════════════
print(f"{'━' * 60}")
print(f"  ► Tuning : XGBoost  (20 tirages × 3 folds)")
print(f"{'━' * 60}")

import xgboost as xgb

xgb_param_dist = {
    'n_estimators':     [300, 500, 700, 1000],
    'max_depth':        [6, 8, 10, 12],
    'learning_rate':    [0.01, 0.03, 0.05, 0.08, 0.1],
    'subsample':        [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'reg_alpha':        [0, 0.01, 0.1, 0.5],
    'reg_lambda':       [0.5, 1.0, 2.0, 5.0],
    'min_child_weight': [3, 5, 10, 20],
}

xgb_base = xgb.XGBRegressor(
    n_jobs=-1, random_state=42, verbosity=0,
    **xgb_device_params
)

t0 = time.time()
xgb_search = RandomizedSearchCV(
    xgb_base,
    param_distributions=xgb_param_dist,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=tscv,
    random_state=42,
    n_jobs=1,
    verbose=0,
)
xgb_search.fit(X_train_ts, y_train_ts)
xgb_time = time.time() - t0

print(f"  ✅ Terminé en {xgb_time/60:.1f} min ({int(xgb_time)} s)")
print(f"  ► Meilleur score CV (neg_MAE) : {xgb_search.best_score_:,.0f}")
print(f"\n  Meilleurs hyperparamètres :")
for k, v in xgb_search.best_params_.items():
    print(f"    {k:20s} = {v}")

y_pred_xgb = xgb_search.best_estimator_.predict(X_test)
res_xgb = evaluate_model('XGBoost (tuné)', y_test.values, y_pred_xgb, is_holiday_test)

# ══════════════════════════════════════════════════════════════════════════════
# 2) LightGBM — RandomizedSearchCV + TimeSeriesSplit   (CPU uniquement)
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n{'━' * 60}")
print(f"  ► Tuning : LightGBM  (20 tirages × 3 folds, CPU)")
print(f"{'━' * 60}")

import lightgbm as lgb

lgb_param_dist = {
    'n_estimators':      [300, 500, 700, 1000],
    'max_depth':         [6, 8, 10, 12, -1],
    'learning_rate':     [0.01, 0.03, 0.05, 0.08, 0.1],
    'num_leaves':        [31, 63, 127, 255],
    'min_child_samples': [20, 50, 100],
    'subsample':         [0.7, 0.8, 0.9],
    'colsample_bytree':  [0.7, 0.8, 0.9],
    'reg_alpha':         [0, 0.01, 0.1, 0.5],
    'reg_lambda':        [0, 0.1, 1.0, 5.0],
}

lgb_base = lgb.LGBMRegressor(
    n_jobs=-1, random_state=42, verbose=-1
)

t0 = time.time()
lgb_search = RandomizedSearchCV(
    lgb_base,
    param_distributions=lgb_param_dist,
    n_iter=20,
    scoring='neg_mean_absolute_error',
    cv=tscv,
    random_state=42,
    n_jobs=1,
    verbose=0,
)
lgb_search.fit(X_train_ts, y_train_ts)
lgb_time = time.time() - t0

print(f"  ✅ Terminé en {lgb_time/60:.1f} min ({int(lgb_time)} s)")
print(f"  ► Meilleur score CV (neg_MAE) : {lgb_search.best_score_:,.0f}")
print(f"\n  Meilleurs hyperparamètres :")
for k, v in lgb_search.best_params_.items():
    print(f"    {k:20s} = {v}")

y_pred_lgb = lgb_search.best_estimator_.predict(X_test)
res_lgb = evaluate_model('LightGBM (tuné)', y_test.values, y_pred_lgb, is_holiday_test)

# ══════════════════════════════════════════════════════════════════════════════
# 3) Tableau comparatif : Avant vs Après tuning
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n\n{'=' * 70}")
print("  COMPARAISON AVANT / APRÈS TUNING")
print(f"{'=' * 70}\n")

comparison = []
for name_key, res_tuned, y_pred_tuned in [
    ('XGBoost', res_xgb, y_pred_xgb),
    ('LightGBM', res_lgb, y_pred_lgb),
]:
    before = baseline_wmae[name_key]
    after  = res_tuned['WMAE']
    delta  = after - before
    pct_change = (delta / before) * 100
    comparison.append({
        'Modèle': name_key,
        'WMAE avant': f"{before:,.0f} $",
        'WMAE après': f"{after:,.0f} $",
        'Δ WMAE': f"{delta:+,.0f} $",
        'Variation': f"{pct_change:+.1f} %",
        'R² après': f"{res_tuned['R2']:.4f}",
    })

df_comparison = pd.DataFrame(comparison)
df_comparison.index = range(1, len(df_comparison) + 1)
display(df_comparison)

# ── Mettre à jour les modèles entraînés ──────────────────────────────────────
trained_models['XGBoost']  = xgb_search.best_estimator_
trained_models['LightGBM'] = lgb_search.best_estimator_

best_tuned = min(
    [('XGBoost', res_xgb), ('LightGBM', res_lgb)],
    key=lambda x: x[1]['WMAE']
)
print(f"\n🏆 Champion après tuning : {best_tuned[0]}")
print(f"   WMAE = {best_tuned[1]['WMAE']:,.0f} $  |  R² = {best_tuned[1]['R2']:.4f}")
````

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.2.4 — Génération des prévisions (modèle champion : XGBoost tuné)
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np
import pandas as pd

print("=" * 70)
print("  GÉNÉRATION DES PRÉVISIONS — XGBoost (tuné)")
print("=" * 70)

# ── 1) Modèle champion ──────────────────────────────────────────────────────
champion_name = best_tuned[0]
champion_model = trained_models[champion_name]
print(f"\n  ► Modèle champion : {champion_name}")
print(f"  ► WMAE (test)     : {best_tuned[1]['WMAE']:,.0f} $")
print(f"  ► R² (test)       : {best_tuned[1]['R2']:.4f}")

# ── 2) Générer les prédictions finales ───────────────────────────────────────
print(f"\n── Prédictions sur l'ensemble de test ({X_test.shape[0]:,} lignes) ──")

y_pred_final = champion_model.predict(X_test)
print(f"  ✅ {len(y_pred_final):,} prévisions générées")

# ── 3) Construire le DataFrame de prévisions ────────────────────────────────
print("\n── Construction du DataFrame de prévisions ──")

df_predictions = pd.DataFrame({
    'Store':          df_train.loc[mask_test, 'Store_orig'].values,
    'Dept':           df_train.loc[mask_test, 'Dept_orig'].values,
    'Date':           df_train.loc[mask_test, 'Date'].values,
    'IsHoliday':      is_holiday_test,
    'Reel':           y_test.values,
    'Prevu':          y_pred_final,
})

# Colonnes dérivées
df_predictions['Erreur']     = df_predictions['Prevu'] - df_predictions['Reel']
df_predictions['Erreur_Abs'] = df_predictions['Erreur'].abs()
df_predictions['Erreur_Pct'] = (df_predictions['Erreur_Abs'] / df_predictions['Reel'].abs().replace(0, np.nan) * 100).round(2)

print(f"  ✅ DataFrame : {df_predictions.shape[0]:,} lignes × {df_predictions.shape[1]} colonnes")
print(f"  ► Colonnes   : {list(df_predictions.columns)}")

# ── 4) Statistiques descriptives des prévisions ─────────────────────────────
print(f"\n{'=' * 70}")
print("  STATISTIQUES DES PRÉVISIONS")
print(f"{'=' * 70}\n")

print(f"  ► Ventes réelles (test) :")
print(f"    Moyenne      : {df_predictions['Reel'].mean():>12,.0f} $")
print(f"    Médiane      : {df_predictions['Reel'].median():>12,.0f} $")
print(f"    Min          : {df_predictions['Reel'].min():>12,.0f} $")
print(f"    Max          : {df_predictions['Reel'].max():>12,.0f} $")

print(f"\n  ► Ventes prévues :")
print(f"    Moyenne      : {df_predictions['Prevu'].mean():>12,.0f} $")
print(f"    Médiane      : {df_predictions['Prevu'].median():>12,.0f} $")
print(f"    Min          : {df_predictions['Prevu'].min():>12,.0f} $")
print(f"    Max          : {df_predictions['Prevu'].max():>12,.0f} $")

print(f"\n  ► Erreurs absolues :")
print(f"    Moyenne (MAE): {df_predictions['Erreur_Abs'].mean():>12,.0f} $")
print(f"    Médiane      : {df_predictions['Erreur_Abs'].median():>12,.0f} $")
print(f"    90e centile  : {df_predictions['Erreur_Abs'].quantile(0.90):>12,.0f} $")
print(f"    95e centile  : {df_predictions['Erreur_Abs'].quantile(0.95):>12,.0f} $")
print(f"    Max          : {df_predictions['Erreur_Abs'].max():>12,.0f} $")

# Erreur médiane en %
median_pct = df_predictions['Erreur_Pct'].median()
print(f"\n  ► Erreur % médiane : {median_pct:.1f} %")

# ── 5) Répartition des erreurs ───────────────────────────────────────────────
print(f"\n{'=' * 70}")
print("  RÉPARTITION DES ERREURS")
print(f"{'=' * 70}\n")

seuils = [500, 1000, 2000, 5000]
for s in seuils:
    n = (df_predictions['Erreur_Abs'] <= s).sum()
    pct = n / len(df_predictions) * 100
    print(f"  ► Erreur ≤ {s:>5,} $ : {n:>6,} lignes ({pct:5.1f} %)")

# ── 6) Prévisions fériés vs normaux ─────────────────────────────────────────
print(f"\n{'=' * 70}")
print("  PRÉVISIONS : JOURS FÉRIÉS vs NORMAUX")
print(f"{'=' * 70}\n")

for label, mask_h in [("Normaux (IsHoliday=0)", df_predictions['IsHoliday'] == 0),
                       ("Fériés  (IsHoliday=1)", df_predictions['IsHoliday'] == 1)]:
    sub = df_predictions[mask_h]
    if len(sub) > 0:
        mae_h = sub['Erreur_Abs'].mean()
        med_h = sub['Erreur_Abs'].median()
        print(f"  ► {label} : {len(sub):,} lignes")
        print(f"    MAE     = {mae_h:,.0f} $")
        print(f"    Médiane = {med_h:,.0f} $\n")

# ── 7) Aperçu : meilleures et pires prévisions ──────────────────────────────
print(f"\n{'=' * 70}")
print("  TOP 10 — MEILLEURES PRÉVISIONS (erreur la plus faible)")
print(f"{'=' * 70}\n")

top10 = df_predictions.nsmallest(10, 'Erreur_Abs')[
    ['Store', 'Dept', 'Date', 'Reel', 'Prevu', 'Erreur', 'Erreur_Pct']
].copy()
top10['Reel']   = top10['Reel'].apply(lambda x: f"{x:,.0f} $")
top10['Prevu']  = top10['Prevu'].apply(lambda x: f"{x:,.0f} $")
top10['Erreur'] = top10['Erreur'].apply(lambda x: f"{x:+,.0f} $")
top10['Erreur_Pct'] = top10['Erreur_Pct'].apply(lambda x: f"{x:.2f} %")
top10.index = range(1, 11)
display(top10)

print(f"\n{'=' * 70}")
print("  TOP 10 — PIRES PRÉVISIONS (erreur la plus élevée)")
print(f"{'=' * 70}\n")

worst10 = df_predictions.nlargest(10, 'Erreur_Abs')[
    ['Store', 'Dept', 'Date', 'Reel', 'Prevu', 'Erreur', 'Erreur_Pct']
].copy()
worst10['Reel']   = worst10['Reel'].apply(lambda x: f"{x:,.0f} $")
worst10['Prevu']  = worst10['Prevu'].apply(lambda x: f"{x:,.0f} $")
worst10['Erreur'] = worst10['Erreur'].apply(lambda x: f"{x:+,.0f} $")
worst10['Erreur_Pct'] = worst10['Erreur_Pct'].apply(lambda x: f"{x:.2f} %")
worst10.index = range(1, 11)
display(worst10)

print(f"\n{'=' * 70}")
print(f"  ✅ DataFrame 'df_predictions' prêt ({df_predictions.shape[0]:,} lignes)")
print(f"{'=' * 70}")
````

### <span id="231" style="color:#2E86C1;">2.3.1 Choix des métriques d'évaluation</span>

<div style="background:#1B2631; border-left:4px solid #2E86C1; padding:14px 18px; border-radius:8px; margin-top:12px; color:#D5D8DC;">

**Objectif :** Définir et justifier les métriques utilisées pour évaluer les modèles de prévision des ventes. Le choix de la **WMAE** comme métrique principale est dicté par le concours Kaggle Walmart, mais nous calculons également des métriques complémentaires pour une vision complète.

**Métriques retenues :**
- **WMAE** (Weighted MAE) — métrique principale, pondère ×5 les semaines fériées
- **MAE** (Mean Absolute Error) — erreur moyenne en dollars, interprétable directement
- **RMSE** (Root Mean Squared Error) — pénalise les grosses erreurs
- **R²** (Coefficient de détermination) — variance expliquée par le modèle
- **MAPE** (Mean Absolute Percentage Error) — erreur relative en %

</div>

````python
# ══════════════════════════════════════════════════════════════════════════════
# 2.3.1 — Choix des métriques d'évaluation
# ══════════════════════════════════════════════════════════════════════════════
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("=" * 70)
print("  MÉTRIQUES D'ÉVALUATION RETENUES")
print("=" * 70)

# ── 1) Rappel des formules ───────────────────────────────────────────────────

# ── 2) Calcul complet pour le champion (XGBoost tuné) ────────────────────────
print(f"── Métriques du champion : {champion_name} (tuné) ──\n")

y_true = y_test.values
y_pred_ch = y_pred_final

# WMAE
w_test = np.where(is_holiday_test == 1, 5, 1)
val_wmae = np.sum(w_test * np.abs(y_true - y_pred_ch)) / np.sum(w_test)

# MAE
val_mae = mean_absolute_error(y_true, y_pred_ch)

# RMSE
val_rmse = np.sqrt(mean_squared_error(y_true, y_pred_ch))

# R²
val_r2 = r2_score(y_true, y_pred_ch)

# MAPE (excluant les ventes nulles/négatives pour éviter division par 0)
mask_valid = np.abs(y_true) > 0
val_mape = np.mean(np.abs(y_true[mask_valid] - y_pred_ch[mask_valid]) / np.abs(y_true[mask_valid])) * 100
n_excluded = (~mask_valid).sum()

print(f"  ► WMAE  = {val_wmae:>10,.0f} $   ← métrique principale (Kaggle)")
print(f"  ► MAE   = {val_mae:>10,.0f} $   ← erreur absolue moyenne")
print(f"  ► RMSE  = {val_rmse:>10,.0f} $   ← sensible aux outliers")
print(f"  ► R²    = {val_r2:>10.4f}     ← variance expliquée")
print(f"  ► MAPE  = {val_mape:>10.2f} %   ← erreur relative moyenne")
if n_excluded > 0:
    print(f"    (MAPE calculé sur {mask_valid.sum():,} lignes, {n_excluded:,} exclues car ventes = 0)")

# ── 3) Tableau récapitulatif formaté ─────────────────────────────────────────
print(f"\n{'=' * 70}")
print("  TABLEAU RÉCAPITULATIF — CHAMPION")
print(f"{'=' * 70}\n")

df_metrics = pd.DataFrame([{
    'Métrique': 'WMAE (pondérée)',
    'Valeur': f"{val_wmae:,.0f} $",
    'Rôle': 'Métrique principale — pondère x5 les fériés',
}, {
    'Métrique': 'MAE',
    'Valeur': f"{val_mae:,.0f} $",
    'Rôle': 'Erreur moyenne en dollars',
}, {
    'Métrique': 'RMSE',
    'Valeur': f"{val_rmse:,.0f} $",
    'Rôle': 'Pénalise les grandes erreurs',
}, {
    'Métrique': 'R²',
    'Valeur': f"{val_r2:.4f}",
    'Rôle': f'{val_r2*100:.2f}% de la variance expliquée',
}, {
    'Métrique': 'MAPE',
    'Valeur': f"{val_mape:.2f} %",
    'Rôle': 'Erreur relative — interprétable par le métier',
}])
df_metrics.index = range(1, len(df_metrics) + 1)
display(df_metrics)

````

<div style="background:#1e1e2e; border-left:4px solid #f9e2af; padding:14px 18px; border-radius:8px; margin-top:12px; color:#cdd6f4;">

**Commentaire — Métriques d'évaluation :**

| Métrique | Valeur | Interprétation |
|:---|:---|:---|
| **WMAE** | **1 332 $** | En moyenne, l'erreur pondérée est de 1 332 $ par magasin/département/semaine |
| **MAE** | **1 263 $** | Sans pondération fériée, l'erreur descend — les fériés sont plus difficiles à prédire |
| **RMSE** | **2 801 $** | Le ratio RMSE/MAE ≈ 2.2 indique la présence de quelques erreurs importantes |
| **R²** | **0.9837** | Le modèle explique **98.4%** de la variance des ventes — excellente qualité |
